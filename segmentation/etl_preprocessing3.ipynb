```python
# –ó–∞–≥—Ä—É–∑–∫–∞ JSON —Ñ–∞–π–ª–æ–≤
with open('translation.json', 'r', encoding='utf-8') as f:
    translation = json.load(f)
    
# –í—ã–≤–æ–¥–∏–º –ø–µ—Ä–≤—ã–µ —ç–ª–µ–º–µ–Ω—Ç—ã –∫–∞–∂–¥–æ–≥–æ —Ä–∞–∑–¥–µ–ª–∞
for dict_name, data in translation.items():
    print(f"\n=== –ü–µ—Ä–≤—ã–µ 3 —ç–ª–µ–º–µ–Ω—Ç–∞ –∏–∑ {dict_name} ===")
    
    # –ü—Ä–µ–æ–±—Ä–∞–∑—É–µ–º —Å–ª–æ–≤–∞—Ä—å –≤ —Å–ø–∏—Å–æ–∫ –ø–∞—Ä (key, value) –∏ –±–µ—Ä–µ–º –ø–µ—Ä–≤—ã–µ 3 —ç–ª–µ–º–µ–Ω—Ç–∞
    first_items = list(data.items())[:3]
    
    for key, value in first_items:
        print(f"{key}: {value}")
```

### –í—ã–ø–æ–ª–Ω–∏–º –ø—Ä–æ–≤–µ—Ä–∫—É –∏ –æ—á–∏—Å—Ç–∫—É –¥–∞–Ω–Ω—ã—Ö.

```python
def validate_and_clean_data(
    dfso: dict, 
    fk_dict: dict,
    verbose: bool = True
) -> tuple[dict, dict]:
    """
    –ê–≤—Ç–æ–º–∞—Ç–∏—á–µ—Å–∫–∞—è –ø—Ä–æ–≤–µ—Ä–∫–∞ –∏ –æ—á–∏—Å—Ç–∫–∞ –¥–∞–Ω–Ω—ã—Ö –Ω–∞ –æ—Å–Ω–æ–≤–µ —Å–ª–æ–≤–∞—Ä—è –≤–Ω–µ—à–Ω–∏—Ö –∫–ª—é—á–µ–π
    
    –ü–∞—Ä–∞–º–µ—Ç—Ä—ã:
        dfso: –°–ª–æ–≤–∞—Ä—å —Å DataFrame
        fk_dict: –°–ª–æ–≤–∞—Ä—å –≤–Ω–µ—à–Ω–∏—Ö –∫–ª—é—á–µ–π –≤ —Ñ–æ—Ä–º–∞—Ç–µ {—Ç–∞–±–ª–∏—Ü–∞: {—Å—Ç–æ–ª–±–µ—Ü: —Å—Å—ã–ª–∫–∞}}
        verbose: –§–ª–∞–≥ –≤—ã–≤–æ–¥–∞ –ø–æ–¥—Ä–æ–±–Ω–æ–π –∏–Ω—Ñ–æ—Ä–º–∞—Ü–∏–∏
        
    –í–æ–∑–≤—Ä–∞—â–∞–µ—Ç:
        –ö–æ—Ä—Ç–µ–∂ (–æ—á–∏—â–µ–Ω–Ω—ã–µ –¥–∞–Ω–Ω—ã–µ, —Å—Ç–∞—Ç–∏—Å—Ç–∏–∫–∞ —É–¥–∞–ª–µ–Ω–∏–π)
    """
    
    stats = {}
    cleaned_dfs = dfso.copy()
    
    # –ü—Ä–æ—Ö–æ–¥ –ø–æ –≤—Å–µ–º —Ç–∞–±–ª–∏—Ü–∞–º —Å FK
    for child_table, fks in fk_dict.items():
        if child_table not in cleaned_dfs:
            if verbose:
                print(f"‚ö†Ô∏è –¢–∞–±–ª–∏—Ü–∞ {child_table} –æ—Ç—Å—É—Ç—Å—Ç–≤—É–µ—Ç –≤ –¥–∞–Ω–Ω—ã—Ö, –ø—Ä–æ–ø—É—Å–∫–∞–µ–º...")
            continue
            
        df = cleaned_dfs[child_table].copy()
        initial_count = len(df)
        
        for fk_column, ref in fks.items():
            # –ü–∞—Ä—Å–∏–º —Å—Å—ã–ª–∫—É: parent_table(parent_column)
            ref_match = re.match(r"(\w+)\((\w+)\)", ref)
            if not ref_match:
                raise ValueError(f"–ù–µ–∫–æ—Ä—Ä–µ–∫—Ç–Ω—ã–π —Ñ–æ—Ä–º–∞—Ç —Å—Å—ã–ª–∫–∏: {ref}")
                
            parent_table, parent_column = ref_match.groups()
            
            # –ü—Ä–æ–≤–µ—Ä–∫–∞ —Å—É—â–µ—Å—Ç–≤–æ–≤–∞–Ω–∏—è —Ä–æ–¥–∏—Ç–µ–ª—å—Å–∫–æ–π —Ç–∞–±–ª–∏—Ü—ã
            if parent_table not in cleaned_dfs:
                raise ValueError(f"–†–æ–¥–∏—Ç–µ–ª—å—Å–∫–∞—è —Ç–∞–±–ª–∏—Ü–∞ {parent_table} –Ω–µ –Ω–∞–π–¥–µ–Ω–∞")
                
            # –ü–æ–ª—É—á–∞–µ–º –≤–∞–ª–∏–¥–Ω—ã–µ –∫–ª—é—á–∏
            valid_keys = set(cleaned_dfs[parent_table][parent_column].unique())
            
            # –§–∏–ª—å—Ç—Ä–∞—Ü–∏—è –¥–∞–Ω–Ω—ã—Ö
            mask = df[fk_column].isin(valid_keys)
            invalid_count = (~mask).sum()
            
            if verbose:
                print(f"üîç {child_table}.{fk_column} ‚Üí {parent_table}.{parent_column}")
                print(f"   –ù–∞–π–¥–µ–Ω–æ –Ω–µ–∫–æ—Ä—Ä–µ–∫—Ç–Ω—ã—Ö –∑–∞–ø–∏—Å–µ–π: {invalid_count} ({invalid_count/len(df):.1%})")
                
            df = df[mask].copy()
            
            # –û–±–Ω–æ–≤–ª—è–µ–º —Å—Ç–∞—Ç–∏—Å—Ç–∏–∫—É
            stats_key = f"{child_table}.{fk_column}"
            stats[stats_key] = stats.get(stats_key, 0) + invalid_count
            
        # –°–æ—Ö—Ä–∞–Ω—è–µ–º –æ—á–∏—â–µ–Ω–Ω—ã–µ –¥–∞–Ω–Ω—ã–µ
        cleaned_dfs[child_table] = df
        if verbose:
            print(f"‚úÖ {child_table}: –£–¥–∞–ª–µ–Ω–æ {initial_count - len(df)} –∑–∞–ø–∏—Å–µ–π")
            
    return cleaned_dfs, stats
```

## 6. ‚òÅÔ∏è –ó–∞–≥—Ä—É–∑–∫–∞ –¥–∞–Ω–Ω—ã—Ö –≤ Supabase (PostgreSQL)

- –ñ—É—Ä–Ω–∞–ª–∏—Ä–æ–≤–∞–Ω–∏–µ:

  - –∫–æ–Ω—Ç—Ä–æ–ª—å —Ç–∏–ø–æ–≤ –∏ —Ä–∞–∑–º–µ—Ä–Ω–æ—Å—Ç–µ–π.

  - –ø—Ä–æ–≤–µ—Ä–∫–∞ –æ—Ç—Å—É—Ç—Å—Ç–≤–∏—è –¥—É–±–ª–∏–∫–∞—Ç–æ–≤.

üìå _–¶–µ–ª—å: —Å–æ–∑–¥–∞—Ç—å —á–∏—Å—Ç—É—é, –Ω–æ—Ä–º–∞–ª–∏–∑–æ–≤–∞–Ω–Ω—É—é, —Å–æ–≥–ª–∞—Å–æ–≤–∞–Ω–Ω—É—é PostgreSQL-–±–∞–∑—É –ø–æ–¥ –∞–Ω–∞–ª–∏—Ç–∏—á–µ—Å–∫–∏–µ –∑–∞–ø—Ä–æ—Å—ã._

```python
# pipeline.py

import pandas as pd
import re
from sqlalchemy import create_engine, text
from sqlalchemy.pool import NullPool
from sqlalchemy.exc import ProgrammingError
from dotenv import load_dotenv
import os
import time
from tqdm import tqdm

from src.db_utils import get_engine

# –ò—Å–ø–æ–ª—å–∑—É–µ–º –∑–∞—Ä–∞–Ω–µ–µ –ø–æ–¥–≥–æ—Ç–æ–≤–ª–µ–Ω–Ω—ã–π —Å–∫—Ä–∏–ø—Ç –¥–ª—è –ø–æ–¥–∫–ª—é—á–µ–Ω–∏—è –∫ –ë–î
engine = get_engine()

# === –£—Ç–∏–ª–∏—Ç—ã === #
def map_dtype(dtype):
    if pd.api.types.is_string_dtype(dtype):
        return 'VARCHAR'
    elif pd.api.types.is_integer_dtype(dtype):
        return 'INTEGER'
    elif pd.api.types.is_float_dtype(dtype):
        return 'FLOAT'
    elif pd.api.types.is_bool_dtype(dtype):
        return 'BOOLEAN'
    elif pd.api.types.is_datetime64_any_dtype(dtype):
        return 'TIMESTAMP'
    else:
        return 'VARCHAR'

def to_snake_case(name):
    return re.sub(r'(?<!^)(?=[A-Z])', '_', name).lower().strip()

# === –ì–µ–Ω–µ—Ä–∞—Ü–∏—è DDL === #
def generate_clean_ddl(df_dict, pk_dict):
    ddl_statements = {}
    for table_name, df in df_dict.items():
        table_snake = to_snake_case(table_name)
        lines = [f"CREATE TABLE IF NOT EXISTS {table_snake} ("]
        col_defs = []

        for col in df.columns:
            col_snake = to_snake_case(col)
            sql_type = map_dtype(df[col])
            col_defs.append(f"    {col_snake} {sql_type}")
        lines.extend([line + "," for line in col_defs])

        # –£—Ç–æ—á–Ω—ë–Ω–Ω–∞—è –ª–æ–≥–∏–∫–∞ –¥–æ–±–∞–≤–ª–µ–Ω–∏—è PK
        if table_name in pk_dict:
            pk_cols = [to_snake_case(col) for col in pk_dict[table_name]]
            pk_line = f"    PRIMARY KEY ({', '.join(pk_cols)})"
            lines.append(pk_line)

        lines[-1] = lines[-1].rstrip(",")
        lines.append(");")
        ddl_statements[table_snake] = "\n".join(lines)
    return ddl_statements

# === –°–æ–∑–¥–∞–Ω–∏–µ —Ç–∞–±–ª–∏—Ü === #
def create_tables_from_ddl(ddl_statements, engine):
    with engine.connect() as conn:
        for table, ddl in tqdm(ddl_statements.items(), desc="üîß Creating tables"):
            try:
                conn.execute(text(ddl))
                print(f"‚úÖ Created: {table}")
            except Exception as e:
                print(f"‚ùå Error creating table {table}: {e}")

# === –ü—Ä–æ–≤–µ—Ä–∫–∞ –Ω–∞ —Å—É—â–µ—Å—Ç–≤–æ–≤–∞–Ω–∏–µ —Ç–∞–±–ª–∏—Ü—ã –∏ –¥–∞–Ω–Ω—ã—Ö === #
def table_has_data(table_name, engine):
    query = f"SELECT COUNT(*) FROM {table_name}"
    try:
        with engine.connect() as conn:
            result = conn.execute(text(query))
            count = result.scalar()
            return count > 0
    except ProgrammingError:
        return False  # —Ç–∞–±–ª–∏—Ü—ã –µ—â—ë –Ω–µ—Ç

# === –ó–∞–≥—Ä—É–∑–∫–∞ –¥–∞–Ω–Ω—ã—Ö —Å –ø—Ä–æ–≤–µ—Ä–∫–æ–π === #
def upload_dfso_to_db_safe(df_dict, engine):
    for table_name, df in tqdm(df_dict.items(), desc="‚¨ÜÔ∏è Uploading data"):
        table_snake = to_snake_case(table_name)
        if table_has_data(table_snake, engine):
            print(f"‚ö†Ô∏è Skipping {table_snake} ‚Äî already filled.")
            continue

        print(f"‚¨ÜÔ∏è Uploading: {table_snake}")
        try:
            df.columns = [to_snake_case(col) for col in df.columns]
            df.to_sql(
                table_snake,
                con=engine,
                if_exists='append',
                index=False,
                chunksize=1000
            )
            print(f"‚úÖ Uploaded: {table_snake}")
        except Exception as e:
            print(f"‚ùå Error uploading {table_snake}: {e}")
            print("‚è≥ Waiting 10 seconds before retry...")
            time.sleep(10)
            try:
                df.to_sql(
                    table_snake,
                    con=engine,
                    if_exists='append',
                    index=False,
                    chunksize=1000
                )
                print(f"‚úÖ Retried and uploaded: {table_snake}")
            except Exception as e:
                print(f"üö´ Second attempt failed for {table_snake}: {e}")

# === –§–∏–Ω–∞–ª—å–Ω–∞—è —Ñ—É–Ω–∫—Ü–∏—è –∑–∞–ø—É—Å–∫–∞ === #
def run_full_pipeline(dfso, engine, pk_dict):
    print("üìå Starting pipeline...")
    ddl_dict = generate_clean_ddl(dfso, pk_dict)
    create_tables_from_ddl(ddl_dict, engine)
    upload_dfso_to_db_safe(dfso, engine)
    print("üèÅ Pipeline completed.")
```

### –í—ã–ø–æ–ª–Ω–∏–º –≤–∞–ª–∏–¥–∞—Ü–∏—é —Ä–µ–∑—É–ª—å—Ç–∞—Ç–æ–≤ 

```python
def validate_tables_exist(engine, table_names):
    with engine.connect() as conn:
        for table_name in table_names:
            result = conn.execute(text(f"SELECT to_regclass('{table_name}');"))
            if result.fetchone()[0] is None:
                print(f"‚ùå Table {table_name} does not exist.")
            else:
                print(f"‚úÖ Table {table_name} exists.")
                
def validate_row_count(engine, df_dict):
    with engine.connect() as conn:
        for table_name, df in df_dict.items():
            table_snake = to_snake_case(table_name)
            result = conn.execute(text(f"SELECT COUNT(*) FROM {table_snake};"))
            db_row_count = result.fetchone()[0]
            df_row_count = len(df)
            if db_row_count == df_row_count:
                print(f"‚úÖ Row count matches for {table_snake}: {db_row_count} rows.")
            else:
                print(f"‚ùå Row count mismatch for {table_snake}: Expected {df_row_count}, got {db_row_count}.")
                
def validate_data_types(engine, df_dict):
    with engine.connect() as conn:
        for table_name, df in df_dict.items():
            table_snake = to_snake_case(table_name)
            for col in df.columns:
                col_snake = to_snake_case(col)
                # –ü–æ–ª—É—á–∞–µ–º —Ç–∏–ø –¥–∞–Ω–Ω—ã—Ö –∏–∑ –±–∞–∑—ã –¥–∞–Ω–Ω—ã—Ö
                result = conn.execute(text(f"SELECT data_type FROM information_schema.columns WHERE table_name = '{table_snake}' AND column_name = '{col_snake}';"))
                db_data_type = result.fetchone()[0]
                # –ü—Ä–æ–≤–µ—Ä—è–µ–º —Å–æ–æ—Ç–≤–µ—Ç—Å—Ç–≤–∏–µ —Ç–∏–ø–æ–≤ –¥–∞–Ω–Ω—ã—Ö
                expected_type = map_dtype(df[col])
                if db_data_type.lower() == expected_type.lower():
                    print(f"‚úÖ Data type matches for {table_snake}.{col_snake}: {db_data_type}.")
                else:
                    print(f"‚ùå Data type mismatch for {table_snake}.{col_snake}: Expected {expected_type}, got {db_data_type}.")
                
def validate_data_types(engine, df_dict):
    with engine.connect() as conn:
        for table_name, df in df_dict.items():
            table_snake = to_snake_case(table_name)
            for col in df.columns:
                col_snake = to_snake_case(col)
                
                # –ü—Ä–æ–≤–µ—Ä—è–µ–º —Å—É—â–µ—Å—Ç–≤–æ–≤–∞–Ω–∏–µ –∫–æ–ª–æ–Ω–∫–∏
                col_exists = conn.execute(text(
                    f"SELECT EXISTS ("
                    f"SELECT 1 FROM information_schema.columns "
                    f"WHERE table_name = '{table_snake}' "
                    f"AND column_name = '{col_snake}'"
                    f");"
                )).scalar()
                
                if not col_exists:
                    print(f"‚ö†Ô∏è Column {table_snake}.{col_snake} does not exist in DB")
                    continue
                
                # –ü–æ–ª—É—á–∞–µ–º —Ç–∏–ø –¥–∞–Ω–Ω—ã—Ö –∏–∑ –ë–î
                result = conn.execute(text(
                    f"SELECT data_type FROM information_schema.columns "
                    f"WHERE table_name = '{table_snake}' "
                    f"AND column_name = '{col_snake}';"
                ))
                db_data_type = result.fetchone()[0].lower()
                
                # –£—Ç–æ—á–Ω—ë–Ω–Ω—ã–π –º–∞–ø–ø–∏–Ω–≥ —Ç–∏–ø–æ–≤
                dtype_map = {
                    'object': 'text',
                    'int64': 'integer',
                    'float64': 'double precision',
                    'bool': 'boolean',
                    'datetime64[ns]': 'timestamp without time zone'
                }
                expected_type = dtype_map.get(str(df[col].dtype), 'text')
                
                # –ù–æ—Ä–º–∞–ª–∏–∑–∞—Ü–∏—è –Ω–∞–∑–≤–∞–Ω–∏–π —Ç–∏–ø–æ–≤
                type_aliases = {
                    'varchar': 'text',
                    'timestamp': 'timestamp without time zone'
                }
                db_data_type = type_aliases.get(db_data_type, db_data_type)
                
                if db_data_type == expected_type:
                    print(f"‚úÖ {table_snake}.{col_snake}: {db_data_type}")
                else:
                    print(f"‚ùå {table_snake}.{col_snake}: "
                          f"DB –∏–º–µ–µ—Ç {db_data_type}, –æ–∂–∏–¥–∞–ª—Å—è {expected_type}")

def validate_missing_values(engine, df_dict):
    with engine.connect() as conn:
        for table_name, df in df_dict.items():
            table_snake = to_snake_case(table_name)
            for col in df.columns:
                col_snake = to_snake_case(col)
                # –ü—Ä–æ–≤–µ—Ä—è–µ–º –Ω–∞–ª–∏—á–∏–µ –ø—Ä–æ–ø—É—Å–∫–æ–≤ –≤ –±–∞–∑–µ –¥–∞–Ω–Ω—ã—Ö
                result = conn.execute(text(f"SELECT COUNT(*) FROM {table_snake} WHERE {col_snake} IS NULL;"))
                db_null_count = result.fetchone()[0]
                df_null_count = df[col].isnull().sum()
                if db_null_count == df_null_count:
                    print(f"‚úÖ Missing values match for {table_snake}.{col_snake}: {db_null_count} missing values.")
                else:
                    print(f"‚ùå Missing values mismatch for {table_snake}.{col_snake}: Expected {df_null_count}, got {db_null_count}.")
                    
                    
                    
def validate_primary_keys(engine, df_dict):
    with engine.connect() as conn:
        for table_name, df in df_dict.items():
            table_snake = to_snake_case(table_name)
            # –ü—Ä–∏–º–µ—Ä –¥–ª—è –ø–µ—Ä–≤–∏—á–Ω–æ–≥–æ –∫–ª—é—á–∞
            primary_key_col = f"{table_snake[:-1]}_id"  # –ü—Ä–µ–¥–ø–æ–ª–∞–≥–∞–µ–º, —á—Ç–æ –ø–µ—Ä–≤–∏—á–Ω—ã–π –∫–ª—é—á –∏–º–µ–µ—Ç —Ç–∞–∫–æ–µ –Ω–∞–∑–≤–∞–Ω–∏–µ
            if primary_key_col in [to_snake_case(c) for c in df.columns]:
                result = conn.execute(text(f"SELECT COUNT(DISTINCT {primary_key_col}) FROM {table_snake};"))
                db_unique_count = result.fetchone()[0]
                df_unique_count = df[primary_key_col].nunique()
                if db_unique_count == df_unique_count:
                    print(f"‚úÖ Unique values match for {table_snake}.{primary_key_col}: {db_unique_count} unique values.")
                else:
                    print(f"‚ùå Unique values mismatch for {table_snake}.{primary_key_col}: Expected {df_unique_count}, got {db_unique_count}.")

                    
def run_validation(dfso, engine):
    print("üìå Starting validation...")
    table_names = [to_snake_case(table_name) for table_name in dfso.keys()]
    
    # –ü—Ä–æ–≤–µ—Ä–∫–∞ –Ω–∞–ª–∏—á–∏—è —Ç–∞–±–ª–∏—Ü
    validate_tables_exist(engine, table_names)
    
    # –ü—Ä–æ–≤–µ—Ä–∫–∞ –∫–æ–ª–∏—á–µ—Å—Ç–≤–∞ —Å—Ç—Ä–æ–∫
    validate_row_count(engine, dfso)
    
    # –ü—Ä–æ–≤–µ—Ä–∫–∞ —Ç–∏–ø–æ–≤ –¥–∞–Ω–Ω—ã—Ö
    validate_data_types(engine, dfso)
    
    # –ü—Ä–æ–≤–µ—Ä–∫–∞ –ø—Ä–æ–ø—É—Å–∫–æ–≤
    validate_missing_values(engine, dfso)
    
    # –ü—Ä–æ–≤–µ—Ä–∫–∞ —É–Ω–∏–∫–∞–ª—å–Ω–æ—Å—Ç–∏ –ø–µ—Ä–≤–∏—á–Ω—ã—Ö –∫–ª—é—á–µ–π
    validate_primary_keys(engine, dfso)

    print("üèÅ Validation completed.")
                    
```

```python
import os
from dotenv import load_dotenv
import subprocess

load_dotenv()

user = os.getenv("user")
password = os.getenv("password")
host = os.getenv("host")
port = os.getenv("port")
dbname = os.getenv("dbname")

db_url = f"postgresql+psycopg2://{user}:{password}@{host}:{port}/{dbname}"

# 1. –ì–µ–Ω–µ—Ä–∞—Ü–∏—è .dot-—Ñ–∞–π–ª–∞
subprocess.run(["eralchemy", "-i", db_url, "-o", "schema.dot", "--exclude-tables", "^pg_.*,^sql_.*"])

# 2. –ö–æ–Ω–≤–µ—Ä—Ç–∞—Ü–∏—è –≤ –∫–∞—Ä—Ç–∏–Ω–∫—É
subprocess.run(["dot", "-Tpng", "schema.dot", "-o", "schema.png"])

print("‚úÖ ER-–¥–∏–∞–≥—Ä–∞–º–º–∞ —Å–æ—Ö—Ä–∞–Ω–µ–Ω–∞ –∫–∞–∫ schema.png")
```
