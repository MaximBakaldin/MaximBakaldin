–£–Ω–∏–≤–µ—Ä—Å–∞–ª—å–Ω–∞—è –æ–±—ë—Ä—Ç–∫–∞ –¥–ª—è –≤—ã–ø–æ–ª–Ω–µ–Ω–∏—è SQL-–∑–∞–ø—Ä–æ—Å–æ–≤ —Å –∏—Å–ø–æ–ª—å–∑–æ–≤–∞–Ω–∏–µ–º SQLAlchemy. –ü–æ–∑–≤–æ–ª—è–µ—Ç –≤—ã–ø–æ–ª–Ω—è—Ç—å –∑–∞–ø—Ä–æ—Å—ã —Å –ø–∞—Ä–∞–º–µ—Ç—Ä–∞–º–∏ –∏ –ø–æ–ª—É—á–∞—Ç—å —Ä–µ–∑—É–ª—å—Ç–∞—Ç –≤ —Ä–∞–∑–ª–∏—á–Ω—ã—Ö —Ñ–æ—Ä–º–∞—Ç–∞—Ö: DataFrame, —Å–ø–∏—Å–æ–∫ —Å—Ç—Ä–æ–∫, –æ–¥–Ω–∞ —Å—Ç—Ä–æ–∫–∞ –∏–ª–∏ —Å–∫–∞–ª—è—Ä–Ω–æ–µ –∑–Ω–∞—á–µ–Ω–∏–µ.

*–ü—Ä–∏–º–µ—á–∞–Ω–∏—è:*
*–ò—Å–ø–æ–ª—å–∑–æ–≤–∞–Ω–∏–µ –ø–∞—Ä–∞–º–µ—Ç—Ä–æ–≤ —á–µ—Ä–µ–∑ :placeholder –∑–∞—â–∏—â–∞–µ—Ç –æ—Ç SQL-–∏–Ω—ä–µ–∫—Ü–∏–π.*

*–§—É–Ω–∫—Ü–∏—è –∞–≤—Ç–æ–º–∞—Ç–∏—á–µ—Å–∫–∏ –æ—Ç–∫—Ä—ã–≤–∞–µ—Ç –∏ –∑–∞–∫—Ä—ã–≤–∞–µ—Ç —Å–æ–µ–¥–∏–Ω–µ–Ω–∏–µ –ø—Ä–∏ –∫–∞–∂–¥–æ–º –≤—ã–∑–æ–≤–µ.*

```python
def run_query(engine, query_str, params=None, fetch="df"):
    """
    –í—ã–ø–æ–ª–Ω—è–µ—Ç SQL-–∑–∞–ø—Ä–æ—Å –∏ –≤–æ–∑–≤—Ä–∞—â–∞–µ—Ç —Ä–µ–∑—É–ª—å—Ç–∞—Ç.

    –ü–∞—Ä–∞–º–µ—Ç—Ä—ã:
    - engine: SQLAlchemy engine
    - query_str: SQL-–∑–∞–ø—Ä–æ—Å (—Å :placeholder –µ—Å–ª–∏ –Ω—É–∂–Ω—ã –ø–∞—Ä–∞–º–µ—Ç—Ä—ã)
    - params: dict —Å –ø–∞—Ä–∞–º–µ—Ç—Ä–∞–º–∏
    - fetch: one / all / scalar / df

    –í–æ–∑–≤—Ä–∞—â–∞–µ—Ç:
    - —Å–ø–∏—Å–æ–∫ —Å—Ç—Ä–æ–∫, –æ–¥–Ω—É —Å—Ç—Ä–æ–∫—É, —Å–∫–∞–ª—è—Ä –∏–ª–∏ DataFrame (–µ—Å–ª–∏ fetch='df')
    """
    query = text(query_str)
    with engine.connect() as conn:
        result = conn.execute(query, params or {})

        if fetch == "df":
            df = pd.DataFrame(result.fetchall(), columns=result.keys())
            return df
        elif fetch == "all":
            return result.fetchall()
        elif fetch == "one":
            return result.fetchone()
        elif fetch == "scalar":
            return result.scalar()
        else:
            raise ValueError("fetch –¥–æ–ª–∂–µ–Ω –±—ã—Ç—å 'all', 'one', 'scalar' –∏–ª–∏ 'df'")
```

## **Acquisition (–ü—Ä–∏–≤–ª–µ—á–µ–Ω–∏–µ)**
*HEART ‚Äì Adoption: - –£—Å–ø–µ—à–Ω–æ—Å—Ç—å –∫–ª—é—á–µ–≤–æ–≥–æ –¥–µ–π—Å—Ç–≤–∏—è: –ø–µ—Ä–≤—ã–π –∑–∞–∫–∞–∑ –ø—Ä–æ–¥–∞–≤—Ü–∞.*


- **–¶–µ–ª—å:** –ü–æ–Ω—è—Ç—å, –∫–∞–∫ –ø—Ä–∏–≤–ª–µ–∫–∞—é—Ç—Å—è –Ω–æ–≤—ã–µ –∫–ª–∏–µ–Ω—Ç—ã –∏ –ø—Ä–æ–¥–∞–≤—Ü—ã, –∏ –∫–∞–∫–∏–µ –∫–∞–Ω–∞–ª—ã / —Å–µ–≥–º–µ–Ω—Ç—ã —ç—Ñ—Ñ–µ–∫—Ç–∏–≤–Ω—ã –¥–ª—è –∞–∫—Ç–∏–≤–∞—Ü–∏–∏ –ª–∏–¥–æ–≤.


- **–ö–ª—é—á–µ–≤—ã–µ –º–µ—Ç—Ä–∏–∫–∏:**  
    - –û–±—â–∏–π –æ—Ö–≤–∞—Ç.
    - –ù–æ–≤—ã–µ –ø–æ–∫—É–ø–∞—Ç–µ–ª–∏ –ø–æ –º–µ—Å—è—Ü–∞–º;
    - –ö–æ–ª–∏—á–µ—Å—Ç–≤–æ –ª–∏–¥–æ–≤ –ø–æ –∫–∞–Ω–∞–ª–∞–º –ø—Ä–∏–≤–ª–µ—á–µ–Ω–∏—è;
    - –ö–æ–Ω–≤–µ—Ä—Å–∏—è –∏–∑ –ª–∏–¥–∞ –≤ –ø–µ—Ä–≤—ã–π –∑–∞–∫–∞–∑;
    - –ö–æ–Ω–≤–µ—Ä—Å–∏—è –ø–æ –∫–∞–Ω–∞–ª–∞–º –ø—Ä–∏–≤–ª–µ—á–µ–Ω–∏—è;

- **–û–≥—Ä–∞–Ω–∏—á–µ–Ω–∏—è –∞–Ω–∞–ª–∏–∑–∞:**

    - –û—Ç—Å—É—Ç—Å—Ç–≤—É–µ—Ç –∏–Ω—Ñ–æ—Ä–º–∞—Ü–∏—è –æ –¥–∞—Ç–µ —Ä–µ–≥–∏—Å—Ç—Ä–∞—Ü–∏ –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª–µ–π.
    - –ò–Ω—Ñ–æ—Ä–º–∞—Ü–∏—è –ø–æ –∫–∞–Ω–∞–ª –ø—Ä–∏–≤–ª–µ—á–µ–Ω–∏—è –ø—Ä–µ–¥—Å—Ç–∞–≤–ª–µ–Ω–∞ –¥–ª—è –æ–≥—Ä–∞–Ω–∏—á–µ–Ω–Ω–æ–π –≤—ã–±–æ—Ä–∫–∏, –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –ø—Ä–æ–¥–∞–≤—Ü–æ–≤ –≤ –∫–æ—Ç–æ—Ä–æ–π –º–µ–Ω—å—à–µ —á–µ–º –≤ –æ—Å–Ω–æ–≤–Ω–æ–º –¥–∞—Ç–∞—Å–µ—Ç–µ.
    - –ù–µ –≤—Å–µ –∫–∞–Ω–∞–ª—ã –ø—Ä–µ–¥—Å—Ç–∞–≤–ª–µ–Ω—ã —Ä–∞–≤–Ω–æ–º–µ—Ä–Ω–æ –ø–æ –æ–±—ä—ë–º—É –ª–∏–¥–æ–≤ ‚Üí —Å—Ä–∞–≤–Ω–µ–Ω–∏–µ CR –ø–æ –∫–∞–Ω–∞–ª–∞–º —Ç—Ä–µ–±—É–µ—Ç –æ—Å—Ç–æ—Ä–æ–∂–Ω–æ—Å—Ç–∏.
    - –û—Ç—Å—É—Ç—Å—Ç–≤—É–µ—Ç –∏–Ω—Ñ–æ—Ä–º–∞—Ü–∏—è –æ –±—é–¥–∂–µ—Ç–µ –Ω–∞ –∫–∞–Ω–∞–ª—ã ‚Äî ROI –∏ CAC –Ω–µ —Ä–∞—Å—Å—á–∏—Ç—ã–≤–∞—é—Ç—Å—è.

- **–í—ã–≤–æ–¥:**  
    - –ü–ª–∞—Ç—Ñ–æ—Ä–º–∞ –¥–µ–º–æ–Ω—Å—Ç—Ä–∏—Ä—É–µ—Ç —É—Å—Ç–æ–π—á–∏–≤—ã–π —Ä–æ—Å—Ç –Ω–æ–≤—ã—Ö –ø–æ–∫—É–ø–∞—Ç–µ–ª–µ–π, –æ—Å–æ–±–µ–Ω–Ω–æ –≤ —Å–µ–∑–æ–Ω–Ω—ã–µ –ø–µ—Ä–∏–æ–¥—ã. 
    - –ö–∞–Ω–∞–ª—ã –æ—Ä–≥–∞–Ω–∏–∫–∏ –∏ –ø–ª–∞—Ç–Ω–æ–≥–æ –ø–æ–∏—Å–∫–∞ –ø—Ä–∏–Ω–æ—Å—è—Ç –Ω–∞–∏–±–æ–ª—å—à–∏–π –æ–±—ä—ë–º –ª–∏–¥–æ–≤ —Å —Ö–æ—Ä–æ—à–µ–π –∫–æ–Ω–≤–µ—Ä—Å–∏–µ–π; –ø—Ä—è–º–æ–π —Ç—Ä–∞—Ñ–∏–∫ –∏ —Ä–µ—Ñ–µ—Ä–∞–ª—ã –ø–æ–∫–∞–∑—ã–≤–∞—é—Ç –Ω–∞–∏–≤—ã—Å—à–∏–π CR, –Ω–æ –∏–º–µ—é—Ç –Ω–∏–∑–∫–∏–π –æ—Ö–≤–∞—Ç. –°–æ—Ü–∏–∞–ª—å–Ω—ã–µ —Å–µ—Ç–∏ –æ–±–µ—Å–ø–µ—á–∏–≤–∞—é—Ç —à–∏—Ä–æ–∫–∏–π —Ç—Ä–∞—Ñ–∏–∫, –Ω–æ —Ç—Ä–µ–±—É—é—Ç –¥–æ–Ω–∞—Å—Ç—Ä–æ–π–∫–∏ –¥–ª—è –ø–æ–≤—ã—à–µ–Ω–∏—è –∫–∞—á–µ—Å—Ç–≤–∞ –ª–∏–¥–æ–≤. 
    - –°—Ä–µ–¥–∏ —Å–∫–æ–Ω–≤–µ—Ä—Ç–∏—Ä–æ–≤–∞–Ω–Ω—ã—Ö –ª–∏–¥–æ–≤ –ø—Ä–µ–æ–±–ª–∞–¥–∞—é—Ç —Å–µ–≥–º–µ–Ω—Ç—ã –¥–æ–º–∞—à–Ω–µ–≥–æ –¥–µ–∫–æ—Ä–∞, –∫—Ä–∞—Å–æ—Ç—ã –∏ –∞–≤—Ç–æ–∞–∫—Å–µ—Å—Å—É–∞—Ä–æ–≤. 
    - –ï—Å—Ç—å –ø–æ—Ç–µ–Ω—Ü–∏–∞–ª –≤ —Ä–∞–∑–≤–∏—Ç–∏–∏ —Ä–µ—Ñ–µ—Ä–∞–ª—å–Ω–æ–π —Å–∏—Å—Ç–µ–º—ã –∏ –æ–ø—Ç–∏–º–∏–∑–∞—Ü–∏–∏ –∫–∞–Ω–∞–ª–æ–≤ —Å –Ω–∏–∑–∫–æ–π –∫–æ–Ω–≤–µ—Ä—Å–∏–µ–π.

### –í—Å–µ–≥–æ —É–Ω–∏–∫–∞–ª—å–Ω—ã—Ö –∫–ª–∏–µ–Ω—Ç–æ–≤ —Å–µ—Ä–≤–∏—Å–∞.

- **–û–ø–∏—Å–∞–Ω–∏–µ –º–µ—Ç—Ä–∏–∫–∏:**  
  –ü–æ–∫–∞–∑—ã–≤–∞–µ—Ç –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ —É–Ω–∏–∫–∞–ª—å–Ω—ã—Ö –ø–æ–∫—É–ø–∞—Ç–µ–ª–µ–π –∏ –ø—Ä–æ–¥–∞–≤—Ü–æ–≤ –∑–∞ –≤–µ—Å—å –ø–µ—Ä–∏–æ–¥ –∏—Å—Å–ª–µ–¥–æ–≤–∞–Ω–∏—è.

- **–ë–∏–∑–Ω–µ—Å-–∏–Ω—Ç–µ—Ä–ø—Ä–µ—Ç–∞—Ü–∏—è:**  
  –ü–æ–∑–≤–æ–ª—è–µ—Ç –æ—Ü–µ–Ω–∏—Ç—å –æ—Ö–≤–∞—Ç –∫–ª–∏–µ–Ω—Ç—Å–∫–æ–π –±–∞–∑—ã –ø–æ —Ä–µ–≥–∏–æ–Ω–∞–º.
- **–î–æ–ø–æ–ª–Ω–∏—Ç–µ–ª—å–Ω—ã–µ –¥–µ—Ç–∞–ª–∏:**  
  –ò—Å–ø–æ–ª—å–∑—É–µ–º –≤–µ—Å—å –æ–±—ä–µ–º –∏–Ω—Ñ–æ—Ä–º–∞—Ü–∏–∏ –æ –¥–æ—Å—Ç–∞–≤–ª–µ–Ω–Ω—ã—Ö –∑–∞–∫–∞–∑–∞—Ö.

```python
query = """
SELECT seller_state AS state,
        COUNT(DISTINCT seller_id) AS count,
        '–ü—Ä–æ–¥–∞–≤—Ü—ã' AS type
FROM sellers
GROUP BY 1

UNION

SELECT customer_state AS state,
        COUNT(DISTINCT customer_unique_id) AS count,
        '–ü–æ–∫—É–ø–∞—Ç–µ–ª–∏' AS type
FROM customers 
GROUP BY 1;
"""
states = run_query(engine, query, fetch='df')

# –ó–∞–≥—Ä—É–∑–∏–º –≥–µ–æ–º–µ—Ç—Ä–∏—é —à—Ç–∞—Ç–æ–≤ –ë—Ä–∞–∑–∏–ª–∏–∏
geo_url = 'https://raw.githubusercontent.com/codeforamerica/click_that_hood/master/public/data/brazil-states.geojson'
geo_json = requests.get(geo_url).json()


# –ü—Ä–∏–≤–æ–¥–∏–º –ö–û–î –∫ –ø–æ–ª–Ω–æ–º—É –Ω–∞–∑–≤–∞–Ω–∏—é
state_name_mapping = {
    'AC': 'Acre', 'AL': 'Alagoas', 'AM': 'Amazonas', 'AP': 'Amap√°', 'BA': 'Bahia',
    'CE': 'Cear√°', 'DF': 'Distrito Federal', 'ES': 'Esp√≠rito Santo', 'GO': 'Goi√°s',
    'MA': 'Maranh√£o', 'MG': 'Minas Gerais', 'MS': 'Mato Grosso do Sul',
    'MT': 'Mato Grosso', 'PA': 'Par√°', 'PB': 'Para√≠ba', 'PE': 'Pernambuco',
    'PI': 'Piau√≠', 'PR': 'Paran√°', 'RJ': 'Rio de Janeiro', 'RN': 'Rio Grande do Norte',
    'RO': 'Rond√¥nia', 'RR': 'Roraima', 'RS': 'Rio Grande do Sul', 'SC': 'Santa Catarina',
    'SE': 'Sergipe', 'SP': 'S√£o Paulo', 'TO': 'Tocantins'
}

states['state_name'] = states['state'].map(state_name_mapping)
# –õ–æ–≥–∞—Ä–∏—Ñ–º–∏—Ä—É–µ–º –∑–Ω–∞—á–µ–Ω–∏—è
states['log_count'] = np.log1p(states['count'])

# –†–∞–∑–¥–µ–ª–∏–º –¥–∞–Ω–Ω—ã–µ
buyers = states[states['type'] == '–ü–æ–∫—É–ø–∞—Ç–µ–ª–∏']
sellers = states[states['type'] == '–ü—Ä–æ–¥–∞–≤—Ü—ã']

# –°–æ–∑–¥–∞—ë–º —Ñ–∏–≥—É—Ä—É —Å –¥–≤—É–º—è –ø–æ–¥–≥—Ä–∞—Ñ–∏–∫–∞–º–∏
fig = make_subplots(
    horizontal_spacing=0.1,
    rows=1, cols=2,
    subplot_titles=(f"–í—Å–µ–≥–æ –ø–æ–∫—É–ø–∞—Ç–µ–ª–µ–π: {buyers['count'].sum()}", f"–í—Å–µ–≥–æ –ø—Ä–æ–¥–∞–≤—Ü–æ–≤: {sellers['count'].sum()}"),
    specs=[[{"type": "mapbox"}, {"type": "mapbox"}]]
)



# –î–æ–±–∞–≤–∏–º –∫–∞—Ä—Ç—É –ø–æ–∫—É–ø–∞—Ç–µ–ª–µ–π
fig.add_trace(
    go.Choroplethmapbox(
        geojson=geo_json,
        locations=buyers['state_name'],
        z=buyers['count'],
        colorscale='Greys',
        #marker_opacity=0.6,
        featureidkey='properties.name',
        name='–ü–æ–∫—É–ø–∞—Ç–µ–ª–∏',
        hovertemplate='<b>%{location}</b><br>–ü–æ–∫—É–ø–∞—Ç–µ–ª–µ–π: %{z}<extra></extra>',
        showscale=False
    ),
    row=1, col=1
)

# –î–æ–±–∞–≤–∏–º –∫–∞—Ä—Ç—É –ø—Ä–æ–¥–∞–≤—Ü–æ–≤
fig.add_trace(
    go.Choroplethmapbox(
        geojson=geo_json,
        locations=sellers['state_name'],
        z=sellers['count'],
        colorscale='Blues',
        #marker_opacity=0.6,
        featureidkey='properties.name',
        name='–ü—Ä–æ–¥–∞–≤—Ü—ã',
        hovertemplate='<b>%{location}</b><br>–ü—Ä–æ–¥–∞–≤—Ü–æ–≤: %{z}<extra></extra>',
        showscale=False),
    row=1, col=2)

# –ù–∞—Å—Ç—Ä–æ–π–∫–∏ –∫–∞—Ä—Ç—ã –∏ layout
fig.update_layout(
    mapbox1=dict(
        style="white-bg",
        zoom=2.7,
        center={"lat": -14.2350, "lon": -51.9253}
    ),
    mapbox2=dict(
        style="white-bg",
        zoom=2.7,
        center={"lat": -14.2350, "lon": -51.9253}
    ),
    height=600,
    margin={"r":0,"t":100,"l":0,"b":0},
    title_text="–†–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ –ø—Ä–æ–¥–∞–≤—Ü–æ–≤ –∏ –ø–æ–∫—É–ø–∞—Ç–µ–ª–µ–π –ø–æ —Ä–µ–≥–∏–æ–Ω–∞–º –ë—Ä–∞–∑–∏–ª–∏–∏"
)

fig.write_image('1.png')
```

### –ù–æ–≤—ã–µ –ø–æ–∫—É–ø–∞—Ç–µ–ª–∏ –∏ –ø—Ä–æ–¥–∞–≤—Ü—ã –ø–æ –º–µ—Å—è—Ü–∞–º

  –ê–Ω–∞–ª–∏–∑–∏—Ä—É–µ—Ç –¥–∏–Ω–∞–º–∏–∫—É –ø—Ä–∏–≤–ª–µ—á–µ–Ω–∏—è –Ω–æ–≤—ã—Ö –∫–ª–∏–µ–Ω—Ç–æ–≤. –í–∞–∂–Ω–æ –¥–ª—è –æ—Ü–µ–Ω–∫–∏ —ç—Ñ—Ñ–µ–∫—Ç–∏–≤–Ω–æ—Å—Ç–∏ –º–∞—Ä–∫–µ—Ç–∏–Ω–≥–æ–≤—ã—Ö —Å—Ç—Ä–∞—Ç–µ–≥–∏–π –∏ –≤—ã—è–≤–ª–µ–Ω–∏—è —Ç—Ä–µ–Ω–¥–æ–≤ –≤ –ø–æ–≤–µ–¥–µ–Ω–∏–∏ –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª–µ–π.

- **–î–æ–ø–æ–ª–Ω–∏—Ç–µ–ª—å–Ω—ã–µ –¥–µ—Ç–∞–ª–∏:**  
  - –ò—Å–ø–æ–ª—å–∑—É–µ—Ç—Å—è —Ç–æ–ª—å–∫–æ –ø–µ—Ä–≤—ã–π –∑–∞–∫–∞–∑ –∏–ª–∏ –ø—Ä–æ–¥–∞–∂–∞ –∫–ª–∏–µ–Ω—Ç–∞.
  - –§–∏–ª—å—Ç—Ä—É–µ–º –∑–∞–∫–∞–∑—ã –ø–æ —Å—Ç–∞—Ç—É—Å—É '–¥–æ—Å—Ç–∞–≤–ª–µ–Ω'.

```python
query =  """
-- –ü–æ–ª—É—á–∞–µ–º –¥–∞—Ç—É –ø–µ—Ä–≤–æ–≥–æ –∑–∞–∫–∞–∑–∞ –¥–ª—è –∫–∞–∂–¥–æ–≥–æ –∫–ª–∏–µ–Ω—Ç–∞
WITH first_orders AS (
  SELECT 
    c.customer_unique_id,
    MIN(o.order_purchase_timestamp)::DATE AS first_order_date
  FROM orders o
  JOIN customers c ON o.customer_id = c.customer_id
  WHERE o.order_status = '–¥–æ—Å—Ç–∞–≤–ª–µ–Ω'
  GROUP BY c.customer_unique_id
),
-- –ü–æ–ª—É—á–∞–µ–º –¥–∞—Ç—É –ø–µ—Ä–≤–æ–π –ø—Ä–æ–¥–∞–∂–∏ –¥–ª—è –∫–∞–∂–¥–æ–≥–æ –ø—Ä–æ–¥–∞–≤—Ü–∞
first_sales AS (
  SELECT 
    oi.seller_id,
    MIN(o.order_purchase_timestamp)::DATE AS first_sale_date
  FROM order_items oi
  JOIN orders o ON oi.order_id = o.order_id
  WHERE o.order_status = '–¥–æ—Å—Ç–∞–≤–ª–µ–Ω'
  GROUP BY oi.seller_id
),
-- –°—á–∏—Ç–∞–µ–º –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –Ω–æ–≤—ã—Ö –∫–ª–∏–µ–Ω—Ç–æ–≤ –ø–æ –º–µ—Å—è—Ü—É
monthly_customers AS (
  SELECT 
    DATE_TRUNC('month', first_order_date) AS event_month,
    COUNT(DISTINCT customer_unique_id) AS new_customers
  FROM first_orders
  GROUP BY event_month
),

-- –°—á–∏—Ç–∞–µ–º –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –Ω–æ–≤—ã—Ö –ø—Ä–æ–¥–∞–≤—Ü–æ–≤ –ø–æ –º–µ—Å—è—Ü—É
monthly_sellers AS (
  SELECT 
    DATE_TRUNC('month', first_sale_date) AS event_month,
    COUNT(DISTINCT seller_id) AS new_sellers
  FROM first_sales
  GROUP BY event_month
)

-- –í—ã–≤–æ–¥–∏–º –æ–±—â—É—é —Ç–∞–±–ª–∏—Ü—É –æ–±—ä–µ–¥–∏–Ω—ë–Ω–Ω—É—é –ø–æ –¥–∞—Ç–∞–º
SELECT 
  COALESCE(mc.event_month, ms.event_month) AS event_month,
  COALESCE(new_customers, 0) AS new_customers,
  COALESCE(new_sellers, 0) AS new_sellers
FROM monthly_customers mc
FULL OUTER JOIN monthly_sellers ms 
  ON mc.event_month = ms.event_month
ORDER BY event_month;

"""
new_sc = run_query(engine, query, fetch='df')

# –ü–æ—Å—á–∏—Ç–∞–µ–º —Å—Ä–µ–¥–Ω–µ–µ –∑–Ω–∞—á–µ–Ω–∏–µ
mean_sellers = new_sc['new_sellers'].mean().round()

# –ü–∞–ª–∏—Ç—Ä–∞ —Ü–≤–µ—Ç–æ–≤: —É–∑–Ω–∞–≤–∞–µ–º–æ—Å—Ç—å
color_customers = 'lightblue'
color_sellers = 'darkgray'

# –°–æ–∑–¥–∞–Ω–∏–µ —Å–∞–±–ø–ª–æ—Ç–∞
fig = make_subplots(
    rows=2, cols=1,
    row_heights=[0.7, 0.3],
    shared_xaxes=True,
    vertical_spacing=0.1,
    subplot_titles=("–ù–æ–≤—ã–µ –ø–æ–∫—É–ø–∞—Ç–µ–ª–∏", "–ù–æ–≤—ã–µ –ø—Ä–æ–¥–∞–≤—Ü—ã")
)

# –í–µ—Ä—Ö–Ω–∏–π –≥—Ä–∞—Ñ–∏–∫ ‚Äî –ø–æ–∫—É–ø–∞—Ç–µ–ª–∏
fig.add_trace(
    go.Scatter(
        x=new_sc['event_month'],
        y=new_sc['new_customers'],
        mode='lines',
        name='–ü–æ–∫—É–ø–∞—Ç–µ–ª–∏',
        line=dict(color=custom_palette[4], width=3),
    ),
    row=1, col=1
)

# –ù–∏–∂–Ω–∏–π –≥—Ä–∞—Ñ–∏–∫ ‚Äî –ø—Ä–æ–¥–∞–≤—Ü—ã
fig.add_trace(
    go.Scatter(
        x=new_sc['event_month'],
        y=new_sc['new_sellers'],
        mode='lines',
        name='–ü—Ä–æ–¥–∞–≤—Ü—ã',
        line=dict(color=custom_palette[0], width=3),
    ),
    row=2, col=1
)

fig.add_hline(
    y=mean_sellers, 
    line_width=2, 
    line_dash="dash", 
    line_color=custom_palette[0],
    row=2, col=1
)


# –ù–∞—Å—Ç—Ä–æ–π–∫–∏
fig.update_layout(
    title_text="<b>–î–∏–Ω–∞–º–∏–∫–∞ –ø—Ä–∏–≤–ª–µ—á–µ–Ω–∏—è: –ø–æ–∫—É–ø–∞—Ç–µ–ª–∏ –∏ –ø—Ä–æ–¥–∞–≤—Ü—ã</b>",
    height=600,
    showlegend=False,
    margin=dict(l=60, r=30, t=100, b=60),
    font=dict(family="Segoe UI", size=14, color="#333"),
    xaxis2=dict(title="–ú–µ—Å—è—Ü"),
    yaxis=dict(title="–ü–æ–∫—É–ø–∞—Ç–µ–ª–∏"),
    yaxis2=dict(title="–ü—Ä–æ–¥–∞–≤—Ü—ã")
)

fig.write_image('1.png')
```

**üìä –ò–Ω—Ç–µ—Ä–ø—Ä–µ—Ç–∞—Ü–∏—è —Ä–µ–∑—É–ª—å—Ç–∞—Ç–∞:**
- –ü—Ä–æ–¥—É–∫—Ç –Ω–∞—Ö–æ–¥–∏—Ç—Å—è –Ω–∞ —Ä–∞–Ω–Ω–µ–π —Å—Ç–∞–¥–∏–∏ –∂–∏–∑–Ω–∏ –∏ –∞–∫—Ç–∏–≤–Ω–æ –ø—Ä–∏–≤–ª–µ–∫–∞–µ—Ç –∫–∞–∫ –ø—Ä–æ–¥–∞–≤—Ü–æ–≤ —Ç–∞–∫ –∏ –ø–æ–∫—É–ø–∞—Ç–µ–ª–µ–π.
- –ö–æ–ª–∏—á–µ—Å—Ç–≤–æ –Ω–æ–≤—ã—Ö –∫–ª–∏–µ–Ω—Ç–æ–≤ –∏–º–µ–µ—Ç —Ä–∞—Å—Ç—É—â–∏–π —Ç—Ä–µ–Ω–¥, —á—Ç–æ —É–∫–∞–∑—ã–≤–∞–µ—Ç –Ω–∞ —É—Å–∏–ª–µ–Ω–∏–µ –º–∞—Ä–∫–µ—Ç–∏–Ω–≥–æ–≤–æ–π –∞–∫—Ç–∏–≤–Ω–æ—Å—Ç–∏ –∏/–∏–ª–∏ —Ä–æ—Å—Ç —É–∑–Ω–∞–≤–∞–µ–º–æ—Å—Ç–∏ –ø–ª–∞—Ç—Ñ–æ—Ä–º—ã. –ü–∏–∫–æ–≤—ã–µ –∑–Ω–∞—á–µ–Ω–∏—è –≤ –∫–æ–Ω—Ü–µ 2017 –≥–æ–¥–∞ –º–æ–≥—É—Ç –±—ã—Ç—å —Å–≤—è–∑–∞–Ω—ã —Å —Å–µ–∑–æ–Ω–Ω—ã–º–∏ —Ä–∞—Å–ø—Ä–æ–¥–∞–∂–∞–º–∏.
- –î–∏–Ω–∞–º–∏–∫–∞ –ø—Ä–∏–≤–ª–µ—á–µ–Ω–∏—è –ø—Ä–æ–¥–∞–≤—Ü–æ–≤ –Ω–µ—Ä–∞–≤–Ω–æ–º–µ—Ä–Ω–∞, —Å–æ —Å—Ä–µ–¥–Ω–µ–º–µ—Å—è—á–Ω—ã–º –ø–æ–∫–∞–∑–∞—Ç–µ–ª–µ–º 129 –Ω–æ–≤—ã—Ö –ø—Ä–æ–¥–∞–≤—Ü–æ–≤.

### –†–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ –ø–æ –∫–∞–Ω–∞–ª–∞–º –ø—Ä–∏–≤–ª–µ—á–µ–Ω–∏—è.


- **–û–ø–∏—Å–∞–Ω–∏–µ –º–µ—Ç—Ä–∏–∫–∏:**  
  –ü–æ–∫–∞–∑—ã–≤–∞–µ—Ç, —Å–∫–æ–ª—å–∫–æ –Ω–æ–≤—ã—Ö –ø–æ—Ç–µ–Ω—Ü–∏–∞–ª—å–Ω—ã—Ö –ø—Ä–æ–¥–∞–≤—Ü–æ–≤ –ø—Ä–∏—à–ª–æ –∏–∑ –∫–∞–∂–¥–æ–≥–æ –∏—Å—Ç–æ—á–Ω–∏–∫–∞.

- **–ë–∏–∑–Ω–µ—Å-–∏–Ω—Ç–µ—Ä–ø—Ä–µ—Ç–∞—Ü–∏—è:**  
  –ü–æ–º–æ–≥–∞–µ—Ç –ø–æ–Ω—è—Ç—å, –∫–∞–∫–∏–µ –º–∞—Ä–∫–µ—Ç–∏–Ω–≥–æ–≤—ã–µ –∫–∞–Ω–∞–ª—ã –≥–µ–Ω–µ—Ä–∏—Ä—É—é—Ç –º–∞–∫—Å–∏–º–∞–ª—å–Ω—ã–π –æ–±—ä—ë–º –ª–∏–¥–æ–≤. –í–∞–∂–Ω–æ –¥–ª—è –ø–ª–∞–Ω–∏—Ä–æ–≤–∞–Ω–∏—è –º–∞—Ä–∫–µ—Ç–∏–Ω–≥–æ–≤–æ–≥–æ –±—é–¥–∂–µ—Ç–∞ –∏ –æ—Ü–µ–Ω–∫–∏ —ç—Ñ—Ñ–µ–∫—Ç–∏–≤–Ω–æ—Å—Ç–∏ —Ä–µ–∫–ª–∞–º–Ω—ã—Ö –∫–∞–º–ø–∞–Ω–∏–π.

- **–î–æ–ø–æ–ª–Ω–∏—Ç–µ–ª—å–Ω—ã–µ –¥–µ—Ç–∞–ª–∏:**  
    –ü—Ä–µ–¥—Å—Ç–∞–≤–ª–µ–Ω–Ω—ã–µ –¥–∞–Ω–Ω—ã–µ –æ—Ö–≤–∞—Ç—ã–≤–∞—é—Ç –Ω–µ –≤—Å–µ—Ö –ø—Ä–æ–¥–∞–≤—Ü–æ–≤, –æ–¥–Ω–∞–∫–æ –ø–æ–∑–≤–æ–ª—è—é—Ç –æ—Ç—Ä–∞–∑–∏—Ç—å –æ–±—â—É—é –∫–∞—Ä—Ç–∏–Ω—É –ø–æ –ø—Ä–∏–≤–ª–µ—á–µ–Ω–∏—é.
---

```python
# –í—ã—á–∏—Å–ª–∏–º –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –ª–∏–¥–æ–≤ –ø–æ –∫–∞–Ω–∞–ª–∞–º.
query = """
SELECT 
  COALESCE(
    origin, '–Ω–µ–∏–∑–≤–µ—Å—Ç–Ω—ã–π'
  ) AS channel, 
  ROUND(COUNT(DISTINCT mql_id) / SUM(COUNT(DISTINCT mql_id)) OVER() * 100, 2) AS leads 
FROM 
  marketing_qualified 
WHERE 
  first_contact_date BETWEEN '2017-01-01' 
  AND '2018-05-31' 
  AND origin != '–Ω–µ–∏–∑–≤–µ—Å—Ç–Ω—ã–π' 
GROUP BY 
  origin 
ORDER BY 
  2 ASC;
"""
leads = run_query(engine, query, fetch="df")

fig = px.bar(leads, y='channel', x='leads', orientation='h',  
             text_auto=True,
             labels ={
                 'channel':'–ö–∞–Ω–∞–ª',
                 'leads':'–î–æ–ª—è, %'
             },
            title="<b>–†–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ –ø–æ –∫–∞–Ω–∞–ª–∞–º –ø—Ä–∏–≤–ª–µ—á–µ–Ω–∏—è</b>",
            color_discrete_sequence=custom_palette[4:5])
fig.update_traces(texttemplate='%{x:.2f}%', textposition='outside')
fig.update_layout(
    showlegend=False,
)
fig.write_image('1.png')
```

**üìä –ò–Ω—Ç–µ—Ä–ø—Ä–µ—Ç–∞—Ü–∏—è —Ä–µ–∑—É–ª—å—Ç–∞—Ç–∞:**
- –û—Ä–≥–∞–Ω–∏–∫–∞ –∏ –ø–ª–∞—Ç–Ω—ã–π –ø–æ–∏—Å–∫ –ø—Ä–∏–Ω–æ—Å—è—Ç —Å—Ç–∞–±–∏–ª—å–Ω—ã–π –ø–æ—Ç–æ–∫ –ª–∏–¥–æ–≤. 
- –°–æ—Ü–∏–∞–ª—å–Ω—ã–µ —Å–µ—Ç–∏ ‚Äî —Ç—Ä–µ—Ç–∏–π –ø–æ –æ–±—ä—ë–º—É –∏—Å—Ç–æ—á–Ω–∏–∫, —á—Ç–æ –º–æ–∂–µ—Ç –≥–æ–≤–æ—Ä–∏—Ç—å –æ –ø–æ—Ç–µ–Ω—Ü–∏–∞–ª–µ –¥–ª—è –¥–∞–ª—å–Ω–µ–π—à–µ–≥–æ –º–∞—Å—à—Ç–∞–±–∏—Ä–æ–≤–∞–Ω–∏—è. 
- –ú–µ–¥–∏–π–Ω–∞—è —Ä–µ–∫–ª–∞–º–∞ –¥–∞—ë—Ç –Ω–∞–∏–º–µ–Ω—å—à–∏–π —Ä–µ–∑—É–ª—å—Ç–∞—Ç ‚Äî –≤–æ–∑–º–æ–∂–Ω–æ, —Ç—Ä–µ–±—É–µ—Ç—Å—è –ø–µ—Ä–µ–æ—Ü–µ–Ω–∫–∞ –∫—Ä–µ–∞—Ç–∏–≤–æ–≤ –∏–ª–∏ –∞—É–¥–∏—Ç–æ—Ä–∏–∏.
---

### –ö–æ–Ω–≤–µ—Ä—Å–∏—è –≤ —Å–¥–µ–ª–∫—É.

- **–û–ø–∏—Å–∞–Ω–∏–µ –º–µ—Ç—Ä–∏–∫–∏:**  
–ü—Ä–æ—Ü–µ–Ω—Ç –ª–∏–¥–æ–≤, –∫–æ—Ç–æ—Ä—ã–µ –≤ –∏—Ç–æ–≥–µ —Å–æ–≤–µ—Ä—à–∏–ª–∏ —Ö–æ—Ç—è –±—ã –æ–¥–∏–Ω –∑–∞–∫–∞–∑.

- **–ë–∏–∑–Ω–µ—Å-–∏–Ω—Ç–µ—Ä–ø—Ä–µ—Ç–∞—Ü–∏—è:**  
–ü–æ–∫–∞–∑—ã–≤–∞–µ—Ç —ç—Ñ—Ñ–µ–∫—Ç–∏–≤–Ω–æ—Å—Ç—å –∫–≤–∞–ª–∏—Ñ–∏–∫–∞—Ü–∏–∏ –ª–∏–¥–æ–≤ –∏ —Ä–∞–±–æ—Ç—ã —Å –Ω–∏–º–∏: –≤—ã—Å–æ–∫–∏–π CR –æ–∑–Ω–∞—á–∞–µ—Ç, —á—Ç–æ –ª–∏–¥—ã –∫–∞—á–µ—Å—Ç–≤–µ–Ω–Ω—ã–µ –∏ —Ö–æ—Ä–æ—à–æ –∫–æ–Ω–≤–µ—Ä—Ç–∏—Ä—É—é—Ç—Å—è, –Ω–∏–∑–∫–∏–π ‚Äî —á—Ç–æ –∫–∞–Ω–∞–ª –∏–ª–∏ –ø—Ä–æ—Ü–µ—Å—Å –ª–∏–¥–æ–≥–µ–Ω–µ—Ä–∞—Ü–∏–∏ –Ω—É–∂–¥–∞—é—Ç—Å—è –≤ –æ–ø—Ç–∏–º–∏–∑–∞—Ü–∏–∏.

- **–î–æ–ø–æ–ª–Ω–∏—Ç–µ–ª—å–Ω—ã–µ –¥–µ—Ç–∞–ª–∏:**    
- –°—á–∏—Ç–∞—é—Ç—Å—è —É–Ω–∏–∫–∞–ª—å–Ω—ã–µ –∫–ª–∏–µ–Ω—Ç—ã: –µ—Å–ª–∏ –æ–¥–∏–Ω –∫–ª–∏–µ–Ω—Ç —Å–¥–µ–ª–∞–ª –Ω–µ—Å–∫–æ–ª—å–∫–æ –∑–∞–∫–∞–∑–æ–≤, –æ–Ω —Å—á–∏—Ç–∞–µ—Ç—Å—è —Ç–æ–ª—å–∫–æ –æ–¥–∏–Ω —Ä–∞–∑.  
- –ü—Ä–∏–≤—è–∑–∫–∞ –∫ –ø–µ—Ä–∏–æ–¥—É: –∏—Å–ø–æ–ª—å–∑—É–µ–º –∏–Ω—Ç–µ—Ä–≤–∞–ª –≤ 60 –¥–Ω–µ–π –ø–æ—Å–ª–µ —Ä–µ–≥–∏—Å—Ç—Ä–∞—Ü–∏–∏.

```python
# –ü–æ–ª—É—á–∏–º –∏–Ω—Ñ–æ—Ä–º–∞—Ü–∏—é –æ –∫–æ–Ω–≤–µ—Ä—Å–∏–∏ —Å –Ω–∞—á–∞–ª–∞ –≥–æ–¥–∞.
days_interval = {"days": "60 day"} # –ó–∞–¥–∞–¥–∏–º –ø–∞—Ä–∞–º–µ—Ç—Ä –¥–Ω–µ–π –∏–Ω—Ç–µ—Ä–≤–∞–ª–∞

query = """
SELECT 
  COUNT(DISTINCT m.mql_id) as leads, 
  COUNT(DISTINCT c.mql_id) as closed_leads 
FROM 
  marketing_qualified AS m 
  LEFT JOIN closed_deals AS c on c.mql_id = m.mql_id 
  AND c.won_date BETWEEN m.first_contact_date 
  AND m.first_contact_date + INTERVAL :days 
WHERE 
  m.first_contact_date BETWEEN '2018-01-01' 
  AND '2018-05-31';
"""
# –ü—Ä–∏–º–µ–Ω—è–µ–º —Ñ—É–Ω–∫—Ü–∏—é
convers = run_query(engine, query, params=days_interval, fetch='df')

# –†–∞—Å—á–∏—Ç—ã–≤–∞–µ–º –æ—Ç–Ω–æ—à–µ–Ω–∏–µ
cr = round(convers['closed_leads'] / convers['leads'] * 100, 2).iloc[0]

# –ü–æ—Å—Ç—Ä–æ–∏–º —Å–∞–±–ø–ª–æ—Ç
fig = make_subplots(
        rows=1, cols=2,
    specs=[
        [{"type": "indicator"}, {"type": "funnel"}] # –î–≤–∞ –∏–Ω–¥–∏–∫–∞—Ç–æ—Ä–∞ –≤ –ø–µ—Ä–≤–æ–π —Å—Ç—Ä–æ–∫–µ
    ],
)

# –û—Ç–æ–±—Ä–∞–∑–∏–º –∏–Ω–¥–∏–∫–∞—Ç–æ—Ä
fig.add_trace(
    go.Indicator(
    mode='number',
    value = cr,    
    title={"text": "<b>Conversion Rate</b><br><span style='font-size:0.8em'>–∑–∞ 60 –¥–Ω–µ–π –ø–æ—Å–ª–µ —Ä–µ–≥–∏—Å—Ç—Ä–∞—Ü–∏–∏</span>"},
    number={'suffix': "%", 'valueformat': ".2f"},
    ), row=1, col=1
)

# –î–æ–±–∞–≤–ª—è–µ–º –≤–æ—Ä–æ–Ω–∫—É –Ω–∞–ø—Ä—è–º—É—é (–±–µ–∑ —Ü–∏–∫–ª–∞)
fig.add_trace(
    go.Funnel(
        y=convers.T.index,
        x=convers.T.values.flatten(),
        name="–ö–æ–Ω–≤–µ—Ä—Å–∏—è",
        marker=dict(color=custom_palette[4])
    ),
    row=1, col=2
)

# –û–±–Ω–æ–≤–ª—è–µ–º layout
fig.update_layout(
    showlegend=False,
    height=500,
    title_text="<b>–ö–æ–Ω–≤–µ—Ä—Å–∏—è –≤ –ø–µ—Ä–≤—É—é –ø—Ä–æ–¥–∞–∂—É</b>",
    yaxis=dict(
        title=None,
        showticklabels=False
    )
)

fig.write_image('1.png')
```

**üìä –ò–Ω—Ç–µ—Ä–ø—Ä–µ—Ç–∞—Ü–∏—è —Ä–µ–∑—É–ª—å—Ç–∞—Ç–∞:**
- –ü—Ä–∏–º–µ—Ä–Ω–æ –∫–∞–∂–¥—ã–π –¥–µ—Å—è—Ç—ã–π –ª–∏–¥ –∫–æ–Ω–≤–µ—Ä—Ç–∏—Ä—É–µ—Ç—Å—è –≤ —Å–¥–µ–ª–∫—É. –ï—Å—Ç—å –ø–æ—Ç–µ–Ω—Ü–∏–∞–ª –¥–ª—è –ø–æ–≤—ã—à–µ–Ω–∏—è ‚Äî –æ—Å–æ–±–µ–Ω–Ω–æ –≤ —Ä–∞–±–æ—Ç–µ —Å –∫–∞–Ω–∞–ª–∞–º–∏ —Å –Ω–∏–∑–∫–æ–π CR –∏–ª–∏ —Å –ø–æ–º–æ—â—å—é –æ–ø—Ç–∏–º–∏–∑–∞—Ü–∏–∏ –ø–µ—Ä–≤—ã—Ö –∫–∞—Å–∞–Ω–∏–π –∏ —Å–∫—Ä–∏–ø—Ç–æ–≤ –ø—Ä–æ–¥–∞–∂.
---

### –ö–æ–Ω–≤–µ—Ä—Å–∏—è –≤ —Å–¥–µ–ª–∫—É –ø–æ –∫–∞–Ω–∞–ª–∞–º –ø—Ä–∏–≤–ª–µ—á–µ–Ω–∏—è

- **–û–ø–∏—Å–∞–Ω–∏–µ –º–µ—Ç—Ä–∏–∫–∏:**  
–ü—Ä–æ—Ü–µ–Ω—Ç –ª–∏–¥–æ–≤, –∫–æ—Ç–æ—Ä—ã–µ —Å–æ–≤–µ—Ä—à–∏–ª–∏ –ø—Ä–æ–¥–∞–∂—É, –ø–æ –∫–∞–∂–¥–æ–º—É –∫–∞–Ω–∞–ª—É –ø—Ä–∏–≤–ª–µ—á–µ–Ω–∏—è.

### –ü–æ–∫—É–ø–∞—Ç–µ–ª–∏.

#### Repeat Purchase Rate.

- **–û–ø–∏—Å–∞–Ω–∏–µ –º–µ—Ç—Ä–∏–∫–∏:**
–ü–æ–∫–∞–∑—ã–≤–∞–µ—Ç, –∫–∞–∫–∞—è —á–∞—Å—Ç—å –∫–ª–∏–µ–Ω—Ç–æ–≤ —Å–¥–µ–ª–∞–ª–∞ —Ö–æ—Ç—è –±—ã –¥–≤–∞ –¥–æ—Å—Ç–∞–≤–ª–µ–Ω–Ω—ã—Ö –∑–∞–∫–∞–∑–∞.

- **–ë–∏–∑–Ω–µ—Å-–∏–Ω—Ç–µ—Ä–ø—Ä–µ—Ç–∞—Ü–∏—è:**
–ß–µ–º –≤—ã—à–µ —ç—Ç–∞ –¥–æ–ª—è ‚Äî —Ç–µ–º –≤—ã—à–µ –≤–æ–≤–ª–µ—á—ë–Ω–Ω–æ—Å—Ç—å –∏ —É—Å–ø–µ—à–Ω–æ—Å—Ç—å "–ø–µ—Ä–≤–∏—á–Ω–æ–≥–æ –æ–Ω–±–æ—Ä–¥–∏–Ω–≥–∞".
–ù–∏–∑–∫–æ–µ –∑–Ω–∞—á–µ–Ω–∏–µ –º–æ–∂–µ—Ç —Å–∏–≥–Ω–∞–ª–∏–∑–∏—Ä–æ–≤–∞—Ç—å –æ –ø–ª–æ—Ö–æ–º –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª—å—Å–∫–æ–º –æ–ø—ã—Ç–µ, –¥–æ–ª–≥–æ–π –¥–æ—Å—Ç–∞–≤–∫–µ –∏–ª–∏ –Ω–µ–ø–æ–¥—Ö–æ–¥—è—â–µ–º –∞—Å—Å–æ—Ä—Ç–∏–º–µ–Ω—Ç–µ.

- **–î–æ–ø–æ–ª–Ω–∏—Ç–µ–ª—å–Ω—ã–µ –¥–µ—Ç–∞–ª–∏:**
–£—á–∏—Ç—ã–≤–∞—é—Ç—Å—è —Ç–æ–ª—å–∫–æ –¥–æ—Å—Ç–∞–≤–ª–µ–Ω–Ω—ã–µ –∑–∞–∫–∞–∑—ã. –ü–æ–≤—Ç–æ—Ä–Ω—ã–º —Å—á–∏—Ç–∞–µ—Ç—Å—è –∑–∞–∫–∞–∑ –æ—Ç —Ç–æ–≥–æ –∂–µ –∫–ª–∏–µ–Ω—Ç–∞ —Å –ª—é–±—ã–º –∏–Ω—Ç–µ—Ä–≤–∞–ª–æ–º –≤—Ä–µ–º–µ–Ω–∏.

```python
query = """
WITH orders_per_customer AS (
  SELECT 
    customer_unique_id, 
    COUNT(DISTINCT order_id) as order_count 
  FROM 
    orders AS o 
    JOIN customers AS c on c.customer_id = o.customer_id
    JOIN order_payments op USING(order_id)
  WHERE 
    order_status = '–¥–æ—Å—Ç–∞–≤–ª–µ–Ω' 
  GROUP BY 
    customer_unique_id
) 
SELECT 
  COUNT(*) AS non_repeat, 
  COUNT(CASE WHEN order_count > 1 THEN 1 END) as repeat_purchase, 
  ROUND(
    COUNT(CASE WHEN order_count > 1 THEN 1 END) * 1.0 / COUNT(*) * 100, 
    2
  ):: float AS repeat_order_rate
FROM 
  orders_per_customer;
"""


repeat_order_rate = run_query(engine, query, fetch='df')

rr = repeat_order_rate.iloc[0, 2]

# –ü–æ—Å—Ç—Ä–æ–∏–º —Å–∞–±–ø–ª–æ—Ç
fig = make_subplots(
        rows=1, cols=2,
    specs=[
        [{"type": "indicator"}, {"type": "funnel"}] # –î–≤–∞ –∏–Ω–¥–∏–∫–∞—Ç–æ—Ä–∞ –≤ –ø–µ—Ä–≤–æ–π —Å—Ç—Ä–æ–∫–µ
    ],
)

# –û—Ç–æ–±—Ä–∞–∑–∏–º –∏–Ω–¥–∏–∫–∞—Ç–æ—Ä
fig.add_trace(
    go.Indicator(
    mode='number',
    value = rr,    
    title={"text": "<b>Repeat Purchase Rate</b><br><span style='font-size:0.8em'></span>"},
    number={'suffix': "%", 'valueformat': ".2f"},
    ), row=1, col=1
)

# –î–æ–±–∞–≤–ª—è–µ–º –≤–æ—Ä–æ–Ω–∫—É –Ω–∞–ø—Ä—è–º—É—é (–±–µ–∑ —Ü–∏–∫–ª–∞)
fig.add_trace(
    go.Funnel(
        y=repeat_order_rate.iloc[:, :2].T.index,
        x=repeat_order_rate.iloc[:, :2].T.values.flatten(),
        name="–ö–æ–Ω–≤–µ—Ä—Å–∏—è",
        marker=dict(color=custom_palette[4])
    ),
    row=1, col=2
)

# –û–±–Ω–æ–≤–ª—è–µ–º layout
fig.update_layout(
    showlegend=False,
    height=500,
    title_text="<b>–î–æ–ª—è –ø–æ–≤—Ç–æ—Ä–Ω—ã—Ö –ø–æ–∫—É–ø–æ–∫</b>",
    yaxis=dict(
        title=None,
        showticklabels=False
    )
)

fig.write_image('1.png')
```

- –ú–µ—Ç—Ä–∏–∫–∞ –ø–æ–∫–∞–∑—ã–≤–∞–µ—Ç, —Å–∫–æ–ª—å–∫–æ –∫–ª–∏–µ–Ω—Ç–æ–≤ –≤–µ—Ä–Ω—É–ª–∏—Å—å –∑–∞ –ø–æ–≤—Ç–æ—Ä–Ω–æ–π –ø–æ–∫—É–ø–∫–æ–π. –†–µ–∑—É–ª—å—Ç–∞—Ç ‚Äî 3%. –£–≤–µ–ª–∏—á–µ–Ω–∏–µ –ø–æ–≤—Ç–æ—Ä–Ω—ã—Ö –ø–æ–∫—É–ø–æ–∫ - –æ—Å–Ω–æ–≤–Ω–æ–π –¥—Ä–∞–π–≤–µ—Ä –¥–ª—è —Ä–æ—Å—Ç–∞ –ø—Ä–æ–¥—É–∫—Ç–∞ –∏ GMV.

#### Average Orders.

- **–û–ø–∏—Å–∞–Ω–∏–µ –º–µ—Ç—Ä–∏–∫–∏:**
   –°—Ä–µ–¥–Ω–µ–µ –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –∑–∞–∫–∞–∑–æ–≤, —Å–æ–≤–µ—Ä—à–µ–Ω–Ω—ã—Ö –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª—è–º–∏, –∫–æ—Ç–æ—Ä—ã–µ —Å–¥–µ–ª–∞–ª–∏ –±–æ–ª–µ–µ –æ–¥–Ω–æ–≥–æ –∑–∞–∫–∞–∑–∞ –Ω–∞ –ø–ª–∞—Ç—Ñ–æ—Ä–º–µ.

- **–ë–∏–∑–Ω–µ—Å-–∏–Ω—Ç–µ—Ä–ø—Ä–µ—Ç–∞—Ü–∏—è:**
    –û—Ç—Å–ª–µ–∂–∏–≤–∞–Ω–∏–µ —ç—Ç–æ–π –º–µ—Ç—Ä–∏–∫–∏ –≤–æ –≤—Ä–µ–º–µ–Ω–∏ –º–æ–∂–µ—Ç –ø–æ–∫–∞–∑–∞—Ç—å, –Ω–∞—Å–∫–æ–ª—å–∫–æ —ç—Ñ—Ñ–µ–∫—Ç–∏–≤–Ω–æ —Ä–∞–±–æ—Ç–∞—é—Ç —Å—Ç—Ä–∞—Ç–µ–≥–∏–∏ —É–¥–µ—Ä–∂–∞–Ω–∏—è –∏ –ø–æ–≤—ã—à–µ–Ω–∏—è –ª–æ—è–ª—å–Ω–æ—Å—Ç–∏ –ø–æ–≤—Ç–æ—Ä–Ω—ã—Ö –ø–æ–∫—É–ø–∞—Ç–µ–ª–µ–π.

- **–î–æ–ø–æ–ª–Ω–∏—Ç–µ–ª—å–Ω—ã–µ –¥–µ—Ç–∞–ª–∏:**
    –£—á–∏—Ç—ã–≤–∞—é—Ç—Å—è —Ç–æ–ª—å–∫–æ –¥–æ—Å—Ç–∞–≤–ª–µ–Ω–Ω—ã–µ –∑–∞–∫–∞–∑—ã. –ü–æ–≤—Ç–æ—Ä–Ω—ã–º —Å—á–∏—Ç–∞–µ—Ç—Å—è –∑–∞–∫–∞–∑ –æ—Ç —Ç–æ–≥–æ –∂–µ –∫–ª–∏–µ–Ω—Ç–∞ —Å –ª—é–±—ã–º –∏–Ω—Ç–µ—Ä–≤–∞–ª–æ–º –≤—Ä–µ–º–µ–Ω–∏.

```python
query = """
SELECT ROUND(AVG(order_count), 2)::float
FROM (SELECT customer_unique_id,
           COUNT(o.order_id) AS order_count
        FROM customers c
        JOIN orders o ON o.customer_id=c.customer_id
        WHERE order_status = '–¥–æ—Å—Ç–∞–≤–ª–µ–Ω'
        GROUP BY customer_unique_id
        HAVING COUNT(o.order_id) > 1) AS cust_orders;
"""

avg_orders = run_query(engine, query, fetch='scalar')

# –í–∏–∑—É–∞–ª–∏–∑–∞—Ü–∏—è (–ø—Ä–æ—Å—Ç–æ–µ –∑–Ω–∞—á–µ–Ω–∏–µ)
fig = go.Figure(go.Indicator(
    mode = "number",
    value = avg_orders,
    title = {
    "text": "<b>–°—Ä–µ–¥–Ω–µ–µ –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –∑–∞–∫–∞–∑–æ–≤</b><br>"
    "<span style='font-size:0.8em'>–Ω–∞ –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª—è —Å –ø–æ–≤—Ç–æ—Ä–Ω—ã–º–∏ –ø–æ–∫—É–ø–∞–º–∏</span>"
}
))
fig.write_image('1.png')
```

**üìä –ò–Ω—Ç–µ—Ä–ø—Ä–µ—Ç–∞—Ü–∏—è —Ä–µ–∑—É–ª—å—Ç–∞—Ç–∞:**
- –°—Ä–µ–¥–Ω–µ–µ —á–∏—Å–ª–æ –∑–∞–∫–∞–∑–æ–≤ —á—É—Ç—å –±–æ–ª—å—à–µ –¥–≤—É—Ö, –¥–æ–ª—è –ø–æ—Å—Ç–æ—è–Ω–Ω—ã—Ö –ø–æ–∫—É–ø–∞—Ç–µ–ª–µ–π –æ—á–µ–Ω—å –Ω–∏–∑–∫–∞—è.

#### –í—Ä–µ–º—è –º–µ–∂–¥—É –ø–æ–∫—É–ø–∫–∞–º–∏.

- **–û–ø–∏—Å–∞–Ω–∏–µ –º–µ—Ç—Ä–∏–∫–∏:**  
–†–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ –∫–ª–∏–µ–Ω—Ç–æ–≤ –ø–æ –¥–Ω—è–º –º–µ–∂–¥—É –ø–µ—Ä–≤—ã–º –∏ –≤—Ç–æ—Ä—ã–º –∑–∞–∫–∞–∑–æ–º.

- **–ë–∏–∑–Ω–µ—Å-–∏–Ω—Ç–µ—Ä–ø—Ä–µ—Ç–∞—Ü–∏—è:**  
–ë—ã—Å—Ç—Ä—ã–π –ø–æ–≤—Ç–æ—Ä–Ω—ã–π –∑–∞–∫–∞–∑ —É–∫–∞–∑—ã–≤–∞–µ—Ç –Ω–∞ –≤—ã—Å–æ–∫–∏–π –∏–Ω—Ç–µ—Ä–µ—Å –∏ —Ä–µ–ª–µ–≤–∞–Ω—Ç–Ω–æ—Å—Ç—å –ø—Ä–µ–¥–ª–æ–∂–µ–Ω–∏—è. –î–æ–ª–≥–∏–π ‚Äî —Ç–æ—á–∫–∞ —Ä–æ—Å—Ç–∞ –¥–ª—è —Ç—Ä–∏–≥–≥–µ—Ä–Ω—ã—Ö –∫–æ–º–º—É–Ω–∏–∫–∞—Ü–∏–π.

- **–î–æ–ø–æ–ª–Ω–∏—Ç–µ–ª—å–Ω—ã–µ –¥–µ—Ç–∞–ª–∏:**  
–°—Ä–∞–≤–Ω–∏–≤–∞—Ç—å –º–µ–¥–∏–∞–Ω—É –∏ —Å—Ä–µ–¥–Ω–µ–µ. –ü–æ—Å—Ç—Ä–æ–∏—Ç—å —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ.


```python
query = """
WITH customer_orders AS (
  SELECT 
    customer_unique_id, 
    order_purchase_timestamp, 
    MIN(order_purchase_timestamp) OVER(PARTITION BY customer_unique_id) AS first_purchase_date, 
    LEAD(order_purchase_timestamp) OVER(
      PARTITION BY customer_unique_id 
      ORDER BY 
        order_purchase_timestamp
    ) AS second_purchase_date, 
    COUNT(order_id) OVER(PARTITION BY customer_unique_id) as order_count 
  FROM 
    orders AS o 
    JOIN customers AS c ON c.customer_id = o.customer_id 
  WHERE 
    order_status = '–¥–æ—Å—Ç–∞–≤–ª–µ–Ω'
) 
SELECT 
  customer_unique_id, 
  first_purchase_date, 
  second_purchase_date, 
  EXTRACT(DAY FROM (second_purchase_date - first_purchase_date)) AS days_between_orders, 
  order_count, 
  ROUND(AVG(EXTRACT(DAY FROM (second_purchase_date - first_purchase_date))) OVER()) AS avg_days 
FROM 
  customer_orders 
WHERE 
  second_purchase_date IS NOT NULL 
  AND first_purchase_date != second_purchase_date 
ORDER BY 
  customer_unique_id;
"""
hist = run_query(engine, query, fetch='df')

# –í—ã—á–∏—Å–ª—è–µ–º —Å—Ä–µ–¥–Ω–µ–µ –∑–Ω–∞—á–µ–Ω–∏–µ
mean_value = hist['avg_days'].max()

# –°–æ–∑–¥–∞–µ–º –≥–∏—Å—Ç–æ–≥—Ä–∞–º–º—É
fig = px.histogram(hist,
                   x="days_between_orders",
                   nbins=50, marginal='box',
                   color_discrete_sequence=custom_palette[4:5],
                   title=f"<b>–ú–µ–∂–¥—É –ø–æ–∫—É–ø–∫–∞–º–∏ –ø—Ä–æ—Ö–æ–¥–∏—Ç –≤ —Å—Ä–µ–¥–Ω–µ–º {mean_value} –¥–Ω–µ–π.</b>")

# –î–æ–±–∞–≤–ª—è–µ–º –ª–∏–Ω–∏—é —Å—Ä–µ–¥–Ω–µ–≥–æ –∑–Ω–∞—á–µ–Ω–∏—è
fig.add_vline(
    x=hist['avg_days'].max(), 
    line_width=3, 
    line_dash="dash", 
    line_color=custom_palette[0],
)

# –ù–∞—Å—Ç—Ä–∞–∏–≤–∞–µ–º –æ—Ç—Å—Ç—É–ø—ã –¥–ª—è –∞–Ω–Ω–æ—Ç–∞—Ü–∏–∏
fig.update_layout(margin=dict(r=120),
                 bargap=0.05)

fig.write_image('1.png')
```

   - –†–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ –ø–æ –¥–Ω—è–º –¥–æ –ø–æ–≤—Ç–æ—Ä–Ω–æ–π –ø–æ–∫—É–ø–∫–∏ —Å–∏–ª—å–Ω–æ —Å–∫–æ—à–µ–Ω–æ, –≤–ø–ª–æ—Ç—å –¥–æ 600 –¥–Ω–µ–π, –ø–∏–∫–æ–≤–æ–µ –∑–Ω–∞—á–µ–Ω–∏–µ —Å–æ—Å—Ç–∞–≤–ª—è–µ—Ç –¥–æ 20 –¥–Ω–µ–π.
   - –°—Ä–µ–¥–Ω–µ–µ –≤—Ä–µ–º—è ‚Äî 96 –¥–Ω–µ–π, –º–µ–¥–∏–∞–Ω–∞ ‚Äî 48. –ë–æ–ª—å—à–æ–π —Ä–∞–∑—Ä—ã–≤ –º–µ–∂–¥—É —Å—Ä–µ–¥–Ω–∏–º –∏ –º–µ–¥–∏–∞–Ω–æ–π —É–∫–∞–∑—ã–≤–∞–µ—Ç –Ω–∞ –≤—ã–±—Ä–æ—Å—ã: –Ω–µ–∫–æ—Ç–æ—Ä—ã–µ –∫–ª–∏–µ–Ω—Ç—ã –¥–µ–ª–∞—é—Ç –ø–æ–≤—Ç–æ—Ä–Ω—ã–µ –∑–∞–∫–∞–∑—ã —á–µ—Ä–µ–∑ –æ—á–µ–Ω—å –¥–æ–ª–≥–∏–π —Å—Ä–æ–∫. 
   - –¢–∞–∫–∂–µ —Å—Ç–æ–∏—Ç —Ä–∞–∑–¥–µ–ª–∏—Ç—å –∞–Ω–∞–ª–∏–∑ –ø–æ –∫–∞—Ç–µ–≥–æ—Ä–∏—è–º —Ç–æ–≤–∞—Ä–æ–≤: –≤–æ–∑–º–æ–∂–Ω–æ, –≤ –Ω–µ–∫–æ—Ç–æ—Ä—ã—Ö –∫–∞—Ç–µ–≥–æ—Ä–∏—è—Ö –ø–æ–≤—Ç–æ—Ä–Ω—ã–µ –ø–æ–∫—É–ø–∫–∏ –ø—Ä–æ–∏—Å—Ö–æ–¥—è—Ç –±—ã—Å—Ç—Ä–µ–µ.
   
---

#### RPR –ø–æ –∫–∞—Ç–µ–≥–æ—Ä–∏—è–º —Ç–æ–≤–∞—Ä–æ–≤

- **–û–ø–∏—Å–∞–Ω–∏–µ –º–µ—Ç—Ä–∏–∫–∏:**  
  –î–æ–ª—è –∫–ª–∏–µ–Ω—Ç–æ–≤, —Å–æ–≤–µ—Ä—à–∏–≤—à–∏—Ö –ø–æ–≤—Ç–æ—Ä–Ω—É—é –ø–æ–∫—É–ø–∫—É, –Ω–µ–∑–∞–≤–∏—Å–∏–º–æ –æ—Ç –∫–∞—Ç–µ–≥–æ—Ä–∏–∏ –≤—Ç–æ—Ä–æ–≥–æ –∑–∞–∫–∞–∑–∞.

- **–ë–∏–∑–Ω–µ—Å-–∏–Ω—Ç–µ—Ä–ø—Ä–µ—Ç–∞—Ü–∏—è:**  
    –í—ã—Å–æ–∫–∏–π RPR –≥–æ–≤–æ—Ä–∏—Ç –æ —Ç–æ–º, —á—Ç–æ –ø–æ–∫—É–ø–∫–∞ –≤ –¥–∞–Ω–Ω–æ–π –∫–∞—Ç–µ–≥–æ—Ä–∏–∏ —Ö–æ—Ä–æ—à–æ ‚Äú–≤–æ–≤–ª–µ–∫–∞–µ—Ç‚Äù –∫–ª–∏–µ–Ω—Ç–∞ –∏ —Å–ø–æ—Å–æ–±—Å—Ç–≤—É–µ—Ç –±—ã—Å—Ç—Ä–æ–º—É –≤–æ–∑–≤—Ä–∞—Ç—É, —á—Ç–æ –∫—Ä–∏—Ç–∏—á–Ω–æ –¥–ª—è –∞–∫—Ç–∏–≤–∞—Ü–∏–∏. 
    –ù–∏–∑–∫–∏–π RPR –º–æ–∂–µ—Ç —Å–∏–≥–Ω–∞–ª–∏–∑–∏—Ä–æ–≤–∞—Ç—å –æ –±–∞—Ä—å–µ—Ä–∞—Ö: –æ–¥–Ω–æ—Ä–∞–∑–æ–≤—ã–π —Ö–∞—Ä–∞–∫—Ç–µ—Ä —Ç–æ–≤–∞—Ä–∞, –ø–ª–æ—Ö–æ–π –∫–ª–∏–µ–Ω—Ç—Å–∫–∏–π –æ–ø—ã—Ç, –≤—ã—Å–æ–∫–∞—è —Ü–µ–Ω–∞, –Ω–µ—É–¥–æ–±–Ω–∞—è –¥–æ—Å—Ç–∞–≤–∫–∞ –∏ —Ç.–¥.

- **–î–æ–ø–æ–ª–Ω–∏—Ç–µ–ª—å–Ω—ã–µ –¥–µ—Ç–∞–ª–∏:**  
  - –ù–µ –∞–Ω–∞–ª–∏–∑–∏—Ä—É–µ—Ç—Å—è –∫–∞—Ç–µ–≥–æ—Ä–∏—è –ø–æ–≤—Ç–æ—Ä–Ω–æ–≥–æ –∑–∞–∫–∞–∑–∞. 
  - –ò—Å–∫–ª—é—á–∞—é—Ç—Å—è –æ—Ç–º–µ–Ω—ë–Ω–Ω—ã–µ –∏ –≤–æ–∑–≤—Ä–∞—Ç–Ω—ã–µ –∑–∞–∫–∞–∑—ã. 

```python
query = """
WITH customer_orders AS (
    SELECT 
        c.customer_unique_id,
        o.order_id,
        o.order_purchase_timestamp,
        MIN(o.order_purchase_timestamp) OVER (PARTITION BY c.customer_unique_id) AS first_purchase_date,
        ROW_NUMBER() OVER (PARTITION BY c.customer_unique_id ORDER BY o.order_purchase_timestamp) AS order_rank
    FROM orders o
    JOIN customers c ON o.customer_id = c.customer_id
    WHERE o.order_status = '–¥–æ—Å—Ç–∞–≤–ª–µ–Ω'
),

-- –ù–∞—Ö–æ–¥–∏–º –∫–∞—Ç–µ–≥–æ—Ä–∏—é —Ç–æ–≤–∞—Ä–∞ –≤ –ø–µ—Ä–≤–æ–º –∑–∞–∫–∞–∑–µ
first_order_items AS (
    SELECT 
        co.customer_unique_id,
        p.category_name_translated
    FROM customer_orders co
    JOIN order_items oi ON co.order_id = oi.order_id
    JOIN products p ON oi.product_id = p.product_id
    WHERE co.order_rank = 1
),

-- –ò—â–µ–º –ø–æ–≤—Ç–æ—Ä–Ω—ã–µ –ø–æ–∫—É–ø–∫–∏ –≤ —Ç–µ—á–µ–Ω–∏–µ 60 –¥–Ω–µ–π
repeat_flags AS (
    SELECT 
        co.customer_unique_id
    FROM customer_orders co
    WHERE co.order_rank > 1
)

-- –ê–≥—Ä–µ–≥–∞—Ü–∏—è TSR –ø–æ –∫–∞—Ç–µ–≥–æ—Ä–∏—è–º
SELECT
    foi.category_name_translated,
    COUNT(DISTINCT foi.customer_unique_id) AS customers_total,
    COUNT(DISTINCT CASE WHEN rf.customer_unique_id IS NOT NULL THEN foi.customer_unique_id END) AS customers_repeat_60d,
    ROUND(
        COUNT(DISTINCT CASE WHEN rf.customer_unique_id IS NOT NULL THEN foi.customer_unique_id END) 
        * 100.0 / COUNT(DISTINCT foi.customer_unique_id), 
        2
    ) AS repeat_rate_60d
FROM first_order_items foi
LEFT JOIN repeat_flags rf ON foi.customer_unique_id = rf.customer_unique_id
GROUP BY foi.category_name_translated
ORDER BY repeat_rate_60d DESC;
"""

tsr_category = run_query(engine, query, fetch='df')



# –ü–æ–¥—Å—á–∏—Ç–∞–µ–º —Å—Ä–µ–¥–Ω–∏–π –ø–æ–∫–∞–∑–∞—Ç–µ–ª—å –≤–æ–∑–≤—Ä–∞—Ç–∞ –ø–æ –∫–∞—Ç–µ–≥–æ—Ä–∏—è–º
mean =  round(tsr_category['customers_repeat_60d'].sum() /tsr_category['customers_total'].sum() * 100, 2)

# –ü—Ä–∏–º–µ–Ω–∏ —Å–∫—Ä–∏–ø—Ç –¥–ª—è –æ—Ç—Ä–∏—Å–æ–≤–∫–∏ –≥—Ä–∞—Ñ–∏–∫–∞
scatter_quadrant_plot(
    df=tsr_category,
    x_col='customers_total',
    y_col='repeat_rate_60d',
    x_mean=None,
    y_mean=mean,
    category_col='category_name_translated',
    size_col='customers_repeat_60d',
    title='<b>RPR 60 –¥–Ω–µ–π: –∫–∞—Ç–µ–≥–æ—Ä–∏–∏ VS –æ—Ö–≤–∞—Ç</b>',
    condition='>',
    show_labels=True,
    show_legend=False,
    labels={
        'customers_total': '–û–±—â–µ–µ —á–∏—Å–ª–æ –∫–ª–∏–µ–Ω—Ç–æ–≤',
        'repeat_rate_60d': '–î–æ–ª—è –ø–æ–≤—Ç–æ—Ä–Ω—ã—Ö –ø–æ–∫—É–ø–æ–∫, %',
        'category_name_translated': '–ö–∞—Ç–µ–≥–æ—Ä–∏—è'
    },
    color_palette=custom_palette[::-1]
)
fig.write_image('1.png')
```

**üìä –ò–Ω—Ç–µ—Ä–ø—Ä–µ—Ç–∞—Ü–∏—è —Ä–µ–∑—É–ª—å—Ç–∞—Ç–∞:**

- –õ—É—á—à–∏–µ –ø–æ–∫–∞–∑–∞—Ç–µ–ª–∏ —Å–æ–≤–æ–∫—É–ø–Ω—ã–µ –ø–æ–∫–∞–∑–∞—Ç–µ–ª–∏ —É –∫–∞—Ç–µ–≥–æ—Ä–∏–π **—Ç–æ–≤–∞—Ä—ã –¥–ª—è –æ—Ç–¥—ã—Ö–∞**, **—Å–ø–æ—Ä—Ç –∏ –æ—Ç–¥—ã—Ö**, **–º–µ–±–µ–ª—å –∏ –¥–µ–∫–æ—Ä**  –∏ **—Å—É–º–∫–∏ –∏ –∞–∫—Å–µ—Å—Å—É–∞—Ä—ã** —Å–æ—á–µ—Ç–∞–Ω–∏–µ –±–æ–ª—å—à–æ–≥–æ –∫–æ–ª–∏—á–µ—Å—Ç–≤–∞ –∑–∞–∫–∞–∑–æ–≤ –∏ –ø–æ–∫–∞–∑–∞—Ç–µ–ª—è RPR –≤—ã—à–µ —Å—Ä–µ–¥–Ω–µ–≥–æ.
- –ö–∞—Ç–µ–≥–æ—Ä–∏–∏ **–†—É–∫–æ–¥–µ–ª–∏–µ** –∏ **–ö—É–ª–∏–Ω–∞—Ä–∏—è** –∏–º–µ—é—Ç –Ω–∏–∑–∫–∏–π –æ—Ö–≤–∞—Ç –Ω–æ –≤—ã—Å–æ–∫–∏–µ –ø–æ–∫–∞–∑–∞—Ç–µ–ª–∏ RPR - –≤—ã—Å–æ–∫–∏–π –ø–æ—Ç–µ–Ω—Ü–∏–∞–ª –¥–ª—è —Ä–∞–∑–≤–∏—Ç–∏—è. 
---

### –ü—Ä–æ–¥–∞–≤—Ü—ã.

#### Repeat Sale Rate.

- **–û–ø–∏—Å–∞–Ω–∏–µ –º–µ—Ç—Ä–∏–∫–∏:**
–ü–æ–∫–∞–∑—ã–≤–∞–µ—Ç, –∫–∞–∫–∞—è —á–∞—Å—Ç—å –ø—Ä–æ–¥–∞–≤—Ü–æ–≤ —Å–æ–≤–µ—Ä—à–∏–ª–∞ –∫–∞–∫ –º–∏–Ω–∏–º—É–º –¥–≤–µ –ø—Ä–æ–¥–∞–∂–∏ —Å –ø–æ–¥—Ç–≤–µ—Ä–∂–¥—ë–Ω–Ω–æ–π –¥–æ—Å—Ç–∞–≤–∫–æ–π.

- **–ë–∏–∑–Ω–µ—Å-–∏–Ω—Ç–µ—Ä–ø—Ä–µ—Ç–∞—Ü–∏—è:**
–ß–µ–º –≤—ã—à–µ —ç—Ç–∞ –¥–æ–ª—è ‚Äî —Ç–µ–º —É—Å–ø–µ—à–Ω–µ–µ –ø—Ä–æ—Ö–æ–¥–∏—Ç –ø–µ—Ä–≤–∏—á–Ω–æ–µ –≤–æ–≤–ª–µ—á–µ–Ω–∏–µ –ø—Ä–æ–¥–∞–≤—Ü–æ–≤, —Ç–µ–º –≤—ã—à–µ –∏—Ö –º–æ—Ç–∏–≤–∞—Ü–∏—è –∏ –¥–æ–≤–µ—Ä–∏–µ –∫ –ø–ª–∞—Ç—Ñ–æ—Ä–º–µ.
–ù–∏–∑–∫–æ–µ –∑–Ω–∞—á–µ–Ω–∏–µ –º–æ–∂–µ—Ç —É–∫–∞–∑—ã–≤–∞—Ç—å –Ω–∞ —Ç—Ä—É–¥–Ω–æ—Å—Ç–∏ —Å –∫–æ–Ω–∫—É—Ä–µ–Ω—Ü–∏–µ–π, –Ω–µ–¥–æ—Å—Ç–∞—Ç–æ—á–Ω—É—é –≤–∏–¥–∏–º–æ—Å—Ç—å —Ç–æ–≤–∞—Ä–æ–≤, –∏–ª–∏ –±–∞—Ä—å–µ—Ä—ã –≤ –∏–Ω—Ç–µ—Ä—Ñ–µ–π—Å–µ/–ª–æ–≥–∏—Å—Ç–∏–∫–µ, –º–µ—à–∞—é—â–∏–µ –ø—Ä–æ–¥–∞–≤–∞—Ç—å –ø–æ–≤—Ç–æ—Ä–Ω–æ.

- **–î–æ–ø–æ–ª–Ω–∏—Ç–µ–ª—å–Ω—ã–µ –¥–µ—Ç–∞–ª–∏:**
–£—á–∏—Ç—ã–≤–∞—é—Ç—Å—è —Ç–æ–ª—å–∫–æ –∑–∞–∫–∞–∑—ã —Å–æ —Å—Ç–∞—Ç—É—Å–æ–º "–¥–æ—Å—Ç–∞–≤–ª–µ–Ω". –ü–æ–≤—Ç–æ—Ä–Ω–æ–π —Å—á–∏—Ç–∞–µ—Ç—Å—è –ª—é–±–∞—è –ø—Ä–æ–¥–∞–∂–∞ –ø–æ—Å–ª–µ –ø–µ—Ä–≤–æ–π, –Ω–µ–∑–∞–≤–∏—Å–∏–º–æ –æ—Ç –∏–Ω—Ç–µ—Ä–≤–∞–ª–∞.

```python
query = """

WITH
seller_order AS (
SELECT seller_id,
        COUNT(DISTINCT order_id) as total_sales
FROM order_items AS oi
JOIN orders USING(order_id)
WHERE order_status = '–¥–æ—Å—Ç–∞–≤–ª–µ–Ω'
GROUP BY 1
ORDER BY 2 DESC)

SELECT COUNT(*) AS all_sellers,
        COUNT(CASE WHEN total_sales > 1 THEN 1 END) AS  repeat_sellers,
        ROUND(COUNT(CASE WHEN total_sales > 1 THEN 100 END) * 1.0 / COUNT(*) * 100, 2)
FROM seller_order;
"""

repeat_sale_rate = run_query(engine, query, fetch='df')


rr = repeat_sale_rate.iloc[0,2]

# –ü–æ—Å—Ç—Ä–æ–∏–º —Å–∞–±–ø–ª–æ—Ç
fig = make_subplots(
        rows=1, cols=2,
    specs=[
        [{"type": "indicator"}, {"type": "funnel"}] # –î–≤–∞ –∏–Ω–¥–∏–∫–∞—Ç–æ—Ä–∞ –≤ –ø–µ—Ä–≤–æ–π —Å—Ç—Ä–æ–∫–µ
    ],
)

# –û—Ç–æ–±—Ä–∞–∑–∏–º –∏–Ω–¥–∏–∫–∞—Ç–æ—Ä
fig.add_trace(
    go.Indicator(
    mode='number',
    value = rr,    
    title={"text": "<b>Repeat Purchase Rate</b><br><span style='font-size:0.8em'></span>"},
    number={'suffix': "%", 'valueformat': ".2f"},
    ), row=1, col=1
)

# –î–æ–±–∞–≤–ª—è–µ–º –≤–æ—Ä–æ–Ω–∫—É –Ω–∞–ø—Ä—è–º—É—é (–±–µ–∑ —Ü–∏–∫–ª–∞)
fig.add_trace(
    go.Funnel(
        y=repeat_sale_rate.iloc[:, :2].T.index,
        x=repeat_sale_rate.iloc[:, :2].T.values.flatten(),
        name="–ö–æ–Ω–≤–µ—Ä—Å–∏—è",
        marker=dict(color=custom_palette[4])
    ),
    row=1, col=2
)

# –û–±–Ω–æ–≤–ª—è–µ–º layout
fig.update_layout(
    showlegend=False,
    height=500,
    title_text="<b>–î–æ–ª—è –ø—Ä–æ–¥–∞–≤—Ü–æ–≤ —Å –ø–æ–≤—Ç–æ—Ä–Ω—ã–º–∏ –ø—Ä–æ–¥–∞–∂–∞–º–∏.</b>",
    yaxis=dict(
        title=None,
        showticklabels=False
    )
)
fig.write_image('1.png')
```

-  –†–µ–∑—É–ª—å—Ç–∞—Ç ‚Äî 82%, –±–æ–ª—å—à–∏–Ω—Å—Ç–≤–æ –ø—Ä–æ–¥–∞–≤—Ü–æ–≤ –∑–∞–¥–µ—Ä–∂–∏–≤–∞—é—Ç—Å—è –Ω–∞ –ø–ª–∞—Ç—Ñ–æ—Ä–º–µ.

#### –í—Ä–µ–º—è –º–µ–∂–¥—É –ø—Ä–æ–¥–∞–∂–∞–º–∏.

- **–û–ø–∏—Å–∞–Ω–∏–µ –º–µ—Ç—Ä–∏–∫–∏:**
–°—Ä–µ–¥–Ω–µ–µ –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –¥–Ω–µ–π –º–µ–∂–¥—É –ø—Ä–æ–¥–∞–∂–∞–º–∏.
- **–ë–∏–∑–Ω–µ—Å-–∏–Ω—Ç–µ—Ä–ø—Ä–µ—Ç–∞—Ü–∏—è:**
–ü–æ–∑–≤–æ–ª—è–µ—Ç –æ—Ü–µ–Ω–∏—Ç—å —Å—Ç–∞–±–∏–ª—å–Ω–æ—Å—Ç—å –ø—Ä–æ–¥–∞–∂.

- **–î–æ–ø–æ–ª–Ω–∏—Ç–µ–ª—å–Ω—ã–µ –¥–µ—Ç–∞–ª–∏:**
–£—á–∏—Ç—ã–≤–∞—é—Ç—Å—è —Ç–æ–ª—å–∫–æ –∑–∞–∫–∞–∑—ã —Å–æ —Å—Ç–∞—Ç—É—Å–æ–º "–¥–æ—Å—Ç–∞–≤–ª–µ–Ω". 


```python
query = """
WITH unique_seller_orders AS (
    SELECT DISTINCT  -- –£–±–∏—Ä–∞–µ–º –¥—É–±–ª–∏–∫–∞—Ç—ã –∑–∞–∫–∞–∑–æ–≤
        s.seller_id,
        s.seller_state,
        o.order_id,
        o.order_purchase_timestamp
    FROM sellers AS s
    JOIN order_items AS oi ON oi.seller_id = s.seller_id
    JOIN orders AS o ON o.order_id = oi.order_id
    WHERE order_status = '–¥–æ—Å—Ç–∞–≤–ª–µ–Ω'
),

seller_order_sequence AS (
    SELECT 
        seller_id,
        seller_state,
        order_id,
        order_purchase_timestamp,
        LAG(order_purchase_timestamp) OVER(
            PARTITION BY seller_id 
            ORDER BY order_purchase_timestamp
        ) AS prev_order_timestamp
    FROM unique_seller_orders
)

SELECT 
    seller_id,
    EXTRACT(DAY FROM (order_purchase_timestamp - prev_order_timestamp)) AS days_between_orders
FROM seller_order_sequence
WHERE prev_order_timestamp IS NOT NULL;
"""

hist = run_query(engine, query, fetch='df')

# –§–∏–ª—å—Ç—Ä—É–µ–º –≤—ã–±—Ä–æ—Å—ã 
hist_filtered = hist[hist['days_between_orders'] < 60]

# –ü–µ—Ä–µ—Å—á–∏—Ç—ã–≤–∞–µ–º —Å—Ä–µ–¥–Ω–µ–µ
mean_value = round(hist_filtered['days_between_orders'].mean())
median_value = round(hist_filtered['days_between_orders'].median())
fig = px.histogram(
    hist_filtered,
    x="days_between_orders",
    nbins=50,
    color_discrete_sequence=custom_palette[4:5],
    marginal='box',
    title=f"<b>–†–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ –∑–∞–∫–∞–∑–æ–≤ –ø–æ –¥–Ω—è–º –ø–æ–≤—Ç–æ—Ä–Ω–æ–π –ø—Ä–æ–¥–∞–∂–∏</b><br>—Å—Ä–µ–¥–Ω–µ–µ –∑–Ω–∞—á–µ–Ω–∏–µ: {mean_value} –¥–Ω—è",
    labels={'days_between_orders': '–î–Ω–µ–π –º–µ–∂–¥—É –ø—Ä–æ–¥–∞–∂–∞–º–∏'}
)

# –î–æ–±–∞–≤–ª—è–µ–º –ª–∏–Ω–∏—é —Å—Ä–µ–¥–Ω–µ–≥–æ –∑–Ω–∞—á–µ–Ω–∏—è
fig.add_vline(
    x=mean_value, 
    line_width=3, 
    line_dash="dash", 
    line_color=custom_palette[0],
)

fig.update_layout(
    bargap=0.05,
    xaxis_title="–î–Ω–µ–π –º–µ–∂–¥—É –ø—Ä–æ–¥–∞–∂–∞–º–∏",
    yaxis_title="–ö–æ–ª–∏—á–µ—Å—Ç–≤–æ –ø—Ä–æ–¥–∞–≤—Ü–æ–≤"
)

fig.write_image('1.png')
```

   - –†–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ –ø–æ –¥–Ω—è–º –¥–æ –ø–æ–≤—Ç–æ—Ä–Ω–æ–π –ø—Ä–æ–¥–∞–∂–∏ —Å–∫–æ—à–µ–Ω–æ, –≤–ø–ª–æ—Ç—å –¥–æ 60 –¥–Ω–µ–π, –ø–∏–∫–æ–≤–æ–µ –∑–Ω–∞—á–µ–Ω–∏–µ —Å–æ—Å—Ç–∞–≤–ª—è–µ—Ç 0 - 1 –¥–µ–Ω—å.

#### –î–∏–Ω–∞–º–∏–∫–∞ –∞–∫—Ç–∏–≤–Ω—ã—Ö –ø—Ä–æ–¥–∞–≤—Ü–æ–≤.

- **–û–ø–∏—Å–∞–Ω–∏–µ –º–µ—Ç—Ä–∏–∫–∏**:
    –ï–∂–µ–º–µ—Å—è—á–Ω–æ–µ –∏–∑–º–µ–Ω–µ–Ω–∏–µ –∫–æ–ª–∏—á–µ—Å—Ç–≤–∞ –∞–∫—Ç–∏–≤–Ω—ã—Ö –ø—Ä–æ–¥–∞–≤—Ü–æ–≤ –Ω–∞ –ø–ª–∞—Ç—Ñ–æ—Ä–º–µ. 
    
- **–ë–∏–∑–Ω–µ—Å-–∏–Ω—Ç–µ—Ä–ø—Ä–µ—Ç–∞—Ü–∏—è**:
    -    –†–æ—Å—Ç –æ–±—â–µ–≥–æ —á–∏—Å–ª–∞ –∞–∫—Ç–∏–≤–Ω—ã—Ö –ø—Ä–æ–¥–∞–≤—Ü–æ–≤: –ü–æ–∑–∏—Ç–∏–≤–Ω—ã–π —Ç—Ä–µ–Ω–¥, —É–∫–∞–∑—ã–≤–∞—é—â–∏–π –Ω–∞ —Ä–∞—Å—à–∏—Ä–µ–Ω–∏–µ –±–∞–∑—ã –ø—Ä–æ–¥–∞–≤—Ü–æ–≤ –Ω–∞ –ø–ª–∞—Ç—Ñ–æ—Ä–º–µ. –≠—Ç–æ –º–æ–∂–µ—Ç –ø—Ä–∏–≤–µ—Å—Ç–∏ –∫ —É–≤–µ–ª–∏—á–µ–Ω–∏—é –∞—Å—Å–æ—Ä—Ç–∏–º–µ–Ω—Ç–∞ —Ç–æ–≤–∞—Ä–æ–≤, –ø–æ–≤—ã—à–µ–Ω–∏—é –∫–æ–Ω–∫—É—Ä–µ–Ω—Ü–∏–∏ –∏ —Ä–æ—Å—Ç—É –ø–æ—Ç–µ–Ω—Ü–∏–∞–ª—å–Ω–æ–≥–æ –æ–±—ä–µ–º–∞ –ø—Ä–æ–¥–∞–∂.
    -    –í—ã—Å–æ–∫–∞—è –¥–æ–ª—è –Ω–æ–≤—ã—Ö –∞–∫—Ç–∏–≤–Ω—ã—Ö –ø—Ä–æ–¥–∞–≤—Ü–æ–≤: –£–∫–∞–∑—ã–≤–∞–µ—Ç –Ω–∞ —É—Å–ø–µ—à–Ω–æ–µ –ø—Ä–∏–≤–ª–µ—á–µ–Ω–∏–µ –Ω–æ–≤—ã—Ö –ø–∞—Ä—Ç–Ω–µ—Ä–æ–≤ –Ω–∞ –ø–ª–∞—Ç—Ñ–æ—Ä–º—É. –û–¥–Ω–∞–∫–æ –≤–∞–∂–Ω–æ –æ—Ç—Å–ª–µ–∂–∏–≤–∞—Ç—å –∏—Ö –¥–æ–ª–≥–æ—Å—Ä–æ—á–Ω—É—é –∞–∫—Ç–∏–≤–Ω–æ—Å—Ç—å –¥–ª—è –æ—Ü–µ–Ω–∫–∏ –∫–∞—á–µ—Å—Ç–≤–∞ –ø—Ä–∏–≤–ª–µ—á–µ–Ω–∏—è.
    -    –ù–∏–∑–∫–∞—è –¥–æ–ª—è –Ω–æ–≤—ã—Ö –∞–∫—Ç–∏–≤–Ω—ã—Ö –ø—Ä–æ–¥–∞–≤—Ü–æ–≤ –ø—Ä–∏ —Ä–∞—Å—Ç—É—â–µ–º –æ–±—â–µ–º —á–∏—Å–ª–µ: –ü—Ä–µ–¥–ø–æ–ª–∞–≥–∞–µ—Ç, —á—Ç–æ —Ä–æ—Å—Ç –æ–±–µ—Å–ø–µ—á–∏–≤–∞–µ—Ç—Å—è –∑–∞ —Å—á–µ—Ç —É–≤–µ–ª–∏—á–µ–Ω–∏—è –∞–∫—Ç–∏–≤–Ω–æ—Å—Ç–∏ —Å—É—â–µ—Å—Ç–≤—É—é—â–∏—Ö –ø—Ä–æ–¥–∞–≤—Ü–æ–≤, —á—Ç–æ —è–≤–ª—è–µ—Ç—Å—è —Ö–æ—Ä–æ—à–∏–º –∑–Ω–∞–∫–æ–º –ª–æ—è–ª—å–Ω–æ—Å—Ç–∏, –Ω–æ –Ω–µ–æ–±—Ö–æ–¥–∏–º–æ –ø—Ä–æ–¥–æ–ª–∂–∞—Ç—å —É—Å–∏–ª–∏—è –ø–æ –ø—Ä–∏–≤–ª–µ—á–µ–Ω–∏—é –Ω–æ–≤—ã—Ö –ø–∞—Ä—Ç–Ω–µ—Ä–æ–≤ –¥–ª—è —É—Å—Ç–æ–π—á–∏–≤–æ–≥–æ —Ä–æ—Å—Ç–∞.
    
- **–î–æ–ø–æ–ª–Ω–∏—Ç–µ–ª—å–Ω—ã–µ –¥–µ—Ç–∞–ª–∏:**
   - –°–æ–ø–æ—Å—Ç–∞–≤–ª—è—Ç—å —Å –¥–∏–Ω–∞–º–∏–∫–æ–π –∫–æ–ª–∏—á–µ—Å—Ç–≤–∞ –∞–∫—Ç–∏–≤–Ω—ã—Ö –ø–æ–∫—É–ø–∞—Ç–µ–ª–µ–π –¥–ª—è –æ—Ü–µ–Ω–∫–∏ —Å–±–∞–ª–∞–Ω—Å–∏—Ä–æ–≤–∞–Ω–Ω–æ—Å—Ç–∏ —Ä–æ—Å—Ç–∞ –ø–ª–∞—Ç—Ñ–æ—Ä–º—ã.
  
    

```python

query_seller_activation = """
WITH 
-- –í—Å–µ —É–Ω–∏–∫–∞–ª—å–Ω—ã–µ –ø—Ä–æ–¥–∞–≤—Ü—ã –ø–æ –º–µ—Å—è—Ü–∞–º –∏—Ö –ø–µ—Ä–≤–æ–π –ø—Ä–æ–¥–∞–∂–∏
first_sales AS (
    SELECT 
        seller_id,
        DATE_TRUNC('month', MIN(order_purchase_timestamp)) AS first_month
    FROM order_items oi
    JOIN orders o ON oi.order_id = o.order_id
    WHERE o.order_purchase_timestamp BETWEEN '2017-01-01' AND '2018-08-01'
    GROUP BY seller_id
),

-- –ú–µ—Å—è—á–Ω–∞—è –∞–∫—Ç–∏–≤–Ω–æ—Å—Ç—å –ø—Ä–æ–¥–∞–≤—Ü–æ–≤
monthly_active_sellers AS (
    SELECT
        DATE_TRUNC('month', o.order_purchase_timestamp) AS month,
        oi.seller_id,
        fs.first_month
    FROM order_items oi
    JOIN orders o ON oi.order_id = o.order_id
    JOIN first_sales fs ON oi.seller_id = fs.seller_id
    WHERE o.order_purchase_timestamp BETWEEN '2017-01-01' AND '2018-08-01'
    GROUP BY 1, 2, 3
)

-- –ò—Ç–æ–≥–æ–≤—ã–µ –º–µ—Ç—Ä–∏–∫–∏
SELECT
    month,
    COUNT(seller_id) AS active_sellers,
    SUM(CASE WHEN month = first_month THEN 1 ELSE 0 END) AS new_sellers,
    ROUND(SUM(CASE WHEN month = first_month THEN 1 ELSE 0 END) * 100.0 / 
          COUNT(seller_id), 1) AS new_sellers_percentage
FROM monthly_active_sellers
GROUP BY month
ORDER BY month;
"""

seller_activation_df = run_query(engine, query_seller_activation, fetch='df')
import plotly.graph_objects as go
from plotly.subplots import make_subplots

# –°–æ–∑–¥–∞–µ–º —Ñ–∏–≥—É—Ä—É —Å –¥–≤—É–º—è –≥—Ä–∞—Ñ–∏–∫–∞–º–∏
fig = make_subplots(rows=2, cols=1, 
                   subplot_titles=('–ö–æ–ª–∏—á–µ—Å—Ç–≤–æ –∞–∫—Ç–∏–≤–Ω—ã—Ö –ø—Ä–æ–¥–∞–≤—Ü–æ–≤ –ø–æ –º–µ—Å—è—Ü–∞–º',
                                   '–î–æ–ª—è –Ω–æ–≤—ã—Ö –ø—Ä–æ–¥–∞–≤—Ü–æ–≤ —Å—Ä–µ–¥–∏ –∞–∫—Ç–∏–≤–Ω—ã—Ö'),
                   vertical_spacing=0.15)


# –ì—Ä–∞—Ñ–∏–∫ –Ω–æ–≤—ã—Ö –ø—Ä–æ–¥–∞–≤—Ü–æ–≤
fig.add_trace(
    go.Bar(
        x=seller_activation_df['month'],
        y=seller_activation_df['new_sellers'],
        name='–ù–æ–≤—ã–µ –ø—Ä–æ–¥–∞–≤—Ü—ã',
        marker_color=custom_palette[0],
        hovertemplate="%{x|%b %Y}<br>–ù–æ–≤—ã—Ö: %{y:,}<extra></extra>"
    ),
    row=1, col=1
)


# –ì—Ä–∞—Ñ–∏–∫ –æ–±—â–µ–≥–æ –∫–æ–ª–∏—á–µ—Å—Ç–≤–∞ –∞–∫—Ç–∏–≤–Ω—ã—Ö –ø—Ä–æ–¥–∞–≤—Ü–æ–≤
fig.add_trace(
    go.Bar(
        x=seller_activation_df['month'],
        y=seller_activation_df['active_sellers'],
        name='–ê–∫—Ç–∏–≤–Ω—ã–µ –ø—Ä–æ–¥–∞–≤—Ü—ã',
        marker_color=custom_palette[4],
        hovertemplate="%{x|%b %Y}<br>–ê–∫—Ç–∏–≤–Ω—ã—Ö: %{y:,}<extra></extra>"
    ),
    row=1, col=1
)


# –ì—Ä–∞—Ñ–∏–∫ –¥–æ–ª–∏ –Ω–æ–≤—ã—Ö –ø—Ä–æ–¥–∞–≤—Ü–æ–≤
fig.add_trace(
    go.Scatter(
        x=seller_activation_df['month'],
        y=seller_activation_df['new_sellers_percentage'],
        name='–î–æ–ª—è –Ω–æ–≤—ã—Ö',
        mode='lines+markers',
        line=dict(color=custom_palette[0], width=3),
        marker=dict(size=8),
        hovertemplate="%{x|%b %Y}<br>–î–æ–ª—è –Ω–æ–≤—ã—Ö: %{y:.1f}%<extra></extra>",
        showlegend=False
    ),
    row=2, col=1
)

# –ù–∞—Å—Ç—Ä–∞–∏–≤–∞–µ–º –æ—Å–∏
fig.update_yaxes(title_text="–ö–æ–ª–∏—á–µ—Å—Ç–≤–æ –ø—Ä–æ–¥–∞–≤—Ü–æ–≤", row=1, col=1)
fig.update_yaxes(title_text="–î–æ–ª—è –Ω–æ–≤—ã—Ö, %", range=[0, 100], row=2, col=1)
fig.update_xaxes(title_text="–ú–µ—Å—è—Ü", row=2, col=1)

# –û–±—â–∏–µ –Ω–∞—Å—Ç—Ä–æ–π–∫–∏
fig.update_layout(
    title_text="<b>–ê–∫—Ç–∏–≤–∞—Ü–∏—è –ø—Ä–æ–¥–∞–≤—Ü–æ–≤ –Ω–∞ –º–∞—Ä–∫–µ—Ç–ø–ª–µ–π—Å–µ</b>",
    height=700,
    margin=dict(t=150),
    barmode='stack',
    legend=dict(orientation="h", yanchor="bottom", y=1.06, xanchor="right", x=1),
    hovermode="x unified"
)

# –î–æ–±–∞–≤–ª—è–µ–º —Å—Ä–µ–¥–Ω–∏–µ –∑–Ω–∞—á–µ–Ω–∏—è
avg_new_percentage = seller_activation_df['new_sellers_percentage'].mean()
fig.add_hline(y=avg_new_percentage, line_dash="dot", line_color=custom_palette[0], 
              annotation_text=f"–°—Ä–µ–¥–Ω–µ–µ: {avg_new_percentage:.1f}%", 
              annotation_position="bottom right",
              row=2, col=1)

fig.write_image('1.png')
```

### üìå –í—ã–≤–æ–¥—ã –ø–æ —ç—Ç–∞–ø—É.

-  **Repeat Purchase Rate** ‚Äî –≤—Å–µ–≥–æ 3%. –≠—Ç–æ –æ—Ç—Ä–∞–∂–∞–µ—Ç –Ω–∞—á–∞–ª—å–Ω—É—é —Å—Ç–∞–¥–∏—é e-commerce –≤ –ë—Ä–∞–∑–∏–ª–∏–∏ –≤ 2017‚Äì2018 –≥–≥., –≥–¥–µ —É –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª–µ–π –µ—â—ë –Ω–µ —Å—Ñ–æ—Ä–º–∏—Ä–æ–≤–∞–Ω–æ –ø—Ä–∏–≤—ã—á–µ–∫ –æ–Ω–ª–∞–π–Ω-—à–æ–ø–∏–Ω–≥–∞. –ü–æ–≤—Ç–æ—Ä–Ω—ã–µ –ø–æ–∫—É–ø–∫–∏ –º–æ–≥–ª–∏ —Å–¥–µ—Ä–∂–∏–≤–∞—Ç—å—Å—è –Ω–µ–¥–æ–≤–µ—Ä–∏–µ–º, —Å–ª–∞–±—ã–º –æ–ø—ã—Ç–æ–º –ø–µ—Ä–≤–æ–π –¥–æ—Å—Ç–∞–≤–∫–∏, –æ—Ç—Å—É—Ç—Å—Ç–≤–∏–µ–º –ø–æ—Å—Ç–ø—Ä–æ–¥–∞–∂–Ω—ã—Ö –∫–æ–º–º—É–Ω–∏–∫–∞—Ü–∏–π.
- **Average Orders** - –°—Ä–µ–¥–Ω–µ–µ —á–∏—Å–ª–æ –∑–∞–∫–∞–∑–æ–≤ —á—É—Ç—å –±–æ–ª—å—à–µ –¥–≤—É—Ö, —ç—Ç–æ –≥–æ–≤–æ—Ä–∏—Ç –æ —Ç–æ–º, —á—Ç–æ –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª–µ–π —Å –±–æ–ª–µ–µ —á–µ–º –¥–≤—É–º—è –∑–∞–∫–∞–∑–∞–º–∏ –æ—á–µ–Ω—å –º–∞–ª–æ.
-  **–í—Ä–µ–º—è –º–µ–∂–¥—É –ø–æ–∫—É–ø–∫–∞–º–∏:** –º–µ–¥–∏–∞–Ω–∞ ‚Äî 48 –¥–Ω–µ–π, —Å—Ä–µ–¥–Ω–µ–µ ‚Äî 96. –≠—Ç–æ —Å–∏–≥–Ω–∞–ª –æ –Ω–∏–∑–∫–æ–π —Å—Ç–µ–ø–µ–Ω–∏ –ø—Ä–∏–≤—è–∑–∞–Ω–Ω–æ—Å—Ç–∏ –∫ –ø–ª–∞—Ç—Ñ–æ—Ä–º–µ, –∞ —Ç–∞–∫–∂–µ –æ –Ω–µ—Ä–µ–≥—É–ª—è—Ä–Ω–æ–º –æ–Ω–±–æ—Ä–¥–∏–Ω–≥–µ.
-  **RSR –ø–æ –∫–∞—Ç–µ–≥–æ—Ä–∏—è–º:** –õ—É—á—à–∏–µ –ø–æ–∫–∞–∑–∞—Ç–µ–ª–∏ —Å–æ–≤–æ–∫—É–ø–Ω—ã–µ –ø–æ–∫–∞–∑–∞—Ç–µ–ª–∏ —É –∫–∞—Ç–µ–≥–æ—Ä–∏–π **—Ç–æ–≤–∞—Ä—ã –¥–ª—è –æ—Ç–¥—ã—Ö–∞**, **—Å–ø–æ—Ä—Ç –∏ –æ—Ç–¥—ã—Ö**, **–º–µ–±–µ–ª—å –∏ –¥–µ–∫–æ—Ä**  –∏ **—Å—É–º–∫–∏ –∏ –∞–∫—Å–µ—Å—Å—É–∞—Ä—ã** —Å–æ—á–µ—Ç–∞–Ω–∏–µ –±–æ–ª—å—à–æ–≥–æ –∫–æ–ª–∏—á–µ—Å—Ç–≤–∞ –∑–∞–∫–∞–∑–æ–≤ –∏ –ø–æ–∫–∞–∑–∞—Ç–µ–ª—è RPR –≤—ã—à–µ —Å—Ä–µ–¥–Ω–µ–≥–æ.

#### Retention

- **–û–ø–∏—Å–∞–Ω–∏–µ –º–µ—Ç—Ä–∏–∫–∏:**
  –ü–æ–∫–∞–∑—ã–≤–∞–µ—Ç, –∫–∞–∫–æ–π –ø—Ä–æ—Ü–µ–Ω—Ç –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª–µ–π –≤–µ—Ä–Ω—É–ª—Å—è –∫ –ø–æ–∫—É–ø–∫–∞–º —Å–ø—É—Å—Ç—è N –ø–µ—Ä–∏–æ–¥–æ–≤ –ø–æ—Å–ª–µ –ø–µ—Ä–≤–æ–π –ø–æ–∫—É–ø–∫–∏.

    - –í—ã—Å–æ–∫–æ–µ –∑–Ω–∞—á–µ–Ω–∏–µ –≥–æ–≤–æ—Ä–∏—Ç –æ –ª–æ—è–ª—å–Ω–æ—Å—Ç–∏ –∫–ª–∏–µ–Ω—Ç–æ–≤ –∏ —Å—Ç–∞–±–∏–ª—å–Ω–æ—Å—Ç–∏ –ø—Ä–æ–¥—É–∫—Ç–∞.
    - –ü–∞–¥–µ–Ω–∏–µ –º–æ–∂–µ—Ç —Å–∏–≥–Ω–∞–ª–∏–∑–∏—Ä–æ–≤–∞—Ç—å –æ –Ω–µ—É–¥–æ–≤–ª–µ—Ç–≤–æ—Ä—ë–Ω–Ω–æ—Å—Ç–∏, –ª–æ–≥–∏—Å—Ç–∏—á–µ—Å–∫–∏—Ö –ø—Ä–æ–±–ª–µ–º–∞—Ö –∏–ª–∏ –Ω–∏–∑–∫–æ–π —Ü–µ–Ω–Ω–æ—Å—Ç–∏ –ø—Ä–µ–¥–ª–æ–∂–µ–Ω–∏—è.

- **–î–æ–ø–æ–ª–Ω–∏—Ç–µ–ª—å–Ω—ã–µ –¥–µ—Ç–∞–ª–∏:**
    - –£—á–∏—Ç—ã–≤–∞—è –Ω–∏–∑–∫–∏–π % –ø–æ–≤—Ç–æ—Ä–Ω—ã—Ö –ø–æ–∫—É–ø–æ–∫ —É–¥–∞–ª–∏–º –º–µ—Å—è—Ü –Ω–∞—á–∞–ª–∞ –∏–∑ –¥–∏–∞–≥—Ä–∞–º–º—ã, —á—Ç–æ–±—ã —É–≤–µ–ª–∏—á–∏—Ç—å —Ä–∞–∑–ª–∏—á–∏—è –º–µ–∂–¥—É –º–µ—Å—è—Ü–∞–º–∏ –Ω–∞ –≥—Ä–∞—Ñ–∏–∫–µ.

```python
query = """
WITH
cohorts AS (
    SELECT 
        c.customer_unique_id,
        DATE_TRUNC('month', MIN(o.order_purchase_timestamp)) AS cohort_month
    FROM orders o
    JOIN customers c ON o.customer_id = c.customer_id
    WHERE o.order_status = '–¥–æ—Å—Ç–∞–≤–ª–µ–Ω' AND order_purchase_timestamp > '2017-01-01'
    GROUP BY 1
),
cohort_activity AS (
    SELECT
        c.cohort_month,
        DATE_TRUNC('month', o.order_purchase_timestamp) AS activity_month,
        COUNT(DISTINCT c.customer_unique_id) AS active_users
    FROM cohorts c
    JOIN customers cust ON c.customer_unique_id = cust.customer_unique_id
    JOIN orders o ON cust.customer_id = o.customer_id
    WHERE o.order_status = '–¥–æ—Å—Ç–∞–≤–ª–µ–Ω'
    GROUP BY 1, 2
),
retention_data AS (
    SELECT
        ca.cohort_month,
        EXTRACT(MONTH FROM AGE(ca.activity_month, ca.cohort_month)) 
            + 12 * EXTRACT(YEAR FROM AGE(ca.activity_month, ca.cohort_month)) 
            AS lifetime_month,
        ca.active_users,
        c.cohort_size,
        ROUND((ca.active_users * 100.0 / c.cohort_size)::numeric, 2) AS retention_rate
    FROM cohort_activity ca
    JOIN (
        SELECT cohort_month, COUNT(DISTINCT customer_unique_id) AS cohort_size
        FROM cohorts
        GROUP BY 1
    ) c USING(cohort_month)
    WHERE ca.activity_month >= ca.cohort_month
)

SELECT *
FROM retention_data
ORDER BY cohort_month, lifetime_month;
"""
df = run_query(engine, query, fetch='df')

# –ü—Ä–∏–º–µ—Ä –∏—Å–ø–æ–ª—å–∑–æ–≤–∞–Ω–∏—è
fig = plot_cohort_analysis(
    df,
    title="RETENTION –ü–û–ö–£–ü–ê–¢–ï–õ–ï–ô –ü–û –ö–û–ì–û–†–¢–ê–ú",
    value_col='retention_rate',
    zmax=100,
    colorscale='Blues',
    exclude_month_zero=True
)
fig.write_image('1.png')
```

- **–î–æ–ø–æ–ª–Ω–∏—Ç–µ–ª—å–Ω—ã–µ –¥–µ—Ç–∞–ª–∏:**  
–ò—Å–∫–ª—é—á–∞—Ç—å –æ—Ç–º–µ–Ω—ë–Ω–Ω—ã–µ –∏ –≤–æ–∑–≤—Ä–∞—Ç–Ω—ã–µ –∑–∞–∫–∞–∑—ã. –í–æ–∑–º–æ–∂–Ω–∞ —Ä–∞–∑–±–∏–≤–∫–∞ –ø–æ –∫–∞—Ç–µ–≥–æ—Ä–∏—è–º.

```python
query = """
WITH customer_orders AS (
    SELECT 
        customer_unique_id,
        order_purchase_timestamp,
        MIN(order_purchase_timestamp) OVER(PARTITION BY customer_unique_id) AS first_purchase_date,
        LEAD(order_purchase_timestamp) OVER(PARTITION BY customer_unique_id ORDER BY order_purchase_timestamp)
        AS second_purchase_date
    FROM orders o
    JOIN customers c ON c.customer_id = o.customer_id
    WHERE order_status = '–¥–æ—Å—Ç–∞–≤–ª–µ–Ω'
),

customers_with_flags AS (
    SELECT 
        customer_unique_id,
        DATE_TRUNC('month', first_purchase_date) AS cohort_month,
        second_purchase_date,
        EXTRACT(DAY FROM second_purchase_date - first_purchase_date) AS days_between_orders
    FROM customer_orders
    WHERE order_purchase_timestamp = first_purchase_date
    AND first_purchase_date >= '2017-01-01' -- —Ñ–∏–ª—å—Ç—Ä—É–µ–º —Ç–æ–ª—å–∫–æ –ø–µ—Ä–≤—É—é –ø–æ–∫—É–ø–∫—É
)

SELECT 
    cohort_month,
    COUNT(DISTINCT customer_unique_id) AS total_customers,
    COUNT(DISTINCT CASE WHEN days_between_orders IS NOT NULL AND days_between_orders <= 60 THEN customer_unique_id END) 
    AS repeat_in_60d,
    ROUND(
        COUNT(DISTINCT CASE WHEN days_between_orders IS NOT NULL AND days_between_orders <= 60 THEN customer_unique_id END)
        * 100.0 / 
        COUNT(DISTINCT customer_unique_id), 
        2
    ) AS repeat_rate_60d
FROM customers_with_flags
GROUP BY cohort_month
ORDER BY cohort_month;
"""

cohort = run_query(engine, query, fetch='df')

# –ü–æ–¥—Å—á–∏—Ç–∞–µ–º —Å—Ä–µ–¥–Ω–∏–π –ø–æ–∫–∞–∑–∞—Ç–µ–ª—å TSR 
mean =  round(cohort['repeat_in_60d'].sum() / cohort['total_customers'].sum() * 100, 2)

# –ü–æ—Å—Ç—Ä–æ–∏–º –≥—Ä–∞—Ñ–∏–∫
fig = px.bar(
    cohort, 
    x='cohort_month',
    y='repeat_rate_60d',
    labels={'cohort_month': '–ú–µ—Å—è—Ü –∫–æ–≥–æ—Ä—Ç—ã', 'repeat_rate_60d': 'TSR (%)'},
    title=f'<b>Retention –∑–∞ 60 –¥–Ω–µ–π - {mean}%</b>',
    color_discrete_sequence=custom_palette[::-1],
)

# –î–æ–±–∞–≤–ª—è–µ–º –ª–∏–Ω–∏—é —Å—Ä–µ–¥–Ω–µ–≥–æ –∑–Ω–∞—á–µ–Ω–∏—è
fig.add_hline(
    y=mean, 
    line_width=2, 
    line_dash="dash", 
    line_color=custom_palette[0],
    annotation_text=f"–°—Ä–µ–¥–Ω–µ–µ: {mean} %", 
)

fig.write_image('1.png')
```

- –ú–∞–ª—ã–µ –∫–æ–≥–æ—Ä—Ç—ã (–≤ –Ω–∞—á–∞–ª–µ) ‚Äî –Ω–µ–∏–Ω—Ñ–æ—Ä–º–∞—Ç–∏–≤–Ω—ã: —Å–ª–∏—à–∫–æ–º –º–∞–ª–æ –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª–µ–π ‚Üí –≤—ã—Å–æ–∫–∞—è –¥–∏—Å–ø–µ—Ä—Å–∏—è.

- –û–±—â–∏–π —É—Ä–æ–≤–µ–Ω—å –≤–æ–∑–≤—Ä–∞—Ç–æ–≤ –≤ 60 –¥–Ω–µ–π –¥–æ–≤–æ–ª—å–Ω–æ –Ω–∏–∑–∫–∏–π: —Ä–µ–¥–∫–æ –≤—ã—à–µ 2%.

- –ö–æ–≥–æ—Ä—Ç—ã  –º–∞–π - –Ω–æ—è–±—Ä—å 2017 –≥–æ–¥–∞, –ø–æ–∫–∞–∑—ã–≤–∞–ª–∏ —Ä–µ–∑—É–ª—å—Ç–∞—Ç TSR –≤—ã—à–µ —Å—Ä–µ–¥–Ω–µ–≥–æ,

- –¢—Ä–µ–Ω–¥ –ø–æ—Å–ª–µ –º–∞—Ä—Ç–∞ 2018 ‚Äî —Å—Ç–∞–±–∏–ª—å–Ω—ã–π —Å–ø–∞–¥, –≤–æ–∑–º–æ–∂–Ω–æ —É—Ö—É–¥—à–µ–Ω–∏–µ CX –∏–ª–∏ UX, —Ä–æ—Å—Ç –∫–æ–Ω–∫—É—Ä–µ–Ω—Ü–∏–∏, –ª–æ–≥–∏—Å—Ç–∏—á–µ—Å–∫–∏–µ —Å–±–æ–∏ –∏ —Ç.–ø.

### –ü—Ä–æ–¥–∞–≤—Ü—ã.

#### Retention

- **–û–ø–∏—Å–∞–Ω–∏–µ –º–µ—Ç—Ä–∏–∫–∏:**
  –ü–æ–∫–∞–∑—ã–≤–∞–µ—Ç, –Ω–∞—Å–∫–æ–ª—å–∫–æ –ø—Ä–æ–¥–∞–≤–µ—Ü —Å–ø–æ—Å–æ–±–µ–Ω —É–¥–µ—Ä–∂–∞—Ç—å –∫–ª–∏–µ–Ω—Ç–æ–≤ –Ω–∞ –ø–ª–∞—Ç—Ñ–æ—Ä–º–µ.

- **–ë–∏–∑–Ω–µ—Å-–∏–Ω—Ç–µ—Ä–ø—Ä–µ—Ç–∞—Ü–∏—è:**
    - –í—ã—Å–æ–∫–æ–µ —É–¥–µ—Ä–∂–∞–Ω–∏–µ —É–∫–∞–∑—ã–≤–∞–µ—Ç –Ω–∞ –¥–æ–≤–µ—Ä–∏–µ, —Ö–æ—Ä–æ—à–µ–µ –æ–±—Å–ª—É–∂–∏–≤–∞–Ω–∏–µ –∏–ª–∏ –∞—Å—Å–æ—Ä—Ç–∏–º–µ–Ω—Ç.
    - –ù–∏–∑–∫–æ–µ ‚Äî –≤–æ–∑–º–æ–∂–Ω–æ, –ø—Ä–æ–±–ª–µ–º—ã —Å –∫–∞—á–µ—Å—Ç–≤–æ–º, –¥–æ—Å—Ç–∞–≤–∫–æ–π –∏–ª–∏ –æ–∂–∏–¥–∞–Ω–∏—è–º–∏.

- **–î–æ–ø–æ–ª–Ω–∏—Ç–µ–ª—å–Ω—ã–µ –¥–µ—Ç–∞–ª–∏:**
    - –ú–æ–∂–µ—Ç –∏—Å–ø–æ–ª—å–∑–æ–≤–∞—Ç—å—Å—è –¥–ª—è —Ä–∞–Ω–∂–∏—Ä–æ–≤–∞–Ω–∏—è –ø—Ä–æ–¥–∞–≤—Ü–æ–≤.
    - –£–º–µ—Å—Ç–Ω–æ —Å–µ–≥–º–µ–Ω—Ç–∏—Ä–æ–≤–∞—Ç—å –ø–æ –∫–∞—Ç–µ–≥–æ—Ä–∏—è–º, —Ä–µ–≥–∏–æ–Ω–∞–º, NPS.

```python
query = """
WITH
cohorts AS (SELECT seller_id,
    DATE_TRUNC('month', MIN(order_purchase_timestamp)) AS cohort_month
FROM order_items 
JOIN orders USING(order_id)
WHERE order_status ='–¥–æ—Å—Ç–∞–≤–ª–µ–Ω' AND order_purchase_timestamp >= '2017-01-01'
GROUP BY 1),

cohort_activity AS(
SELECT cohort_month,
       DATE_TRUNC('month', order_purchase_timestamp) AS activity_month,
       COUNT(DISTINCT c.seller_id) as active_sellers
FROM cohorts AS c
JOIN order_items AS oi ON c.seller_id = oi.seller_id
JOIN orders AS o ON o.order_id = oi.order_id
WHERE order_status ='–¥–æ—Å—Ç–∞–≤–ª–µ–Ω'
GROUP BY 1, 2)

SELECT ca.cohort_month,
       EXTRACT(MONTH FROM AGE(activity_month, cohort_month)) + 12 * 
       EXTRACT(YEAR FROM AGE(activity_month, cohort_month)) AS lifetime_month,
       ca.active_sellers,
       cs.cohort_size,
       ROUND((ca.active_sellers * 100.0 / cs.cohort_size)::numeric, 2) as retention_rate,
       cs.cohort_size AS cohort_size_value
FROM cohort_activity ca
JOIN(SELECT cohort_month,
            COUNT(DISTINCT seller_id) AS cohort_size
    FROM cohorts
    GROUP BY 1) AS cs USING(cohort_month)
WHERE ca.activity_month >= ca.cohort_month
ORDER BY 1,2;
"""

df = run_query(engine, query, fetch='df')

# –ü—Ä–∏–º–µ—Ä –∏—Å–ø–æ–ª—å–∑–æ–≤–∞–Ω–∏—è
fig = plot_cohort_analysis(
    df,
    title="<b>Retention –ø—Ä–æ–¥–∞–≤—Ü–æ–≤ –ø–æ –∫–æ–≥–æ—Ä—Ç–∞–º</b>",
    value_col='retention_rate',
    zmax=80,
    zmin=20,
    colorscale='Blues',
    exclude_month_zero=False
)
fig.write_image('1.png')
```

**–ù–∞–±–ª—é–¥–µ–Ω–∏—è:**
- –°—Ç–∞–±–∏–ª—å–Ω–æ –≤—ã—Å–æ–∫–∏–µ –ø–æ–∫–∞–∑–∞—Ç–µ–ª–∏: 40-68% —á–µ—Ä–µ–∑ 1 –º–µ—Å—è—Ü
- –ß–µ—Ä–µ–∑ 6 –º–µ—Å—è—Ü–µ–≤ —É–¥–µ—Ä–∂–∏–≤–∞–µ—Ç—Å—è 33-52% –ø—Ä–æ–¥–∞–≤—Ü–æ–≤
- –õ—É—á—à–∏–µ –ø–æ–∫–∞–∑–∞—Ç–µ–ª–∏: –∫–æ–≥–æ—Ä—Ç–∞ **2017-01** (75.8% –Ω–∞ 1-–π –º–µ—Å—è—Ü –∏ 219 –Ω–æ–≤—ã—Ö –ø—Ä–æ–¥–∞–≤—Ü–æ–≤)
- –í—ã–≤–æ–¥: –ü—Ä–æ–¥–∞–≤—Ü—ã –¥–µ–º–æ–Ω—Å—Ç—Ä–∏—Ä—É—é—Ç —Ö–æ—Ä–æ—à—É—é –ª–æ—è–ª—å–Ω–æ—Å—Ç—å, –Ω–æ —Ç—Ä–µ–±—É–µ—Ç—Å—è –ø–æ–¥–¥–µ—Ä–∂–∫–∞ –Ω–æ–≤—ã—Ö —É—á–∞—Å—Ç–Ω–∏–∫–æ–≤

**–†–µ–∫–æ–º–µ–Ω–¥–∞—Ü–∏–∏:**
- –ê–∫—Ü–µ–Ω—Ç –Ω–∞ onboarding –¥–ª—è –Ω–æ–≤—ã—Ö –ø—Ä–æ–¥–∞–≤—Ü–æ–≤ –∏–∑ –º–∞–ª—ã—Ö –≥–æ—Ä–æ–¥–æ–≤
- –§–∏–Ω–∞–Ω—Å–æ–≤—ã–µ —Å—Ç–∏–º—É–ª—ã –¥–ª—è –ø–µ—Ä–≤—ã—Ö —É—á–∞—Å—Ç–Ω–∏–∫–æ–≤

---

#### Retention –≤ —Ç–µ—á–µ–Ω–∏–∏ 60 –¥–Ω–µ–π.

- **–û–ø–∏—Å–∞–Ω–∏–µ –º–µ—Ç—Ä–∏–∫–∏:**  
–î–æ–ª—è –ø—Ä–æ–¥–∞–≤—Ü–æ–≤, –ø–æ–ª—É—á–∏–≤—à–∏—Ö –≤—Ç–æ—Ä–æ–π –∑–∞–∫–∞–∑ –≤ —Ç–µ—á–µ–Ω–∏–µ 60 –¥–Ω–µ–π –ø–æ—Å–ª–µ –ø–µ—Ä–≤–æ–π –ø—Ä–æ–¥–∞–∂–∏.

- **–ë–∏–∑–Ω–µ—Å-–∏–Ω—Ç–µ—Ä–ø—Ä–µ—Ç–∞—Ü–∏—è:**  
–†–∞–Ω–Ω–∏–π –ø–æ–≤—Ç–æ—Ä–Ω—ã–π –∑–∞–∫–∞–∑ —É–∫–∞–∑—ã–≤–∞–µ—Ç –Ω–∞ –∫–∞—á–µ—Å—Ç–≤–æ —Ç–æ–≤–∞—Ä–Ω–æ–π –º–∞—Ç—Ä–∏—Ü—ã –∏ —Å–ø—Ä–æ—Å. –ù–∏–∑–∫–∞—è –º–µ—Ç—Ä–∏–∫–∞ ‚Äî —Å–∏–≥–Ω–∞–ª –æ —Å–ª–æ–∂–Ω–æ—Å—Ç–∏ —Å—Ç–∞—Ä—Ç–∞ –¥–ª—è –ø—Ä–æ–¥–∞–≤—Ü–æ–≤.

- **–î–æ–ø–æ–ª–Ω–∏—Ç–µ–ª—å–Ω—ã–µ –¥–µ—Ç–∞–ª–∏:**  
–ò—Å–∫–ª—é—á–∞—Ç—å –æ—Ç–º–µ–Ω—ë–Ω–Ω—ã–µ –∑–∞–∫–∞–∑—ã. –í—ã–¥–µ–ª—è—Ç—å –ø–æ –∫–∞—Ç–µ–≥–æ—Ä–∏—è–º –∏ —Ä–µ–≥–∏–æ–Ω–∞–º –ø—Ä–æ–¥–∞–≤—Ü–æ–≤. –£—á–∏—Ç—ã–≤–∞—Ç—å —Å–µ–∑–æ–Ω–Ω–æ—Å—Ç—å.

```python
query = """
WITH seller_orders AS (
    SELECT 
        oi.seller_id
        , o.order_purchase_timestamp::date AS order_date
        , MIN(o.order_purchase_timestamp) OVER(PARTITION BY oi.seller_id) AS first_sale_date
        , LEAD(o.order_purchase_timestamp) OVER(PARTITION BY oi.seller_id ORDER BY o.order_purchase_timestamp)
        AS second_sale_date
    FROM order_items oi
    JOIN orders o ON o.order_id = oi.order_id
    WHERE o.order_status = '–¥–æ—Å—Ç–∞–≤–ª–µ–Ω' AND order_purchase_timestamp BETWEEN '2017-01-01' AND '2018-08-01'
),

filtered AS (
    SELECT 
        seller_id
        , first_sale_date::date
        , second_sale_date::date
        , DATE_TRUNC('month', first_sale_date) AS cohort_month
        , EXTRACT(DAY FROM (second_sale_date - first_sale_date)) AS days_between_sales
    FROM seller_orders
    WHERE second_sale_date IS NOT NULL AND first_sale_date >= '2017-01-01'
),

cohort_stats AS (
    SELECT 
        cohort_month
        , COUNT(DISTINCT seller_id) AS total_sellers_with_2_sales
        , COUNT(DISTINCT CASE WHEN days_between_sales <= 30 THEN seller_id END) AS repeated_within_60d
    FROM filtered
    GROUP BY cohort_month
),

all_new_sellers AS (
    SELECT 
        DATE_TRUNC('month', MIN(o.order_purchase_timestamp)) AS cohort_month
        , oi.seller_id
    FROM order_items oi
    JOIN orders o ON o.order_id = oi.order_id
    WHERE o.order_status = '–¥–æ—Å—Ç–∞–≤–ª–µ–Ω'
    GROUP BY oi.seller_id
)

SELECT 
    cs.cohort_month
    , COUNT(DISTINCT ans.seller_id) AS total_new_sellers
    , cs.repeated_within_60d 
    , ROUND(cs.repeated_within_60d * 100.0 / COUNT(DISTINCT ans.seller_id), 2) AS repeat_rate_60d
FROM cohort_stats cs
JOIN all_new_sellers ans ON cs.cohort_month = ans.cohort_month
GROUP BY cs.cohort_month, cs.repeated_within_60d
ORDER BY cs.cohort_month;
"""

cohort_sales = run_query(engine, query, fetch='df')

# –ü–æ–¥—Å—á–∏—Ç–∞–µ–º —Å—Ä–µ–¥–Ω–∏–π –ø–æ–∫–∞–∑–∞—Ç–µ–ª—å TSR 
mean =  round(cohort_sales['repeated_within_60d'].sum() / cohort_sales['total_new_sellers'].sum() * 100, 2)

# –°—Ç—Ä–æ–∏–º area chart
fig = px.bar(
    cohort_sales,
    color_discrete_sequence=custom_palette[4:5],
    x='cohort_month',
    y='repeat_rate_60d',
    title=f'<b>Retention 30 –¥–Ω–µ–π –≤ —Å—Ä–µ–¥–Ω–µ–º - {mean}%</b>',
    labels={'cohort_month': '–ú–µ—Å—è—Ü –æ–Ω–±–æ—Ä–¥–∏–Ω–≥–∞', 'repeat_rate_60d': '–ü–æ–≤—Ç–æ—Ä–Ω–∞—è –ø—Ä–æ–¥–∞–∂–∞ (%)'},
)

# –î–æ–±–∞–≤–ª—è–µ–º –ª–∏–Ω–∏—é —Å—Ä–µ–¥–Ω–µ–≥–æ –∑–Ω–∞—á–µ–Ω–∏—è
fig.add_hline(
    y=mean, 
    line_width=2, 
    line_dash="dash", 
    line_color=custom_palette[0],
    annotation_text=f"–°—Ä–µ–¥–Ω–µ–µ: {mean} %", 
)


# –î–æ–±–∞–≤–ª—è–µ–º —Ü–≤–µ—Ç–æ–≤—É—é —à–∫–∞–ª—É
fig.update_layout(
    coloraxis_colorbar=dict(
        title="–†–∞–∑–º–µ—Ä –∫–æ–≥–æ—Ä—Ç—ã"
    )
)
fig.write_image('1.png')
```

**üìä –ò–Ω—Ç–µ—Ä–ø—Ä–µ—Ç–∞—Ü–∏—è —Ä–µ–∑—É–ª—å—Ç–∞—Ç–∞:**
- –ü–æ–∫–∞–∑–∞—Ç–µ–ª–∏ –≥–æ–≤–æ—Ä—è—Ç –æ —Ç–æ–º, —á—Ç–æ –±–æ–ª—å—à–∏–Ω—Å—Ç–≤–æ –ø—Ä–æ–¥–∞–≤—Ü–æ–≤ –ø–æ–ª—É—á–∞—é—Ç –ø–æ–≤—Ç–æ—Ä–Ω—ã–µ –∑–∞–∫–∞–∑—ã –≤ —Ç–µ—á–µ–Ω–∏–µ 30 –¥–Ω–µ–π. –û–¥–Ω–∞–∫–æ –≤ –Ω–µ–∫–æ—Ç–æ—Ä—ã—Ö –º–µ—Å—è—Ü–∞—Ö (–Ω–∞–ø—Ä–∏–º–µ—Ä, –∏—é–ª—å 2018 ‚Äî 52.17%) –Ω–∞–±–ª—é–¥–∞–µ—Ç—Å—è —Å–ø–∞–¥.

#### Retention –ø–æ –∫–∞–Ω–∞–ª–∞–º –ø—Ä–∏–≤–ª–µ—á–µ–Ω–∏—è.
- **–û–ø–∏—Å–∞–Ω–∏–µ:**  –ú–µ—Ç—Ä–∏–∫–∞ –æ–ø–∏—Å—ã–≤–∞–µ—Ç —ç—Ñ—Ñ–µ–∫—Ç–∏–≤–Ω–æ—Å—Ç—å –∫–∞–Ω–∞–ª–æ–≤ –ø—Ä–∏–≤–ª–µ—á–µ–Ω–∏—è –ø—Ä–æ–¥–∞–≤—Ü–æ–≤.

- **–ë–∏–∑–Ω–µ—Å –∏–Ω—Ç–µ—Ä–ø—Ä–µ—Ç–∞—Ü–∏—è:**  
    - **–ü—Ä–∞–≤—ã–π –≤–µ—Ä—Ö–Ω–∏–π –∫–≤–∞–¥—Ä–∞–Ω—Ç** = –∫–∞–Ω–∞–ª—ã, –ø—Ä–∏–≤–ª–µ–∫–∞—é—â–∏–µ –∫–∞—á–µ—Å—Ç–≤–µ–Ω–Ω—ã—Ö, –∞–∫—Ç–∏–≤–Ω—ã—Ö –∏ —É—Å—Ç–æ–π—á–∏–≤—ã—Ö –ø—Ä–æ–¥–∞–≤—Ü–æ–≤ ‚Äî –ø—Ä–∏–æ—Ä–∏—Ç–µ—Ç –¥–ª—è –º–∞—Å—à—Ç–∞–±–∏—Ä–æ–≤–∞–Ω–∏—è;
    - **–õ–µ–≤—ã–π –Ω–∏–∂–Ω–∏–π** = —Å–ª–∞–±—ã–µ –∫–∞–Ω–∞–ª—ã: –Ω–∏–∑–∫–æ–µ —É–¥–µ—Ä–∂–∞–Ω–∏–µ, –Ω–∏–∑–∫–∞—è —á–∞—Å—Ç–æ—Ç–∞;
    - **–ü—Ä–∞–≤—ã–π –Ω–∏–∂–Ω–∏–π** = –≤—ã—Å–æ–∫–∞—è –∞–∫—Ç–∏–≤–Ω–æ—Å—Ç—å, –Ω–æ –º–∞–ª–æ –∫—Ç–æ –¥–µ–ª–∞–µ—Ç –±–æ–ª—å—à–µ –æ–¥–Ω–æ–≥–æ –∑–∞–∫–∞–∑–∞ ‚Äî –≤–æ–∑–º–æ–∂–Ω–æ, –º–æ—Ç–∏–≤–∞—Ü–∏–æ–Ω–Ω—ã–µ –∫–∞–º–ø–∞–Ω–∏–∏;
    - **–õ–µ–≤—ã–π –≤–µ—Ä—Ö–Ω–∏–π** = —É–¥–µ—Ä–∂–∞–Ω–∏–µ –µ—Å—Ç—å, –Ω–æ –ø—Ä–æ–¥–∞–≤—Ü—ã –º–∞–ª–æ–ø—Ä–æ–¥—É–∫—Ç–∏–≤–Ω—ã ‚Äî –≤–æ–∑–º–æ–∂–Ω—ã –ø—Ä–æ–±–ª–µ–º—ã —Å –ª–æ–≥–∏—Å—Ç–∏–∫–æ–π –∏–ª–∏ –∞—Å—Å–æ—Ä—Ç–∏–º–µ–Ω—Ç–æ–º.

```python
query = """
WITH channel_stats AS (
    SELECT 
        CASE 
            WHEN mql.origin = '–¥—Ä—É–≥–∏–µ –∏—Å—Ç–æ—á–Ω–∏–∫–∏' THEN '–¥—Ä—É–≥–∏–µ –∫–∞–Ω–∞–ª—ã'
            ELSE mql.origin
        END AS acquisition_channel,
        COUNT(DISTINCT cd.seller_id) AS total_sellers,
        COUNT(DISTINCT CASE WHEN o.order_count >= 2 THEN cd.seller_id END) AS retained_sellers,
        COALESCE(ROUND(AVG(o.order_count), 2), 0) AS avg_orders_per_seller
    FROM closed_deals cd
    JOIN marketing_qualified mql 
        ON cd.mql_id = mql.mql_id
    LEFT JOIN (
        SELECT 
            seller_id, 
            COUNT(DISTINCT order_id) AS order_count
        FROM order_items 
        JOIN orders USING(order_id)
        WHERE order_status = '–¥–æ—Å—Ç–∞–≤–ª–µ–Ω'
        GROUP BY 1
    ) o ON cd.seller_id = o.seller_id
    GROUP BY 1
)

SELECT 
    acquisition_channel,
    total_sellers::int,
    retained_sellers::int,
    ROUND(100.0 * retained_sellers / total_sellers, 1)::float AS retention_rate,
    avg_orders_per_seller::float
FROM channel_stats
ORDER BY total_sellers DESC;

"""

df = run_query(engine, query, fetch='df')

df['avg_orders_per_seller'].median()

scatter_quadrant_plot(
    df=df,
    x_col='avg_orders_per_seller',
    y_col='retention_rate',
    x_mean=df['avg_orders_per_seller'].median(),
    y_mean=None,
    category_col='acquisition_channel',
    size_col='retained_sellers',
    title='<b>–≠—Ñ—Ñ–µ–∫—Ç–∏–≤–Ω–æ—Å—Ç—å –∫–∞–Ω–∞–ª–æ–≤: —É–¥–µ—Ä–∂–∞–Ω–∏–µ VS —á–∏—Å—Ç–æ—Ç–∞ </b>',
    condition='>',
    labels = {
        'retention_rate':'–î–æ–ª—è –ø–æ–≤—Ç–æ—Ä–Ω—ã—Ö –ø–æ–∫—É–ø–æ–∫, %',
        'avg_orders_per_seller':'–°—Ä–µ–¥–Ω–µ–µ –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –∑–∞–∫–∞–∑–æ–≤'
    },
    color_palette=custom_palette[::-1]
)
fig.write_image('1.png')
```

### NPS-–ø—Ä–æ–∫—Å–∏.

- **–û–ø–∏—Å–∞–Ω–∏–µ –º–µ—Ç—Ä–∏–∫–∏:** –ò–Ω–¥–µ–∫—Å —É–¥–æ–≤–ª–µ—Ç–≤–æ—Ä–µ–Ω–Ω–æ—Å—Ç–∏ –∫–ª–∏–µ–Ω—Ç–æ–≤, —Ä–∞—Å—Å—á–∏—Ç–∞–Ω–Ω—ã–π –∫–∞–∫ —Ä–∞–∑–Ω–æ—Å—Ç—å –º–µ–∂–¥—É –¥–æ–ª–µ–π –ø–æ–ª–æ–∂–∏—Ç–µ–ª—å–Ω—ã—Ö (5) –∏ –æ—Ç—Ä–∏—Ü–∞—Ç–µ–ª—å–Ω—ã—Ö (–æ—Ü–µ–Ω–∫–∞ 1‚Äì3) –æ—Ç–∑—ã–≤–æ–≤, –æ—Å—Ç–∞–≤–ª–µ–Ω–Ω—ã—Ö –ø–æ—Å–ª–µ –¥–æ—Å—Ç–∞–≤–∫–∏ –∑–∞–∫–∞–∑–∞. –ò—Å–ø–æ–ª—å–∑—É–µ—Ç—Å—è –∫–∞–∫ –ø—Ä–æ–∫—Å–∏ –¥–ª—è –æ—Ü–µ–Ω–∫–∏ —É—Ä–æ–≤–Ω—è –ª–æ—è–ª—å–Ω–æ—Å—Ç–∏ –≤ —É—Å–ª–æ–≤–∏—è—Ö –æ—Ç—Å—É—Ç—Å—Ç–≤–∏—è –ø—Ä—è–º—ã—Ö –æ–ø—Ä–æ—Å–æ–≤.
- **–ë–∏–∑–Ω–µ—Å-–∏–Ω—Ç–µ—Ä–ø—Ä–µ—Ç–∞—Ü–∏—è:**
    –Ø–≤–ª—è–µ—Ç—Å—è –∫–æ—Å–≤–µ–Ω–Ω—ã–º –∏–Ω–¥–∏–∫–∞—Ç–æ—Ä–æ–º –∫–ª–∏–µ–Ω—Ç—Å–∫–æ–≥–æ –æ–ø—ã—Ç–∞ –∏ –≥–æ—Ç–æ–≤–Ω–æ—Å—Ç–∏ —Ä–µ–∫–æ–º–µ–Ω–¥–æ–≤–∞—Ç—å –ø–ª–∞—Ç—Ñ–æ—Ä–º—É. –í—ã—Å–æ–∫–∏–π NPS-–ø—Ä–æ–∫—Å–∏ –º–æ–∂–µ—Ç —É–∫–∞–∑—ã–≤–∞—Ç—å –Ω–∞ –∫–∞—á–µ—Å—Ç–≤–µ–Ω–Ω–æ–µ –æ–±—Å–ª—É–∂–∏–≤–∞–Ω–∏–µ, –Ω–∞–¥—ë–∂–Ω—É—é –¥–æ—Å—Ç–∞–≤–∫—É –∏ —Å–æ–æ—Ç–≤–µ—Ç—Å—Ç–≤–∏–µ –æ–∂–∏–¥–∞–Ω–∏—è–º. –ù–∏–∑–∫–∏–π ‚Äî –Ω–∞ –ø—Ä–æ–±–ª–µ–º—ã –≤ –ª–æ–≥–∏—Å—Ç–∏–∫–µ, –ø–æ–¥–¥–µ—Ä–∂–∫–µ, –∞—Å—Å–æ—Ä—Ç–∏–º–µ–Ω—Ç–µ –∏–ª–∏ UX.

- **–î–æ–ø–æ–ª–Ω–∏—Ç–µ–ª—å–Ω—ã–µ –¥–µ—Ç–∞–ª–∏:**
    - –í —Ä–∞—Å—á—ë—Ç –≤–∫–ª—é—á–∞—é—Ç—Å—è —Ç–æ–ª—å–∫–æ –∑–∞–∫–∞–∑—ã —Å–æ —Å—Ç–∞—Ç—É—Å–æ–º "–¥–æ—Å—Ç–∞–≤–ª–µ–Ω", –ø–æ –∫–æ—Ç–æ—Ä—ã–º –æ—Å—Ç–∞–≤–ª–µ–Ω—ã –æ—Ü–µ–Ω–∫–∏.
    - –ú–æ–∂–µ—Ç –±—ã—Ç—å –∏—Å–∫–∞–∂–µ–Ω –∏–∑-–∑–∞ –Ω–∏–∑–∫–æ–π –¥–æ–ª–∏ –æ—Ç–∑—ã–≤–æ–≤ –∏ –æ—Ç—Å—É—Ç—Å—Ç–≤–∏—è –æ–±—è–∑–∞—Ç–µ–ª—å–Ω–æ—Å—Ç–∏ –æ—Ü–µ–Ω–∫–∏.
    - –ù–µ —É—á–∏—Ç—ã–≤–∞–µ—Ç –Ω–µ–π—Ç—Ä–∞–ª—å–Ω—ã–µ –æ—Ü–µ–Ω–∫–∏ (4).


```python
query = """
WITH base_data AS (
  SELECT 
    DATE_TRUNC('month', o.order_purchase_timestamp) AS month,
    r.review_score,
    COUNT(DISTINCT o.order_id) AS total_orders,
    COUNT(r.review_id) AS reviews_count
  FROM orders o
  LEFT JOIN order_reviews r USING(order_id)
  WHERE o.order_purchase_timestamp BETWEEN '2017-01-01' AND '2018-08-01'
    AND o.order_status = '–¥–æ—Å—Ç–∞–≤–ª–µ–Ω'
  GROUP BY 1, 2
)
SELECT 
  month,
  ROUND(
    (SUM(CASE WHEN review_score = 5 THEN reviews_count ELSE 0 END) -
    SUM(CASE WHEN review_score <= 3 THEN reviews_count ELSE 0 END)
  ) * 100.0 / NULLIF(SUM(reviews_count), 0)) AS nps_proxy,
  -- –î–æ–ø–æ–ª–Ω–∏—Ç–µ–ª—å–Ω—ã–µ –º–µ—Ç—Ä–∏–∫–∏
  SUM(reviews_count) AS total_reviews,
  ROUND(SUM(reviews_count)*100.0 / SUM(total_orders), 1) AS response_rate,
  SUM(CASE WHEN review_score = 5 THEN reviews_count ELSE 0 END) AS promoters,
  SUM(CASE WHEN review_score <= 3 THEN reviews_count ELSE 0 END) AS detractors,
  SUM(CASE WHEN review_score = 4 THEN reviews_count ELSE 0 END) AS neutrals
FROM base_data
GROUP BY 1
ORDER BY 1;
"""

nps = run_query(engine, query, fetch='df')


# –ò—Å–ø–æ–ª—å–∑—É–µ–º –∫–∞—Å—Ç–æ–º–Ω—ã–π —Å–∫—Ä–∏–ø—Ç
plot_nps_analysis(nps)

fig.write_image('1.png')
```

- –°—Ä–µ–¥–Ω–∏–π —É—Ä–æ–≤–µ–Ω—å NPS: 38% ‚Äî –∏ –∏–º–µ–µ—Ç —Ç—Ä–µ–Ω–¥ –∫ —Ä–æ—Å—Ç—É.

**–†–µ–∑–∫–æ–µ –ø–∞–¥–µ–Ω–∏–µ NPS –≤ –Ω–∞—á–∞–ª–µ 2018**
   - –§–µ–≤—Ä–∞–ª—å‚Äì–º–∞—Ä—Ç 2018: –ø—Ä–æ–≤–∞–ª –¥–æ 23% ‚Üí 20%, —Å–∞–º—ã–π –Ω–∏–∑–∫–∏–π —É—Ä–æ–≤–µ–Ω—å –∑–∞ –ø–µ—Ä–∏–æ–¥.

*–û—Ö–≤–∞—Ç –æ—Ç–∑—ã–≤–∞–º–∏ —Å—Ç–∞–±–∏–ª—å–Ω–æ –≤—ã—Å–æ–∫–∏–π: 98‚Äì99% –∑–∞–∫–∞–∑–æ–≤.*

### –ü—Ä–æ–∫—Å–∏ NPS –ø–æ –∫–∞—Ç–µ–≥–æ—Ä–∏—è–º.

- **–û–ø–∏—Å–∞–Ω–∏–µ –º–µ—Ç—Ä–∏–∫–∏:** –ò—Å–ø–æ–ª—å–∑—É–µ—Ç—Å—è –∫–∞–∫ –ø—Ä–æ–∫—Å–∏ –¥–ª—è –æ—Ü–µ–Ω–∫–∏ —É—Ä–æ–≤–Ω—è —É–¥–æ–≤–ª–µ—Ç–≤–æ—Ä–µ–Ω–Ω–æ—Å—Ç–∏ –ø–æ–∫—É–ø–∞—Ç–µ–ª–µ–π —Ç–æ–≤–∞—Ä–∞–º–∏ —Ä–∞–∑–ª–∏—á–Ω—ã—Ö –∫–∞—Ç–µ–≥–æ—Ä–∏–π.
- **–ë–∏–∑–Ω–µ—Å-–∏–Ω—Ç–µ—Ä–ø—Ä–µ—Ç–∞—Ü–∏—è:**
    –í—ã—Å–æ–∫–∏–µ —Å—Ä–µ–¥–Ω–∏–µ –æ—Ü–µ–Ω–∫–∏ –∫–∞—Ç–µ–≥–æ—Ä–∏–∏: –£–∫–∞–∑—ã–≤–∞—é—Ç –Ω–∞ –æ–±—â–µ–µ —É–¥–æ–≤–ª–µ—Ç–≤–æ—Ä–µ–Ω–∏–µ –ø–æ–∫—É–ø–∞—Ç–µ–ª–µ–π —Ç–æ–≤–∞—Ä–∞–º–∏ –¥–∞–Ω–Ω–æ–π –∫–∞—Ç–µ–≥–æ—Ä–∏–∏. –≠—Ç–æ –º–æ–∂–µ—Ç —Å–≤–∏–¥–µ—Ç–µ–ª—å—Å—Ç–≤–æ–≤–∞—Ç—å –æ –≤—ã—Å–æ–∫–æ–º –∫–∞—á–µ—Å—Ç–≤–µ —Ç–æ–≤–∞—Ä–æ–≤, –∞–¥–µ–∫–≤–∞—Ç–Ω–æ–º –æ–ø–∏—Å–∞–Ω–∏–∏, —Å–æ–æ—Ç–≤–µ—Ç—Å—Ç–≤–∏–∏ –æ–∂–∏–¥–∞–Ω–∏—è–º –∏ —Ö–æ—Ä–æ—à–µ–º –æ–ø—ã—Ç–µ –ø–æ–∫—É–ø–∫–∏.
    –ù–∏–∑–∫–∏–µ —Å—Ä–µ–¥–Ω–∏–µ –æ—Ü–µ–Ω–∫–∏ –∫–∞—Ç–µ–≥–æ—Ä–∏–∏: –°–∏–≥–Ω–∞–ª–∏–∑–∏—Ä—É—é—Ç –æ –ø–æ—Ç–µ–Ω—Ü–∏–∞–ª—å–Ω—ã—Ö –ø—Ä–æ–±–ª–µ–º–∞—Ö —Å —Ç–æ–≤–∞—Ä–∞–º–∏ –¥–∞–Ω–Ω–æ–π –∫–∞—Ç–µ–≥–æ—Ä–∏–∏. –¢—Ä–µ–±—É–µ—Ç—Å—è –¥–æ–ø–æ–ª–Ω–∏—Ç–µ–ª—å–Ω–æ–µ –∏—Å—Å–ª–µ–¥–æ–≤–∞–Ω–∏–µ –¥–ª—è –≤—ã—è–≤–ª–µ–Ω–∏—è –ø—Ä–∏—á–∏–Ω (–Ω–∞–ø—Ä–∏–º–µ—Ä, –Ω–∏–∑–∫–æ–µ –∫–∞—á–µ—Å—Ç–≤–æ —Ç–æ–≤–∞—Ä–æ–≤, –Ω–µ—Å–æ–æ—Ç–≤–µ—Ç—Å—Ç–≤–∏–µ –æ–ø–∏—Å–∞–Ω–∏—é, –ø—Ä–æ–±–ª–µ–º—ã —Å –ø—Ä–æ–¥–∞–≤—Ü–∞–º–∏, –ø—Ä–µ–¥–ª–∞–≥–∞—é—â–∏–º–∏ —Ç–æ–≤–∞—Ä—ã —ç—Ç–æ–π –∫–∞—Ç–µ–≥–æ—Ä–∏–∏).
    –°—Ä–∞–≤–Ω–µ–Ω–∏–µ –ø—Ä–æ–∫—Å–∏ NPS –º–µ–∂–¥—É –∫–∞—Ç–µ–≥–æ—Ä–∏—è–º–∏ –ø–æ–º–æ–≥–∞–µ—Ç –≤—ã—è–≤–∏—Ç—å –Ω–∞–∏–±–æ–ª–µ–µ –∏ –Ω–∞–∏–º–µ–Ω–µ–µ —É–¥–æ–≤–ª–µ—Ç–≤–æ—Ä—è—é—â–∏–µ –∫–∞—Ç–µ–≥–æ—Ä–∏–∏, —á—Ç–æ –º–æ–∂–µ—Ç –ø–æ–≤–ª–∏—è—Ç—å –Ω–∞ —Å—Ç—Ä–∞—Ç–µ–≥–∏—é —Ä–∞–∑–≤–∏—Ç–∏—è –∞—Å—Å–æ—Ä—Ç–∏–º–µ–Ω—Ç–∞ –∏ —Ä–∞–±–æ—Ç—É —Å –ø–æ—Å—Ç–∞–≤—â–∏–∫–∞–º–∏.

- **–î–æ–ø–æ–ª–Ω–∏—Ç–µ–ª—å–Ω—ã–µ –¥–µ—Ç–∞–ª–∏:**
    - –í —Ä–∞—Å—á—ë—Ç –≤–∫–ª—é—á–∞—é—Ç—Å—è —Ç–æ–ª—å–∫–æ –∑–∞–∫–∞–∑—ã —Å–æ —Å—Ç–∞—Ç—É—Å–æ–º "–¥–æ—Å—Ç–∞–≤–ª–µ–Ω", –ø–æ –∫–æ—Ç–æ—Ä—ã–º –æ—Å—Ç–∞–≤–ª–µ–Ω—ã –æ—Ü–µ–Ω–∫–∏.
    - –û—Ç—Å–ª–µ–∂–∏–≤–∞—Ç—å –¥–∏–Ω–∞–º–∏–∫—É –ø—Ä–æ–∫—Å–∏ NPS –ø–æ –∫–∞—Ç–µ–≥–æ—Ä–∏—è–º –≤–æ –≤—Ä–µ–º–µ–Ω–∏, –æ—Å–æ–±–µ–Ω–Ω–æ –ø–æ—Å–ª–µ –∏–∑–º–µ–Ω–µ–Ω–∏–π –≤ –∞—Å—Å–æ—Ä—Ç–∏–º–µ–Ω—Ç–µ –∏–ª–∏ —Ä–∞–±–æ—Ç–µ —Å –ø–æ—Å—Ç–∞–≤—â–∏–∫–∞–º–∏.

```python
query = """

WITH base_data AS (
  SELECT 
    DATE_TRUNC('month', o.order_purchase_timestamp) AS month,
    p.category_name_translated,
    r.review_score,
    COUNT(DISTINCT o.order_id) AS total_orders,
    COUNT(r.review_id) AS reviews_count
  FROM orders o
  LEFT JOIN order_reviews r USING(order_id)
  LEFT JOIN order_items i USING(order_id)
  LEFT JOIN products p USING(product_id)
  WHERE o.order_purchase_timestamp BETWEEN '2017-01-01' AND '2018-08-01'
    AND o.order_status = '–¥–æ—Å—Ç–∞–≤–ª–µ–Ω'
    AND p.category_name_translated IS NOT NULL
  GROUP BY 1, 2, 3
),
category_nps AS (
  SELECT 
    month,
    category_name_translated,
    ROUND(
      (SUM(CASE WHEN review_score = 5 THEN reviews_count ELSE 0 END) -
      SUM(CASE WHEN review_score <= 3 THEN reviews_count ELSE 0 END)
    ) * 100.0 / NULLIF(SUM(reviews_count), 0)) AS nps_proxy,
    SUM(reviews_count) AS total_reviews
  FROM base_data
  GROUP BY 1, 2
  HAVING SUM(reviews_count) >= 50  -- –ú–∏–Ω–∏–º—É–º 50 –æ—Ç–∑—ã–≤–æ–≤ –¥–ª—è —Å—Ç–∞—Ç–∏—Å—Ç–∏—á–µ—Å–∫–æ–π –∑–Ω–∞—á–∏–º–æ—Å—Ç–∏
)
SELECT 
  month,
  category_name_translated,
  nps_proxy,
  total_reviews,
  -- –û–ø—Ä–µ–¥–µ–ª—è–µ–º —Ç–æ–ø –∏ —Ö—É–¥—à–∏–µ –∫–∞—Ç–µ–≥–æ—Ä–∏–∏
  CASE 
    WHEN nps_proxy >= 70 THEN 'TOP'
    WHEN nps_proxy <= 30 THEN 'BOTTOM'
    ELSE 'MIDDLE'
  END AS category_group
FROM category_nps
ORDER BY 1, 2;
"""
category_nps_df = run_query(engine, query, fetch='df')



# –†–∞—Å—Å—á–∏—Ç–∞–µ–º —Å—Ç–∞—Ç–∏—Å—Ç–∏–∫—É –ø–æ –∫–∞—Ç–µ–≥–æ—Ä–∏—è–º
category_stats = category_nps_df.groupby('category_name_translated').agg(
    avg_nps=('nps_proxy', 'mean'),
    months_count=('month', 'nunique'),
    total_reviews=('total_reviews', 'sum')
).reset_index()

# –û—Ç—Ñ–∏–ª—å—Ç—Ä—É–µ–º –∫–∞—Ç–µ–≥–æ—Ä–∏–∏ —Å –Ω–µ–¥–æ—Å—Ç–∞—Ç–æ—á–Ω—ã–º–∏ –¥–∞–Ω–Ω—ã–º–∏
MIN_MONTHS = 6  # –ú–∏–Ω–∏–º—É–º 6 –º–µ—Å—è—Ü–µ–≤ —Å –¥–∞–Ω–Ω—ã–º–∏
MIN_REVIEWS = 100  # –ú–∏–Ω–∏–º—É–º 100 –æ—Ç–∑—ã–≤–æ–≤
valid_categories = category_stats[
    (category_stats['months_count'] >= MIN_MONTHS) & 
    (category_stats['total_reviews'] >= MIN_REVIEWS)
]

# 3. –ù–∞–π–¥–µ–º –ª—É—á—à—É—é –∏ —Ö—É–¥—à—É—é –∫–∞—Ç–µ–≥–æ—Ä–∏–∏ —Å—Ä–µ–¥–∏ –≤–∞–ª–∏–¥–Ω—ã—Ö
top_category = valid_categories.loc[valid_categories['avg_nps'].idxmax()]
bottom_category = valid_categories.loc[valid_categories['avg_nps'].idxmin()]

# 4. –ü–æ–¥–≥–æ—Ç–æ–≤–∏–º –¥–∞–Ω–Ω—ã–µ
top_data = category_nps_df[category_nps_df['category_name_translated'] == top_category['category_name_translated']]
bottom_data = category_nps_df[category_nps_df['category_name_translated'] == bottom_category['category_name_translated']]
other_data = category_nps_df[
    ~category_nps_df['category_name_translated'].isin([top_category['category_name_translated'],
                                                       bottom_category['category_name_translated']]) &
    category_nps_df['category_name_translated'].isin(valid_categories['category_name_translated'])
]

# 5. –°–æ–∑–¥–∞–µ–º –≤–∏–∑—É–∞–ª–∏–∑–∞—Ü–∏—é
fig = go.Figure()

# –í—Å–µ –æ—Å—Ç–∞–ª—å–Ω—ã–µ –≤–∞–ª–∏–¥–Ω—ã–µ –∫–∞—Ç–µ–≥–æ—Ä–∏–∏ (–ø–æ–ª—É–ø—Ä–æ–∑—Ä–∞—á–Ω—ã–µ)
for category in other_data['category_name_translated'].unique():
    cat_data = other_data[other_data['category_name_translated'] == category]
    fig.add_trace(
        go.Scatter(
            x=cat_data['month'],
            y=cat_data['nps_proxy'],
            name=category,
            line=dict(color='lightgray', width=1),
            showlegend=False,
            hovertemplate="%{x|%b %Y}<br><b>" + category + "</b><br>NPS: %{y}%<br>–û—Ç–∑—ã–≤–æ–≤: " + 
                         cat_data['total_reviews'].astype(str)
        )
    )

# –õ—É—á—à–∞—è –∫–∞—Ç–µ–≥–æ—Ä–∏—è
fig.add_trace(
    go.Scatter(
        x=top_data['month'],
        y=top_data['nps_proxy'],
        name=f"–õ—É—á—à–∞—è: {top_category['category_name_translated']}",
        line=dict(color='#2ca02c', width=4),
        mode='lines+markers',
        hovertemplate="%{x|%b %Y}<br><b>–õ—É—á—à–∞—è: " + top_category['category_name_translated'] + 
                     "</b><br>NPS: %{y}%<br>–û—Ç–∑—ã–≤–æ–≤: " + top_data['total_reviews'].astype(str)
    )
)

# –•—É–¥—à–∞—è –∫–∞—Ç–µ–≥–æ—Ä–∏—è
fig.add_trace(
    go.Scatter(
        x=bottom_data['month'],
        y=bottom_data['nps_proxy'],
        name=f"–•—É–¥—à–∞—è: {bottom_category['category_name_translated']}",
        line=dict(color='#d62728', width=4),
        mode='lines+markers',
        hovertemplate="%{x|%b %Y}<br><b>–•—É–¥—à–∞—è: " + bottom_category['category_name_translated'] + 
                     "</b><br>NPS: %{y}%<br>–û—Ç–∑—ã–≤–æ–≤: " + bottom_data['total_reviews'].astype(str)
    )
)

# –î–æ–±–∞–≤–∏–º –∑–æ–Ω—ã NPS
fig.add_hrect(y0=70, y1=100, fillcolor="green", opacity=0.1, line_width=0,
             annotation_text="–í—ã—Å–æ–∫–∏–π NPS", annotation_position="top left")
fig.add_hrect(y0=-50, y1=30, fillcolor="red", opacity=0.1, line_width=0,
             annotation_text="–ù–∏–∑–∫–∏–π NPS", annotation_position="bottom left")

fig.update_layout(
    title=f"<b>–ê–Ω–∞–ª–∏–∑ NPS –ø–æ –∫–∞—Ç–µ–≥–æ—Ä–∏—è–º</b>",
    yaxis_title="NPS, %",
    xaxis_title="–ú–µ—Å—è—Ü",
    height=600,
    legend=dict(
        orientation="h",
        yanchor="bottom",
        y=1.02,
        xanchor="right",
        x=1
    ),
    margin=dict(t=100)
)

fig.write_image('1.png')
```

- –í –∏—é–ª–µ 2018 –ª—É—á—à–∏–π –ø–æ–∫–∞–∑–∞—Ç–µ–ª—å NPS –∑–∞ –≤—Å—ë –≤—Ä–µ–º—è –∏—Å—Å–ª–µ–¥–æ–≤–∞–Ω–∏—è –ø–æ–ª—É—á–∏–ª–∞ –∫–∞—Ç–µ–≥–æ—Ä–∏—è **–ø–∞—Ä—Ñ—é–º–µ—Ä–∏—è**.

```python
import plotly.graph_objects as go
import pandas as pd

# 1. –°–Ω–∞—á–∞–ª–∞ –ø–æ–ª—É—á–∏–º –¥–∞–Ω–Ω—ã–µ NPS –ø–æ —à—Ç–∞—Ç–∞–º (–∑–∞–º–µ–Ω–∏—Ç–µ –Ω–∞ –≤–∞—à —Ä–µ–∞–ª—å–Ω—ã–π –∑–∞–ø—Ä–æ—Å)
query = """
WITH customer_state_stats AS (
  SELECT 
    c.customer_state AS state,
    COUNT(DISTINCT o.order_id) AS total_orders,
    SUM(CASE WHEN r.review_score = 5 THEN 1 ELSE 0 END) AS promoters,
    SUM(CASE WHEN r.review_score <= 3 THEN 1 ELSE 0 END) AS detractors,
    COUNT(r.review_id) AS total_reviews
  FROM orders o
  LEFT JOIN order_reviews r USING(order_id)
  JOIN customers c USING(customer_id)
  WHERE o.order_purchase_timestamp BETWEEN '2017-01-01' AND '2018-08-01'
    AND o.order_status = '–¥–æ—Å—Ç–∞–≤–ª–µ–Ω'
  GROUP BY 1
)
SELECT 
  state,
  total_orders,
  ROUND((promoters - detractors) * 100.0 / NULLIF(total_reviews, 0), 1) AS nps_score,
  total_reviews
FROM customer_state_stats
WHERE total_reviews >= 30  -- –§–∏–ª—å—Ç—Ä –ø–æ –º–∏–Ω–∏–º–∞–ª—å–Ω–æ–º—É –∫–æ–ª–∏—á–µ—Å—Ç–≤—É –æ—Ç–∑—ã–≤–æ–≤
ORDER BY nps_score DESC;
"""


states = run_query(engine, query, fetch='df')

# –î–æ–±–∞–≤–ª—è–µ–º –ø–æ–ª–Ω—ã–µ –Ω–∞–∑–≤–∞–Ω–∏—è —à—Ç–∞—Ç–æ–≤
states['state_name'] = states['state'].map(state_name_mapping)

# –°–æ–∑–¥–∞–µ–º —Ö–æ—Ä–æ–ø–ª–µ—Ç
fig = go.Figure(go.Choropleth(
    locations=states['state_name'],
    z=states['nps_score'],
    geojson=geo_json,
    featureidkey='properties.name',
    colorscale='RdYlGn',
    zmid=25,  # –¶–µ–Ω—Ç—Ä —à–∫–∞–ª—ã –Ω–∞ 0 (NPS –º–æ–∂–µ—Ç –±—ã—Ç—å –æ—Ç—Ä–∏—Ü–∞—Ç–µ–ª—å–Ω—ã–º)
    marker_line_color='white',
    marker_line_width=0.5,
    colorbar_title='NPS, %',
    customdata=states[['total_orders', 'total_reviews']],
    hovertemplate=(
        "<b>%{location}</b><br>"
        "NPS: %{z:.0f}%<br>"
        "–ó–∞–∫–∞–∑–æ–≤: %{customdata[0]:,}<br>"
        "–û—Ç–∑—ã–≤–æ–≤: %{customdata[1]:,}<extra></extra>"
    )
))

# –ù–∞—Å—Ç—Ä–æ–π–∫–∏ –∫–∞—Ä—Ç—ã
fig.update_geos(
    fitbounds="locations",
    visible=False,
    projection_scale=3.5  # –ù–µ–±–æ–ª—å—à–æ–π –∑—É–º –Ω–∞ –ë—Ä–∞–∑–∏–ª–∏—é
)

# –ù–∞—Å—Ç—Ä–æ–π–∫–∏ –º–∞–∫–µ—Ç–∞
fig.update_layout(
    title_text='<b>–ò–Ω–¥–µ–∫—Å proxy NPS –ø–æ —à—Ç–∞—Ç–∞–º –ë—Ä–∞–∑–∏–ª–∏–∏</b>',
    margin=dict(l=0, r=0, t=40, b=0),
    height=700,
    geo=dict(
        lakecolor='rgba(0,125,255,0.1)'
    )
)

fig.write_image('1.png')
```

- **–ë–∏–∑–Ω–µ—Å-–∏–Ω—Ç–µ—Ä–ø—Ä–µ—Ç–∞—Ü–∏—è:**
    –ü–æ–∑–≤–æ–ª—è–µ—Ç —Å—É–¥–∏—Ç—å –æ –µ—Å—Ç–µ—Å—Ç–≤–µ–Ω–Ω–æ–º —Ä–æ—Å—Ç–µ –ø–ª–∞—Ç—Ñ–æ—Ä–º—ã –±–ª–∞–≥–æ–¥–∞—Ä—è –ª–æ—è–ª—å–Ω—ã–º –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª—è–º. –†–æ—Å—Ç –¥–æ–ª–∏ —Ä–µ—Ñ–µ—Ä–∞–ª—å–Ω—ã—Ö –∫–∞–Ω–∞–ª–æ–≤ ‚Äî –ø—Ä–∏–∑–Ω–∞–∫ –¥–æ–≤–µ—Ä–∏—è –∫ –ø–ª–∞—Ç—Ñ–æ—Ä–º–µ –∏ –µ—ë "–≤–∏—Ä—É—Å–Ω–æ–≥–æ" —ç—Ñ—Ñ–µ–∫—Ç–∞.

    - –í –æ—Ç—Å—É—Ç—Å—Ç–≤–∏–µ –ø—Ä—è–º—ã—Ö UTM-–º–µ—Ç–æ–∫ –∏ –ø—Ä–æ–º–æ–∫–æ–¥–æ–≤ –∞–Ω–∞–ª–∏–∑ –±–∞–∑–∏—Ä—É–µ—Ç—Å—è –Ω–∞ –∫–∞–Ω–∞–ª–∞—Ö –ø—Ä–∏–≤–ª–µ—á–µ–Ω–∏—è.
    - –ö–∞–Ω–∞–ª—ã —Å –Ω–∏–∑–∫–æ–π –¥–æ–ª–µ–π –∑–∞—Ç—Ä–∞—Ç, –Ω–æ –≤—ã—Å–æ–∫–∏–º retention, —Ç–∞–∫–∂–µ –º–æ–≥—É—Ç –≤—ã—Å—Ç—É–ø–∞—Ç—å –∫–∞–∫ —Ä–µ—Ñ–µ—Ä–∞–ª—å–Ω—ã–µ.
    - –í–∞–∂–Ω–æ –∫–æ—Ä—Ä–µ–∫—Ç–Ω–æ –∫–ª–∞—Å—Å–∏—Ñ–∏—Ü–∏—Ä–æ–≤–∞—Ç—å —Ç—Ä–∞—Ñ–∏–∫, —á—Ç–æ–±—ã –æ—Ç–¥–µ–ª–∏—Ç—å –æ—Ä–≥–∞–Ω–∏—á–µ—Å–∫–∏–π —Ä–æ—Å—Ç –æ—Ç –ø–ª–∞—Ç–Ω–æ–≥–æ.

```python
query = """
SELECT 
  DATE_TRUNC('month', first_contact_date) AS month,
  COUNT(*) FILTER (WHERE origin = '—Ä–µ—Ñ–µ—Ä–∞–ª—å–Ω—ã–π —Ç—Ä–∞—Ñ–∏–∫' OR origin = '—Å–æ—Ü–∏–∞–ª—å–Ω—ã–µ —Å–µ—Ç–∏') AS referral_clients,
  COUNT(*) AS total_clients,
  ROUND(
    COUNT(*) FILTER (WHERE origin = '—Ä–µ—Ñ–µ—Ä–∞–ª—å–Ω—ã–π —Ç—Ä–∞—Ñ–∏–∫' OR origin = '—Å–æ—Ü–∏–∞–ª—å–Ω—ã–µ —Å–µ—Ç–∏') * 100.0 / COUNT(*),
    1
  ) AS referral_ratio
FROM marketing_qualified
WHERE first_contact_date > '2017-07-01'
GROUP BY 1
ORDER BY 1;
"""
df = run_query(engine, query, fetch='df')

df['other_clients'] = df['total_clients'] - df['referral_clients']

round(df['referral_ratio'].mean())

fig = go.Figure()
fig.add_trace(go.Bar(
    x=df['month'], y=df['referral_clients'],
    name='–†–µ—Ñ–µ—Ä–∞–ª—å–Ω—ã–µ –∫–ª–∏–µ–Ω—Ç—ã',
    marker_color=custom_palette[0],
    hovertemplate='%{x|%b %Y}<br>–†–µ—Ñ–µ—Ä–∞–ª—å–Ω—ã–µ: %{y}<br>–î–æ–ª—è: %{customdata}%',
    customdata=df['referral_ratio']
))
fig.add_trace(go.Bar(
    x=df['month'], y=df['other_clients'],
    name='–û—Å—Ç–∞–ª—å–Ω—ã–µ –∫–ª–∏–µ–Ω—Ç—ã',
    marker_color='rgba(200,200,200,0.5)',
    hoverinfo='skip'
))

fig.update_layout(
    barmode='stack',
    title=f"–î–ò–ù–ê–ú–ò–ö–ê –ü–†–ò–í–õ–ï–ß–ï–ù–ò–Ø –†–ï–§–ï–†–ï–ê–õ–¨–ù–û–ì–û –¢–†–ê–§–ò–ö–ê",
    xaxis_title='–ú–µ—Å—è—Ü',
    yaxis_title='–ö–æ–ª–∏—á–µ—Å—Ç–≤–æ –∫–ª–∏–µ–Ω—Ç–æ–≤',
    hovermode='x unified',
    legend=dict(orientation='h', y=1.1, x=0)
)

fig.write_image('1.png')
```

###  –í–ª–∏—è–Ω–∏–µ –æ—Ç–∑—ã–≤–æ–≤ –Ω–∞ –ø–æ–≤—Ç–æ—Ä–Ω—ã–µ –ø–æ–∫—É–ø–∫–∏.

- **–û–ø–∏—Å–∞–Ω–∏–µ –º–µ—Ç—Ä–∏–∫–∏:** –î–æ–ª—è –ø–æ–∫—É–ø–∞—Ç–µ–ª–µ–π, —Å–æ–≤–µ—Ä—à–∏–≤—à–∏—Ö –ø–æ–≤—Ç–æ—Ä–Ω—É—é –ø–æ–∫—É–ø–∫—É –≤ —Ç–µ—á–µ–Ω–∏–µ 60 –¥–Ω–µ–π –ø–æ—Å–ª–µ –∑–∞–∫–∞–∑–∞, –≤ –∑–∞–≤–∏—Å–∏–º–æ—Å—Ç–∏ –æ—Ç –æ—Ü–µ–Ω–∫–∏, –æ—Å—Ç–∞–≤–ª–µ–Ω–Ω–æ–π –∏–º–∏ —Ä–∞–Ω–µ–µ.
- **–ë–∏–∑–Ω–µ—Å-–∏–Ω—Ç–µ—Ä–ø—Ä–µ—Ç–∞—Ü–∏—è:**
    –ü–æ–∑–≤–æ–ª—è–µ—Ç –æ–ø—Ä–µ–¥–µ–ª–∏—Ç—å, –Ω–∞—Å–∫–æ–ª—å–∫–æ –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª—å—Å–∫–∏–π –æ–ø—ã—Ç –≤–ª–∏—è–µ—Ç –Ω–∞ –ª–æ—è–ª—å–Ω–æ—Å—Ç—å. –í—ã—Å–æ–∫–∏–µ –æ—Ü–µ–Ω–∫–∏ (4‚Äì5) –¥–æ–ª–∂–Ω—ã –∫–æ—Ä—Ä–µ–ª–∏—Ä–æ–≤–∞—Ç—å —Å –ø–æ–≤—ã—à–µ–Ω–Ω–æ–π –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç—å—é –ø–æ–≤—Ç–æ—Ä–Ω–æ–π –ø–æ–∫—É–ø–∫–∏. –û—Ç—Ä–∏—Ü–∞—Ç–µ–ª—å–Ω—ã–µ –æ—Ü–µ–Ω–∫–∏ ‚Äî —Å –æ—Ç—Ç–æ–∫–æ–º.

- **–î–æ–ø–æ–ª–Ω–∏—Ç–µ–ª—å–Ω—ã–µ –¥–µ—Ç–∞–ª–∏:**
    - –ú–µ—Ç—Ä–∏–∫–∞ –æ–≥—Ä–∞–Ω–∏—á–µ–Ω–∞ –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª—è–º–∏, –æ—Å—Ç–∞–≤–∏–≤—à–∏–º–∏ –æ—Ç–∑—ã–≤.
    - –£—á–∏—Ç—ã–≤–∞—é—Ç—Å—è —Ç–æ–ª—å–∫–æ –∑–∞–≤–µ—Ä—à—ë–Ω–Ω—ã–µ –∑–∞–∫–∞–∑—ã.
    - –ù–∏–∑–∫–∞—è –ø–æ–≤—Ç–æ—Ä–Ω–∞—è –∞–∫—Ç–∏–≤–Ω–æ—Å—Ç—å –ø–æ—Å–ª–µ –Ω–µ–≥–∞—Ç–∏–≤–Ω—ã—Ö –æ—Ü–µ–Ω–æ–∫ –º–æ–∂–µ—Ç —É–∫–∞–∑—ã–≤–∞—Ç—å –Ω–∞ –∫—Ä–∏—Ç–∏—á–µ—Å–∫–∏–µ –¥–µ—Ñ–µ–∫—Ç—ã –≤ –ª–æ–≥–∏—Å—Ç–∏–∫–µ, —Å–∞–ø–ø–æ—Ä—Ç–µ –∏–ª–∏ –ø—Ä–æ–¥—É–∫—Ç–µ.

```python
query = """
WITH user_activity AS (
  SELECT 
    c.customer_unique_id,
    o.order_purchase_timestamp,
    r.review_score,
    LEAD(o.order_purchase_timestamp) OVER(
        PARTITION BY c.customer_unique_id 
        ORDER BY o.order_purchase_timestamp
    ) AS next_order_date
  FROM orders o
  JOIN customers c USING(customer_id)
  LEFT JOIN order_reviews r USING(order_id)
  WHERE o.order_status = '–¥–æ—Å—Ç–∞–≤–ª–µ–Ω'
    AND r.review_score IS NOT NULL
),
repeat_stats AS (
  SELECT 
    review_score,
    COUNT(DISTINCT customer_unique_id) AS total_customers,
    COUNT(DISTINCT CASE 
                    WHEN next_order_date 
                      BETWEEN order_purchase_timestamp 
                      AND order_purchase_timestamp + INTERVAL '60 days' 
                    THEN customer_unique_id 
                  END) AS repeat_customers
  FROM user_activity
  GROUP BY 1
)
SELECT 
  CAST(review_score AS text),
  ROUND(repeat_customers::numeric * 100.0 / NULLIF(total_customers, 0), 2) AS repeat_rate,
  total_customers
FROM repeat_stats
ORDER BY review_score;
"""
df = run_query(engine, query, fetch='df')

fig = px.bar(df,
       x='review_score',
       text='repeat_rate',
       #color='repeat_rate',
       color_discrete_sequence=custom_palette[4:5],
       opacity=0.8,
       labels={
            'review_score': '–û—Ü–µ–Ω–∫–∞ –æ—Ç–∑—ã–≤–∞',
            'repeat_rate': '–î–æ–ª—è –ø–æ–≤—Ç–æ—Ä–Ω—ã—Ö –ø–æ–∫—É–ø–æ–∫ (%)'
       },
       title='<b>–ó–∞–≤–∏—Å–∏–º–æ—Å—Ç—å –ø–æ–≤—Ç–æ—Ä–Ω—ã—Ö –ø–æ–∫—É–ø–æ–∫ –æ—Ç –æ—Ü–µ–Ω–æ–∫</b>',
       y='repeat_rate'
      )
fig.update_traces(texttemplate='%{text}%', textposition='outside')

fig.update_layout(
    yaxis_tickformat=',.1f',
    coloraxis_showscale=True,
    bargap=0.4,
)
fig.write_image('1.png')
```

- –î–∞–∂–µ –Ω–µ–±–æ–ª—å—à–æ–π —Ä–æ—Å—Ç –æ—Ü–µ–Ω–∫–∏ (—Å 4 –¥–æ 5) –¥–∞—ë—Ç –ø—Ä–∏—Ä–æ—Å—Ç –ø–æ–≤—Ç–æ—Ä—è–µ–º–æ—Å—Ç–∏.
- –°—Ç–æ–∏—Ç –ø—Ä–æ–≤–µ—Ä–∏—Ç—å —Å—Ç–∞—Ç–∏—Å—Ç–∏—á—É—Å–∫—É—é –∑–Ω–∞—á–∏–º–æ—Å—Ç—å —ç—Ç–∏—Ö —Ä–∞–∑–ª–∏—á–∏–π.
- –ù–µ—Å–º–æ—Ç—Ä—è –Ω–∞ –Ω–∏–∑–∫–∏–µ –∞–±—Å–æ–ª—é—Ç–Ω—ã–µ –∑–Ω–∞—á–µ–Ω–∏—è (~1.5‚Äì1.8%), —Ç—Ä–µ–Ω–¥ —Å—Ç–∞–±–∏–ª–µ–Ω.

- –≠—Ç–æ –∫–æ—Å–≤–µ–Ω–Ω–æ —É–∫–∞–∑—ã–≤–∞–µ—Ç: –æ—Ç–∑—ã–≤—ã –≤–ª–∏—è—é—Ç –Ω–∞ –≤–æ–∑–≤—Ä–∞—Ç, –Ω–æ, –≤–æ–∑–º–æ–∂–Ω–æ, –µ—Å—Ç—å –∏ –¥—Ä—É–≥–∏–µ –±–∞—Ä—å–µ—Ä—ã (–∞—Å—Å–æ—Ä—Ç–∏–º–µ–Ω—Ç, –ª–æ–≥–∏—Å—Ç–∏–∫–∞, –æ—Ñ–ª–∞–π–Ω –∫–æ–Ω–∫—É—Ä–µ–Ω—Ü–∏—è).

---

### üìå –í—ã–≤–æ–¥—ã –ø–æ —ç—Ç–∞–ø—É.

-  **–û–±—â–∏–π —É—Ä–æ–≤–µ–Ω—å NPS-–ø—Ä–æ–∫—Å–∏**:

###  GMV (Gross Merchandise Value).

- **–û–ø–∏—Å–∞–Ω–∏–µ –º–µ—Ç—Ä–∏–∫–∏:** –û–±—â–∞—è —Å—É–º–º–∞ –ø—Ä–æ–¥–∞–∂ –Ω–∞ –ø–ª–∞—Ç—Ñ–æ—Ä–º–µ –∑–∞ –ø–µ—Ä–∏–æ–¥ (—Å—Ç–æ–∏–º–æ—Å—Ç—å –≤—Å–µ—Ö –¥–æ—Å—Ç–∞–≤–ª–µ–Ω–Ω—ã—Ö –∑–∞–∫–∞–∑–æ–≤, –±–µ–∑ —É—á—ë—Ç–∞ –≤–æ–∑–≤—Ä–∞—Ç–æ–≤ –∏ –∏–∑–¥–µ—Ä–∂–µ–∫).
- **–ë–∏–∑–Ω–µ—Å-–∏–Ω—Ç–µ—Ä–ø—Ä–µ—Ç–∞—Ü–∏—è:**
    - –†–æ—Å—Ç GMV = —Ä–æ—Å—Ç –¥–æ—Ö–æ–¥–∞ –≤—Å–µ–π —ç–∫–æ—Å–∏—Å—Ç–µ–º—ã –ø–ª–∞—Ç—Ñ–æ—Ä–º—ã

    - –°–Ω–∏–∂–µ–Ω–∏–µ –º–æ–∂–µ—Ç —Å–∏–≥–Ω–∞–ª–∏–∑–∏—Ä–æ–≤–∞—Ç—å –æ –ø–∞–¥–µ–Ω–∏–∏ —Å–ø—Ä–æ—Å–∞ –∏–ª–∏ –ø—Ä–æ–±–ª–µ–º–∞—Ö –≤ –ª–æ–≥–∏—Å—Ç–∏–∫–µ

    - –ò—Å–ø–æ–ª—å–∑—É–µ—Ç—Å—è –¥–ª—è –æ—Ü–µ–Ω–∫–∏ –º–∞—Å—à—Ç–∞–±–∞ –±–∏–∑–Ω–µ—Å–∞ –∏ —Ä–∞—Å—á—ë—Ç–∞ —é–Ω–∏—Ç-—ç–∫–æ–Ω–æ–º–∏–∫–∏

- **–î–æ–ø–æ–ª–Ω–∏—Ç–µ–ª—å–Ω—ã–µ –¥–µ—Ç–∞–ª–∏:**
    - –†–∞—Å—Å—á–∏—Ç—ã–≤–∞–µ—Ç—Å—è –ø–æ —Å—Ç–æ–∏–º–æ—Å—Ç–∏ –¥–æ—Å—Ç–∞–≤–ª–µ–Ω–Ω—ã—Ö –∑–∞–∫–∞–∑–æ–≤

    - –í–æ–∑–º–æ–∂–Ω–∞ —Ä–∞–∑–±–∏–≤–∫–∞ –ø–æ –∫–∞—Ç–µ–≥–æ—Ä–∏—è–º, –ø—Ä–æ–¥–∞–≤—Ü–∞–º, —Ä–µ–≥–∏–æ–Ω–∞–º

    - –Ø–≤–ª—è–µ—Ç—Å—è –æ—Å–Ω–æ–≤–æ–π –¥–ª—è —Ä–∞—Å—á—ë—Ç–∞ Take Rate –∏ –æ–±—â–µ–π –≤—ã—Ä—É—á–∫–∏ –ø–ª–∞—Ç—Ñ–æ—Ä–º—ã


```python
query = """
WITH gmv_data AS (
  SELECT
    DATE_TRUNC('month', o.order_purchase_timestamp) AS month,
    SUM(p.payment_value * 14)::numeric AS gmv
  FROM orders o
  JOIN order_payments p USING(order_id)
  WHERE o.order_status = '–¥–æ—Å—Ç–∞–≤–ª–µ–Ω' AND order_purchase_timestamp BETWEEN '2017-01-01' AND '2018-08-01'
  GROUP BY 1
),
gmv_with_growth AS (
  SELECT
    month,
    gmv,
    LAG(gmv) OVER (ORDER BY month) AS prev_gmv
  FROM gmv_data
)
SELECT 
  month,
  gmv,
  COALESCE(ROUND(
    CASE WHEN prev_gmv > 0 THEN (gmv - prev_gmv) * 100.0 / prev_gmv ELSE NULL END, 
    2
  ), 0) AS gmv_mom_growth
FROM gmv_with_growth
ORDER BY month;

"""

# –ó–∞–≥—Ä—É–∑–∫–∞ –¥–∞–Ω–Ω—ã—Ö
df_growth = run_query(engine, query, fetch='df')


# –ò—Å–ø–æ–ª—å–∑—É–µ–º –∫–∞—Ç–æ–º–Ω—ã–π —Å–∫—Ä–∏–ø—Ç –¥–ª—è –≤–∏–∑—É–∞–ª–∏–∑–∞—Ü–∏–∏
plot_gmv_dynamics(df_growth, custom_palette=custom_palette)
fig.write_image('1.png')
```

- –ó–∞ –ø–æ–ª—Ç–æ—Ä–∞ –≥–æ–¥–∞ —Å —è–Ω–≤–∞—Ä—è 2017 –ø–æ –∏—é–Ω—å 2018 –µ–∂–µ–º–µ—Å—è—á–Ω—ã–π –ø–æ–∫–∞–∑–∞—Ç–µ–ª—å. GMV –≤—ã—Ä–æ—Å —Å 1.5 –¥–æ 14 –º–∏–ª–ª–∏–æ–Ω–æ–≤ —Ä—É–±–ª–µ–π.

- –ü–∏–∫–æ–≤—ã–µ –ø–æ–∫–∞–∑–∞—Ç–µ–ª–∏ GMV¬† –Ω–æ—è–±—Ä—å 2017 –≥–æ–¥–∞, —Ñ–µ–≤—Ä–∞–ª—å 2018 - –≤–µ—Ä–æ—è—Ç–Ω–æ –∞–∫—Ü–∏–∏, –º–∞—Ä–∫–µ—Ç–∏–Ω–≥–æ–≤—ã–µ –∫–∞–º–ø–∞–Ω–∏–∏ (—á–µ—Ä–Ω–∞—è –ø—è—Ç–Ω–∏—Ü–∞, –Ω–æ–≤–æ–≥–æ–¥–Ω–∏–µ –ø—Ä–∞–∑–¥–Ω–∏–∫–∏)

- –°–ø–∞–¥—ã –Ω–∞ –≥—Ä–∞—Ñ–∏–∫–æ–≤ –≤–µ—Ä–æ—è—Ç–Ω–æ –æ–±—É—Å–ª–æ–≤–ª–µ–Ω—ã –æ–∫–æ–Ω—á–∞–Ω–∏–µ –∞–∫—Ü–∏–π –∏ —Ä–µ–∫–ª–∞–º–Ω—ã—Ö –∫–æ–º–ø–∞–Ω–∏–π –∞ –Ω–µ¬† —Ä–µ–∞–ª—å–Ω—ã–º —Å–Ω–∏–∂–µ–Ω–∏–µ–º –∏–Ω—Ç–µ—Ä–µ—Å–∞ –∫ –ø–ª–∞—Ç—Ñ–æ—Ä–º–µ.

###  AOV & ARPPU

- **–û–ø–∏—Å–∞–Ω–∏–µ –º–µ—Ç—Ä–∏–∫–∏:** –í —ç—Ç–æ–º —Ä–∞–∑–¥–µ–ª–µ –æ–±—ä–µ–¥–∏–Ω–µ–Ω—ã –¥–≤–µ –∫–ª—é—á–µ–≤—ã–µ –º–µ—Ç—Ä–∏–∫–∏ –º–æ–Ω–µ—Ç–∏–∑–∞—Ü–∏–∏:
    - **ARPPU (Average Revenue Per Paying User)** ‚Äî –æ—Ç—Ä–∞–∂–∞–µ—Ç, —Å–∫–æ–ª—å–∫–æ GMV –ø—Ä–∏—Ö–æ–¥–∏—Ç—Å—è –Ω–∞ –æ–¥–Ω–æ–≥–æ —É–Ω–∏–∫–∞–ª—å–Ω–æ–≥–æ –ø–æ–∫—É–ø–∞—Ç–µ–ª—è –≤ –º–µ—Å—è—Ü. –†–æ—Å—Ç ARPPU –≥–æ–≤–æ—Ä–∏—Ç –æ–± —É–≤–µ–ª–∏—á–µ–Ω–∏–∏ –≤—ã—Ä—É—á–∫–∏ –Ω–∞ –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª—è (–Ω–∞–ø—Ä–∏–º–µ—Ä, –∑–∞ —Å—á—ë—Ç –∞–ø—Å–µ–ª–ª–æ–≤ –∏–ª–∏ —Å–Ω–∏–∂–µ–Ω–∏—è –Ω–µ–∞–∫—Ç–∏–≤–Ω—ã—Ö –∫–ª–∏–µ–Ω—Ç–æ–≤).
    - **AOV (Average Order Value)**  –ø–æ–∫–∞–∑—ã–≤–∞–µ—Ç —Å—Ä–µ–¥–Ω–∏–π —á–µ–∫ –æ–¥–Ω–æ–≥–æ –∑–∞–∫–∞–∑–∞. –†–æ—Å—Ç AOV –º–æ–∂–µ—Ç –æ–∑–Ω–∞—á–∞—Ç—å –∞–ø—Å–µ–ª–ª, —Å–º–µ–Ω—É —Ç–æ–≤–∞—Ä–Ω–æ–≥–æ –º–∏–∫—Å–∞ –∏–ª–∏ —Å–Ω–∏–∂–µ–Ω–∏–µ –ø—Ä–æ–º–æ-–∞–∫—Ç–∏–≤–Ω–æ—Å—Ç–µ–π.
    
- **–ë–∏–∑–Ω–µ—Å-–∏–Ω—Ç–µ—Ä–ø—Ä–µ—Ç–∞—Ü–∏—è:** –°–æ–≤–º–µ—Å—Ç–Ω—ã–π –∞–Ω–∞–ª–∏–∑ —ç—Ç–∏—Ö –º–µ—Ç—Ä–∏–∫ –ø–æ–∑–≤–æ–ª—è–µ—Ç:
    - –û—Ü–µ–Ω–∏—Ç—å –ø–æ—Ç—Ä–µ–±–∏—Ç–µ–ª—å—Å–∫—É—é –∞–∫—Ç–∏–≤–Ω–æ—Å—Ç—å –∏ –ø–æ–≤–µ–¥–µ–Ω–∏–µ;

    - –û—Ç—Å–ª–µ–∂–∏–≤–∞—Ç—å —ç—Ñ—Ñ–µ–∫—Ç –æ—Ç –∞–∫—Ü–∏–π, —Å–∫–∏–¥–æ–∫ –∏–ª–∏ –∏–∑–º–µ–Ω–µ–Ω–∏–π —Ü–µ–Ω–æ–≤–æ–π –ø–æ–ª–∏—Ç–∏–∫–∏;

    - –í—ã—è–≤–∏—Ç—å –∏–∑–º–µ–Ω–µ–Ω–∏—è –≤ —Å—Ç—Ä—É–∫—Ç—É—Ä–µ —Å–ø—Ä–æ—Å–∞.
    
  *–ò–∑–º–µ–Ω–µ–Ω–∏—è ARPU –∏ AOV –≤ –ø—Ä–æ—Ç–∏–≤–æ–ø–æ–ª–æ–∂–Ω—ã–µ —Å—Ç–æ—Ä–æ–Ω—ã —á–∞—Å—Ç–æ —Å–∏–≥–Ω–∞–ª–∏–∑–∏—Ä—É—é—Ç –æ —Å–¥–≤–∏–≥–∞—Ö –≤ –ø—Ä–∏–≤—ã—á–∫–∞—Ö –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª–µ–π, —á—Ç–æ –º–æ–∂–µ—Ç —Ç—Ä–µ–±–æ–≤–∞—Ç—å –≤–Ω–∏–º–∞–Ω–∏—è —Å–æ —Å—Ç–æ—Ä–æ–Ω—ã –º–∞—Ä–∫–µ—Ç–∏–Ω–≥–∞ –∏–ª–∏ –ø—Ä–æ–¥—É–∫—Ç–æ–≤–æ–π –∫–æ–º–∞–Ω–¥—ã.*


- **–î–æ–ø–æ–ª–Ω–∏—Ç–µ–ª—å–Ω—ã–µ –¥–µ—Ç–∞–ª–∏:**
    - –í —Ä–∞—Å—á—ë—Ç –±–µ—Ä—É—Ç—Å—è —Ç–æ–ª—å–∫–æ –¥–æ—Å—Ç–∞–≤–ª–µ–Ω–Ω—ã–µ –∑–∞–∫–∞–∑—ã

```python
query = """
WITH monthly_metrics AS (
  SELECT
    DATE_TRUNC('month', o.order_purchase_timestamp) AS month,
    COUNT(DISTINCT c.customer_unique_id) AS unique_customers,
    COUNT(DISTINCT o.order_id) AS total_orders,
    SUM(op.payment_value * 14)::numeric AS gmv
  FROM orders o
  JOIN customers c ON c.customer_id = o.customer_id
  JOIN order_payments op ON o.order_id = op.order_id
  WHERE o.order_status = '–¥–æ—Å—Ç–∞–≤–ª–µ–Ω'
    AND order_purchase_timestamp BETWEEN '2017-01-01' AND '2018-08-01'
  GROUP BY 1
)

SELECT
  month,
  unique_customers,
  total_orders,
  gmv,
  ROUND(gmv / NULLIF(unique_customers, 0), 2) AS arpu,
  ROUND(gmv / NULLIF(total_orders, 0), 2) AS aov
FROM monthly_metrics
ORDER BY month;
"""

df_arpu = run_query(engine, query, fetch='df')

# –†–∞—Å—á—ë—Ç –ø—Ä–∏—Ä–æ—Å—Ç–æ–≤
df_arpu['arpu_mom_growth'] = (df_arpu['arpu'].pct_change().fillna(0) * 100).astype('float').round(2)
df_arpu['aov_mom_growth'] = (df_arpu['aov'].pct_change().fillna(0) * 100).astype('float').round(2)

# –¶–≤–µ—Ç–∞ –ø—Ä–∏—Ä–æ—Å—Ç–∞
df_arpu['color_arpu'] = df_arpu['arpu_mom_growth'].apply(lambda x: custom_palette[0] if x >= 0 else custom_palette[3])

# –°–∞–±–ø–ª–æ—Ç—ã
fig = make_subplots(
    rows=2, cols=1,
    shared_xaxes=True,
    row_heights=[0.5, 0.5],
    vertical_spacing=0.1,
    subplot_titles=[
        "–î–∏–Ω–∞–º–∏–∫–∞ ARPPU –∏ AOV –ø–æ –º–µ—Å—è—Ü–∞–º",
        "–ü—Ä–∏—Ä–æ—Å—Ç ARPPU (MoM, %)"
    ]
)

# –ª–∏–Ω–∏–∏ ARPU –∏ AOV
fig.add_trace(
    go.Scatter(
        x=df_arpu['month'],
        y=df_arpu['arpu'],
        name='ARPU',
        mode='lines+markers',
        line=dict(color=custom_palette[0], width=3)
    ),
    row=1, col=1
)

fig.add_trace(
    go.Scatter(
        x=df_arpu['month'],
        y=df_arpu['aov'],
        name='AOV',
        mode='lines+markers',
        line=dict(color=custom_palette[2], width=3)
    ),
    row=1, col=1
)

# –ø—Ä–∏—Ä–æ—Å—Ç—ã ARPU –∏ AOV
fig.add_trace(
    go.Bar(
        x=df_arpu['month'],
        y=df_arpu['arpu_mom_growth'],
        name='ARPPU MoM',
        text=df_arpu['arpu_mom_growth'],
        textposition='outside',
        marker_color=df_arpu['color_arpu'],
        showlegend=False
    ),
    row=2, col=1
)

# –ù–∞—Å—Ç—Ä–æ–π–∫–∏
fig.update_layout(
    height=700,
    title_text='<b>–î–∏–Ω–∞–º–∏–∫–∞ ARPPU –∏ AOV</b>',
    hovermode='x unified',
    legend=dict(orientation='h', yanchor='bottom', y=0.92, xanchor='right', x=1)
)

fig.update_yaxes(range=[1500, 3000], title_text='ARPPU / AOV (—Ä—É–±)', row=1, col=1)
fig.update_yaxes(range=[-15, 12], title_text='–ü—Ä–∏—Ä–æ—Å—Ç (%)', row=2, col=1, )

fig.write_image('1.png')
```

###  LTV (Lifetime Value).

- **–û–ø–∏—Å–∞–Ω–∏–µ –º–µ—Ç—Ä–∏–∫–∏:** –°–æ–≤–æ–∫—É–ø–Ω—ã–π –¥–æ—Ö–æ–¥ –æ—Ç –æ–¥–Ω–æ–≥–æ –∫–ª–∏–µ–Ω—Ç–∞–∑–∞ –≤–µ—Å—å –∂–∏–∑–Ω–µ–Ω–Ω—ã–π —Ü–∏–∫–ª: —Å—É–º–º–∞ GMV, –ø—Ä–∏–Ω–µ—Å—ë–Ω–Ω–∞—è –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª–µ–º —Å –º–æ–º–µ–Ω—Ç–∞ –ø–µ—Ä–≤–æ–π –ø–æ–∫—É–ø–∫–∏/–ø—Ä–æ–¥–∞–∂–∏.
- **–ë–∏–∑–Ω–µ—Å-–∏–Ω—Ç–µ—Ä–ø—Ä–µ—Ç–∞—Ü–∏—è:**
    - –í—ã—Å–æ–∫–∏–π LTV –≥–æ–≤–æ—Ä–∏—Ç –æ –ª–æ—è–ª—å–Ω–æ–π –∏ –ø—Ä–∏–±—ã–ª—å–Ω–æ–π –∞—É–¥–∏—Ç–æ—Ä–∏–∏

- **–î–æ–ø–æ–ª–Ω–∏—Ç–µ–ª—å–Ω—ã–µ –¥–µ—Ç–∞–ª–∏:**

    - –û—Å–Ω–æ–≤–∞–Ω –Ω–∞ –∫–æ–≥–æ—Ä—Ç–∞—Ö –ø–æ –º–µ—Å—è—Ü—É –ø–µ—Ä–≤–æ–π —É—Å–ø–µ—à–Ω–æ–π –ø–æ–∫—É–ø–∫–∏

    - –ú–æ–∂–µ—Ç —Ä–∞—Å—Å—á–∏—Ç—ã–≤–∞—Ç—å—Å—è –∫—É–º—É–ª—è—Ç–∏–≤–Ω–æ –ø–æ –º–µ—Å—è—Ü–∞–º –∂–∏–∑–Ω–∏



```python
query = """
WITH customer_first_purchase AS (
  SELECT
    c.customer_unique_id,
    DATE_TRUNC('month', MIN(o.order_purchase_timestamp)) AS cohort_month
  FROM orders o
  JOIN customers c ON o.customer_id = c.customer_id
  WHERE o.order_status = '–¥–æ—Å—Ç–∞–≤–ª–µ–Ω' 
    AND o.order_purchase_timestamp BETWEEN '2017-01-01' AND '2018-08-01'
  GROUP BY c.customer_unique_id
),

customer_monthly_spend AS (
  SELECT
    cfp.customer_unique_id,
    cfp.cohort_month,
    DATE_TRUNC('month', o.order_purchase_timestamp) AS order_month,
    EXTRACT(YEAR FROM AGE(DATE_TRUNC('month', o.order_purchase_timestamp), cfp.cohort_month)) * 12 +
    EXTRACT(MONTH FROM AGE(DATE_TRUNC('month', o.order_purchase_timestamp), cfp.cohort_month)) AS lifetime_month,
    SUM(op.payment_value) AS spend
  FROM customer_first_purchase cfp
  JOIN customers c ON cfp.customer_unique_id = c.customer_unique_id
  JOIN orders o ON c.customer_id = o.customer_id
  JOIN order_payments op ON o.order_id = op.order_id
  WHERE o.order_status = '–¥–æ—Å—Ç–∞–≤–ª–µ–Ω' 
    AND o.order_purchase_timestamp BETWEEN '2017-01-01' AND '2018-08-01'
  GROUP BY cfp.customer_unique_id, cfp.cohort_month, order_month, lifetime_month
),

cohort_spend AS (
  SELECT
    cohort_month,
    lifetime_month,
    SUM(spend)::numeric AS total_spend,
    COUNT(DISTINCT customer_unique_id) AS cohort_size
  FROM customer_monthly_spend
  GROUP BY cohort_month, lifetime_month
)

SELECT
  cohort_month,
  lifetime_month,
  SUM(total_spend) OVER (PARTITION BY cohort_month ORDER BY lifetime_month) AS cumulative_gmv,
  cohort_size,
  ROUND(SUM(total_spend) OVER (PARTITION BY cohort_month ORDER BY lifetime_month) / NULLIF(cohort_size, 0), 2) AS cumulative_ltv
FROM cohort_spend
ORDER BY cohort_month, lifetime_month;
"""
sf = run_query(engine, query, fetch='df')

plot_cohort_ltv_analysis(sf, custom_palette=custom_palette)
fig.write_image('1.png')
```

- **–ë–∏–∑–Ω–µ—Å-–∏–Ω—Ç–µ—Ä–ø—Ä–µ—Ç–∞—Ü–∏—è:**

- **–î–æ–ø–æ–ª–Ω–∏—Ç–µ–ª—å–Ω—ã–µ –¥–µ—Ç–∞–ª–∏:**
    - –ê–Ω–∞–ª–∏–∑ –ø–æ –ø—Ä–æ–¥–∞–≤—Ü–∞–º, –∫–∞—Ç–µ–≥–æ—Ä–∏—è–º, —Ä–µ–≥–∏–æ–Ω–∞–º: –ö–æ–Ω—Ü–µ–Ω—Ç—Ä–∞—Ü–∏—è GMV –º–æ–∂–µ—Ç —Å–∏–ª—å–Ω–æ –≤–∞—Ä—å–∏—Ä–æ–≤–∞—Ç—å—Å—è –≤ –∑–∞–≤–∏—Å–∏–º–æ—Å—Ç–∏ –æ—Ç –ø—Ä–æ–¥–∞–≤—Ü–∞, –∫–∞—Ç–µ–≥–æ—Ä–∏–∏ —Ç–æ–≤–∞—Ä–∞ –∏–ª–∏ —Ä–µ–≥–∏–æ–Ω–∞, —á—Ç–æ –ø–æ–∑–≤–æ–ª—è–µ—Ç –≤—ã–¥–µ–ª–∏—Ç—å —Å—Ç—Ä–∞—Ç–µ–≥–∏—á–µ—Å–∫–∏ –≤–∞–∂–Ω—ã–µ —Å–µ–≥–º–µ–Ω—Ç—ã –¥–ª—è –±–∏–∑–Ω–µ—Å–∞.

    - –ò—Å–ø–æ–ª—å–∑–æ–≤–∞–Ω–∏–µ –∏–Ω–¥–µ–∫—Å–∞ –î–∂–∏–Ω–∏ –¥–ª—è –æ—Ü–µ–Ω–∫–∏ —Å—Ç–µ–ø–µ–Ω–∏ –∫–æ–Ω—Ü–µ–Ω—Ç—Ä–∞—Ü–∏–∏. –ò–Ω–¥–µ–∫—Å –î–∂–∏–Ω–∏ –∏—Å–ø–æ–ª—å–∑—É–µ—Ç—Å—è –¥–ª—è –∏–∑–º–µ—Ä–µ–Ω–∏—è –Ω–µ—Ä–∞–≤–Ω–æ–º–µ—Ä–Ω–æ—Å—Ç–∏ —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—è, –≥–¥–µ –∑–Ω–∞—á–µ–Ω–∏—è –±–ª–∏–∑–∫–∏–µ –∫ 1 –≥–æ–≤–æ—Ä—è—Ç –æ –≤—ã—Å–æ–∫–æ–º —É—Ä–æ–≤–Ω–µ –∫–æ–Ω—Ü–µ–Ω—Ç—Ä–∞—Ü–∏–∏, –∞ –∑–Ω–∞—á–µ–Ω–∏—è –±–ª–∏–∑–∫–∏–µ –∫ 0 ‚Äî –æ–± –∏–¥–µ–∞–ª—å–Ω–æ–π –¥–∏–≤–µ—Ä—Å–∏—Ñ–∏–∫–∞—Ü–∏–∏.

```python
# –û–ø—Ç–∏–º–∏–∑–∏—Ä–æ–≤–∞–Ω–Ω—ã–π –∑–∞–ø—Ä–æ—Å —Å —Ä–∞—Å—á–µ—Ç–æ–º –∏–Ω–¥–µ–∫—Å–∞ –î–∂–∏–Ω–∏
query = """
WITH seller_gmv AS (
    SELECT
        seller_id,
        SUM(payment_value)::numeric AS gmv
    FROM order_items
    JOIN order_payments USING(order_id)
    GROUP BY seller_id
),
ranked_sellers AS (
    SELECT
        seller_id,
        gmv,
        ROW_NUMBER() OVER(ORDER BY gmv DESC) AS seller_rank,
        COUNT(*) OVER() AS total_sellers,
        SUM(gmv) OVER() AS total_gmv
    FROM seller_gmv
),
cumulative_data AS (
    SELECT
        seller_id,
        gmv,
        seller_rank,
        total_sellers,
        total_gmv,
        -- –†–∞—Å—á–µ—Ç –¥–æ–ª–∏ GMV –∏ –ø—Ä–æ–¥–∞–≤—Ü–æ–≤
        ROUND(100 * gmv / total_gmv, 2) AS percent_gmv,
        -- –ò–Ω–¥–µ–∫—Å –î–∂–∏–Ω–∏
        (ROW_NUMBER() OVER (ORDER BY gmv DESC) - 0.5) * gmv AS weighted_gmv
    FROM ranked_sellers
)
SELECT 
    seller_rank,
    seller_id,
    gmv::float,
    percent_gmv::float,
    SUM(percent_gmv) OVER(ORDER BY seller_rank)::float AS cumulative_gmv_percent,
    100.0 * seller_rank / total_sellers::float AS cumulative_seller_percent,
    ROUND(1.0 - (2.0 * SUM(weighted_gmv) OVER() / (total_sellers * total_gmv)), 4) AS gini_index
FROM cumulative_data
ORDER BY seller_rank;

"""

# –ó–∞–ø—É—Å–∫ –∑–∞–ø—Ä–æ—Å–∞
df_concentration = run_query(engine, query, fetch='df')


plot_gmv_concentration(df_concentration, custom_palette=custom_palette[0:1])

fig.write_image('1.png')
```

**üìä –ò–Ω—Ç–µ—Ä–ø—Ä–µ—Ç–∞—Ü–∏—è —Ä–µ–∑—É–ª—å—Ç–∞—Ç–∞:**
- –û–ø—Ä–µ–¥–µ–ª–∏–ª–∏ —á—Ç–æ –∫–æ–Ω—Ü–µ–Ω—Ç—Ä–∞—Ü–∏—è gmv —Å–æ—Å—Ç–∞–≤–ª—è–µ—Ç 0.79 –ø–æ –∏–Ω–¥–µ–∫—Å—É –î–∂–∏–Ω–∏, 20 % –ø—Ä–æ–¥–∞–≤—Ü–æ–≤ –¥–µ–ª–∞–µ—Ç 82 % GMV, —á—Ç–æ —Å–æ–æ—Ç–≤–µ—Ç—Å—Ç–≤—É–µ—Ç –ø—Ä–∞–≤–∏–ª—É –ü–∞—Ä–µ—Ç–æ.
- —Ä–∞–∑–¥–µ–ª–∏–ª–∏ –ø—Ä–æ–¥–∞–≤—Ü–æ–≤ –Ω–∞ –¥–≤–µ –≥—Ä—É–ø–ø—ã (—Ç–æ–ø -20 –∏ –æ—Å—Ç–∞–ª—å–Ω—ã–µ)
    
    
---

### –ö–æ–Ω—Ü–µ–Ω—Ç—Ä–∞—Ü–∏—è –≤ –¥–∏–Ω–∞–º–∏–∫–µ
- **–û–ø–∏—Å–∞–Ω–∏–µ –º–µ—Ç—Ä–∏–∫–∏** - –ø–æ–∫–∞–∑—ã–≤–∞–µ—Ç –∫–∞–∫ —Ä–∞—Å–ø—Ä–µ–¥–µ–ª—è—é—Ç—Å—è –ø–æ–∫–∞–∑–∞—Ç–µ–ª–∏ GMV —Å—Ä–µ–¥–∏ –ø—Ä–æ–¥–∞—Ü–æ–≤ –ø–æ –º–µ—Å—è—Ü–∞–º.
- **–ë–∏–∑–Ω–µ—Å –∏–Ω—Ç–µ—Ä–ø—Ä–µ—Ç–∞—Ü–∏—è** -  —É–≤–µ–ª–∏—á–µ–Ω–∏–µ –∫–æ–Ω—Ü–µ–Ω—Ç—Ä–∞—Ü–∏–∏ –ø–æ–≤—ã—à–∞–µ—Ç —Ä–∏—Å–∫–∏ –∏ —Å—Ç–∞–±–∏–ª—å–Ω–æ—Å—Ç—å –ø—Ä–æ–¥—É–∫—Ç–∞, –∏–∑ –∑–∞ –∑–∞–≤–∏—Å–∏–º–æ—Å—Ç–∏ –æ—Ç –¢–û–ü –ø—Ä–æ–¥–∞–≤—Ü–æ–≤.

```python
query = """
WITH 
seller_gmv AS (
    SELECT
        seller_id,
        SUM(payment_value) AS gmv
    FROM order_items
    JOIN order_payments USING(order_id)
    GROUP BY seller_id
),
ranked_sellers AS (
    SELECT
        seller_id,
        gmv,
        NTILE(5) OVER (ORDER BY gmv DESC) AS gmv_ntile
    FROM seller_gmv
),

seller_groups AS (
SELECT
    seller_id,
    CASE WHEN gmv_ntile = 1 THEN 'top_20' ELSE 'bottom_80' END AS seller_group
FROM ranked_sellers
),

orders_with_groups AS (
    SELECT
        o.order_id,
        o.order_approved_at,
        sg.seller_group,
        op.payment_value
    FROM orders o
    JOIN order_payments op ON o.order_id = op.order_id
    JOIN order_items oi ON o.order_id=oi.order_id
    JOIN seller_groups sg ON oi.seller_id = sg.seller_id
    WHERE order_purchase_timestamp BETWEEN '2017-01-01' AND '2018-08-01'
)
SELECT
    DATE_TRUNC('month', order_approved_at) AS month,
    seller_group,
    SUM(payment_value) AS monthly_gmv
FROM orders_with_groups
GROUP BY month , seller_group
ORDER BY month ASC, seller_group DESC;
"""

df_gmv = run_query(engine, query, fetch ='df')


# –í–∏–∑—É–∞–ª–∏–∑–∏—Ä—É–µ–º —Ä–µ–∑—É–ª—å—Ç–∞—Ç
fig = px.area(
    df_gmv,
    x='month',
    y='monthly_gmv',
    color='seller_group',
    color_discrete_sequence=custom_palette,
    title='<b>–î–∏–Ω–∞–º–∏–∫–∞ GMV: –¢–æ–ø-20% vs –û—Å—Ç–∞–ª—å–Ω—ã–µ –ø—Ä–æ–¥–∞–≤—Ü—ã</b>',
    labels={
        'month': '–ú–µ—Å—è—Ü',
        'monthly_gmv': 'GMV',
        'seller_group': ''
    },
    groupnorm='percent'
)
fig.update_layout(hovermode='x unified', height=600,
                 legend=dict(orientation='h', yanchor='bottom', y=0.96, xanchor='right', x=1))
fig.write_image('1.png')
```

**üìä –ò–Ω—Ç–µ—Ä–ø—Ä–µ—Ç–∞—Ü–∏—è —Ä–µ–∑—É–ª—å—Ç–∞—Ç–∞:**
- –î–æ–ª—è GMV –ø—Ä–∏—Ö–æ–¥—è—â–∞—è—Å—è –Ω–∞ –¢–û–ü-20% –ø—Ä–æ–¥–∞–≤—Ü–æ–≤ –ø–æ—Å—Ç–µ–ø–µ–Ω–Ω–æ —Å–Ω–∏–∂–∞–µ—Ç—Å—è, —á—Ç–æ —è–≤–ª—è–µ—Ç—Å—è —Ö–æ—Ä–æ—à–∏–º –∑–Ω–∞–∫–æ–º –¥–ª—è –ø—Ä–æ–¥—É–∫—Ç–∞.

###  GMV –ø–æ –∫–∞—Ç–µ–≥–æ—Ä–∏—è–º —Ç–æ–≤–∞—Ä–æ–≤.

- **–û–ø–∏—Å–∞–Ω–∏–µ –º–µ—Ç—Ä–∏–∫–∏:** –ü–æ–∫–∞–∑–∞—Ç–µ–ª–∏ GMV –≤ —Ä–∞–∑—Ä–µ–∑–µ –∫–∞—Ç–µ–≥–æ—Ä–∏–π, –∞ —Ç–∞–∫–∂–µ –≥—Ä—É–ø–ø –ø—Ä–æ–¥–∞–≤—Ü–æ–≤ **–¢–û–ü 20 %** –∏ –æ—Å—Ç–∞–ª—å–Ω—ã–µ.
- **–ë–∏–∑–Ω–µ—Å-–∏–Ω—Ç–µ—Ä–ø—Ä–µ—Ç–∞—Ü–∏—è:**
    - –°–∫–∞—á–∫–∏ GMV = —É—Å–ø–µ—Ö –∫–∞–º–ø–∞–Ω–∏–π, –ø—Ä–æ–º–æ, –≤–Ω–µ—à–Ω–∏–µ —ç—Ñ—Ñ–µ–∫—Ç—ã

    - –°–µ–∑–æ–Ω–Ω—ã–µ –ø—Ä–æ–≤–∞–ª—ã ‚Äî –∫—Ä–∏—Ç–∏—á–µ—Å–∫–∏–µ –∑–æ–Ω—ã —Ä–æ—Å—Ç–∞

    - –ü–æ–º–æ–≥–∞–µ—Ç –≤ –ø–ª–∞–Ω–∏—Ä–æ–≤–∞–Ω–∏–∏ –∏ –ø—Ä–æ–≥–Ω–æ–∑–∏—Ä–æ–≤–∞–Ω–∏–∏ –≤—ã—Ä—É—á–∫–∏

- **–î–æ–ø–æ–ª–Ω–∏—Ç–µ–ª—å–Ω—ã–µ –¥–µ—Ç–∞–ª–∏:**
    - –ò—Å–ø–æ–ª—å–∑—É—é—Ç—Å—è –≤—Ä–µ–º–µ–Ω–Ω—ã–µ —Ä—è–¥—ã GMV


```python
query = """
WITH 
 seller_gmv AS (
   SELECT seller_id, SUM(payment_value) AS gmv
   FROM order_items
   JOIN order_payments USING(order_id)
   GROUP BY seller_id
 ),
 ranked_s AS (
   SELECT seller_id, gmv,
          NTILE(5) OVER(ORDER BY gmv DESC) AS quintile
   FROM seller_gmv
 ),
 seller_groups AS (
   SELECT seller_id,
          CASE WHEN quintile = 1 THEN 'top_20' ELSE 'bottom_80' END AS seller_group
   FROM ranked_s
 ),
 orders_labeled AS (
   SELECT 
     DATE_TRUNC('month', o.order_approved_at)::date AS month,
     p.category_name_translated,
     op.payment_value,
     sg.seller_group
   FROM order_items oi
   JOIN orders o ON oi.order_id = o.order_id
   JOIN order_payments op ON oi.order_id = op.order_id
   JOIN products p ON oi.product_id = p.product_id
   JOIN seller_groups sg ON oi.seller_id = sg.seller_id
   WHERE o.order_status = '–¥–æ—Å—Ç–∞–≤–ª–µ–Ω'
 )
SELECT 
  month,
  category_name_translated,
  SUM(payment_value) AS category_gmv,
  seller_group
FROM orders_labeled
WHERE month BETWEEN '2017-01-01' AND '2018-07-01'
GROUP BY month, category_name_translated, seller_group
ORDER BY month, seller_group, category_gmv DESC;
"""
df = run_query(engine, query, fetch='df')

df_top20    = df[df.seller_group=='top_20'].copy()
df_bottom80 = df[df.seller_group=='bottom_80'].copy()

# –°–æ–±–∏—Ä–∞–µ–º —Å–∞–±–ø–ª–æ—Ç—ã
fig = make_subplots(
    rows=3, cols=2,
    specs=[
        [{"type":"xy","colspan":2}, None],
        [{"type":"xy","colspan":2}, None],
        [{"type":"bar"}, {"type":"bar"}],
    ],
    shared_xaxes=True,
    subplot_titles=(
        "–¢–û–ü‚Äë20% –ø—Ä–æ–¥–∞–≤—Ü–æ–≤: —Ä–∞—Å—Ç—É—â–∏–µ –∫–∞—Ç–µ–≥–æ—Ä–∏–∏ ",
        "–ù–∏–∂–Ω–∏–µ‚Äë80% –ø—Ä–æ–¥–∞–≤—Ü–æ–≤: —Ä–∞—Å—Ç—É—â–∏–µ –∫–∞—Ç–µ–≥–æ—Ä–∏–∏",
        "–¢–û–ü‚Äë3 –ö–∞—Ç–µ–≥–æ—Ä–∏–∏ (–í–µ—Ä—Ö–Ω–∏–µ‚Äë20%)",
        "–¢–û–ü‚Äë3 –ö–∞—Ç–µ–≥–æ—Ä–∏–∏ (–ù–∏–∂–Ω–∏–µ‚Äë80%)"
    ),
    vertical_spacing=0.12,
    horizontal_spacing=0.08
)

# –ü–µ—Ä–≤—ã–µ –¥–≤–µ —Å—Ç—Ä–æ–∫–∏ —Å–µ–∑–æ–Ω–Ω–æ—Å—Ç—å
add_seasonality_traces(fig, df_top20,    row=1, title="Top20")
add_seasonality_traces(fig, df_bottom80, row=2, title="Bottom80")

# –¢—Ä–µ—Ç—å—è —Å—Ç—Ä–æ–∫–∞ –¥–≤–∞ –±–∞—Ä–ø–ª–æ—Ç–∞
add_top5_bar(fig, df_top20,    row=3, col=1, palette=custom_palette[0])
add_top5_bar(fig, df_bottom80, row=3, col=2, palette=custom_palette[0])

# –û–∫–æ–Ω—á–∞—Ç–µ–ª—å–Ω–æ–µ –æ—Ñ–æ—Ä–º–ª–µ–Ω–∏–µ
fig.update_layout(
    height=1000,
    title_text="–î–∏–Ω–∞–º–∏–∫–∞ GMV –ø–æ –∫–∞—Ç–µ–≥–æ—Ä–∏—è–º",
    margin=dict(t=100, l=10, r=10, b=60),
    showlegend=False,
    bargap=0.4
)
fig.write_image('1.png')
```

- –ò–∑—É—á–∏–ª–∏ GMV –≤ –¥–∏–Ω–∞–º–∏–∫–µ –ø–æ –∫–∞—Ç–µ–≥–æ—Ä–∏—è–º —Ç–æ–≤–∞—Ä–æ–≤ —Å —Ä–∞–∑–±–∏–≤–∫–æ–π –ø–æ¬† –∫–∞—Ç–µ–≥–æ—Ä–∏—è–º –ø—Ä–æ–¥–∞–≤—Ü–æ–≤ (—Ç–æ–ø 20 –∏ –¥—Ä—É–≥–∏–µ)

- –°—Ä–µ–¥–∏ –æ—Å—Ç–∞–ª—å–Ω—ã—Ö –ø—Ä–æ–¥–∞–≤—Ü–æ–≤ —Ç–æ–ø- 3 –∫–∞—Ç–µ–≥–æ—Ä–∏–∏ –ø–æ GMV –∑–∞ –≤—Å–µ –≤—Ä–µ–º—è - **—Å–ø–æ—Ä—Ç**, **–∫—Ä–∞—Å–æ—Ç–∞ –∏ –∑–¥–æ—Ä–æ–≤—å–µ** –∏ **—Ç–æ–≤–∞—Ä—ã –¥–ª—è –¥–æ–º–∞**.

- –¢—Ä–µ–Ω–¥ –ø–æ—Å–ª–µ–¥–Ω–µ–≥–æ –∫–≤–∞—Ä—Ç–∞–ª–∞ - **—Ç–æ–≤–∞—Ä—ã –¥–ª—è –¥–æ–º–∞**, **–∞–≤—Ç–æ**, **–∫—Ä–∞—Å–æ—Ç–∞ –∏ –∑–¥–æ—Ä–æ–≤—å–µ**.

- –ö–∞—Ç–µ–≥–æ—Ä–∏–∏ –ø–µ—Ä–µ—Å–µ–∫–∞—é—Ç—Å—è.


### –ü–æ–∫–∞–∑–∞—Ç–µ–ª–∏ –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª–µ–π —Å –ø–æ–≤—Ç–æ—Ä–Ω—ã–º–∏ –ø–æ–∫—É–ø–∫–∞–º–∏ (–ª–æ—è–ª—å–Ω—ã—Ö)

**–û–ø–∏—Å–∞–Ω–∏–µ –≥—Ä—É–ø–ø—ã:**
- –°–µ–≥–º–µ–Ω—Ç –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª–µ–π, —Å–æ–≤–µ—Ä—à–∏–≤—à–∏—Ö –Ω–∞ –ø–ª–∞—Ç—Ñ–æ—Ä–º–µ –±–æ–ª–µ–µ –æ–¥–Ω–æ–π —É—Å–ø–µ—à–Ω–æ–π –ø–æ–∫—É–ø–∫–∏. –≠—Ç–∏ –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª–∏ –¥–µ–º–æ–Ω—Å—Ç—Ä–∏—Ä—É—é—Ç –æ–ø—Ä–µ–¥–µ–ª–µ–Ω–Ω—ã–π —É—Ä–æ–≤–µ–Ω—å –ª–æ—è–ª—å–Ω–æ—Å—Ç–∏ –∏ —è–≤–ª—è—é—Ç—Å—è –∫–ª—é—á–µ–≤–æ–π –∞—É–¥–∏—Ç–æ—Ä–∏–µ–π –¥–ª—è —É–¥–µ—Ä–∂–∞–Ω–∏—è –∏ —Å—Ç–∏–º—É–ª–∏—Ä–æ–≤–∞–Ω–∏—è –¥–∞–ª—å–Ω–µ–π—à–∏—Ö –ø–æ–∫—É–ø–æ–∫.

**–ö–ª—é—á–µ–≤—ã–µ –ø–æ–∫–∞–∑–∞—Ç–µ–ª–∏ –¥–ª—è –∞–Ω–∞–ª–∏–∑–∞:**


   **–°—Ä–µ–¥–Ω–µ–µ —á–∏—Å–ª–æ –∑–∞–∫–∞–∑–æ–≤ –Ω–∞ –ø–æ–≤—Ç–æ—Ä–Ω–æ–≥–æ –ø–æ–∫—É–ø–∞—Ç–µ–ª—è**: –û–±—â–µ–µ –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –∑–∞–∫–∞–∑–æ–≤, —Å–¥–µ–ª–∞–Ω–Ω—ã—Ö –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª—è–º–∏ –∏–∑ —ç—Ç–æ–≥–æ —Å–µ–≥–º–µ–Ω—Ç–∞, –¥–µ–ª–µ–Ω–Ω–æ–µ –Ω–∞ –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ —É–Ω–∏–∫–∞–ª—å–Ω—ã—Ö –ø–æ–≤—Ç–æ—Ä–Ω—ã—Ö –ø–æ–∫—É–ø–∞—Ç–µ–ª–µ–π.
   **LTV (Lifetime Value) –ø–æ–≤—Ç–æ—Ä–Ω—ã—Ö –ø–æ–∫—É–ø–∞—Ç–µ–ª–µ–π**: –û–±—â–∞—è –≤—ã—Ä—É—á–∫–∞, –∫–æ—Ç–æ—Ä—É—é –ø—Ä–∏–Ω–æ—Å–∏—Ç —Å—Ä–µ–¥–Ω–∏–π –ø–æ–≤—Ç–æ—Ä–Ω—ã–π –ø–æ–∫—É–ø–∞—Ç–µ–ª—å –∑–∞ –≤—Å–µ –≤—Ä–µ–º—è –≤–∑–∞–∏–º–æ–¥–µ–π—Å—Ç–≤–∏—è —Å –ø–ª–∞—Ç—Ñ–æ—Ä–º–æ–π.
   **–°—Ä–µ–¥–Ω–∏–π —á–µ–∫ –ø–æ–≤—Ç–æ—Ä–Ω–æ–≥–æ –ø–æ–∫—É–ø–∞—Ç–µ–ª—è (AOV)**: –°—Ä–µ–¥–Ω—è—è —Å—Ç–æ–∏–º–æ—Å—Ç—å –∑–∞–∫–∞–∑–∞, —Å–¥–µ–ª–∞–Ω–Ω–æ–≥–æ –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª—è–º–∏ –∏–∑ —ç—Ç–æ–≥–æ —Å–µ–≥–º–µ–Ω—Ç–∞.

**–ë–∏–∑–Ω–µ—Å-–∏–Ω—Ç–µ—Ä–ø—Ä–µ—Ç–∞—Ü–∏—è:**
    –ê–Ω–∞–ª–∏–∑ —ç—Ç–∏—Ö –ø–æ–∫–∞–∑–∞—Ç–µ–ª–µ–π –ø–æ–∑–≤–æ–ª—è–µ—Ç –æ—Ü–µ–Ω–∏—Ç—å —Ü–µ–Ω–Ω–æ—Å—Ç—å –ª–æ—è–ª—å–Ω—ã—Ö –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª–µ–π –¥–ª—è –±–∏–∑–Ω–µ—Å–∞, –ø–æ–Ω—è—Ç—å –∏—Ö –ø–æ–∫—É–ø–∞—Ç–µ–ª—å—Å–∫–æ–µ –ø–æ–≤–µ–¥–µ–Ω–∏–µ, –≤—ã—è–≤–∏—Ç—å –≤–æ–∑–º–æ–∂–Ω–æ—Å—Ç–∏ –¥–ª—è —É–≤–µ–ª–∏—á–µ–Ω–∏—è –∏—Ö –≤–æ–≤–ª–µ—á–µ–Ω–Ω–æ—Å—Ç–∏ –∏ —Å—Ç–∏–º—É–ª–∏—Ä–æ–≤–∞–Ω–∏—è –¥–∞–ª—å–Ω–µ–π—à–∏—Ö –ø–æ–∫—É–ø–æ–∫. –ë–æ–ª–µ–µ –≤—ã—Å–æ–∫–∏–µ –∑–Ω–∞—á–µ–Ω–∏—è —ç—Ç–∏—Ö –ø–æ–∫–∞–∑–∞—Ç–µ–ª–µ–π –æ–±—ã—á–Ω–æ –∫–æ—Ä—Ä–µ–ª–∏—Ä—É—é—Ç —Å —É—Å—Ç–æ–π—á–∏–≤—ã–º —Ä–æ—Å—Ç–æ–º –∏ –ø—Ä–∏–±—ã–ª—å–Ω–æ—Å—Ç—å—é –ø–ª–∞—Ç—Ñ–æ—Ä–º—ã.


```python
query = """
WITH clean_orders AS (
  SELECT DISTINCT 
    o.order_id, 
    c.customer_unique_id, 
    o.order_status, 
    SUM(op.payment_value) AS payment_value
  FROM orders o
  JOIN customers c USING(customer_id)
  JOIN order_payments op USING(order_id)
  WHERE o.order_status = '–¥–æ—Å—Ç–∞–≤–ª–µ–Ω'
  GROUP BY 1, 2 ,3
),
customer_cat AS (
  SELECT 
    customer_unique_id,
    COUNT(DISTINCT order_id) AS order_count,
    SUM(payment_value) * 14 AS gmv,
    CASE WHEN COUNT(order_id) > 1 THEN 'loyal' ELSE 'other' END AS customer_category
  FROM clean_orders
  GROUP BY customer_unique_id
)
SELECT 
  customer_category,
  COUNT(DISTINCT customer_unique_id)::float AS customers_count,
  SUM(order_count)::int AS order_count,
  ROUND(SUM(order_count) /  COUNT(DISTINCT customer_unique_id), 2) AS avg_orders_per_customer,
  SUM(gmv) as sum_gmv,
  ROUND(SUM(gmv)::numeric /COUNT(DISTINCT customer_unique_id), 2) AS avg_gmv_per_customer,
  ROUND(SUM(gmv)::numeric / SUM(order_count), 2) AS aov,
  ROUND(SUM(gmv)::numeric / SUM(SUM(gmv)) OVER()::numeric * 100, 2) AS pct_gmv
FROM customer_cat
GROUP BY customer_category;
"""

df = run_query(engine, query, fetch='df')



# –°–æ–∑–¥–∞–µ–º –º–∞—Ç—Ä–∏—Ü—É 2x2
fig = make_subplots(
    rows=2, cols=2,
    subplot_titles=(
        '–î–æ–ª—è GMV –ø–æ –∫–∞—Ç–µ–≥–æ—Ä–∏—è–º –∫–ª–∏–µ–Ω—Ç–æ–≤', 
        '–°—Ä–µ–¥–Ω–∏–π —á–µ–∫ (AOV)',
        '–°—Ä–µ–¥–Ω–∏–π LTV –Ω–∞ –∫–ª–∏–µ–Ω—Ç–∞', 
        '–ó–∞–∫–∞–∑–æ–≤ –Ω–∞ –∫–ª–∏–µ–Ω—Ç–∞'
    ),
    vertical_spacing=0.15
)

# –î–æ–±–∞–≤–ª—è–µ–º –≥—Ä–∞—Ñ–∏–∫–∏
categories = df['customer_category']
colors = ['#1f77b4', '#ff7f0e']  # –¶–≤–µ—Ç–∞ –¥–ª—è loyal/other

# –î–æ–ª—è GMV
fig.add_trace(
    go.Bar(
        x=categories,
        y=df['pct_gmv'],
        text=df['pct_gmv'].round(2).astype(str) + '%',
        marker_color=custom_palette,
        name='GMV %'
    ),
    row=1, col=1
)

# –°—Ä–µ–¥–Ω–∏–π —á–µ–∫
fig.add_trace(
    go.Bar(
        x=categories,
        y=df['aov'],
        text=df['aov'].round(2),
        marker_color=custom_palette,
        name='AOV'
    ),
    row=1, col=2
)

# –°—Ä–µ–¥–Ω–∏–π LTV –Ω–∞ –∫–ª–∏–µ–Ω—Ç–∞
fig.add_trace(
    go.Bar(
        x=categories,
        y=df['avg_gmv_per_customer'],
        text=df['avg_gmv_per_customer'].round(2),
        marker_color=custom_palette,
        name='GMV/–∫–ª–∏–µ–Ω—Ç'
    ),
    row=2, col=1
)


# –û–±—â–∏–µ –Ω–∞—Å—Ç—Ä–æ–π–∫–∏
fig.update_layout(
    height=600,
    showlegend=False,
    title_text="<b>–°—Ä–∞–≤–Ω–µ–Ω–∏–µ –ª–æ—è–ª—å–Ω—ã—Ö –∏ –æ—Å—Ç–∞–ª—å–Ω—ã—Ö –∫–ª–∏–µ–Ω—Ç–æ–≤</b>",
)

# –§–æ—Ä–º–∞—Ç–∏—Ä–æ–≤–∞–Ω–∏–µ –ø–æ–¥–ø–∏—Å–µ–π
fig.update_traces(
    textposition='inside',
    textfont_size=12
)

# –ù–∞—Å—Ç—Ä–æ–π–∫–∞ –æ—Å–µ–π
fig.update_yaxes(title_text="–î–æ–ª—è GMV (%)", row=1, col=1)
fig.update_yaxes(title_text="–°—Ä–µ–¥–Ω–∏–π —á–µ–∫ (—Ä—É–±.)", row=1, col=2)
fig.update_yaxes(title_text="GMV/–∫–ª–∏–µ–Ω—Ç (—Ä—É–±.)", row=2, col=1)
fig.update_yaxes(title_text="–ó–∞–∫–∞–∑–æ–≤", row=2, col=2)

fig.write_image('1.png')
```

###  Order Cancellation Rate –∏ Lost GMV

- **–û–ø–∏—Å–∞–Ω–∏–µ –º–µ—Ç—Ä–∏–∫–∏:**
    –ü—Ä–æ—Ü–µ–Ω—Ç –∑–∞–∫–∞–∑–æ–≤, –∫–æ—Ç–æ—Ä—ã–µ –±—ã–ª–∏ –æ—Ç–º–µ–Ω–µ–Ω—ã –æ—Ç –æ–±—â–µ–≥–æ —á–∏—Å–ª–∞ –∑–∞–∫–∞–∑–æ–≤ –∏ –Ω–µ–¥–æ–ø–æ–ª—É—á–µ–Ω–Ω–∞—è –ø—Ä–∏–±—ã–ª—å.
 
- **–ë–∏–∑–Ω–µ—Å-–∏–Ω—Ç–µ—Ä–ø—Ä–µ—Ç–∞—Ü–∏—è:**
    - **–í—ã—Å–æ–∫–∏–π –ø–æ–∫–∞–∑–∞—Ç–µ–ª—å:** –ú–æ–∂–µ—Ç —Å–≤–∏–¥–µ—Ç–µ–ª—å—Å—Ç–≤–æ–≤–∞—Ç—å –æ –ø—Ä–æ–±–ª–µ–º–∞—Ö –Ω–∞ —Ä–∞–∑–ª–∏—á–Ω—ã—Ö —ç—Ç–∞–ø–∞—Ö:
        - **–ù–∞ —ç—Ç–∞–ø–µ –¥–æ –æ–ø–ª–∞—Ç—ã:** –ü—Ä–æ–±–ª–µ–º—ã —Å –ø—Ä–æ—Ü–µ—Å—Å–æ–º –æ—Ñ–æ—Ä–º–ª–µ–Ω–∏—è –∑–∞–∫–∞–∑–∞, –Ω–µ–¥–æ—Å—Ç–∞—Ç–æ–∫ –∏–Ω—Ñ–æ—Ä–º–∞—Ü–∏–∏ –æ —Ç–æ–≤–∞—Ä–µ, –≤—ã—Å–æ–∫–∏–µ —Ü–µ–Ω—ã, –Ω–µ–¥–æ—Å—Ç—É–ø–Ω–æ—Å—Ç—å —Å–ø–æ—Å–æ–±–æ–≤ –æ–ø–ª–∞—Ç—ã.
        - **–ù–∞ —ç—Ç–∞–ø–µ –ø–æ—Å–ª–µ –æ–ø–ª–∞—Ç—ã, –Ω–æ –¥–æ –æ—Ç–ø—Ä–∞–≤–∫–∏:** –ü—Ä–æ–±–ª–µ–º—ã —Å –Ω–∞–ª–∏—á–∏–µ–º —Ç–æ–≤–∞—Ä–∞ —É –ø—Ä–æ–¥–∞–≤—Ü–∞, –∑–∞–¥–µ—Ä–∂–∫–∏ —Å –æ–±—Ä–∞–±–æ—Ç–∫–æ–π –∑–∞–∫–∞–∑–∞, –Ω–µ—É–¥–æ–≤–ª–µ—Ç–≤–æ—Ä–µ–Ω–Ω–æ—Å—Ç—å –∫–ª–∏–µ–Ω—Ç–∞ —É—Å–ª–æ–≤–∏—è–º–∏ –¥–æ—Å—Ç–∞–≤–∫–∏ –∏–ª–∏ —Å—Ä–æ–∫–∞–º–∏.
        - **–ü–æ—Å–ª–µ –æ—Ç–ø—Ä–∞–≤–∫–∏ (—Ä–µ–¥–∫–æ):** –õ–æ–≥–∏—Å—Ç–∏—á–µ—Å–∫–∏–µ –ø—Ä–æ–±–ª–µ–º—ã, –ø–æ–≤—Ä–µ–∂–¥–µ–Ω–∏–µ —Ç–æ–≤–∞—Ä–∞ –≤ –ø—É—Ç–∏ (–æ–±—ã—á–Ω–æ —Ñ–∏–∫—Å–∏—Ä—É–µ—Ç—Å—è –∫–∞–∫ –≤–æ–∑–≤—Ä–∞—Ç, –Ω–æ –º–æ–∂–µ—Ç –±—ã—Ç—å –∏ –æ—Ç–º–µ–Ω–∞).
    –í—ã—Å–æ–∫–∏–π OCR –Ω–µ–≥–∞—Ç–∏–≤–Ω–æ –≤–ª–∏—è–µ—Ç –Ω–∞ GMV, –æ–ø–µ—Ä–∞—Ü–∏–æ–Ω–Ω—ã–µ —Ä–µ—Å—É—Ä—Å—ã –∏ –º–æ–∂–µ—Ç —É–∫–∞–∑—ã–≤–∞—Ç—å –Ω–∞ –ø–ª–æ—Ö–æ–µ –∫–∞—á–µ—Å—Ç–≤–æ —Å–µ—Ä–≤–∏—Å–∞ –∏–ª–∏ –ø—Ä–æ–±–ª–µ–º—ã —Å –∞—Å—Å–æ—Ä—Ç–∏–º–µ–Ω—Ç–æ–º/–Ω–∞–ª–∏—á–∏–µ–º.

- **–î–æ–ø–æ–ª–Ω–∏—Ç–µ–ª—å–Ω—ã–µ –¥–µ—Ç–∞–ª–∏:**
    - –û—Ç–º–µ–Ω—ë–Ω–Ω—ã–º–∏ —Å—á–∏—Ç–∞–µ–º –∑–∞–∫–∞–∑—ã —Å–æ —Å—Ç–∞—Ç—É—Å–æ–º "–æ—Ç–º–µ–Ω—ë–Ω" –∏ "–Ω–µ–¥–æ—Å—Ç—É–ø–µ–Ω" - —Ç–∞–∫ –∫–∞–∫ —Å—Ç–∞—Ç—É—Å "–Ω–µ–¥–æ—Å—Ç—É–ø–µ–Ω" –º–æ–∂–µ—Ç —Å–≤–∏–¥–µ—Ç–µ–ª—å—Å—Ç–≤–æ–≤–∞—Ç—å –æ–± –æ—Ç–º–µ–Ω–µ —Å–æ —Å—Ç–æ—Ä–æ–Ω—ã –ø—Ä–æ–¥–∞–≤—Ü–∞.

```python
# –†–∞—Å—Å—á–∏—Ç—ã–≤–∞–µ–º –æ–±—â–∏–π –ø–æ–∫–∞–∑–∞—Ç–µ–ª—å OCR
query_order_cancellation_rate = """
SELECT
    CAST(SUM(CASE WHEN order_status = '–æ—Ç–º–µ–Ω–µ–Ω' THEN 1
                  WHEN order_status = '–Ω–µ–¥–æ—Å—Ç—É–ø–µ–Ω' THEN 1 ELSE 0 END) AS REAL) * 100 
                  / COUNT(*) AS order_cancellation_rate
FROM orders;
"""

order_cancellation_rate = run_query(engine, query_order_cancellation_rate, fetch='scalar')

# –†–∞—Å—Å—á–∏—Ç—ã–≤–∞–µ–º –æ–±—â–∏–π –ø–æ–∫–∞–∑–∞—Ç–µ–ª—å Lost GMV
query_lost_gmv = """
SELECT
    ROUND((SUM(payment_value)::numeric * 14)/ 1000000 , 2)::float AS lost_gmv
FROM
    orders o
JOIN
    order_payments op ON o.order_id = op.order_id
WHERE
    o.order_status IN ('–æ—Ç–º–µ–Ω–µ–Ω', '–Ω–µ–¥–æ—Å—Ç—É–ø–µ–Ω');
"""
lost_gmv = run_query(engine, query_lost_gmv, fetch='scalar')


# –î–æ–ø–æ–ª–Ω–∏—Ç–µ–ª—å–Ω—ã–π –∞–Ω–∞–ª–∏–∑ OCR –ø–æ –º–µ—Å—è—Ü–∞–º
query_ocr_monthly = """
SELECT
    DATE_TRUNC('month', order_purchase_timestamp)::DATE AS month,
    CAST(SUM(CASE WHEN order_status = '–æ—Ç–º–µ–Ω–µ–Ω' THEN 1
                  WHEN order_status = '–Ω–µ–¥–æ—Å—Ç—É–ø–µ–Ω' THEN 1
                        ELSE 0 END) AS REAL) * 100 / COUNT(*) AS monthly_ocr
FROM orders
WHERE order_purchase_timestamp BETWEEN '2017-01-01' AND '2018-07-01'
GROUP BY 1
ORDER BY 1;
"""
ocr_monthly = run_query(engine, query_ocr_monthly, fetch='df')

# –î–æ–ø–æ–ª–Ω–∏—Ç–µ–ª—å–Ω—ã–π –∞–Ω–∞–ª–∏–∑ lost gmv –ø–æ –º–µ—Å—è—Ü–∞–º 
query_lost_monthly = """
SELECT
    DATE_TRUNC('month', o.order_purchase_timestamp)::DATE AS month,
    ROUND(SUM(op.payment_value)::numeric * 14 / 1000000, 2)::float AS monthly_lost_gmv
FROM
    orders o
JOIN
    order_payments op ON o.order_id = op.order_id
WHERE
    o.order_status IN ('–æ—Ç–º–µ–Ω–µ–Ω', '–Ω–µ–¥–æ—Å—Ç—É–ø–µ–Ω')
    AND o.order_purchase_timestamp BETWEEN '2017-01-01' AND '2018-07-01'
GROUP BY 1
ORDER BY 1;
"""
lost_gmv_monthly = run_query(engine, query_lost_monthly, fetch='df')


# –°–æ–∑–¥–∞–µ–º —Ñ–∏–≥—É—Ä—É —Å 3 –æ–±–ª–∞—Å—Ç—è–º–∏: 2 –∏–Ω–¥–∏–∫–∞—Ç–æ—Ä–∞ –≤–≤–µ—Ä—Ö—É –∏ –≥—Ä–∞—Ñ–∏–∫ –≤–Ω–∏–∑—É
fig = make_subplots(
    rows=3, cols=2,
    specs=[
        [{"type": "indicator", "colspan": 1}, {"type": "indicator", "colspan": 1}],  
        [{"type": "xy", "colspan": 2}, None],
        [{"type": "xy", "colspan": 2}, None],
    ],
    column_widths=[0.5, 0.5],  # –†–∞–≤–Ω—ã–µ –¥–æ–ª–∏ —à–∏—Ä–∏–Ω—ã –¥–ª—è –∏–Ω–¥–∏–∫–∞—Ç–æ—Ä–æ–≤
    row_heights=[0.4, 0.3, 0.3],
    vertical_spacing=0.2,
    shared_xaxes=True,
    horizontal_spacing=0.01
)

# –ü–µ—Ä–≤—ã–π –∏–Ω–¥–∏–∫–∞—Ç–æ—Ä 
fig.add_trace(
    go.Indicator(
        mode="number",
        value=float(order_cancellation_rate),
        title={"text": "<b>Order Cancellation Rate</b><br><span style='font-size:0.8em'>–ü—Ä–æ—Ü–µ–Ω—Ç –æ—Ç–º–µ–Ω–µ–Ω–Ω—ã—Ö –∑–∞–∫–∞–∑–æ–≤</span>"},
        number={'suffix': "%", 'valueformat': ".2f"},
    ),
    row=1, col=1
)

# –í—Ç–æ—Ä–æ–π –∏–Ω–¥–∏–∫–∞—Ç–æ—Ä 
fig.add_trace(
    go.Indicator(
        mode="number",
        value=float(lost_gmv),
        title={"text": "<b>Lost GMV</b><br><span style='font-size:0.8em'>–ù–µ–¥–æ–ø–æ–ª—É—á–µ–Ω–Ω–∞—è –ø—Ä–∏–±—ã–ª—å</span>"},
        number={'suffix': " –º–ª–Ω", 'valueformat': ".2f"},
    ),
    
    row=1, col=2
)

# –õ–∏–Ω–µ–π–Ω—ã–π –≥—Ä–∞—Ñ–∏–∫
fig.add_trace(
    go.Scatter(
        x=ocr_monthly['month'],
        y=ocr_monthly['monthly_ocr'],
        mode='lines+markers',
        line=dict(color=custom_palette[2], width=3),
        name="–£—Ä–æ–≤–µ–Ω—å –æ—Ç–º–µ–Ω",
        hovertemplate="<b>%{x|%b %Y}</b><br>%{y:.2f}%"
    ),
    row=2, col=1
)

fig.add_trace(
    go.Scatter(
        x=lost_gmv_monthly['month'],
        y=lost_gmv_monthly['monthly_lost_gmv'],
        mode='lines+markers',
        line=dict(color=custom_palette[1], width=2,),
        name="–ù–µ–¥–æ–ø–æ–ª—É—á–µ–Ω–Ω–æ–µ GMV",
        yaxis="y2",
        hovertemplate="<b>%{x|%b %Y}</b><br>%{y:.2f} –º–ª–Ω —Ä—É–±."
    ),
    row=3, col=1
)


# –û–±—â–∏–µ –Ω–∞—Å—Ç—Ä–æ–π–∫–∏
fig.update_layout(
    height=700,
    title_text="<b>–ê–Ω–∞–ª–∏–∑ –æ—Ç–º–µ–Ω –∑–∞–∫–∞–∑–æ–≤ –∏ –Ω–µ–ø–æ–ª—É—á–µ–Ω–Ω–æ–π –ø—Ä–∏–±—ã–ª–∏</b>",
 #   title_x=0.5,
    margin=dict(t=150, b=10, l=50, r=50),
    showlegend=False
)

# –ù–∞—Å—Ç—Ä–æ–π–∫–∏ –æ—Å–µ–π
fig.update_xaxes(title_text="–ú–µ—Å—è—Ü", row=3, col=1)
fig.update_xaxes(title_text="", showticklabels=False, row=2, col=1)  # –£–±–∏—Ä–∞–µ–º –ø–æ–¥–ø–∏—Å–∏ X –≤–æ –≤—Ç–æ—Ä–æ–º —Ä—è–¥—É
fig.update_yaxes(title_text="–£—Ä–æ–≤–µ–Ω—å –æ—Ç–º–µ–Ω, %", row=2, col=1)
fig.update_yaxes(title_text="–ü–æ—Ç–µ—Ä—è GMV, –º–ª–Ω", row=3, col=1)
fig.write_image('1.png')
```

- **–ë–∏–∑–Ω–µ—Å-–∏–Ω—Ç–µ—Ä–ø—Ä–µ—Ç–∞—Ü–∏—è:**
    - –†–µ–∑–∫–∏–µ –≤—Å–ø–ª–µ—Å–∫–∏ OCR –≤ –æ–ø—Ä–µ–¥–µ–ª–µ–Ω–Ω—ã—Ö –∫–∞—Ç–µ–≥–æ—Ä–∏—è—Ö: –ú–æ–≥—É—Ç —É–∫–∞–∑—ã–≤–∞—Ç—å –Ω–∞ –≤—Ä–µ–º–µ–Ω–Ω—ã–µ –ø—Ä–æ–±–ª–µ–º—ã, —Å–ø–µ—Ü–∏—Ñ–∏—á–Ω—ã–µ –¥–ª—è —ç—Ç–∏—Ö –∫–∞—Ç–µ–≥–æ—Ä–∏–π, —Ç–∞–∫–∏–µ –∫–∞–∫ –ø—Ä–æ–±–ª–µ–º—ã —Å –ø–æ—Å—Ç–∞–≤–∫–∞–º–∏, –∏–∑–º–µ–Ω–µ–Ω–∏—è –≤ –æ–ø–∏—Å–∞–Ω–∏–∏ —Ç–æ–≤–∞—Ä–æ–≤, —Å–µ–∑–æ–Ω–Ω–æ–µ –æ—Ç—Å—É—Ç—Å—Ç–≤–∏–µ —Å–ø—Ä–æ—Å–∞ –∏–ª–∏ –ø—Ä–æ–±–ª–µ–º—ã —Å —Ü–µ–Ω–æ–æ–±—Ä–∞–∑–æ–≤–∞–Ω–∏–µ–º.
    - –ü–æ—Å—Ç–µ–ø–µ–Ω–Ω—ã–π —Ä–æ—Å—Ç OCR –≤ –∫–∞—Ç–µ–≥–æ—Ä–∏–∏: –ú–æ–∂–µ—Ç —Å–≤–∏–¥–µ—Ç–µ–ª—å—Å—Ç–≤–æ–≤–∞—Ç—å –æ —Å–∏—Å—Ç–µ–º–Ω—ã—Ö –ø—Ä–æ–±–ª–µ–º–∞—Ö, —Ç—Ä–µ–±—É—é—â–∏—Ö –±–æ–ª–µ–µ –≥–ª—É–±–æ–∫–æ–≥–æ –∏–∑—É—á–µ–Ω–∏—è (–Ω–∞–ø—Ä–∏–º–µ—Ä, –Ω–µ–∫–æ–Ω–∫—É—Ä–µ–Ω—Ç–Ω—ã–µ —Ü–µ–Ω—ã, –Ω–∏–∑–∫–æ–µ –∫–∞—á–µ—Å—Ç–≤–æ —Ç–æ–≤–∞—Ä–æ–≤, –ø—Ä–æ–±–ª–µ–º—ã —Å –ª–æ–≥–∏—Å—Ç–∏–∫–æ–π –¥–ª—è –¥–∞–Ω–Ω–æ–π –∫–∞—Ç–µ–≥–æ—Ä–∏–∏).
    - –°–Ω–∏–∂–µ–Ω–∏–µ OCR –≤ –∫–∞—Ç–µ–≥–æ—Ä–∏–∏: –£–∫–∞–∑—ã–≤–∞–µ—Ç –Ω–∞ –ø–æ–ª–æ–∂–∏—Ç–µ–ª—å–Ω—ã–µ –∏–∑–º–µ–Ω–µ–Ω–∏—è, –≤–æ–∑–º–æ–∂–Ω–æ, –≤ —Ä–µ–∑—É–ª—å—Ç–∞—Ç–µ –ø—Ä–µ–¥–ø—Ä–∏–Ω—è—Ç—ã—Ö –º–µ—Ä –ø–æ —É–ª—É—á—à–µ–Ω–∏—é –∫–∞—á–µ—Å—Ç–≤–∞ —Ç–æ–≤–∞—Ä–æ–≤, –ª–æ–≥–∏—Å—Ç–∏–∫–∏ –∏–ª–∏ –∫–ª–∏–µ–Ω—Ç—Å–∫–æ–≥–æ —Å–µ—Ä–≤–∏—Å–∞ –¥–ª—è –¥–∞–Ω–Ω–æ–π –∫–∞—Ç–µ–≥–æ—Ä–∏–∏.
    - –°—Ä–∞–≤–Ω–µ–Ω–∏–µ –¥–∏–Ω–∞–º–∏–∫–∏ OCR –º–µ–∂–¥—É –∫–∞—Ç–µ–≥–æ—Ä–∏—è–º–∏ –ø–æ–∑–≤–æ–ª—è–µ—Ç –≤—ã—è–≤–∏—Ç—å –Ω–∞–∏–±–æ–ª–µ–µ –ø—Ä–æ–±–ª–µ–º–Ω—ã–µ –æ–±–ª–∞—Å—Ç–∏ –∏ —Å—Ñ–æ–∫—É—Å–∏—Ä–æ–≤–∞—Ç—å —É—Å–∏–ª–∏—è –Ω–∞ –∏—Ö —É–ª—É—á—à–µ–Ω–∏–∏.


- **–î–æ–ø–æ–ª–Ω–∏—Ç–µ–ª—å–Ω—ã–µ –¥–µ—Ç–∞–ª–∏:**
¬† ¬† - –û—Ç–º–µ–Ω—ë–Ω–Ω—ã–º–∏ —Å—á–∏—Ç–∞–µ–º –∑–∞–∫–∞–∑—ã —Å–æ —Å—Ç–∞—Ç—É—Å–æ–º "–æ—Ç–º–µ–Ω—ë–Ω" –∏ "–Ω–µ–¥–æ—Å—Ç—É–ø–µ–Ω" - —Ç–∞–∫ –∫–∞–∫ —Å—Ç–∞—Ç—É—Å "–Ω–µ–¥–æ—Å—Ç—É–ø–µ–Ω" –º–æ–∂–µ—Ç —Å–≤–∏–¥–µ—Ç–µ–ª—å—Å—Ç–≤–æ–≤–∞—Ç—å –æ–± –æ—Ç–º–µ–Ω–µ —Å–æ —Å—Ç–æ—Ä–æ–Ω—ã –ø—Ä–æ–¥–∞–≤—Ü–∞.



```python
# 1. –û–±—â–∏–π —É—Ä–æ–≤–µ–Ω—å –æ—Ç–º–µ–Ω –∑–∞–∫–∞–∑–æ–≤ (KPI)
query = """
WITH monthly_category_cancellations AS (
    SELECT
        DATE_TRUNC('month', o.order_purchase_timestamp)::DATE AS month,
        p.category_name_translated,
        COUNT(DISTINCT o.order_id) AS total_orders,
        SUM(CASE WHEN o.order_status = '–æ—Ç–º–µ–Ω–µ–Ω' THEN 1 ELSE 0 END) AS cancelled_orders,
        ROUND(
            CAST(SUM(CASE WHEN o.order_status = '–æ—Ç–º–µ–Ω–µ–Ω' THEN 1 ELSE 0 END) AS FLOAT)::numeric * 100 / 
            COUNT(DISTINCT o.order_id),
        2
        ) AS cancellation_rate
    FROM orders o
    JOIN order_items oi ON o.order_id = oi.order_id
    JOIN products p ON oi.product_id = p.product_id
    WHERE o.order_purchase_timestamp BETWEEN '2017-01-01' AND '2018-08-01'
      AND p.category_name_translated IS NOT NULL
    GROUP BY 1, 2
    HAVING COUNT(DISTINCT o.order_id) > 10  -- –ò—Å–∫–ª—é—á–∞–µ–º –∫–∞—Ç–µ–≥–æ—Ä–∏–∏ —Å –º–∞–ª—ã–º –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ–º –∑–∞–∫–∞–∑–æ–≤
)

SELECT 
    month,
    category_name_translated,
    cancellation_rate::float
FROM monthly_category_cancellations
ORDER BY month, cancellation_rate DESC;
"""
ocr_data = run_query(engine, query, fetch='df')

def add_cancellation_traces(fig, dataframe, row, title):
    # –ê–≥—Ä–µ–≥–∞—Ü–∏—è –¥–∞–Ω–Ω—ã—Ö
    df_cancel = (
        dataframe
        .groupby(['month', 'category_name_translated'], as_index=False)
        ['cancellation_rate'].mean()
    )
    
    # –ü–æ–ª—É—á–∞–µ–º –ø–æ—Å–ª–µ–¥–Ω–∏–π –º–µ—Å—è—Ü –∏ —Ç–æ–ø-3 –∫–∞—Ç–µ–≥–æ—Ä–∏–∏
    last_month = df_cancel['month'].max()
    top3 = (
        df_cancel#[df_cancel['month'] == last_month]
        .nlargest(5, 'cancellation_rate')['category_name_translated']
        .tolist()
    )
    
    # –¶–≤–µ—Ç–æ–≤–∞—è —Å—Ö–µ–º–∞ (–∫—Ä–∞—Å–Ω—ã–µ –æ—Ç—Ç–µ–Ω–∫–∏ –¥–ª—è –æ—Ç–º–µ–Ω)
    gradient = px.colors.sequential.Reds_r[:5]
    all_cats = df_cancel['category_name_translated'].unique()
    cmap = {c: (gradient[top3.index(c)] if c in top3 else '#CCCCCC') for c in all_cats}
    
    # –î–æ–±–∞–≤–ª—è–µ–º –ª–∏–Ω–∏–∏ –¥–ª—è –≤—Å–µ—Ö –∫–∞—Ç–µ–≥–æ—Ä–∏–π
    for cat in all_cats:
        d = df_cancel[df_cancel['category_name_translated'] == cat]
        fig.add_trace(
            go.Scatter(
                x=d.month, y=d.cancellation_rate,
                mode='lines',
                line=dict(color=cmap[cat], width=(3 if cat in top3 else 1)),
                name=cat, 
                showlegend=(cat in top3),
                legendgroup=title,
                hovertemplate=f"{cat}<br>%{{y:.1f}}%"
            ), 
            row=row, col=1
        )
        
        
    # –î–æ–±–∞–≤–ª—è–µ–º –º–∞—Ä–∫–µ—Ä—ã –∏ –ø–æ–¥–ø–∏—Å–∏ –¥–ª—è top-3 –∫–∞—Ç–µ–≥–æ—Ä–∏–π:
    for i, cat in enumerate(top3):
        cat_df = df_cancel[df_cancel['category_name_translated'] == cat]
        
        
        # –ü–∏–∫–æ–≤–æ–µ –∑–Ω–∞—á–µ–Ω–∏–µ –∑–∞ –≤—Å—ë –≤—Ä–µ–º—è
        peak_point = cat_df.loc[cat_df['cancellation_rate'].idxmax()]
        fig.add_trace(
            go.Scatter(
                x=[peak_point['month']], y=[peak_point['cancellation_rate']],
                mode='markers+text',
                marker=dict(color=gradient[i], size=10, symbol='diamond'),
                text=[f"{cat} {peak_point['cancellation_rate']:.1f}%"],
                textposition="top center",
                textfont=dict(color=gradient[i]),
                showlegend=False,
                hoverinfo='skip'
            ),
            row=row, col=1
        ) 
        
    # –ù–∞—Å—Ç—Ä–æ–π–∫–∞ –æ—Å–µ–π
    fig.update_xaxes(
        range=[df_cancel.month.min() - pd.Timedelta(days=30), last_month],
        title_text='–ú–µ—Å—è—Ü', 
        row=row, col=1
    )
    fig.update_yaxes(
        title_text='–£—Ä–æ–≤–µ–Ω—å –æ—Ç–º–µ–Ω (%)',
        range=[0, df_cancel.cancellation_rate.max() * 1.2],
        row=row, col=1
    )

# –°–æ–∑–¥–∞–µ–º —Ñ–∏–≥—É—Ä—É
fig = make_subplots(rows=1, cols=1)
add_cancellation_traces(fig, ocr_data, row=1, title="–¢–æ–ø-3 –∫–∞—Ç–µ–≥–æ—Ä–∏–π –ø–æ –æ—Ç–º–µ–Ω–∞–º")

# –û–±—â–∏–µ –Ω–∞—Å—Ç—Ä–æ–π–∫–∏
fig.update_layout(
    height=500,
    title_text="<b>–î–∏–Ω–∞–º–∏–∫–∞ –æ—Ç–º–µ–Ω –∑–∞–∫–∞–∑–æ–≤ –ø–æ –∫–∞—Ç–µ–≥–æ—Ä–∏—è–º</b><br>–≤—ã–¥–µ–ª–µ–Ω—ã —Ç–æ–ø-5 –∫–∞—Ç–µ–≥–æ—Ä–∏–π –ø–æ OCR</br>",
    margin=dict(t=100, b=50, l=50, r=50),
    showlegend=False
)

fig.write_image('1.png')
```

**üìä –ò–Ω—Ç–µ—Ä–ø—Ä–µ—Ç–∞—Ü–∏—è —Ä–µ–∑—É–ª—å—Ç–∞—Ç–∞:**

- –î–æ–ª—è –æ—Ç–º–µ–Ω –º–µ–Ω—è–µ—Ç—Å—è –æ—Ç –º–µ—Å—è—Ü–∞ –∫ –º–µ—Å—è—Ü—É –ø–æ –∫–∞—Ç–µ–≥–æ—Ä–∏—è–º, —Ç–∞–∫ –±–æ–ª—å—à–µ –≤—Å–µ–≥–æ –æ—Ç–º–µ–Ω –≤ —Ñ–µ—Ä–∞–ª–µ 2017 –±—ã–ª–æ –≤ –∫–∞—Ç–µ–≥–æ—Ä–∏–∏ **–º–µ–±–µ–ª—å –¥–ª—è –≥–æ—Å—Ç–∏–Ω–Ω–æ–π**, –∞ –≤ –∞–≤–≥—É—Å—Ç–µ —É–∂–µ –≤ **–º—É–∑—ã–∫–∞–ª—å–Ω—ã—Ö –∏–Ω—Å—Ç—Ä—É–º–µ–Ω—Ç–æ–≤**.
- **–º—É–∂—Å–∫–∞—è –æ–¥–µ–∂–¥–∞**, **—Å—Ç—Ä–æ–π–±–µ–∑–æ–ø–∞—Å–Ω–æ—Å—Ç—å**, **–∏–≥—Ä—ã –∏ –ø—Ä–∏—Å—Ç–∞–≤–∫–∏** - –≤—Å–µ —ç—Ç–∏ –∫–∞—Ç–µ–≥–æ—Ä–∏–∏ –ø–æ–ø–∞–¥–∞–ª–∏ –≤ —Å–ø–∏—Å–æ–∫ –∫–∞—Ç–µ–≥–æ—Ä–∏–π —Å –≤—ã—Å–æ–∫–∏–º–∏ –æ—Ç–º–µ–Ω–∞–º–∏.

¬†

- **–ë–∏–∑–Ω–µ—Å-–∏–Ω—Ç–µ—Ä–ø—Ä–µ—Ç–∞—Ü–∏—è:**
    - –ê–Ω–∞–ª–∏–∑ OCR –ø–æ —Å—Ç–∞–¥–∏—è–º –ø–æ–º–æ–≥–∞–µ—Ç —Ç–æ—á–Ω–æ –æ–ø—Ä–µ–¥–µ–ª–∏—Ç—å, –Ω–∞ –∫–∞–∫–æ–º —ç—Ç–∞–ø–µ –≤–æ–∑–Ω–∏–∫–∞—é—Ç –æ—Å–Ω–æ–≤–Ω—ã–µ –ø—Ä–æ–±–ª–µ–º—ã —Å –æ—Ç–º–µ–Ω–∞–º–∏, –∏ —Ä–∞–∑—Ä–∞–±–æ—Ç–∞—Ç—å —Ü–µ–ª–µ–Ω–∞–ø—Ä–∞–≤–ª–µ–Ω–Ω—ã–µ —Ä–µ—à–µ–Ω–∏—è.
    - –í—ã—Å–æ–∫–∏–π OCR –¥–æ –æ–ø–ª–∞—Ç—ã: –ú–æ–∂–µ—Ç —É–∫–∞–∑—ã–≤–∞—Ç—å –Ω–∞ –ø—Ä–æ–±–ª–µ–º—ã —Å –ø—Ä–æ—Ü–µ—Å—Å–æ–º –æ—Ñ–æ—Ä–º–ª–µ–Ω–∏—è –∑–∞–∫–∞–∑–∞ (—Å–ª–æ–∂–Ω–æ—Å—Ç—å, –±–æ–ª—å—à–æ–µ –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ —à–∞–≥–æ–≤), –Ω–µ–¥–æ—Å—Ç–∞—Ç–æ–∫ –∏–Ω—Ñ–æ—Ä–º–∞—Ü–∏–∏ –æ —Ç–æ–≤–∞—Ä–µ, –Ω–µ–ø—Ä–∏–≤–ª–µ–∫–∞—Ç–µ–ª—å–Ω—ã–µ —Ü–µ–Ω—ã –∏–ª–∏ –æ–≥—Ä–∞–Ω–∏—á–µ–Ω–Ω—ã–µ —Å–ø–æ—Å–æ–±—ã –æ–ø–ª–∞—Ç—ã.
    - –í—ã—Å–æ–∫–∏–π OCR –ø–æ—Å–ª–µ –æ–ø–ª–∞—Ç—ã, –Ω–æ –¥–æ –æ—Ç–ø—Ä–∞–≤–∫–∏: –í–µ—Ä–æ—è—Ç–Ω–æ, —Å–≤–∏–¥–µ—Ç–µ–ª—å—Å—Ç–≤—É–µ—Ç –æ –ø—Ä–æ–±–ª–µ–º–∞—Ö —Å –Ω–∞–ª–∏—á–∏–µ–º —Ç–æ–≤–∞—Ä–∞ —É –ø—Ä–æ–¥–∞–≤—Ü–∞, –∑–∞–¥–µ—Ä–∂–∫–∞—Ö —Å –æ–±—Ä–∞–±–æ—Ç–∫–æ–π –∑–∞–∫–∞–∑–∞, –∏–∑–º–µ–Ω–µ–Ω–∏–∏ —É—Å–ª–æ–≤–∏–π –¥–æ—Å—Ç–∞–≤–∫–∏ –∏–ª–∏ —Å—Ä–æ–∫–æ–≤, –∫–æ—Ç–æ—Ä—ã–µ –Ω–µ —É—Å—Ç—Ä–∞–∏–≤–∞—é—Ç –∫–ª–∏–µ–Ω—Ç–∞.
    - OCR –ø–æ—Å–ª–µ –æ—Ç–ø—Ä–∞–≤–∫–∏ (–æ–±—ã—á–Ω–æ –Ω–∏–∑–∫–∏–π): –ú–æ–∂–µ—Ç –±—ã—Ç—å —Å–≤—è–∑–∞–Ω —Å –ª–æ–≥–∏—Å—Ç–∏—á–µ—Å–∫–∏–º–∏ –ø—Ä–æ–±–ª–µ–º–∞–º–∏ (–Ω–∞–ø—Ä–∏–º–µ—Ä, –Ω–µ–≤–æ–∑–º–æ–∂–Ω–æ—Å—Ç—å –¥–æ—Å—Ç–∞–≤–∫–∏ –≤ —Ä–µ–≥–∏–æ–Ω –∫–ª–∏–µ–Ω—Ç–∞) –∏–ª–∏ –ø–æ–≤—Ä–µ–∂–¥–µ–Ω–∏–µ–º —Ç–æ–≤–∞—Ä–∞ –≤ –ø—É—Ç–∏ (—Ö–æ—Ç—è —Ç–∞–∫–∏–µ —Å–ª—É—á–∞–∏ —á–∞—â–µ –ø—Ä–∏–≤–æ–¥—è—Ç –∫ –≤–æ–∑–≤—Ä–∞—Ç–∞–º).


- **–î–æ–ø–æ–ª–Ω–∏—Ç–µ–ª—å–Ω—ã–µ –¥–µ—Ç–∞–ª–∏:**
¬†   - –û—Ç—Å–ª–µ–∂–∏–≤–∞—Ç—å –∏–∑–º–µ–Ω–µ–Ω–∏—è OCR –ø–æ —Å—Ç–∞–¥–∏—è–º –ø–æ—Å–ª–µ –≤–Ω–µ–¥—Ä–µ–Ω–∏—è —É–ª—É—á—à–µ–Ω–∏–π –≤ –ø—Ä–æ—Ü–µ—Å—Å –∑–∞–∫–∞–∑–∞ –∏–ª–∏ –ª–æ–≥–∏—Å—Ç–∏–∫—É.

```python
# –í–∞—à –∏—Å—Ö–æ–¥–Ω—ã–π –∑–∞–ø—Ä–æ—Å –¥–ª—è –æ–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—è —Å—Ç–∞–¥–∏–π –æ—Ç–º–µ–Ω—ã
query_cancellation_stage = """
SELECT
    CASE
        WHEN order_status = '–æ—Ç–º–µ–Ω–µ–Ω' AND order_approved_at IS NULL THEN '–û—Ç–º–µ–Ω–µ–Ω –¥–æ –æ–¥–æ–±—Ä–µ–Ω–∏—è –æ–ø–ª–∞—Ç—ã'
        WHEN order_status = '–æ—Ç–º–µ–Ω–µ–Ω' AND order_approved_at IS NOT NULL
            AND order_delivered_carrier_date IS NULL THEN '–û—Ç–º–µ–Ω–µ–Ω –ø–æ—Å–ª–µ –æ–ø–ª–∞—Ç—ã, –¥–æ –ø–µ—Ä–µ–¥–∞—á–∏ –≤ –¥–æ—Å—Ç–∞–≤–∫—É'
        WHEN order_status = '–æ—Ç–º–µ–Ω–µ–Ω' AND order_delivered_carrier_date IS NOT NULL
            AND order_delivered_customer_date IS NULL THEN '–û—Ç–º–µ–Ω–µ–Ω –ø–æ—Å–ª–µ –ø–µ—Ä–µ–¥–∞—á–∏ –≤ –¥–æ—Å—Ç–∞–≤–∫—É, –¥–æ –≤—Ä—É—á–µ–Ω–∏—è'
        WHEN order_status = '–Ω–µ–¥–æ—Å—Ç—É–ø–µ–Ω' THEN '–ù–µ–¥–æ—Å—Ç—É–ø–µ–Ω (–æ—Ç–º–µ–Ω–∞ —Å–æ —Å—Ç–æ—Ä–æ–Ω—ã –ø—Ä–æ–¥–∞–≤—Ü–∞/–ø–ª–∞—Ç—Ñ–æ—Ä–º—ã)'
        ELSE '–î—Ä—É–≥–æ–µ (–≤–∫–ª—é—á–∞—è –æ—Ç–º–µ–Ω—ã –ø–æ—Å–ª–µ –¥–æ—Å—Ç–∞–≤–∫–∏ - —Ä–µ–¥–∫–æ)'
    END AS cancellation_stage,
    COUNT(order_id) AS count_of_cancellations
FROM
    orders
WHERE
    order_status IN ('–æ—Ç–º–µ–Ω–µ–Ω', '–Ω–µ–¥–æ—Å—Ç—É–ø–µ–Ω')
GROUP BY
    1
ORDER BY
    2 DESC;
"""
df_cancellation_stage = run_query(engine, query_cancellation_stage, fetch='df')


# –°–æ–∑–¥–∞–µ–º –∫—Ä—É–≥–æ–≤—É—é –¥–∏–∞–≥—Ä–∞–º–º—É
fig = px.pie(df_cancellation_stage,
             names='cancellation_stage',
             values='count_of_cancellations',
             title='<b>–†–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ –æ—Ç–º–µ–Ω –ø–æ —Å—Ç–∞–¥–∏—è–º –∑–∞–∫–∞–∑–∞</b>',
             hole=0.45, # –ù–µ–º–Ω–æ–≥–æ —É–≤–µ–ª–∏—á–∏–º –æ—Ç–≤–µ—Ä—Å—Ç–∏–µ –¥–ª—è —Ç–µ–∫—Å—Ç–∞
             color_discrete_sequence=custom_palette) # –ü—Ä–∏–º–µ—Ä –¥—Ä—É–≥–æ–π —Ü–≤–µ—Ç–æ–≤–æ–π —Å—Ö–µ–º—ã

# –û–±–Ω–æ–≤–ª—è–µ–º –æ—Ç–æ–±—Ä–∞–∂–µ–Ω–∏–µ —Ç–µ–∫—Å—Ç–∞ –Ω–∞ —Å–µ–∫—Ç–æ—Ä–∞—Ö (–Ω–∞–ø—Ä–∏–º–µ—Ä, —Å–Ω–∞—Ä—É–∂–∏, –µ—Å–ª–∏ —É–¥–æ–±–Ω–µ–µ)
fig.update_traces(textposition='inside', textinfo='percent', pull=[0.02]*len(df_cancellation_stage))


# –î–æ–±–∞–≤–ª—è–µ–º –∞–Ω–Ω–æ—Ç–∞—Ü–∏–∏ –≤ —Ü–µ–Ω—Ç—Ä "–±—É–±–ª–∏–∫–∞"
fig.update_layout(
    uniformtext_minsize=10, # –ú–∏–Ω–∏–º–∞–ª—å–Ω—ã–π —Ä–∞–∑–º–µ—Ä —Ç–µ–∫—Å—Ç–∞ –¥–ª—è –ª—É—á—à–µ–π —á–∏—Ç–∞–µ–º–æ—Å—Ç–∏
    uniformtext_mode='hide', # –°–∫—Ä—ã–≤–∞—Ç—å —Ç–µ–∫—Å—Ç, –µ—Å–ª–∏ –æ–Ω –Ω–µ –ø–æ–º–µ—â–∞–µ—Ç—Å—è
    legend_title_text='–°—Ç–∞–¥–∏–∏ –æ—Ç–º–µ–Ω—ã',
    legend=dict(
        orientation="v",
        yanchor="auto",
        y=1.05,
        xanchor="left", # –õ–µ–≥–µ–Ω–¥–∞ –ø–æ —Ü–µ–Ω—Ç—Ä—É –Ω–∞–¥ –≥—Ä–∞—Ñ–∏–∫–æ–º
        x=0.8
    ),

)

fig.write_image('1.png')
```

    - 49 % –æ—Ç–º–µ–Ω—ã —Å–æ —Å—Ç–æ—Ä–æ–Ω—ã –ø—Ä–æ–¥–∞–≤—Ü–∞ (—Ç–æ–≤–∞—Ä –Ω–µ–¥–æ—Å—Ç—É–ø–µ–Ω);

### On-Time Delivery Rate
- **–û–ø–∏—Å–∞–Ω–∏–µ –º–µ—Ç—Ä–∏–∫–∏:**
    –ü—Ä–æ—Ü–µ–Ω—Ç –∑–∞–∫–∞–∑–æ–≤, —É –∫–æ—Ç–æ—Ä—ã—Ö —Ñ–∞–∫—Ç–∏—á–µ—Å–∫–∞—è –¥–∞—Ç–∞ –¥–æ—Å—Ç–∞–≤–∫–∏ –ø–æ–∫—É–ø–∞—Ç–µ–ª—é  –±—ã–ª–∞ —Ä–∞–Ω—å—à–µ –∏–ª–∏ —Ä–∞–≤–Ω–∞ –æ–∂–∏–¥–∞–µ–º–æ–π –¥–∞—Ç–µ –¥–æ—Å—Ç–∞–≤–∫–∏.

- **–ë–∏–∑–Ω–µ—Å-–∏–Ω—Ç–µ—Ä–ø—Ä–µ—Ç–∞—Ü–∏—è:**
    - **–í—ã—Å–æ–∫–∏–π –ø–æ–∫–∞–∑–∞—Ç–µ–ª—å (>90-95%):** –°–≤–∏–¥–µ—Ç–µ–ª—å—Å—Ç–≤—É–µ—Ç –æ —Ö–æ—Ä–æ—à–æ –æ—Ç–ª–∞–∂–µ–Ω–Ω–æ–π –ª–æ–≥–∏—Å—Ç–∏–∫–µ –∏ –≤—ã–ø–æ–ª–Ω–µ–Ω–∏–∏ –æ–±–µ—â–∞–Ω–∏–π –ø–µ—Ä–µ–¥ –∫–ª–∏–µ–Ω—Ç–∞–º–∏, —á—Ç–æ –ø–æ–ª–æ–∂–∏—Ç–µ–ª—å–Ω–æ –≤–ª–∏—è–µ—Ç –Ω–∞ —É–¥–æ–≤–ª–µ—Ç–≤–æ—Ä–µ–Ω–Ω–æ—Å—Ç—å –∏ –ª–æ—è–ª—å–Ω–æ—Å—Ç—å.
    - **–°—Ä–µ–¥–Ω–∏–π –ø–æ–∫–∞–∑–∞—Ç–µ–ª—å (80-90%):** –£–∫–∞–∑—ã–≤–∞–µ—Ç –Ω–∞ –Ω–∞–ª–∏—á–∏–µ –Ω–µ–∫–æ—Ç–æ—Ä—ã—Ö –ø—Ä–æ–±–ª–µ–º –≤ –ø—Ä–æ—Ü–µ—Å—Å–µ –¥–æ—Å—Ç–∞–≤–∫–∏, –∫–æ—Ç–æ—Ä—ã–µ –º–æ–≥—É—Ç –ø—Ä–∏–≤–æ–¥–∏—Ç—å –∫ –Ω–µ–¥–æ–≤–æ–ª—å—Å—Ç–≤—É —á–∞—Å—Ç–∏ –∫–ª–∏–µ–Ω—Ç–æ–≤. –¢—Ä–µ–±—É–µ—Ç –∞–Ω–∞–ª–∏–∑–∞ –ø—Ä–∏—á–∏–Ω –∑–∞–¥–µ—Ä–∂–µ–∫.
    - **–ù–∏–∑–∫–∏–π –ø–æ–∫–∞–∑–∞—Ç–µ–ª—å (<80%):** –°–∏–≥–Ω–∞–ª–∏–∑–∏—Ä—É–µ—Ç –æ —Å–µ—Ä—å–µ–∑–Ω—ã—Ö –ø—Ä–æ–±–ª–µ–º–∞—Ö –≤ –ª–æ–≥–∏—Å—Ç–∏–∫–µ, —á—Ç–æ –º–æ–∂–µ—Ç –ø—Ä–∏–≤–æ–¥–∏—Ç—å –∫ –≤—ã—Å–æ–∫–æ–º—É –æ—Ç—Ç–æ–∫—É –∫–ª–∏–µ–Ω—Ç–æ–≤, –Ω–µ–≥–∞—Ç–∏–≤–Ω—ã–º –æ—Ç–∑—ã–≤–∞–º –∏ —Å–Ω–∏–∂–µ–Ω–∏—é —Ä–µ–ø—É—Ç–∞—Ü–∏–∏.
    –≠—Ç–∞ –º–µ—Ç—Ä–∏–∫–∞ –Ω–∞–ø—Ä—è–º—É—é –≤–ª–∏—è–µ—Ç –Ω–∞ –≤–æ—Å–ø—Ä–∏—è—Ç–∏–µ –Ω–∞–¥–µ–∂–Ω–æ—Å—Ç–∏ —Å–µ—Ä–≤–∏—Å–∞.

- **–î–æ–ø–æ–ª–Ω–∏—Ç–µ–ª—å–Ω—ã–µ –¥–µ—Ç–∞–ª–∏:**
    - –ê–Ω–∞–ª–∏–∑–∏—Ä–æ–≤–∞—Ç—å –≤ –¥–∏–Ω–∞–º–∏–∫–µ (–ø–æ –º–µ—Å—è—Ü–∞–º, –∫–≤–∞—Ä—Ç–∞–ª–∞–º) –¥–ª—è –æ—Ç—Å–ª–µ–∂–∏–≤–∞–Ω–∏—è —É–ª—É—á—à–µ–Ω–∏–π –∏–ª–∏ —É—Ö—É–¥—à–µ–Ω–∏–π.
---

```python
query_on_time_delivery_rate = """
SELECT
    CAST(SUM(CASE WHEN order_delivered_customer_date <= order_estimated_delivery_date THEN 1 ELSE 0 END) AS REAL) * 100
    / COUNT(*) AS on_time_delivery_rate
FROM orders
WHERE order_status = '–¥–æ—Å—Ç–∞–≤–ª–µ–Ω';
"""

on_time_delivery_rate = run_query(engine, query_on_time_delivery_rate, fetch='df')

# –í–∏–∑—É–∞–ª–∏–∑–∞—Ü–∏—è (–ø—Ä–æ—Å—Ç–æ–µ –∑–Ω–∞—á–µ–Ω–∏–µ)
fig_otdr = go.Figure(go.Indicator(
    mode = "number",
    value = on_time_delivery_rate['on_time_delivery_rate'].iloc[0],
    title = {"text": "<b>On-Time Delivery Rate</b> (%)"},
    number = {'suffix': "%"}
))
fig_otdr.write_image('1.png')
```

**üìä –ò–Ω—Ç–µ—Ä–ø—Ä–µ—Ç–∞—Ü–∏—è —Ä–µ–∑—É–ª—å—Ç–∞—Ç–∞:**
- –ü–ª–∞—Ç—Ñ–æ—Ä–º–∞ –∏–º–µ–µ—Ç –≤—ã—Å–æ–∫–∏–π –ø–æ–∫–∞–∑–∞—Ç–µ–ª—å –º–µ—Ç—Ä–∏–∫–∏ - 91.9 % –∏ –∫–∞–∫ –ø—Ä–∞–≤–∏–ª–æ –¥–æ—Å—Ç–∞–≤–ª—è–µ—Ç –∑–∞–∫–∞–∑—ã –≤–æ–≤—Ä–µ–º—è.

```python
query = """
SELECT
    DATE_TRUNC('month', order_purchase_timestamp) AS month,
    COUNT(*) AS total_orders,
    SUM(CASE WHEN order_delivered_customer_date <= order_estimated_delivery_date THEN 1 ELSE 0 END) AS on_time_orders,
    CAST(SUM(CASE WHEN order_delivered_customer_date <= order_estimated_delivery_date THEN 1 ELSE 0 END) AS REAL) * 100
    / COUNT(*) AS on_time_delivery_rate
FROM orders
WHERE order_status = '–¥–æ—Å—Ç–∞–≤–ª–µ–Ω' AND order_purchase_timestamp BETWEEN '2017-01-01' AND '2018-08-01'
GROUP BY DATE_TRUNC('month', order_purchase_timestamp)
ORDER BY month;
"""
on_time_delivery_dynamic = run_query(engine, query, fetch='df')

# –í–∏–∑—É–∞–ª–∏–∑–∞—Ü–∏—è –¥–∏–Ω–∞–º–∏–∫–∏
fig_dynamic = go.Figure()

fig_dynamic.add_trace(go.Scatter(
    x=on_time_delivery_dynamic['month'],
    y=on_time_delivery_dynamic['on_time_delivery_rate'],
    mode='lines+markers',
    name='On-Time Delivery Rate',
    line=dict(color=custom_palette[0], width=2),
    marker=dict(size=8),
    hovertemplate='<b>%{x|%b %Y}</b><br>Rate: %{y:.2f}%<extra></extra>'
))

# –î–æ–±–∞–≤–ª—è–µ–º —Å—Ä–µ–¥–Ω–µ–µ –∑–Ω–∞—á–µ–Ω–∏–µ –≤ –≤–∏–¥–µ –≥–æ—Ä–∏–∑–æ–Ω—Ç–∞–ª—å–Ω–æ–π –ª–∏–Ω–∏–∏
mean_rate = on_time_delivery_dynamic['on_time_delivery_rate'].mean()
fig_dynamic.add_shape(
    type="line",
    x0=on_time_delivery_dynamic['month'].min(),
    y0=mean_rate,
    x1=on_time_delivery_dynamic['month'].max(),
    y1=mean_rate,
    line=dict(color=custom_palette[3], width=2, dash="dash"),
)

fig_dynamic.update_layout(
    title='<b>On-Time Delivery Rate –≤ –¥–∏–Ω–∞–º–∏–∫–µ</b>',
    xaxis_title='Month',
    yaxis_title='On-Time Delivery Rate (%)',
    yaxis=dict(ticksuffix='%'),
    hovermode='x unified',
    height=500,
    annotations=[
        dict(
            x=on_time_delivery_dynamic['month'].max(),
            y=mean_rate,
            xref="x",
            yref="y",
            text=f"Average: {mean_rate:.2f}%",
            arrowhead=1,
            ax=-50,
            ay=0
        )
    ]
)

fig_dynamic.write_image('1.png')
```

**üìä –ò–Ω—Ç–µ—Ä–ø—Ä–µ—Ç–∞—Ü–∏—è —Ä–µ–∑—É–ª—å—Ç–∞—Ç–∞:**
- –í —Ü–µ–ª–æ–º —É –º–∞—Ä–∫–µ—Ç–ø–ª–µ–π—Å–∞ —Ö–æ—Ä–æ—à–∏–µ –ø–æ–∫–∞–∑–∞—Ç–µ–ª–∏ OTDR, –æ–¥–Ω–∞–∫–æ –≤ –ø–µ—Ä–∏–æ–¥ —Å –Ω–æ—è–±—Ä—è –ø–æ –º–∞—Ä—Ç 2018 –±—ã–ª —Ä–µ–∑–∫–∏–π —Å–ø–∞–¥ –ø–æ–∫–∞–∑–∞—Ç–µ–ª—è –¥–æ 78 %, –≤–µ—Ä–æ—è—Ç–Ω–æ –∏–∑ –∑–∞ –º–∞—Ä–∫–µ—Ç–∏–Ω–≥–æ–≤—ã—Ö –∫–∞–º–ø–∞–Ω–∏–π –∏ –∞–∫—Ü–∏–π, –¥–æ—Å—Ç–∞–≤–∫–∞ –Ω–µ —Å–ø—Ä–∞–≤–ª—è–ª–∞—Å—å —Å –æ–±—ä–µ–º–æ–º.

    - –†–∞—Å—Å—á–∏—Ç—ã–≤–∞—Ç—å —Ç–∞–∫–∂–µ —Å—Ä–µ–¥–Ω–µ–µ –≤—Ä–µ–º—è –¥–æ—Å—Ç–∞–≤–∫–∏ (`Average Delivery Time`) –∏ —Å—Ä–∞–≤–Ω–∏–≤–∞—Ç—å —Å –º–µ–¥–∏–∞–Ω–Ω—ã–º –¥–ª—è –ø–æ–Ω–∏–º–∞–Ω–∏—è –Ω–∞–ª–∏—á–∏—è –∏ –≤–ª–∏—è–Ω–∏—è –≤—ã–±—Ä–æ—Å–æ–≤.

```python
query_median_delivery_time = """
SELECT
    PERCENTILE_CONT(0.5) WITHIN GROUP (ORDER BY (order_delivered_customer_date - order_purchase_timestamp)) AS median_delivery_time
FROM orders
WHERE order_status = '–¥–æ—Å—Ç–∞–≤–ª–µ–Ω';
"""

median_delivery_time = run_query(engine, query_median_delivery_time, fetch='df')
median_days = median_delivery_time['median_delivery_time'].iloc[0].days

# –í–∏–∑—É–∞–ª–∏–∑–∞—Ü–∏—è (–ø—Ä–æ—Å—Ç–æ–µ –∑–Ω–∞—á–µ–Ω–∏–µ)
fig_mdt = go.Figure(go.Indicator(
    mode = "number",
    value = median_days,
    title = {"text": "<b>Median Delivery Time</b> (days)"}
))
fig_mdt.show()
```

```python
query_delivery_analysis = """
WITH delivery_stats AS (
    SELECT
        -- –§–∞–∫—Ç–∏—á–µ—Å–∫–æ–µ –≤—Ä–µ–º—è –¥–æ—Å—Ç–∞–≤–∫–∏ (–ø–æ–∫—É–ø–∫–∞ -> –¥–æ—Å—Ç–∞–≤–∫–∞)
        (o.order_delivered_customer_date - o.order_purchase_timestamp) AS actual_delivery_interval,
        
        -- –û—Ç–∫–ª–æ–Ω–µ–Ω–∏–µ –æ—Ç –ø–ª–∞–Ω–æ–≤–æ–π –¥–∞—Ç—ã (–¥–æ—Å—Ç–∞–≤–∫–∞ - –æ–±–µ—â–∞–Ω–Ω–∞—è –¥–∞—Ç–∞)
        (o.order_delivered_customer_date - o.order_estimated_delivery_date) AS delivery_deviation_interval
    FROM orders o
    WHERE o.order_status = '–¥–æ—Å—Ç–∞–≤–ª–µ–Ω'
      AND o.order_delivered_customer_date IS NOT NULL
      AND o.order_purchase_timestamp IS NOT NULL
      AND o.order_estimated_delivery_date IS NOT NULL
)

SELECT
    EXTRACT(DAY FROM actual_delivery_interval)::INTEGER AS actual_delivery_days,
    EXTRACT(DAY FROM delivery_deviation_interval)::INTEGER AS delivery_deviation_days,
    COUNT(*) AS order_count
FROM delivery_stats
GROUP BY actual_delivery_days, delivery_deviation_days
ORDER BY actual_delivery_days, delivery_deviation_days;
"""

# –ü–æ–ª—É—á–∞–µ–º –¥–∞–Ω–Ω—ã–µ
delivery_data = run_query(engine, query_delivery_analysis, fetch='df')

# –°–æ–∑–¥–∞–µ–º —Ñ–∏–≥—É—Ä—É —Å –¥–≤—É–º—è –ø–æ–¥–≥—Ä–∞—Ñ–∏–∫–∞–º–∏
fig = make_subplots(
    rows=2, cols=1,
    subplot_titles=(
        '–†–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ —Ñ–∞–∫—Ç–∏—á–µ—Å–∫–æ–≥–æ –≤—Ä–µ–º–µ–Ω–∏ –¥–æ—Å—Ç–∞–≤–∫–∏ (–æ—Ç –ø–æ–∫—É–ø–∫–∏ –¥–æ –¥–æ—Å—Ç–∞–≤–∫–∏)',
        '–†–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ –æ—Ç–∫–ª–æ–Ω–µ–Ω–∏—è –æ—Ç –ø–ª–∞–Ω–æ–≤–æ–π –¥–∞—Ç—ã –¥–æ—Å—Ç–∞–≤–∫–∏'
    ),
    vertical_spacing=0.15
)

# –ì–∏—Å—Ç–æ–≥—Ä–∞–º–º–∞ —Ñ–∞–∫—Ç–∏—á–µ—Å–∫–æ–≥–æ –≤—Ä–µ–º–µ–Ω–∏ –¥–æ—Å—Ç–∞–≤–∫–∏
fig.add_trace(
    go.Histogram(
        x=delivery_data['actual_delivery_days'],
        y=delivery_data['order_count'],
        name='–§–∞–∫—Ç–∏—á–µ—Å–∫–æ–µ –≤—Ä–µ–º—è –¥–æ—Å—Ç–∞–≤–∫–∏',
        marker_color=custom_palette[4],
        nbinsx=50,
        hovertemplate='<b>%{x} –¥–Ω–µ–π</b><br>–ó–∞–∫–∞–∑–æ–≤: %{y:,}<extra></extra>'
    ),
    row=1, col=1
)

# –ì–∏—Å—Ç–æ–≥—Ä–∞–º–º–∞ –æ—Ç–∫–ª–æ–Ω–µ–Ω–∏—è –æ—Ç –ø–ª–∞–Ω–æ–≤–æ–π –¥–∞—Ç—ã
fig.add_trace(
    go.Histogram(
        x=delivery_data['delivery_deviation_days'],
        y=delivery_data['order_count'],
        name='–û—Ç–∫–ª–æ–Ω–µ–Ω–∏–µ –æ—Ç –ø–ª–∞–Ω–∞',
        histnorm='probability density',
        marker_color=custom_palette[4],
        nbinsx=50,
        hovertemplate='<b>%{x} –¥–Ω–µ–π</b><br>–ó–∞–∫–∞–∑–æ–≤: %{y:,}<extra></extra>'
    ),
    row=2, col=1
)

# –†–∞—Å—Å—á–∏—Ç—ã–≤–∞–µ–º —Å—Ä–µ–¥–Ω–∏–µ –∑–Ω–∞—á–µ–Ω–∏—è
mean_actual = np.average(delivery_data['actual_delivery_days'], weights=delivery_data['order_count'])
mean_deviation = np.average(delivery_data['delivery_deviation_days'], weights=delivery_data['order_count'])

# –î–æ–±–∞–≤–ª—è–µ–º –ª–∏–Ω–∏–∏ —Å—Ä–µ–¥–Ω–∏—Ö –∑–Ω–∞—á–µ–Ω–∏–π
for row, mean_val, color in zip([1, 2], [mean_actual, mean_deviation], custom_palette):
    fig.add_vline(
        x=mean_val,
        line_dash="dash",
        line_color=color,
        row=row, col=1,
        annotation_text=f"–°—Ä–µ–¥–Ω–µ–µ: {mean_val:.1f} –¥–Ω–µ–π",
        annotation_position="top right",
        annotation_font_size=12
    )

# –ù–∞—Å—Ç—Ä–∞–∏–≤–∞–µ–º –æ—Å–∏
fig.update_xaxes(
    title_text='–î–Ω–∏ (–æ—Ç –ø–æ–∫—É–ø–∫–∏ –¥–æ –¥–æ—Å—Ç–∞–≤–∫–∏)',
    row=1, col=1,
    range=[0, delivery_data['actual_delivery_days'].quantile(0.99)]  # –û–±—Ä–µ–∑–∞–µ–º –≤—ã–±—Ä–æ—Å—ã
)

fig.update_xaxes(
    title_text='–î–Ω–∏ (–æ—Ç—Ä–∏—Ü–∞—Ç–µ–ª—å–Ω—ã–µ - —Ä–∞–Ω—å—à–µ —Å—Ä–æ–∫–∞, –ø–æ–ª–æ–∂–∏—Ç–µ–ª—å–Ω—ã–µ - –∑–∞–¥–µ—Ä–∂–∫–∞)',
    row=2, col=1,
    range=[delivery_data['delivery_deviation_days'].quantile(0.01), 
          delivery_data['delivery_deviation_days'].quantile(0.99)]
)

fig.update_yaxes(title_text='–ö–æ–ª–∏—á–µ—Å—Ç–≤–æ –∑–∞–∫–∞–∑–æ–≤', row=1, col=1)
fig.update_yaxes(title_text='–ö–æ–ª–∏—á–µ—Å—Ç–≤–æ –∑–∞–∫–∞–∑–æ–≤', row=2, col=1)

# –û–±—â–∏–π layout
fig.update_layout(
    title_text='<b>–ê–Ω–∞–ª–∏–∑ –≤—Ä–µ–º–µ–Ω–∏ –¥–æ—Å—Ç–∞–≤–∫–∏ –∑–∞–∫–∞–∑–æ–≤</b>',
    height=700,
    bargap=0.1,
    showlegend=False,
    hovermode='x',
    margin=dict(t=100),
    #title_x=0.5
)


fig.write_image('1.png')
```

```python
query_payment_type_distribution = """
SELECT
    payment_type,
    COUNT(*) AS order_count
FROM order_payments
GROUP BY payment_type
ORDER BY order_count DESC;
"""

payment_type_distribution = run_query(engine, query_payment_type_distribution, fetch='df')

# –í–∏–∑—É–∞–ª–∏–∑–∞—Ü–∏—è (–∫—Ä—É–≥–æ–≤–∞—è –¥–∏–∞–≥—Ä–∞–º–º–∞)
fig_ptd = px.pie(payment_type_distribution, names='payment_type', values='order_count',
                 title='<b>–†–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ –ø–æ —Å–ø–æ—Å–æ–±–∞–º –æ–ø–ª–∞—Ç—ã</b>',
                 hole=0.35,
                 color_discrete_sequence=custom_palette)
fig_ptd.update_traces(textinfo='percent')
fig_ptd.write_image('1.png')
```

- –ö—Ä–µ–¥–∏—Ç–Ω—ã–µ –∫–∞—Ä—Ç—ã ‚Äî –¥–æ–º–∏–Ω–∏—Ä—É—é—â–∏–π —Å–ø–æ—Å–æ–± –æ–ø–ª–∞—Ç—ã (73‚ÄØ% –∑–∞–∫–∞–∑–æ–≤), —á—Ç–æ —É–∫–∞–∑—ã–≤–∞–µ—Ç –Ω–∞ –≤—ã—Å–æ–∫–∏–π —É—Ä–æ–≤–µ–Ω—å –¥–æ–≤–µ—Ä–∏—è –∏ –∑—Ä–µ–ª–æ—Å—Ç—å –∞—É–¥–∏—Ç–æ—Ä–∏–∏.

- Boleto –∑–∞–Ω–∏–º–∞–µ—Ç 19‚ÄØ% ‚Äî —Ç—Ä–∞–¥–∏—Ü–∏–æ–Ω–Ω—ã–π –æ—Ñ–ª–∞–π–Ω-–∏–Ω—Å—Ç—Ä—É–º–µ–Ω—Ç, –≤—Å—ë –µ—â—ë –≤–æ—Å—Ç—Ä–µ–±–æ–≤–∞–Ω —É —á–∞—Å—Ç–∏ –∫–ª–∏–µ–Ω—Ç–æ–≤ (–≤–æ–∑–º–æ–∂–Ω–æ, –∏–∑-–∑–∞ –Ω–µ–¥–æ–≤–µ—Ä–∏—è –∫ –æ–Ω–ª–∞–π–Ω-–æ–ø–ª–∞—Ç–µ –∏–ª–∏ –æ–≥—Ä–∞–Ω–∏—á–µ–Ω–Ω–æ–≥–æ –¥–æ—Å—Ç—É–ø–∞ –∫ –±–∞–Ω–∫–æ–≤—Å–∫–∏–º –ø—Ä–æ–¥—É–∫—Ç–∞–º).

- –í–∞—É—á–µ—Ä—ã –∏ –¥–µ–±–µ—Ç–æ–≤—ã–µ –∫–∞—Ä—Ç—ã –≤ —Å—É–º–º–µ —Å–æ—Å—Ç–∞–≤–ª—è—é—Ç –º–µ–Ω–µ–µ 7‚ÄØ%, –∏—Ö –Ω–∏–∑–∫–æ–µ –ø—Ä–æ–Ω–∏–∫–Ω–æ–≤–µ–Ω–∏–µ –º–æ–∂–µ—Ç —Å–≤–∏–¥–µ—Ç–µ–ª—å—Å—Ç–≤–æ–≤–∞—Ç—å –æ —Å–ª–∞–±–æ–π –∏–Ω—Ç–µ–≥—Ä–∞—Ü–∏–∏ –∞–ª—å—Ç–µ—Ä–Ω–∞—Ç–∏–≤–Ω—ã—Ö —Å–ø–æ—Å–æ–±–æ–≤ –∏–ª–∏ –æ –Ω–∏–∑–∫–æ–π –º–æ—Ç–∏–≤–∞—Ü–∏–∏ –∏—Å–ø–æ–ª—å–∑–æ–≤–∞—Ç—å –∏—Ö.

- **–ë–∏–∑–Ω–µ—Å-–∏–Ω—Å–∞–π—Ç:** –µ—Å—Ç—å –ø–æ—Ç–µ–Ω—Ü–∏–∞–ª –¥–ª—è –æ–ø—Ç–∏–º–∏–∑–∞—Ü–∏–∏ –ø–ª–∞—Ç—ë–∂–Ω–æ–π –≤–æ—Ä–æ–Ω–∫–∏ ‚Äî –Ω–∞–ø—Ä–∏–º–µ—Ä, —Å—Ç–∏–º—É–ª–∏—Ä–æ–≤–∞—Ç—å –ø—Ä–µ–¥–æ–ø–ª–∞—Ç—É –¥–µ–±–µ—Ç–æ–≤–æ–π –∫–∞—Ä—Ç–æ–π —á–µ—Ä–µ–∑ cashback –∏–ª–∏ —É—Å–∫–æ—Ä–µ–Ω–Ω—É—é –¥–æ—Å—Ç–∞–≤–∫—É.

### –ê–Ω–∞–ª–∏–∑ –æ–ø–ª–∞—Ç—ã –≤ —Ä–∞—Å—Å—Ä–æ—á–∫—É –∫—Ä–µ–¥–∏—Ç–Ω–æ–π –∫–∞—Ä—Ç–æ–π

- **–û–ø–∏—Å–∞–Ω–∏–µ –º–µ—Ç—Ä–∏–∫–∏:**
    - **–î–æ–ª—è –∑–∞–∫–∞–∑–æ–≤ –≤ —Ä–∞—Å—Å—Ä–æ—á–∫—É:** –ü—Ä–æ—Ü–µ–Ω—Ç –∑–∞–∫–∞–∑–æ–≤, –æ–ø–ª–∞—á–µ–Ω–Ω—ã—Ö –∫—Ä–µ–¥–∏—Ç–Ω–æ–π –∫–∞—Ä—Ç–æ–π —Å –∏—Å–ø–æ–ª—å–∑–æ–≤–∞–Ω–∏–µ–º —Ä–∞—Å—Å—Ä–æ—á–∫–∏  –æ—Ç –æ–±—â–µ–≥–æ —á–∏—Å–ª–∞ –∑–∞–∫–∞–∑–æ–≤, –æ–ø–ª–∞—á–µ–Ω–Ω—ã—Ö –∫—Ä–µ–¥–∏—Ç–Ω–æ–π –∫–∞—Ä—Ç–æ–π, –∏–ª–∏ –æ—Ç –≤—Å–µ—Ö –∑–∞–∫–∞–∑–æ–≤.
    - **–†–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ –ø–æ –∫–æ–ª–∏—á–µ—Å—Ç–≤—É –ø–ª–∞—Ç–µ–∂–µ–π:** –ö–∞–∫ —Ä–∞—Å–ø—Ä–µ–¥–µ–ª—è—é—Ç—Å—è –∑–∞–∫–∞–∑—ã –ø–æ —á–∏—Å–ª—É –µ–∂–µ–º–µ—Å—è—á–Ω—ã—Ö –ø–ª–∞—Ç–µ–∂–µ–π.

- **–ë–∏–∑–Ω–µ—Å-–∏–Ω—Ç–µ—Ä–ø—Ä–µ—Ç–∞—Ü–∏—è:**
    - **–í—ã—Å–æ–∫–∞—è –¥–æ–ª—è –∏ –±–æ–ª—å—à–æ–µ –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –ø–ª–∞—Ç–µ–∂–µ–π:** –£–∫–∞–∑—ã–≤–∞–µ—Ç –Ω–∞ —Ç–æ, —á—Ç–æ –æ–ø—Ü–∏—è —Ä–∞—Å—Å—Ä–æ—á–∫–∏ –≤–æ—Å—Ç—Ä–µ–±–æ–≤–∞–Ω–∞ –∏, –≤–æ–∑–º–æ–∂–Ω–æ, —Å—Ç–∏–º—É–ª–∏—Ä—É–µ—Ç –ø–æ–∫—É–ø–∫–∏ —Ç–æ–≤–∞—Ä–æ–≤ —Å –±–æ–ª–µ–µ –≤—ã—Å–æ–∫–∏–º —á–µ–∫–æ–º. –û–¥–Ω–∞–∫–æ, —ç—Ç–æ —Ç–∞–∫–∂–µ –º–æ–∂–µ—Ç –≤–ª–∏—è—Ç—å –Ω–∞ –¥–µ–Ω–µ–∂–Ω—ã–π –ø–æ—Ç–æ–∫ Olist (–µ—Å–ª–∏ Olist –Ω–µ –ø–æ–ª—É—á–∞–µ—Ç –≤—Å—é —Å—É–º–º—É —Å—Ä–∞–∑—É –æ—Ç –±–∞–Ω–∫–∞-—ç–∫–≤–∞–π–µ—Ä–∞).


```python
query_credit_card_installments = """
SELECT
    payment_installments,
    COUNT(*) AS order_count
FROM order_payments
WHERE payment_type = '–∫—Ä–µ–¥–∏—Ç–Ω–∞—è –∫–∞—Ä—Ç–∞' AND payment_installments > 0
GROUP BY payment_installments
ORDER BY payment_installments;
"""

credit_card_installments = run_query(engine, query_credit_card_installments, fetch='df')

# –†–∞—Å—á—ë—Ç —Å—Ä–µ–¥–Ω–µ–≥–æ –∫–æ–ª–∏—á–µ—Å—Ç–≤–∞ –ø–ª–∞—Ç–µ–∂–µ–π
mean_installments = credit_card_installments.apply(
    lambda row: row['payment_installments'] * row['order_count'], axis=1).sum() / credit_card_installments['order_count'].sum()

# –ü–æ—Å—Ç—Ä–æ–µ–Ω–∏–µ –≥—Ä–∞—Ñ–∏–∫–∞
fig_cci = px.bar(
    credit_card_installments,
    x='payment_installments',
    y='order_count',
    title='<b>–ö—Ä–µ–¥–∏—Ç–Ω–∞—è –∫–∞—Ä—Ç–∞: —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ –ø–æ –∫–æ–ª–∏—á–µ—Å—Ç–≤—É –ø–ª–∞—Ç–µ–∂–µ–π</b>',
    labels={
        'payment_installments': '–ö–æ–ª–∏—á–µ—Å—Ç–≤–æ –ø–ª–∞—Ç–µ–∂–µ–π',
        'order_count': '–ö–æ–ª–∏—á–µ—Å—Ç–≤–æ –∑–∞–∫–∞–∑–æ–≤'
    },
    color_discrete_sequence=custom_palette[4:5]
)

# –î–æ–±–∞–≤–ª–µ–Ω–∏–µ –≤–µ—Ä—Ç–∏–∫–∞–ª—å–Ω–æ–π –ª–∏–Ω–∏–∏ —Å—Ä–µ–¥–Ω–µ–≥–æ
fig_cci.add_shape(
    type='line',
    x0=mean_installments,
    x1=mean_installments,
    y0=0,
    y1=credit_card_installments['order_count'].max(),
    line=dict(color='red', width=2, dash='dash'),
)

# –î–æ–±–∞–≤–ª–µ–Ω–∏–µ –ø–æ–¥–ø–∏—Å–∏
fig_cci.add_annotation(
    x=mean_installments,
    y=credit_card_installments['order_count'].max(),
    text=f"–°—Ä–µ–¥–Ω–µ–µ: {mean_installments:.2f}",
    showarrow=True,
    arrowhead=1,
    ax=0,
    ay=-40
)
fig_cci.write_image('1.png')
```

**üìä –ò–Ω—Ç–µ—Ä–ø—Ä–µ—Ç–∞—Ü–∏—è —Ä–µ–∑—É–ª—å—Ç–∞—Ç–∞:**
- –†–∞—Å—Å—Ä–æ—á–∫–∞ –≤–∞—Ä—å–∏—Ä—É–µ—Ç—Å—è –æ—Ç 1 –¥–æ 25 –ø–ª–∞—Ç–µ–∂–µ–π, –Ω–æ —Å—Ä–µ–¥–Ω–µ–µ –∑–Ω–∞—á–µ–Ω–∏–µ 3.5.

- –≠—Ç–æ –≥–æ–≤–æ—Ä–∏—Ç –æ –ø—Ä–µ–æ–±–ª–∞–¥–∞–Ω–∏–∏ –∫–æ—Ä–æ—Ç–∫–∏—Ö –ø–ª–∞–Ω–æ–≤ —Ä–∞—Å—Å—Ä–æ—á–∫–∏, –∫–ª–∏–µ–Ω—Ç—ã –∏–∑–±–µ–≥–∞—é—Ç –¥–æ–ª–≥–æ–≤–æ–π –Ω–∞–≥—Ä—É–∑–∫–∏ –Ω–∞ –¥–ª–∏—Ç–µ–ª—å–Ω—ã–π —Å—Ä–æ–∫.

- –£–≤–µ–ª–∏—á–µ–Ω–∏–µ –¥–æ–ª–∏ –∑–∞–∫–∞–∑–æ–≤ —Å –≤—ã—Å–æ–∫–∏–º —á–∏—Å–ª–æ–º –ø–ª–∞—Ç–µ–∂–µ–π –º–æ–∂–µ—Ç —Å–∏–≥–Ω–∞–ª–∏–∑–∏—Ä–æ–≤–∞—Ç—å –æ —Ñ–∏–Ω–∞–Ω—Å–æ–≤–æ–º –Ω–∞–ø—Ä—è–∂–µ–Ω–∏–∏ —É –ø–æ–∫—É–ø–∞—Ç–µ–ª–µ–π.

- **–ë–∏–∑–Ω–µ—Å-–∏–Ω—Å–∞–π—Ç:** —Ü–µ–ª–µ—Å–æ–æ–±—Ä–∞–∑–Ω–æ –ø—Ä–µ–¥–ª–∞–≥–∞—Ç—å –≥–∏–±–∫–∏–µ —É—Å–ª–æ–≤–∏—è —Ä–∞—Å—Å—Ä–æ—á–∫–∏ –∏ A/B —Ç–µ—Å—Ç–∏—Ä–æ–≤–∞—Ç—å –≤–ª–∏—è–Ω–∏–µ —Ä–∞—Å—à–∏—Ä–µ–Ω–Ω—ã—Ö –ø–ª–∞–Ω–æ–≤ –Ω–∞ GMV –∏ –≤–æ–∑–≤—Ä–∞—Ç—ã.

### –ò–Ω–¥–µ–∫—Å –¥–∏–≤–µ—Ä—Å–∏—Ñ–∏–∫–∞—Ü–∏–∏ –∫–∞—Ç–µ–≥–æ—Ä–∏–π.

- **–û–ø–∏—Å–∞–Ω–∏–µ –º–µ—Ç—Ä–∏–∫–∏:**  
–ö–æ–ª–∏—á–µ—Å—Ç–≤–æ —É–Ω–∏–∫–∞–ª—å–Ω—ã—Ö —Ç–æ–≤–∞—Ä–Ω—ã—Ö –∫–∞—Ç–µ–≥–æ—Ä–∏–π, –∫–æ—Ç–æ—Ä—ã–µ –≤—ã—Å—Ç–∞–≤–ª—è–µ—Ç –Ω–∞ –ø—Ä–æ–¥–∞–∂—É –∫–ª–∏–µ–Ω—Ç.

- **–ë–∏–∑–Ω–µ—Å-–∏–Ω—Ç–µ—Ä–ø—Ä–µ—Ç–∞—Ü–∏—è:**  
    - 1 –∫–∞—Ç–µ–≥–æ—Ä–∏—è: –£–∑–∫–æ—Å–ø–µ—Ü–∏–∞–ª–∏–∑–∏—Ä–æ–≤–∞–Ω–Ω—ã–π —Å–ø—Ä–æ—Å (–Ω–∏—à–∞)  
    - 3-5 –∫–∞—Ç–µ–≥–æ—Ä–∏–π: –£–Ω–∏–≤–µ—Ä—Å–∞–ª—å–Ω—ã–π –ø—Ä–æ–¥–∞–≤–µ—Ü (–ø–µ—Ä–µ–∫—Ä–µ—Å—Ç–Ω—ã–µ –ø—Ä–æ–¥–∞–∂–∏)  

- **–î–æ–ø–æ–ª–Ω–∏—Ç–µ–ª—å–Ω—ã–µ –¥–µ—Ç–∞–ª–∏:**  
    - –í–∑–≤–µ—à–µ–Ω–Ω–∞—è –≤–µ—Ä—Å–∏—è: `SUM(category_gmv_share * ln(category_gmv_share))`  

```python
query = """
WITH category_gmv AS (
  SELECT
    s.seller_id,
    p.category_name_translated,
    SUM(oi.price + oi.freight_value) AS category_gmv
  FROM order_items oi
  JOIN products p ON oi.product_id = p.product_id
  JOIN sellers s ON oi.seller_id = s.seller_id
  GROUP BY 1, 2
),
category_share AS (
  SELECT
    seller_id,
    category_name_translated,
    category_gmv,
    category_gmv * 1.0 / SUM(category_gmv) OVER (PARTITION BY seller_id) AS gmv_share
  FROM category_gmv
)
SELECT
  seller_id,
  COUNT(DISTINCT category_name_translated) AS category_count,
  ROUND(SUM(gmv_share * LN(gmv_share))::numeric * -1, 3) AS cdi_weighted
FROM category_share
GROUP BY seller_id
ORDER BY cdi_weighted DESC;

"""
seller_cdi_df = run_query(engine, query, fetch='df')
fig = px.histogram(
    seller_cdi_df,
    x="category_count",
    nbins=20,
    title="–†–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ –ø—Ä–æ–¥–∞–≤—Ü–æ–≤ –ø–æ —á–∏—Å–ª—É –∫–∞—Ç–µ–≥–æ—Ä–∏–π",
    labels={"category_count": "–ö–æ–ª–∏—á–µ—Å—Ç–≤–æ —É–Ω–∏–∫–∞–ª—å–Ω—ã—Ö –∫–∞—Ç–µ–≥–æ—Ä–∏–π"}
)
fig.update_layout(bargap=0.2)
fig.write_image('1.png')
```

**üìä –ò–Ω—Ç–µ—Ä–ø—Ä–µ—Ç–∞—Ü–∏—è —Ä–µ–∑—É–ª—å—Ç–∞—Ç–∞:**
- –ë–æ–ª—å—à–∏–Ω—Å—Ç–≤–æ –ø—Ä–æ–¥–∞–≤—Ü–æ–≤ –æ–≥—Ä–∞–Ω–∏—á–∏–≤–∞—é—Ç—Å—è –æ–¥–Ω–æ–π —Ç–æ–≤–∞—Ä–Ω–æ–π –∫–∞—Ç–µ–≥–æ—Ä–∏–µ–π, —á—Ç–æ —É–∫–∞–∑—ã–≤–∞–µ—Ç –Ω–∞ —Å–ø–µ—Ü–∏–∞–ª–∏–∑–∞—Ü–∏—é.

- –û–¥–Ω–∞–∫–æ —Å—É—â–µ—Å—Ç–≤—É–µ—Ç –¥–ª–∏–Ω–Ω—ã–π —Ö–≤–æ—Å—Ç –ø—Ä–æ–¥–∞–≤—Ü–æ–≤, –ø–æ–∫—Ä—ã–≤–∞—é—â–∏—Ö –¥–æ 26 –∫–∞—Ç–µ–≥–æ—Ä–∏–π, —á—Ç–æ –º–æ–∂–µ—Ç –±—ã—Ç—å –∏–Ω–¥–∏–∫–∞—Ç–æ—Ä–æ–º –º—É–ª—å—Ç–∏–±—Ä–µ–Ω–¥–æ–≤–æ—Å—Ç–∏, –¥—Ä–æ–ø—à–∏–ø–ø–∏–Ω–≥–∞ –∏–ª–∏ –≤—ã—Å–æ–∫–æ–π –æ–ø–µ—Ä–∞—Ü–∏–æ–Ω–Ω–æ–π –∑—Ä–µ–ª–æ—Å—Ç–∏.

- **–ë–∏–∑–Ω–µ—Å-–∏–Ω—Å–∞–π—Ç:** —Å—Ç–æ–∏—Ç –≤—ã–¥–µ–ª–∏—Ç—å –≥—Ä—É–ø–ø—É –º—É–ª—å—Ç–∏–∫–∞—Ç–µ–≥–æ—Ä–∏–π–Ω—ã—Ö –ø—Ä–æ–¥–∞–≤—Ü–æ–≤ –∫–∞–∫ —É—Å—Ç–æ–π—á–∏–≤—É—é –∫ —Å–µ–∑–æ–Ω–Ω–æ—Å—Ç–∏ –∏ –∏—Å–ø–æ–ª—å–∑–æ–≤–∞—Ç—å –≤ –∫–∞—á–µ—Å—Ç–≤–µ –±–µ–Ω—á–º–∞—Ä–∫–∞ –ø–æ —Ä–∞–∑–≤–∏—Ç–∏—é –∞—Å—Å–æ—Ä—Ç–∏–º–µ–Ω—Ç–∞.

###  –ò–Ω–¥–µ–∫—Å —Å—Ç–∞–±–∏–ª—å–Ω–æ—Å—Ç–∏ –ø—Ä–æ–¥–∞–≤—Ü–æ–≤ (Seller Stability Index)

**–û–ø–∏—Å–∞–Ω–∏–µ:**

–ú–µ—Ç—Ä–∏–∫–∞ –ø–æ–∑–≤–æ–ª—è–µ—Ç —Å–µ–≥–º–µ–Ω—Ç–∏—Ä–æ–≤–∞—Ç—å –ø—Ä–æ–¥–∞–≤—Ü–æ–≤ –ø–æ —Å—Ç–∞–±–∏–ª—å–Ω–æ—Å—Ç–∏ –∏ –æ–±—ä—ë–º—É –∏—Ö –∞–∫—Ç–∏–≤–Ω–æ—Å—Ç–∏.  
–ò–Ω–¥–µ–∫—Å —Ä–∞—Å—Å—á–∏—Ç—ã–≤–∞–µ—Ç—Å—è –∫–∞–∫ —Ñ—É–Ω–∫—Ü–∏—è —Å—Ä–µ–¥–Ω–µ–π –∞–∫—Ç–∏–≤–Ω–æ—Å—Ç–∏ (`avg_orders`) –∏ –≤–æ–ª–∞—Ç–∏–ª—å–Ω–æ—Å—Ç–∏ (`stddev_orders`) –ø–æ –º–µ—Å—è—Ü–∞–º:  

> **Stability Index = (–ö–æ–ª–∏—á–µ—Å—Ç–≤–æ –∞–∫—Ç–∏–≤–Ω—ã—Ö –º–µ—Å—è—Ü–µ–≤ √ó –°—Ä–µ–¥–Ω–µ–µ —á–∏—Å–ª–æ –∑–∞–∫–∞–∑–æ–≤) / (–°—Ç–∞–Ω–¥–∞—Ä—Ç–Ω–æ–µ –æ—Ç–∫–ª–æ–Ω–µ–Ω–∏–µ + 1)**

–≠—Ç–æ –ø–æ–∑–≤–æ–ª—è–µ—Ç —Ä–∞–Ω–∂–∏—Ä–æ–≤–∞—Ç—å –ø—Ä–æ–¥–∞–≤—Ü–æ–≤ –ø–æ —Å—Ç–µ–ø–µ–Ω–∏ —É—Å—Ç–æ–π—á–∏–≤–æ—Å—Ç–∏: —á–µ–º –≤—ã—à–µ –∏–Ω–¥–µ–∫—Å, —Ç–µ–º —Ä–µ–≥—É–ª—è—Ä–Ω–µ–µ –∏ –ø—Ä–µ–¥—Å–∫–∞–∑—É–µ–º–µ–µ –ø—Ä–æ–¥–∞–∂–∏.

**–ü–æ—è—Å–Ω–µ–Ω–∏–µ –∫ —Ä–∞—Å—á—ë—Ç—É:**

- –ò—Å–ø–æ–ª—å–∑—É—é—Ç—Å—è —Ç–æ–ª—å–∫–æ —É—Å–ø–µ—à–Ω–æ –¥–æ—Å—Ç–∞–≤–ª–µ–Ω–Ω—ã–µ –∑–∞–∫–∞–∑—ã (`order_status = '–¥–æ—Å—Ç–∞–≤–ª–µ–Ω'`).
- –ü—Ä–æ–¥–∞–≤–µ—Ü —Å—á–∏—Ç–∞–µ—Ç—Å—è –∞–∫—Ç–∏–≤–Ω—ã–º, –µ—Å–ª–∏ —É –Ω–µ–≥–æ –º–∏–Ω–∏–º—É–º **3 –∞–∫—Ç–∏–≤–Ω—ã—Ö –º–µ—Å—è—Ü–∞** –∏ –≤ —Å—Ä–µ–¥–Ω–µ–º **‚â• 5 –∑–∞–∫–∞–∑–æ–≤ –≤ –º–µ—Å—è—Ü**.
- –î–ª—è –∫–∞–∂–¥–æ–≥–æ –ø—Ä–æ–¥–∞–≤—Ü–∞ —Ç–∞–∫–∂–µ –æ–ø—Ä–µ–¥–µ–ª—è–µ—Ç—Å—è –µ–≥–æ –æ—Å–Ω–æ–≤–Ω–∞—è —Ç–æ–≤–∞—Ä–Ω–∞—è –∫–∞—Ç–µ–≥–æ—Ä–∏—è (`top_category`).
- –í —Ñ–∏–Ω–∞–ª—å–Ω–æ–º —à–∞–≥–µ –ø—Ä–æ–¥–∞–≤—Ü—ã —Å–µ–≥–º–µ–Ω—Ç–∏—Ä—É—é—Ç—Å—è –Ω–∞ 4 –≥—Ä—É–ø–ø—ã –Ω–∞ –æ—Å–Ω–æ–≤–µ —Å—Ä–∞–≤–Ω–µ–Ω–∏–π —Å–æ —Å—Ä–µ–¥–Ω–∏–º–∏/–º–µ–¥–∏–∞–Ω–Ω—ã–º–∏ –∑–Ω–∞—á–µ–Ω–∏—è–º–∏:
    - **–õ–∏–¥–µ—Ä—ã** ‚Äî –≤—ã—Å–æ–∫–∞—è —Å—Ç–∞–±–∏–ª—å–Ω–æ—Å—Ç—å –∏ –≤—ã—Å–æ–∫–∏–π –æ–±—ä—ë–º.
    - **–°—Ç–∞–±–∏–ª—å–Ω—ã–µ –º–∞–ª—ã–µ** ‚Äî —Å—Ç–∞–±–∏–ª—å–Ω—ã, –Ω–æ —Å –Ω–µ–≤—ã—Å–æ–∫–∏–º –æ–±—ä—ë–º–æ–º.
    - **–†–∏—Å–∫–æ–≤—ã–µ –∫—Ä—É–ø–Ω—ã–µ** ‚Äî –æ–±—ä—ë–º –≤—ã—Å–æ–∫, –Ω–æ –ø—Ä–æ–¥–∞–∂–∏ –Ω–µ—Å—Ç–∞–±–∏–ª—å–Ω—ã.
    - **–ù–∏–∑–∫–∏–π –ø—Ä–∏–æ—Ä–∏—Ç–µ—Ç** ‚Äî –Ω–µ—Å—Ç–∞–±–∏–ª—å–Ω—ã –∏ –ø—Ä–æ–¥–∞—é—Ç –º–∞–ª–æ.

**–ü—Ä–∏–º–µ–Ω–µ–Ω–∏–µ:**

- –ü–æ–º–æ–≥–∞–µ—Ç –≤—ã—è–≤–∏—Ç—å –Ω–∞–∏–±–æ–ª–µ–µ –Ω–∞–¥—ë–∂–Ω—ã—Ö –ø–∞—Ä—Ç–Ω—ë—Ä–æ–≤ –¥–ª—è —Ä–∞–∑–≤–∏—Ç–∏—è –ø—Ä–æ–≥—Ä–∞–º–º—ã –ª–æ—è–ª—å–Ω–æ—Å—Ç–∏ –∏–ª–∏ –ø–æ–¥–¥–µ—Ä–∂–∫–∏.
- –ò—Å–ø–æ–ª—å–∑—É–µ—Ç—Å—è –¥–ª—è —Ç–∞—Ä–≥–µ—Ç–∏—Ä–æ–≤–∞–Ω–∏—è: –∫–∞–∫–∏–µ –ø—Ä–æ–¥–∞–≤—Ü—ã –Ω—É–∂–¥–∞—é—Ç—Å—è –≤ —Å–æ–ø—Ä–æ–≤–æ–∂–¥–µ–Ω–∏–∏ –∏–ª–∏ —Ä–µ–∞–Ω–∏–º–∞—Ü–∏–∏.

```python
query = """
WITH seller_monthly_activity AS (
  SELECT
    oi.seller_id,
    DATE_TRUNC('month', o.order_purchase_timestamp) AS month,
    COUNT(DISTINCT oi.order_id) AS orders_count,
    SUM(op.payment_value) AS monthly_gmv
  FROM order_items oi
  JOIN orders o USING(order_id)
  JOIN order_payments op ON o.order_id = op.order_id
  WHERE o.order_status = '–¥–æ—Å—Ç–∞–≤–ª–µ–Ω'
  GROUP BY 1, 2
),
seller_top_category AS (
  SELECT 
    oi.seller_id,
    p.category_name_grouped AS top_category,
    COUNT(*) AS category_count,
    ROW_NUMBER() OVER(PARTITION BY oi.seller_id ORDER BY COUNT(*) DESC) AS category_rank
  FROM order_items oi
  JOIN products p USING(product_id)
  GROUP BY 1, 2
),
seller_stats AS (
  SELECT
    sma.seller_id,
    COUNT(DISTINCT sma.month) AS active_months,
    COALESCE(STDDEV(sma.orders_count), 0) AS volatility,
    COALESCE(AVG(sma.orders_count), 0) AS avg_orders,
    SUM(sma.monthly_gmv) AS total_gmv,
    AVG(sma.monthly_gmv) AS avg_monthly_gmv,
    (COUNT(DISTINCT sma.month) * COALESCE(AVG(sma.orders_count), 0)) 
      / (COALESCE(STDDEV(sma.orders_count), 0) + 1) AS stability_index,
    stc.top_category
  FROM seller_monthly_activity sma
  LEFT JOIN seller_top_category stc 
    ON sma.seller_id = stc.seller_id
    AND stc.category_rank = 1
  GROUP BY 1, stc.top_category
  HAVING COUNT(DISTINCT sma.month) >= 1 
    AND AVG(sma.orders_count) >= 1
),
medians AS (
  SELECT
    AVG(stability_index) as mean_stability,
    AVG(avg_orders) as mean_orders,
    PERCENTILE_CONT(0.5) WITHIN GROUP (ORDER BY stability_index) AS median_stability,
    PERCENTILE_CONT(0.5) WITHIN GROUP (ORDER BY avg_orders) AS median_orders
  FROM seller_stats
)
SELECT
  s.seller_id,
  s.active_months::int,
  s.avg_orders::float,
  s.stability_index::float,
  s.total_gmv::float,
  s.avg_monthly_gmv::float,
  s.top_category,
  CASE
    WHEN s.stability_index >= m.median_stability AND s.avg_orders >= m.mean_orders THEN '–õ–∏–¥–µ—Ä—ã'
    WHEN s.stability_index >= m.median_stability AND s.avg_orders < m.mean_orders THEN '–°—Ç–∞–±–∏–ª—å–Ω—ã–µ –º–∞–ª—ã–µ'
    WHEN s.stability_index < m.median_stability AND s.avg_orders >= m.mean_orders THEN '–†–∏—Å–∫–æ–≤—ã–µ –∫—Ä—É–ø–Ω—ã–µ'
    ELSE '–ù–∏–∑–∫–∏–π –ø—Ä–∏–æ—Ä–∏—Ç–µ—Ç'
  END AS quad_category
FROM seller_stats s
CROSS JOIN medians m;
"""

df_sellers = run_query(engine, query, fetch='df')
scatter_quadrant_plot(
    df=df_sellers,
    x_col='avg_orders',
    y_col='stability_index',
    x_mean=df_sellers['avg_orders'].mean(),
    y_mean=df_sellers['stability_index'].median(),
    category_col='top_category',
    size_col='total_gmv',
    title='–ê–Ω–∞–ª–∏–∑ –ø—Ä–æ–¥–∞–≤—Ü–æ–≤: –∑–∞–∫–∞–∑—ã vs —Å—Ç–∞–±–∏–ª—å–Ω–æ—Å—Ç—å (—Ä–∞–∑–º–µ—Ä –ø–æ GMV)',
    color='quad_category',
    color_palette=custom_palette,
    labels={
        'avg_orders': '–°—Ä–µ–¥–Ω–µ–µ –∫–æ–ª-–≤–æ –∑–∞–∫–∞–∑–æ–≤/–º–µ—Å',
        'stability_index': '–ò–Ω–¥–µ–∫—Å —Å—Ç–∞–±–∏–ª—å–Ω–æ—Å—Ç–∏',
        'total_gmv': '–û–±—â–∏–π GMV',
        'top_category': '–ö–∞—Ç–µ–≥–æ—Ä–∏—è'
    },
)
fig.write_image('1.png')
```

- –†–∞–∑–¥–µ–ª–∏–ª–∏ –ø—Ä–æ–¥–∞–≤—Ü–æ–≤ –Ω–∞ 4 –≥—Ä—É–ø–ø—ã –ø–æ –º–µ–¥–∏–∞–Ω–Ω—ã–º –ø–æ–∫–∞–∑–∞—Ç–µ–ª—è–º —Å—Ç–∞–±–∏–ª—å–Ω–æ—Å—Ç–∏ –∏ –∫–æ–ª–∏—á–µ—Å—Ç–≤–∞ –∑–∞–∫–∞–∑–æ–≤.

```python
# –ì—Ä—É–ø–ø–∏—Ä—É–µ–º –∏ —Å–æ—Ä—Ç–∏—Ä—É–µ–º
bar_data = (
    df_sellers.groupby('quad_category', as_index=False)
    .agg({'seller_id': 'count'})
    .sort_values('seller_id', ascending=False)
)

# –ë–∞—Ä—á–∞—Ä—Ç –ø–æ –∫–æ–ª–∏—á–µ—Å—Ç–≤—É –ø—Ä–æ–¥–∞–≤—Ü–æ–≤ –≤ –∫–∞–∂–¥–æ–π –≥—Ä—É–ø–ø–µ 
fig_bar = px.bar(
    bar_data,
    x='quad_category',
    y='seller_id',
    color='quad_category',
    color_discrete_sequence=custom_palette,
    title='–†–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ –ø—Ä–æ–¥–∞–≤—Ü–æ–≤ –ø–æ –≥—Ä—É–ø–ø–∞–º',
    labels={'seller_id': '–ö–æ–ª–∏—á–µ—Å—Ç–≤–æ –ø—Ä–æ–¥–∞–≤—Ü–æ–≤', 'quad_category': '–ì—Ä—É–ø–ø–∞ –ø—Ä–æ–¥–∞–≤—Ü–æ–≤'},
    text='seller_id',
    category_orders={'quad_category': bar_data['quad_category'].tolist()}  # –°–æ—Ö—Ä–∞–Ω—è–µ–º –ø–æ—Ä—è–¥–æ–∫ —Å–æ—Ä—Ç–∏—Ä–æ–≤–∫–∏
)

fig_bar.update_traces(
    texttemplate='%{y}',
    textposition='outside',
    hovertemplate='<b>%{x}</b><br>–ö–æ–ª–∏—á–µ—Å—Ç–≤–æ: %{y}'
)

fig_bar.update_layout(
    showlegend=False,
    yaxis_title='–ö–æ–ª–∏—á–µ—Å—Ç–≤–æ –ø—Ä–æ–¥–∞–≤—Ü–æ–≤',
    xaxis_title=None,
    xaxis={'categoryorder': 'total descending'}  # –î–æ–ø–æ–ª–Ω–∏—Ç–µ–ª—å–Ω–∞—è —Å–æ—Ä—Ç–∏—Ä–æ–≤–∫–∞ –ø–æ –æ—Å–∏ X
)

fig_bar.write_image('1.png')

# –û—Å—Ç–∞–ª—å–Ω—ã–µ –≥—Ä–∞—Ñ–∏–∫–∏ –æ—Å—Ç–∞—é—Ç—Å—è –±–µ–∑ –∏–∑–º–µ–Ω–µ–Ω–∏–π
df_treemap = df_sellers.groupby(['top_category', 'quad_category']).agg(
    total_sellers=('seller_id', 'count'),
    total_gmv=('total_gmv', 'sum'),
    avg_stability=('stability_index', 'mean'),
    avg_monthly_gmv=('avg_monthly_gmv', 'mean')
).reset_index()

fig = px.treemap(
    df_treemap,
    path=[px.Constant("–í—Å–µ –ø—Ä–æ–¥–∞–≤—Ü—ã"), 'quad_category', 'top_category'],
    values='total_gmv',
    color='avg_stability',
    color_continuous_scale='RdYlGn',
    hover_data=['total_sellers', 'avg_monthly_gmv'],
    title='–†–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ –ø—Ä–æ–¥–∞–≤—Ü–æ–≤ –ø–æ GMV –∏ —Å—Ç–∞–±–∏–ª—å–Ω–æ—Å—Ç–∏',
    labels={
        'total_sellers': '–ö–æ–ª-–≤–æ –ø—Ä–æ–¥–∞–≤—Ü–æ–≤',
        'total_gmv': '–°—É–º–º–∞—Ä–Ω—ã–π GMV',
        'avg_stability': '–°—Ä–µ–¥–Ω—è—è —Å—Ç–∞–±–∏–ª—å–Ω–æ—Å—Ç—å',
        'avg_monthly_gmv': '–°—Ä. GMV/–º–µ—Å',
        'quad_category': '–ö–≤–∞–¥—Ä–∞–Ω—Ç'
    }
)

fig.update_layout(
    margin=dict(t=50, l=25, r=25, b=25),
    coloraxis_colorbar=dict(
        title="–°—Ç–∞–±–∏–ª—å–Ω–æ—Å—Ç—å",
        thickness=20,
        tickvals=[df_treemap['avg_stability'].min(), df_treemap['avg_stability'].max()],
        ticktext=['–ù–∏–∑–∫–∞—è', '–í—ã—Å–æ–∫–∞—è']
    )
)

fig.update_traces(
    texttemplate='%{label}<br>%{value:,.0f} GMV',
    hovertemplate='<b>%{label}</b><br>'
                 '–°—É–º–º–∞—Ä–Ω—ã–π GMV: %{value:,.0f}<br>'
                 '–ü—Ä–æ–¥–∞–≤—Ü–æ–≤: %{customdata[0]}<br>'
                 '–°—Ä. —Å—Ç–∞–±–∏–ª—å–Ω–æ—Å—Ç—å: %{color:.2f}<br>'
                 '–°—Ä. GMV/–º–µ—Å: %{customdata[1]:,.0f}'
)

fig.write_image('1.png')
```

- –õ–∏–¥–µ—Ä—ã —Å–æ—Å—Ç–∞–≤–ª—è—é—Ç –≤—Å–µ–≥–æ 621 –ø—Ä–æ–¥–∞–≤—Ü–∞, –Ω–æ –ø—Ä–∏–Ω–æ—Å—è—Ç –Ω–∞–∏–±–æ–ª—å—à—É—é –¥–æ–ª—é GMV.
