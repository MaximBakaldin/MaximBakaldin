#  **Olist data preprocessing**

## Введение.

В этом разделе мы производим системную подготовку данных из публичного e-commerce датасета **Brazilian Olist**. Цель — подготовка чистых, структурированных таблиц для загрузки в PostgreSQL (через Supabase) и последующего аналитического использования.

**Ключевые задачи:**

- предварительное знакомство со структурой данных и взаимосвязями таблиц;
- стандартизация наименований и типов признаков;
- первичная очистка: пропуски, дубликаты, аномалии;
- первичный исследовательский анализ данных;
- идентификация первичных (`Primary Key`) и потенциальных внешних ключей (`Foreign Key`);
- валидация связей между таблицами (soft-check на уровне pandas);
- загрузка данных в в СУБД.

## Описание данных.

Исходный датасет включает **9 таблиц**:

| Таблица                    | Назначение |
|----------------------------|------------|
| `orders`                  | Основные сведения о заказах |
| `order_items`             | Позиции товаров в заказах |
| `order_reviews`           | Отзывы по заказам |
| `order_payments`          | Платежи по заказам |
| `customers`               | Покупатели |
| `sellers`                 | Продавцы |
| `products`                | Каталог товаров |
| `geolocation`             | Географические координаты |
| `product_category_name_translation` | Переводы категорий товаров |
| `closed_deals` | Таблица маркетинговых сделок, связанная с продавцами |
| `product_category_name_translation` | Лиды из маркетинга, потенциально связанные через `mql_id` |

Мы рассматриваем датасет с точки зрения построения **аналитической витрины**: наша задача — не только сохранить полезные данные, но и понять бизнес-логику связей.

- pd.DataFrame с результатами анализа пропусков, если return_df=True, иначе None.
