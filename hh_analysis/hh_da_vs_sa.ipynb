# Анализ вакансий аналитиков на hh.ru

## Описание проблемы:
Рынок труда в IT сфере динамически меняется, продавцы курсов зазывают обещаниями и ставшими уже крылатыми фразами «Войти в IT за 6 месяцев! Зарплата 300 тысяч и удалёнка ждут Вас!» Начинающим специалистам и тем, кто хочет сменить работу, достаточно сложно сориентироваться и найти ответы на насущные вопросы:
- Какой спрос на специалистов различного уровня?
- Сколько зарабатывают аналитики данных и системные аналитики?
- Какими ключевыми навыками следует обзавестись соискателю, чтобы соответствовать требованиям работодателя?
- Какие условия могут предложить работодатели?
В этом проекте мы попытаемся самостоятельно разобраться и найти ответы на эти вопросы.
Проанализируем и выявим различия в предлагаемых вакансиях для Аналитиков данных и системных аналитиков.

**Цель проекта**: Выявить различия в предлагаемых вакансиях для Аналитиков данных и Системных аналитиков.

**Откуда данные для этого исследования?**

Мы получим всю необходимую информацию посредством парсинга сайта hh.ru.

Поля датафрейма после выполнения парсинга должны выглядеть следующим образом:

- `id` — Уникальный идентификатор вакансии.

- `name` — Название вакансии.

- `published_at` — Дата публикации.

- `alternate_url` — Ссылка на вакансию.

- `type` — Статус вакансии на момент получения данных от api и передачи
их в базу.

- `employer` — Работодатель.

- `department` — Работодатель, отдел.

- `area` — Регион места работы.

- `experience` — Требуемый опыт работы.

- `schedule` — График работы.

- `employment` — Тип занятости.

- `salary_from` — Нижняя граница предлагаемой заработной платы.

### Парсинг данных hh.ru

Всю информацию о получении данных по API можно найти в [документации](https://github.com/hhru/api). Зададим функции используя доступ по API и библиотеку [BeautifulSoup](https://pypi.org/project/beautifulsoup4/#description) для текста описания вакансий.


```python
# Зададим функции для парсинга данных
'''
В ДАННОЙ ЯЧЕЙКЕ ИСПОЛЬЗУЕМ «ДЕМО-ВЕРСИЮ» ПОЛУЧИВШЕЙСЯ ФУНКЦИИ ДЛЯ СБОРКИ ДАТАСЕТА,
АТРИБУТЫ ФУНКЦИИ ИЗМЕНЕНЫ С ЦЕЛЬЮ УМЕНЬШЕНИЯ ЗАДЕРЖКИ ВЫВОДА 
ФУНКЦИЯ ВЫВОДИТ ПЕРВУЮ СТРАНИЦУ ДАННЫХ ПО ВАКАНСИЯМ ЗА ВЧЕРАШНЮЮ ДАТУ
'''
BASE_URL = "https://api.hh.ru/vacancies"
HEADERS = {"HH-User-Agent": "my_"}


# Зададим функцию для определения диапазоны дат за которые мы хотим получить данные
def get_vacancies_by_date_range(position, date_from, date_to, max_pages=1):
    """
    Получение вакансий за указанный диапазон дат.
    """
    vacancies = []
    for page in range(max_pages):
        params = {
            "text": position,
            "page": page,
            "per_page": 20,  # Максимальное количество вакансий на странице
            "date_from": date_from,
            "date_to": date_to,
        }

        response = requests.get(BASE_URL, headers=HEADERS, params=params)
        
        if response.status_code != 200:
            print(f"Ошибка запроса на странице {page} с кодом {response.status_code}")
            break

        data = response.json()
        vacancies.extend(data["items"])  # Добавляем вакансии из текущей страницы

        # Если достигнут конец доступных страниц, выходим из цикла
        if page >= data.get("pages", 1) - 1:
            break

        # Пауза, чтобы избежать ограничения по количеству запросов к API HH.ru
        time.sleep(1)

    return vacancies
# 
def get_vacancies_data(position, start_date="2024-11-15", end_date="2024-11-24"):
    """
    Основная функция для получения вакансий с разбивкой по диапазонам дат.
    """
    step = timedelta(weeks=1)
    current_date = datetime.strptime(start_date, "%Y-%m-%d")
    end_date = datetime.strptime(end_date, "%Y-%m-%d")
    
    all_vacancies = []
    
    while current_date < end_date:
        next_date = min(current_date + step, end_date)
        date_from = current_date.strftime("%Y-%m-%d")
        date_to = next_date.strftime("%Y-%m-%d")
        
        vacancies = get_vacancies_by_date_range(position, date_from, date_to)
        all_vacancies.extend(vacancies)
        
        current_date = next_date

    # Преобразуем в формат, соответствующий вашей структуре
    relevant_keywords = [
        "аналитик данных", "data analyst", "системный аналитик", 
        "system analyst", "bi-analyst", "дата аналитик", "бизнес/системный аналитик"
    ]

    parsed_vacancies = []
    # Прогресс-бар для обработки вакансий
    with tqdm(total=len(all_vacancies), desc="Обработка вакансий") as pbar:
        for item in all_vacancies:
            # Фильтрация по ключевым словам в названии вакансии
            if not any(keyword.lower() in item["name"].lower() for keyword in relevant_keywords):
                continue

            salary = item.get("salary")
            vacancy_data = {
                "id": item.get("id"),
                "name": item.get("name"),
                "published_at": item.get("published_at"),
                "alternate_url": item.get("alternate_url"),
                "type": item.get("type"),
                "employer": item.get("employer", {}).get("name"),
                "department": item.get("department", {}).get("name") if item.get("department") else None,
                "area": item.get("area", {}).get("name"),
                "experience": item.get("experience", {}).get("name"),
                "schedule": item.get("schedule", {}).get("name"),
                "employment": item.get("employment", {}).get("name"),
                "salary_from": salary["from"] if salary else None,
                "salary_to": salary["to"] if salary else None,
            }
            parsed_vacancies.append(vacancy_data)
            pbar.update(1)

    return pd.DataFrame(parsed_vacancies)

def enrich_vacancies_data(df):
    """
    Дополняем вакансии описанием и ключевыми навыками.
    """
    def description(vacancy_id):
        url = f'https://api.hh.ru/vacancies/{vacancy_id}'
        response = requests.get(url, headers=HEADERS)
        if response.status_code == 200:
            description_html = response.json().get('description', '')
            return BeautifulSoup(description_html, "html.parser").get_text().lower()
        else:
            print(f"Ошибка при получении описания для вакансии ID: {vacancy_id}")
            return None

    tqdm.pandas(desc="Загрузка описаний вакансий")
    df["description"] = df["id"].progress_apply(description)

    def get_vacancy_skills(vacancy_id):
        url = f'https://api.hh.ru/vacancies/{vacancy_id}'
        response = requests.get(url, headers=HEADERS)
        if response.status_code == 200:
            data = response.json()
            skills = [skill['name'] for skill in data.get('key_skills', [])]
            return ', '.join(skills) if skills else None
        return None

    tqdm.pandas(desc="Загрузка ключевых навыков")
    df["key_skills"] = df["id"].progress_apply(get_vacancy_skills)

    return df

# Сбор данных
# Зададим значение текущей даты для демо-режима функции
today = datetime.today().strftime('%Y-%m-%d')
yesterday = (datetime.today() - timedelta(days=1)).strftime('%Y-%m-%d')

df_analyst = get_vacancies_data(position="Аналитик данных", start_date=yesterday, end_date=today)
df_system_analyst = get_vacancies_data(position="Системный аналитик", start_date=yesterday, end_date=today)

# Объединение результатов
df = pd.concat([df_analyst, df_system_analyst], ignore_index=True)

# Обогащение данных
df = enrich_vacancies_data(df)

# Вывод и сохранение результатов
print(f"Общее количество вакансий на первой странице за {yesterday}: {len(df)}")
df.head()
```

#### Выполним лемматизацию текста в поле `description`
С целью упрощения анализа воспользуемся библиотекой [pymorphy2](https://github.com/pymorphy2/pymorphy2) для лемматизации и нормализации текста  в поле `description` и создадим поле `description_lemmatized`

```python
# Инициализация морфологического анализатора
morph = pymorphy2.MorphAnalyzer()

# Функция для лемматизации текста с удалением знаков препинания, но с сохранением дефиса и "/"
def lemmatize_text(text):
    if not isinstance(text, str):  # Проверка, что текст является строкой
        return ''  # Возвращаем пустую строку для не строковых значений
    # Удаляем все знаки препинания, кроме дефиса и "/"
    text = re.sub(r'[^\w\s\-\/]', '', text)  # Экранируем дефис и косую черту
    # Разделяем текст на слова
    words = text.split()
    # Лемматизируем каждое слово
    lemmatized_words = [morph.parse(word)[0].normal_form for word in words]
    # Собираем слова обратно в строку
    return ' '.join(lemmatized_words)

# Применяем лемматизацию с прогресс-баром
tqdm.pandas(desc="Лемматизация текста")
df['description_lemmatized'] = df['description'].progress_apply(lemmatize_text)

# Печать результата для проверки
df[['description', 'description_lemmatized']].head(5)
```

Мы привели текст в поле описания вакансий к его лемматизированному что позволило:
1. Улучшить качество поиска и анализа текста

2. Повысить точность классификации текста

3. Улучшить работы методов, основанных на частотах слов (поиск уникальных навыков в описаниях вакансий)
При подсчёте частоты слов в текстах (например, в анализе TF-IDF) разные формы одного и того же слова будут рассматриваться как одно слово, что делает результаты более интерпретируемыми.
*Операция произведена на демо-версии датафрейма с целью снижения ожидания. Загрузим заранее обработанный датасет и продолжим работу.*

Получим список уникальных навыков из поля key_skills, для этого преобразуем строку в список сохраним его в отдельную переменную и выведем на экран.

```python
# Зададим функцию для получения списка уникальных навыков
def parse_skills(skills):
    try:
        # Преобразуем строку в список, если она не пуста
        skills_list = eval(skills)
        return skills_list if isinstance(skills_list, list) else []
    except:
        return []

# Применяем функцию 
df['key_skills_list'] = df['key_skills'].apply(parse_skills)

# Используем set для сбора уникальных значений
unique_skills = set()
for skills_list in df['key_skills_list']:
    unique_skills.update(skills_list)

# Преобразуем set в отсортированный список
unique_skills = sorted(unique_skills)

# Выводим список уникальных скиллов
print(f" Количество уникальных навыков в поле key_skills: {len(unique_skills)}")
```

```python
# Для поиска навыков в поле description
all_skills_pattern = re.compile(r"\b(" + "|".join(re.escape(skill) for skill in unique_skills) + r")\b", re.IGNORECASE)

# Функция для поиска навыков в описании вакансии
def find_skills_in_description(description):
    if description:
        found_skills = all_skills_pattern.findall(description)
        return list(set(found_skills))  # Возвращаем уникальные найденные навыки
    return []
# Подключаем tqdm для pandas
tqdm.pandas(desc="Поиск навыков в описании")
# Применяем функцию для поиска навыков в поле 'description_lemmatized' и создаем столбец 'regex_key_skills_lem'
df['regex_key_skills_lem'] = df['description_lemmatized'].progress_apply(find_skills_in_description)
```

#### Очистим поле key_skills от лишних знаков препинания

```python
# Функция для обработки и преобразования списка навыков
def process_skills(skills):
    if isinstance(skills, list):  # Если это список
        # Разворачиваем вложенные списки и приводим к нижнему регистру
        flat_skills = [skill.strip().lower() for sublist in skills for skill in (sublist if isinstance(sublist, list) else [sublist])]
        # Возвращаем уникальные навыки, разделенные запятыми
        return ', '.join(sorted(set(flat_skills)))
    elif isinstance(skills, str):  # Если это строка
        # Сохраняем "1с" как есть, убираем все лишние символы, кроме букв, запятых и дефисов
        skills = re.sub(r'[^a-zа-я1с:, -]', '', skills.lower())  # Убираем лишние символы, но оставляем "1с", ":"
        
        # Обрабатываем случай с "1с" и его вариантами, например, "1с программирование"
        skills = re.sub(r'\b1с [a-zа-я -]+\b', '1с', skills)  # Заменяем "1с" + слово/слова на "1с"
        
        # Используем регулярное выражение для обрезки всего после двоеточия (и до первой запятой или конца строки)
        skills = re.sub(r'1с:[^,]*', '1с', skills)  # Заменяем "1с:" и всё, что после, на "1с"
        
        return skills
    return np.nan  # Если данных нет, возвращаем NaN

# Применяем функцию к столбцу key_skills
df['key_skills_clear'] = df['key_skills'].apply(process_skills)
# Проверяем результат
df[['key_skills', 'key_skills_clear']].sample(5)
```

##### Воспользуемся библиотекой sklearn для кластеризации навыков  методом TF-IDF.
Принцип метода
- TF-IDF: Преобразует текстовые данные (навыки) в числовое представление, учитывая важность слова относительно других (реже встречающиеся слова получают больший вес).


- MiniBatchKMeans: Алгоритм делит данные на заданное число кластеров (n_clusters) по схожести, измеряемой метрикой (например, косинусное расстояние).Использует небольшие мини-пакеты данных для повышения производительности на больших наборах.


- Результат: Навыки группируются в кластеры, где навыки внутри одного кластера схожи по смыслу. Это позволяет создать более структурированный и унифицированный набор данных.

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.cluster import MiniBatchKMeans
from tqdm import tqdm
import pandas as pd

# Разделяем данные на отдельные скиллы
skills_list = df_copy['key_skills_clear'].str.split(',').explode().str.strip().unique()

# Преобразуем скиллы в TF-IDF представление
vectorizer = TfidfVectorizer(stop_words='english')
X = vectorizer.fit_transform(skills_list)

# Параметры кластеризации
num_clusters = 1850
batch_size = 50

# Инициализируем MiniBatchKMeans
kmeans = MiniBatchKMeans(n_clusters=num_clusters,
                         random_state=42,
                         batch_size=batch_size,
                         max_iter=500,
                         n_init=20)

# Создаем прогресс-бар
with tqdm(total=10, desc="KMeans кластеризация", unit="итерация") as pbar:
    for _ in range(10): 
        kmeans.partial_fit(X)  
        pbar.update(1)  
        
# Получаем метки кластеров
clusters = kmeans.labels_

# Создаем DataFrame с результатами
clustered_skills = pd.DataFrame({'skill': skills_list, 'cluster': clusters})

# Группируем скиллы по кластеру
clustered_skills_sorted = clustered_skills.groupby('cluster')['skill'].apply(lambda x: ', '.join(x)).reset_index()

# Подсчитаем количество элементов в каждом кластере
cluster_sizes = (
    clustered_skills.groupby('cluster')['skill']
    .apply(len)
    .reset_index(name='size')
)

# Объединяем информацию о кластерах и их размерах
clustered_skills_with_sizes = clustered_skills.merge(cluster_sizes, on='cluster')

# Сортируем кластеры по размеру в порядке убывания
sorted_clusters = (
    clustered_skills_with_sizes.groupby('cluster')
    .agg({'size': 'first', 'skill': lambda x: ', '.join(x)})
    .reset_index()
    .sort_values(by='size', ascending=False)
)

# Получаем первые 20 самых крупных кластеров
top_20_clusters = sorted_clusters.head(20)

# Выводим результат
for _, row in top_20_clusters.iterrows():
    print(f"Кластер {row['cluster']} (размер {row['size']}):")
    print(f"Элементы: {row['skill']}\n")
```

##### Удалим навыки не имеющие какого либо смысла.

```python
# Список навыков, которые нужно удалить
trash_skills = ['it', 'us', 'c', '', ',', 'с', 'б', 'н', '-п', '1-и', '-у', 'g']

# Зададим функцию для удаления трешовых навыков
def remove_trash_skills(skill_string, trash_skills):
    # Разделяем строку на отдельные навыки
    skills = skill_string.split(',')
    # Фильтруем навыки, исключая те, что есть в списке 
    filtered_skills = [skill.strip() for skill in skills if skill.strip() not in trash_skills]
    # Объединяем оставшиеся навыки обратно в строку
    return ', '.join(filtered_skills)

# Применяем функцию к столбцу навыков
df_copy['key_skills_clear'] = df_copy['key_skills_clear'].apply(lambda x: remove_trash_skills(x, trash_skills))
```

##### Заменим навыки для каждого кластера на их стандартизированное значение

```python
# Зададим функцию, которая выполнит замену навыков нашего датафрейма на их стандартизованное значеие
def standardize_skills(df, skills_column, clustered_skills, new_column_name='key_skills_standardized'):
    """
    Заменяет навыки в на стандартизированные названия из кластеров.
     df: DataFrame с данными
     skills_column: Название столбца с навыками
     clustered_skills: DataFrame с кластеризованными навыками (столбцы: 'skill' и 'cluster')
     new_column_name: Название нового столбца для очищенных навыков
    
    """
    # Находим самое длинное название навыка в каждом кластере
    cluster_to_standard_skill = (
        clustered_skills.groupby('cluster')['skill']
        .apply(lambda skills: min(skills, key=len))  # Выбираем самое длинное название
        .to_dict()
    )

    # Создаем словарь соответствий между всеми навыками и их стандартизированным названием
    skill_to_standard_skill = {
        skill: cluster_to_standard_skill[cluster]
        for cluster, skills in clustered_skills.groupby('cluster')['skill']
        for skill in skills
    }

    # Функция для замены навыков в одной строке
    def replace_skills(skill_string):
        skills = skill_string.split(',')  # Разделяем навыки
        standardized_skills = [
            skill_to_standard_skill.get(skill.strip(), skill.strip())  # Заменяем на стандартизированные
            for skill in skills
        ]
        return ', '.join(sorted(set(standardized_skills)))  # Объединяем обратно и убираем дубли

    # Создаем новое поле с очищенными навыками
    df[new_column_name] = df[skills_column].apply(replace_skills)

    # Сравниваем количество уникальных навыков
    original_skills = df[skills_column].str.split(',').explode().str.strip().unique()
    standardized_skills = df[new_column_name].str.split(',').explode().str.strip().unique()

    print(f"Количество уникальных навыков до стандартизации: {len(original_skills)}")
    print(f"Количество уникальных навыков после стандартизации: {len(standardized_skills)}")

    return df

# Применяем функцию к DataFrame
df_copy = standardize_skills(df_copy, 'key_skills_clear', clustered_skills)

# Проверяем результат
df_copy[['key_skills_clear', 'key_skills_standardized']].head()
```

#### Получим списки самых популярных навыков для каждой профессии.

```python
# Загрузим расширение для подсчёта
from collections import Counter
# Разделяем навыки с учетом нескольких возможных разделителей: запятая, точка с запятой, пробел
df_copy['key_skills_standardized'] = df_copy['key_skills_standardized'].apply(
    lambda x: re.split(r'[,\|;]', x) if isinstance(x, str) else x
)

# Убираем лишние пробелы и приводим все навыки к нижнему регистру
df_copy['key_skills_standardized'] = df_copy['key_skills_standardized'].apply(
    lambda skills: [skill.strip().lower() for skill in skills if skill.strip()]
)
# Функция для подсчета частоты навыков по профессии
def get_top_skills_by_profession(df, profession_column, skills_column, top_n=300):
    # Группируем данные по профессии
    profession_groups = df.groupby(profession_column)
    
    # Списки для хранения данных по навыкам и их частоте для da и sa
    da_skills_list = []
    count_da_list = []
    sa_skills_list = []
    count_sa_list = []
    
    # Для каждой профессии считаем частоту навыков
    for profession, group in profession_groups:
        # Объединяем списки навыков для подсчета частоты
        all_skills = [skill for skills_list in group[skills_column] for skill in skills_list]
        
        # Подсчитываем частоту каждого навыка
        skills_counter = Counter(all_skills)
        
        # Получаем топ N самых частых навыков
        top_skills = skills_counter.most_common(top_n)
        
        # Проверяем, если профессия соответствует da или sa
        if 'аналитик данных' in profession.lower():  # Профессия da
            for skill, count in top_skills:
                da_skills_list.append(skill)
                count_da_list.append(count)

        
        # Профессия System Analyst
        if 'системный аналитик' in profession.lower():  # Профессия sa
            for skill, count in top_skills:
                sa_skills_list.append(skill)
                count_sa_list.append(count)

    
    # Создаем DataFrame с результатами
    result_df = pd.DataFrame({
        'da_skills': da_skills_list,
        'count_da': count_da_list,
        'sa_skills': sa_skills_list,
        'count_sa': count_sa_list
    })
    
    return result_df

# Получаем топ-200 навыков для da и sa
top_skills_by_profession_df = get_top_skills_by_profession(df_copy, 'profession', 'key_skills_standardized', top_n=300)

# Выводим результат
display(top_skills_by_profession_df.head(10))
print(f"Список популярных навыков da: {top_skills_by_profession_df.da_skills.unique()[:20]}")

print(f"Список популярных навыков sa: {top_skills_by_profession_df.sa_skills.unique()[:20]}")
```

```python
# Зададим функцию для группировки полей
def prepare_data(data, grouping_field):
    grouped = (
        data.groupby(['experience', grouping_field])['id']
        .count()
        .reset_index(name='count')
    )
    grouped['fraction'] = grouped.groupby('experience')['count'].apply(lambda x: x / x.sum())
    # Сортируем по убыванию доли
    grouped = grouped.sort_values(by=['experience','fraction'], ascending=[True, False])
    return grouped

# Зададим функцию для построения графика
def plot_distribution(data, title, ax, hue='education_level',add_legend=False , legend_title=False):
    sns.barplot(
        data=data,
        x='experience',
        y='fraction',
        hue=hue,
        palette=custom_palette, 
        ax=ax
    )
    # Добавляем текст с долями над столбцами
    for p in ax.patches:
        height = p.get_height()
        if height > 0:
            ax.text(
                p.get_x() + p.get_width() / 2,
                height,
                f'{height:.2f}',
                ha='center',
                va='bottom',
                fontsize=10
            )
    # Настройка осей и заголовков
    ax.set_title(title, fontsize=14, pad=10)
    ax.set_xlabel('Опыт работы', fontsize=12)
    ax.set_ylabel('Доля вакансий', fontsize=12)
    ax.tick_params(axis='x', labelsize=12)
    ax.tick_params(axis='y', labelsize=12)
    sns.despine()
    
    
    # Убираем легенду
    axes[0].legend().set_visible(False)
    
    # Добавляем легенду, если указан параметр add_legend
    if add_legend:
        ax.legend(
            title=legend_title, 
            fontsize=12,
            loc='upper center', 
            bbox_to_anchor=(0.5, -0.15),  # Позиция легенды под графиками (можно корректировать)
            ncol=4, 
            frameon=False
        )
        
# Подготавливаем данные для SA и DA
sa_distribution = prepare_data(sa_data, 'education_level')
da_distribution = prepare_data(da_data, 'education_level')

# Создаем два графика
fig, axes = plt.subplots(2, 1, figsize=(16, 12))

# График для системных аналитиков
plot_distribution(
    sa_distribution,
    'Системные аналитики: распределение требуемого образования по опыту',
    axes[0],
    add_legend=False,
    hue='education_level',
    legend_title='Образование'
)

# График для аналитиков данных
plot_distribution(
    da_distribution,
    'Аналитики данных: распределение требуемого образования по опыту',
    axes[1],
    hue='education_level',
    add_legend=True,
    legend_title='Уровень образования'
)

# Общий заголовок
fig.suptitle('Распределение вакансий по образованию и опыту\n(Системные аналитики vs Аналитики данных)', 
             fontsize=16, fontweight='bold')

#Изменяем расстояние между графиками
plt.subplots_adjust(hspace=0.3) 
# Отображение графиков
plt.show()
```

```python
# Зададим фунцкцию для подготовки данных
def prepare_salary_data_by_profession(data, profession, field='area',
                                      salary_from='salary_from',
                                      salary_to='salary_to', top_n=10):
    """
    Подготавливает данные о зарплатах для заданной профессии в топ-N городах.
    """
    # Фильтруем данные по профессии
    filtered_data = data[data['profession'] == profession]

    # Определяем топ-N городов по количеству вакансий
    top_cities = filtered_data[field].value_counts().head(top_n).index

    # Фильтруем данные по топ-городам
    top_data = filtered_data[filtered_data[field].isin(top_cities)]

    # Считаем медианные зарплаты
    salary_data = top_data.groupby(field).agg({
        salary_from: 'median',
        salary_to: 'median'
    }).reset_index()

    # Сортируем по медианной зарплате "от" (возрастание)
    return salary_data.sort_values(by=salary_from, ascending=True).reset_index(drop=True)

# Подготовка данных для DA и SA
salary_da = prepare_salary_data_by_profession(df_copy, profession='Аналитик данных')
salary_sa = prepare_salary_data_by_profession(df_copy, profession='Системный аналитик')

# Построение графика
fig, ax = plt.subplots(figsize=(14, 10))
bar_height = 0.30  # Высота одного бара
y_positions = np.arange(len(salary_da)) 
shift = bar_height + 0.005  
'#403d39', '#eb5e28'
# «аналитики данных»
ax.barh(
    y=y_positions,
    width=salary_da['salary_to'] - salary_da['salary_from'],
    left=salary_da['salary_from'],
    color=da_color,
    height=bar_height,
    label='Аналитик данных'
)
for i, row in salary_da.iterrows():
    ax.text(
        x=row['salary_to'] + 1000,
        y=i,
        s=f"{int(row['salary_from'])} - {int(row['salary_to'])}",
        va='center',
        fontsize=10
    )

# «системные аналитики»
ax.barh(
    y=y_positions - shift,
    width=salary_sa['salary_to'] - salary_sa['salary_from'],
    left=salary_sa['salary_from'],
    color=sa_color,
    height=bar_height,
    label='Системный аналитик'
)
for i, row in salary_sa.iterrows():
    ax.text(
        x=row['salary_to'] + 1000,
        y=i - shift,
        s=f"{int(row['salary_from'])} - {int(row['salary_to'])}",
        va='center',
        fontsize=10
    )

# Настройки графика
ax.set_yticks(y_positions - shift / 2)
ax.set_yticklabels(salary_da['area'], fontsize=12)
ax.set_xlabel('Зарплата (руб.)', fontsize=12)
ax.set_ylabel('Города', fontsize=12)
ax.set_title('Зарплатный диапазон \n Системные аналитики vs Аналитики данных', fontsize=16, fontweight='bold', pad=10)

# Перемещение легенды под график
ax.legend(
    title='', 
    fontsize=12, 
    loc='upper center', 
    bbox_to_anchor=(0.5, -0.05),  
    ncol=2,  
    frameon=False
)

sns.despine()
# Отобразим график
plt.show()
```

### Какие работодатели чаще ищут аналитиков?

```python
# Создадим градиентные цветовые карты вручную.

from matplotlib.colors import LinearSegmentedColormap

# Функция для создания градиентной палитры от белого к заданному цвету
def create_gradient_palette(start_color, end_color="white", num_colors=256):
    """
    Создает градиентную палитру от белого к начальной цветовой точке.
    """
    return LinearSegmentedColormap.from_list("custom_gradient", [end_color, start_color], N=num_colors)

# Создаем кастомные палитры с нужным направлением градиента
sa_cmap = create_gradient_palette(sa_color)  # Градиент от белого к красному (для системных аналитиков)
da_cmap = create_gradient_palette(da_color)  # Градиент от белого к зеленому (для аналитиков данных)

def prepare_heatmap_data(data, profession, top_n=10):
    # Определяем топ-N работодателей для конкретной профессии
    top_employers = (
        data[data['profession'] == profession]['employer']
        .value_counts()
        .nlargest(top_n)
        .index
    )

    # Фильтруем данные для тепловой карты
    heatmap_data = (
        data[(data['profession'] == profession) & (data['employer'].isin(top_employers))]
        .groupby(['employer', 'experience'])['id']
        .count()
        .reset_index()
        .pivot(index='employer', columns='experience', values='id')
        .fillna(0)
    )

    # Считаем общую сумму вакансий для каждой компании
    heatmap_data['total'] = heatmap_data.sum(axis=1)

    # Сортируем компании по убыванию общего количества вакансий
    heatmap_data = heatmap_data.sort_values(by='total', ascending=False)

    # Убираем столбец total (он не нужен для отображения)
    heatmap_data = heatmap_data.drop(columns='total')

    # Рассчитываем проценты от общего числа вакансий для каждой компании
    heatmap_data = heatmap_data.div(heatmap_data.sum(axis=1), axis=0) * 100

    return heatmap_data

# Подготовка данных для тепловой карты
sa_heatmap = prepare_heatmap_data(df_filtered, 'Системный аналитик')
da_heatmap = prepare_heatmap_data(df_filtered, 'Аналитик данных')

# Построение тепловых карт
fig, axes = plt.subplots(2, 1, figsize=(16, 16), sharey=False)

# Тепловая карта для системных аналитиков
sns.heatmap(
    da_heatmap,
    annot=True,
    fmt=".0f",
    cmap=da_cmap,  # Используем градиент от белого к красному
    ax=axes[0],
    cbar=False,
    linewidths=0.5,
    annot_kws={"fontsize": 12},  # Увеличиваем шрифт значений внутри ячеек
)

axes[1].set_title(
    "Топ-10 работодателей для системных аналитиков (распределение по опыту в %)",
    fontsize=16,
    pad=20,
    fontweight="bold",
)
axes[0].set_xlabel("Опыт работы", fontsize=14)
axes[1].set_ylabel("", fontsize=14)
axes[1].tick_params(axis="both", which="major", labelsize=12)

# Тепловая карта для аналитиков данных
sns.heatmap(
    sa_heatmap,
    annot=True,
    fmt=".0f",
    cmap=sa_cmap,  # Используем градиент от белого к зеленому
    ax=axes[1],
    cbar=False,
    linewidths=0.5,
    annot_kws={"fontsize": 12},  # Увеличиваем шрифт значений внутри ячеек
)

axes[0].set_title(
    "Топ-10 работодателей для аналитиков данных (распределение по грейдам в %)",
    fontsize=16,
    pad=20,
    fontweight="bold",
)
axes[1].set_xlabel("Опыт работы", fontsize=14)
axes[0].set_ylabel("", fontsize=14)
axes[0].tick_params(axis="both", which="major", labelsize=12)

plt.tight_layout()
plt.show()
```

```python
# Функция для подготовки данных и построения тепловой карты медианных значений зарплаты
def prepare_and_plot_salary_heatmap(data, profession, top_n=10, cmap=None):
    # Фильтруем данные, удаляя строки, где нет информации о зарплате
    data = data.dropna(subset=['salary_from', 'salary_to'], how='all')
    
    # Определяем топ-N работодателей по количеству вакансий
    top_employers = data[data['profession'] == profession]['employer'].value_counts().nlargest(top_n).index
    
    # Создание нового столбца для зарплаты, используя salary_from или salary_to
    data['salary'] = data['salary_from'].combine_first(data['salary_to'])

    # Данные по зарплатам
    salary_data = (
        data[(data['profession'] == profession) & (data['employer'].isin(top_employers))]
        .groupby(['employer', 'experience'])['salary']
        .median()  # Используем медиану для нового столбца salary
        .reset_index()
        .pivot(index='employer', columns='experience', values='salary')
        .replace(0, np.nan)  # Заменяем 0 на NaN
    )
    
    # Сортировка по топ-N работодателям по количеству вакансий
    salary_data = salary_data.loc[top_employers]

    # Построение тепловой карты
    sns.heatmap(
        salary_data, annot=True, fmt=".0f", cmap=cmap, cbar=False, linewidths=0.5,
        annot_kws={"fontsize": 12}, mask=salary_data.isna(), ax=plt.gca()
    )
    plt.title(f"Топ-{top_n} работодателей для {profession} (медианные зарплаты по опыту)", fontsize=16, fontweight="bold")
    plt.xlabel("Опыт работы", fontsize=14)
    plt.ylabel("", fontsize=14)
    plt.tick_params(axis="both", which="major", labelsize=12)

# Настройка палитры
sa_cmap = create_gradient_palette(sa_color)
da_cmap = create_gradient_palette(da_color)

# Построение графиков для обоих типов аналитиков
fig, axes = plt.subplots(2, 1, figsize=(16, 16), sharey=False)

plt.subplot(212)
prepare_and_plot_salary_heatmap(df_filtered.query('experience !="Неизвестно"'), 'Системный аналитик', cmap=sa_cmap)

plt.subplot(211)
prepare_and_plot_salary_heatmap(df_filtered.query('experience !="Неизвестно"'), 'Аналитик данных', cmap=da_cmap)

plt.tight_layout()
plt.show()
```

### Какие навыки самые важные?

```python
# Зададим функцию для расчета количества навыков и формирования таблиц с ТОП навыками.
def format_top_skills_with_counts(data, skill_column):
    """
    Создаёт таблицу навыков и их количества по уровням опыта.
    В результате каждый уровень опыта представлен колонками: skill, count.
    """
    # Фильтруем данные: оставляем только те строки, где указаны навыки
    data_filtered = data[data[skill_column] != 'не указаны'].copy()
    data_filtered['skills'] = data_filtered[skill_column].str.split(',')  # Разделяем навыки
    data_exploded = data_filtered.explode('skills')  # Разворачиваем список навыков
    data_exploded['skills'] = data_exploded['skills'].str.strip()  # Убираем лишние пробелы

    # Группировка по уровню опыта и навыку
    skill_counts = (
        data_exploded.groupby(['experience', 'skills'])
        .size()
        .reset_index(name='count')
        .sort_values(by=['experience', 'count'], ascending=[True, False])
    )

    # Выбираем топ-10 навыков для каждого уровня опыта
    top_skills = (
        skill_counts.groupby('experience')
        .apply(lambda x: x.nlargest(15, 'count'))
        .reset_index(drop=True)
    )

    # Создаём таблицу с колонками "skill, count" для каждого уровня опыта
    result = pd.DataFrame()
    for experience_level in top_skills['experience'].unique():
        subset = top_skills[top_skills['experience'] == experience_level]
        subset = subset[['skills', 'count']].reset_index(drop=True)
        subset.columns = [experience_level, f"{experience_level}_count"]
        result = pd.concat([result, subset], axis=1)  # Объединяем по колонкам

    # Переименуем колонки в нужный формат
    rename_mapping = {
        "Junior (no experience)": "Junior skill name",
        "Junior (no experience)_count": "Junior count",
        "Junior+ (1-3 years)": "Junior+ skill name",
        "Junior+ (1-3 years)_count": "Junior+ count",
        "Middle (3-6 years)": "Middle skill name",
        "Middle (3-6 years)_count": "Middle count",
        "Senior (6+ years)": "Senior skill name",
        "Senior (6+ years)_count": "Senior count"
    }

    # Применяем переименование
    result.columns = [
        rename_mapping.get(col, col) for col in result.columns
    ]

    return result


# Создаём таблицы для каждой комбинации профессии и типа навыков
sa_hard_skills_table = format_top_skills_with_counts(sa_data, 'hard_skills')
sa_soft_skills_table = format_top_skills_with_counts(sa_data, 'soft_skills')
da_hard_skills_table = format_top_skills_with_counts(da_data, 'hard_skills')
da_soft_skills_table = format_top_skills_with_counts(da_data, 'soft_skills')
```

```python
# Зададим функцию для стилизации итоговых таблиц.
def style_table_with_bars(df, title, color='#4A90E2', column_width='200px', bar_color='#4A90E2'):
    """
    Форматирует таблицу и добавляет стили, включая визуальные бары для числовых значений.
    """
    # Сбросить индекс и сохранить таблицу для отображения
    df = df.reset_index(drop=True)

    # Преобразуем столбцы с количеством в числовой тип
    numeric_columns = [col for col in df.columns if 'count' in col]
    df[numeric_columns] = df[numeric_columns].apply(pd.to_numeric, errors='coerce')

    # Настроить стиль
    styled_table = df.style.set_properties(
        **{
            'font-size': '10pt',        # Размер шрифта
            'font-family': 'Arial',     # Шрифт
            'width': column_width       # Ширина столбцов
        }
    ).set_caption(title).set_table_styles(
        [
            # Стили для заголовка таблицы
            {
                'selector': 'caption',
                'props': [
                    ('text-align', 'center'),        # Центрируем заголовок
                    ('font-size', '12pt'),           # Размер шрифта заголовка
                    ('font-weight', 'bold'),         # Жирный шрифт заголовка
                    ('font-family', 'Arial'),        # Шрифт заголовка
                    ('color', 'white'),              # Цвет шрифта заголовка
                    ('background-color', color),     # Цвет фона заголовка
                    ('padding', '6px')              # Отступы вокруг заголовка
                ]
            },
            # Стили для заголовков столбцов
            {
                'selector': 'th',
                'props': [
                    ('font-size', '10pt'),            # Размер шрифта для заголовков столбцов
                    ('font-weight', 'bold'),          # Толщина шрифта для заголовков
                    ('font-family', 'Arial'),         # Шрифт заголовков
                    ('border', 'none'),              # Убираем границу для столбцов
                    ('padding', '8px'),              # Отступы вокруг текста в заголовках столбцов
                    ('width', column_width)          # Устанавливаем фиксированную ширину столбцов
                ]
            },
            # Стили для строк данных
            {
                'selector': 'td',
                'props': [
                    ('padding', '8px 4px'),         # Увеличиваем отступ между строками
                    ('border', 'none'),            # Убираем границу для ячеек данных
                    ('width', column_width)        # Устанавливаем фиксированную ширину столбцов
                ]
            }
        ]
    )

    # Применяем выравнивание для нечетных и четных столбцов
    for i, col in enumerate(df.columns):
        if i % 2 != 0:
            styled_table = styled_table.set_properties(subset=[col], **{'text-align': 'left'})
        else:
            styled_table = styled_table.set_properties(subset=[col], **{'text-align': 'right'})

    # Добавляем бары в числовые колонки
    styled_table = styled_table.bar(
        subset=numeric_columns,         # Применяем только к столбцам с количеством
        color=bar_color,                # Цвет полос
        align='mid'                     # Полосы начинаются с нуля
    )

    return styled_table

# Пример использования функции:
da_hard_styled = style_table_with_bars(da_hard_skills_table, 
                                       title="Топ хард скиллы для аналитика данных", 
                                       color=da_color, 
                                       bar_color=da_color)
da_soft_styled = style_table_with_bars(da_soft_skills_table, 
                                       title="Топ софт скиллы для аналитика данных", 
                                       color=da_color, 
                                       bar_color=da_color)

# Применим функцию к данным по ДА
display(da_hard_styled)
display(da_soft_styled)
```
