Определение факторов влияющих на рыночную стоимость объектов недвижимости, для построения ситемы отслеживания аномалий и мошеннической деятельности.

- **last_price** — цена на момент снятия с публикации
- **living_area** — жилая площадь в квадратных метрах (м²)

Заменим значения размера кухни для студий на нулевоё, проверим наличие зависимости размера кухни от общей площади и выполним замену пропусков на групповое медианное значение по общей площади квартиры, в виду наличия стандартных проектных решений.

```python
# Установим размер кухни для студий равный нулю
df.loc[df['studio'] == True, 'kitchen_area'] = 0 
```

```python
# Выполним замену на групповое медианное значение
# Создадим функцию которая возвращает пропуски, если для данного значения нет соответсвующей групппы
def safe_median(series):
    if series.dropna().empty:
        return np.nan
    else:
        return series.median()
df['kitchen_area'] = df.groupby('total_area')['kitchen_area'].transform(lambda x: x.fillna(safe_median(x)))
```

Аномальные значения, также известные как выбросы или аномалии, существенно отличаются от основной массы данных и могут искажать результаты анализа. Они могут возникать из-за ошибок ввода, технических сбоев, неправильной обработки данных или даже реальных отклонений от нормы. Такие значения могут негативно повлиять на статистический анализ и машинное обучение, поэтому важно их своевременно выявлять и корректировать.

Для фильтрации данных от выбросов на данном этапе мы можем использовать следующие методы:
- 3QR (трёхквантильный диапазон);
- 1,5QR (полутораквантильный размах);
- визуально определить наличие выбросов на графике, исходя из контекста каждой переменной.

Так как данный метод основан на предположении о нормальном распределении значений в выборке, мы можем использовать его для тех переменных в которых распределение будет нормальным.
Если данные имеют тяжёлое хвостовое распределение или содержат большое количество выбросов, метод 1.5IQR может быть предпочтительнее. Этот метод более устойчив к выбросам, чем метод 3QR, и может лучше работать с данными, которые не следуют строгому нормальному распределению.


```python
# Создадим функцию для построения диаграммы размаха и гистограммы с возможностью задать границы построения гистограммы
# Для лучшего восприятия будем использовать логорифмическую шкалу.
def plot_box_and_hist(dataframe, column, hist_xlim=None):
# создадим два подграфика ax_box и ax_hist
# кроме того, укажем, что нам нужны:
    fig, (ax_box, ax_hist) = plt.subplots(2, figsize=(10 , 4),# две строки в сетке подграфиков,
                                        sharex=True, gridspec_kw={'height_ratios': (.15, .85)})# единая шкала по оси x и
    # затем создадим графики, указав через параметр ax, в какой подграфик поместить каждый из них
    sns.boxplot(x=dataframe[column], ax=ax_box)
    sns.histplot(x=dataframe[column], ax=ax_hist, bins=100)
    
    ax_box.set(xlabel='')
    ax_hist.set(xlabel=column)
    plt.yscale('log')
    ax_hist.set(ylabel='count')
    
    if hist_xlim:
        ax_hist.set_xlim(hist_xlim)
    
    plt.show()
```

```python
# Cоздадим функцию для среза и построения графиков
def slice_and_plot(df, column_name, lower_bound=None, upper_bound=None, hist_xlim=None):
    # Копируем DataFrame, чтобы не менять оригинальный
    result_df = df.copy()

    # Проверяем, является ли столбец числовым
    if pd.api.types.is_numeric_dtype(result_df[column_name]):
        # Применяем пользовательские границы среза, если они заданы
        if lower_bound is not None:
            result_df = result_df[result_df[column_name] >= lower_bound]
        if upper_bound is not None:
            result_df = result_df[result_df[column_name] <= upper_bound]
        # Если не заданы применять правило 1.5*IQR
        else:
            # Вычисляем квартили
            Q1 = result_df[column_name].quantile(0.25)
            Q3 = result_df[column_name].quantile(0.75)
            # Вычисляем интерквартильный размах
            IQR = Q3 - Q1
            # Определяем границы выбросов
            lower_bound = Q1 - 1.5 * IQR
            upper_bound = Q3 + 1.5 * IQR
            # Выводим значения границ выбросов на экран
            print("Lower bound for outliers:", lower_bound)
            print("Upper bound for outliers:", upper_bound)
            # Удаляем значения, находящиеся за пределами границ выбросов
            result_df = result_df[(result_df[column_name] >= lower_bound) & (result_df[column_name] <= upper_bound)]
            
        # Строим графики
        fig, (ax_box, ax_hist) = plt.subplots(2, figsize=(10 , 4),
                                        sharex=True, gridspec_kw={'height_ratios': (.15, .85)})
        sns.boxplot(x=result_df[column_name], ax=ax_box)
        sns.histplot(x=result_df[column_name], ax=ax_hist, bins=100)
        ax_box.set(xlabel='')
        ax_hist.set(xlabel=column_name)
        plt.yscale('log')
        ax_hist.set(ylabel='count')
        if hist_xlim:
            ax_hist.set_xlim(hist_xlim)
        plt.show()
        
    else:
        print("Столбец не является числовым.")

 #   return result_df
```

### Создадим срез данных без выбросов.

Определение выбросов и аномальных значений необходимо для того что бы отбросить лишнюю инорфмацию, которая будет являться помехой для построения модели. 
Избавившись от явных аномальных значений, таких как высота потолков более 6 метров или размер квартиры меньше 5 м2, мы улучшили наши данные, очень редкие значения находящиеся на краю гистограмм, как правило снижают точность нашего исследования.
Дальнейшие наши шаги по определению интересующего среза и выбора методов для отбрасывания выбросов, очень зависят от следующих факторов:

- Цель исследования: определение влияния различных факторов на стоимость жилья. Построение системы для выявления аномалий.

- Тип данных: Для некоторых типов данных, таких как непрерывные или нормально распределённые данные, 1,5QR может дать более точные результаты. 

- Контекст исследования: В зависимости от контекста исследования, один из методов может быть предпочтительнее другого. 

В зависимости от того какая информация нас интересует, хотим ли мы обработать информацию по бюджетному жилью или  по жилью высшего класса с высокой стоимостью, будет определять, то какие значения мы будем отбрасывать считая за выбросы.  Также не маловажно иметь ввиду репресентативность данных. Если данных по дорогому жилью слишком мало информация для прогнозирования будет очень не надёжной.


Создадим функцию, для применения среза согласно правилу 1.5*IQR межквартильного размаха и применим к интересующим нас столбцам

```python
def slice_and_dice(df, columns=None):
    # Если столбцы не указаны, обрабатываем все столбцы
    if columns is None:
        columns = df.columns

    # Инициализируем маску для всех строк
    mask = pd.Series([True] * len(df), index=df.index)

    # Проходим по каждому столбцу
    for column in columns:
        # Проверяем, что столбец числовой и не булев
        if pd.api.types.is_numeric_dtype(df[column]) and not pd.api.types.is_bool_dtype(df[column]):
            # Преобразуем данные в числовой тип, заменяя некорректные значения на NaN
            df[column] = pd.to_numeric(df[column], errors='coerce')
            
            # Удаляем строки с NaN перед вычислением квартилей
            non_na_column = df[column].dropna()
            
            if not non_na_column.empty:
                # Вычисляем квартили
                Q1 = non_na_column.quantile(0.25)
                Q3 = non_na_column.quantile(0.75)
                
                # Вычисляем интерквартильный размах
                IQR = Q3 - Q1
                
                # Определяем границы выбросов
                lower_bound = Q1 - 1.5 * IQR
                upper_bound = Q3 + 1.5 * IQR
                
                # Обновляем маску для текущего столбца
                column_mask = (df[column] >= lower_bound) & (df[column] <= upper_bound)
                mask &= column_mask
            else:
                # Если столбец полностью пустой после удаления NaN, пропускаем его
                continue

    # Применяем итоговую маску ко всему DataFrame
    result_df = df[mask]

    return result_df
```

```python
# объединим столбцы в один датафрейф с методом 1.5 межквартильного размаха и вызовем функцию decribe
df_sorted = slice_and_dice(df, ['total_area', 'living_area', 'kitchen_area', 'price', 'rooms'])
df_sorted.describe()
```

```python
# Добавим еще одну функцию с измененными границаи среза согласно правила 3QR.
def slice_three_qr(df, columns=None):
    # Если столбцы не указаны, обрабатываем все столбцы
    if columns is None:
        columns = df.columns

    # Инициализируем маску для всех строк
    mask = pd.Series([True] * len(df), index=df.index)

    # Проходим по каждому столбцу
    for column in columns:
        # Проверяем, что столбец числовой и не булев
        if pd.api.types.is_numeric_dtype(df[column]) and not pd.api.types.is_bool_dtype(df[column]):
            # Преобразуем данные в числовой тип, заменяя некорректные значения на NaN
            df[column] = pd.to_numeric(df[column], errors='coerce')
            
            # Удаляем строки с NaN перед вычислением квартилей
            non_na_column = df[column].dropna()
            
            if not non_na_column.empty:
                # Вычисляем квартили
                Q1 = non_na_column.quantile(0.25)
                Q3 = non_na_column.quantile(0.75)
                
                # Вычисляем интерквартильный размах
                IQR = Q3 - Q1
                
                # Определяем границы выбросов
                lower_bound = Q1 - 3 * IQR
                upper_bound = Q3 + 3 * IQR
                
                # Обновляем маску для текущего столбца
                column_mask = (df[column] >= lower_bound) & (df[column] <= upper_bound)
                mask &= column_mask
            else:
                # Если столбец полностью пустой после удаления NaN, пропускаем его
                continue

    # Применяем итоговую маску ко всему DataFrame
    result_df = df[mask]

    return result_df
```
