# Анализ результатов A/B теста

## Описание проекта

Сегодня мы выступаем в роли продуктового аналитика крупного интернет-магазина. Вместе с отделом маркетинга был подготовлен список гипотез для увеличения выручки компании.



**Цели проекта**

Первая часть проекта направлена на определение приоритета гипотез для проведения экспериментов с целью максимизации ожидаемого улучшения ключевых бизнес-метрик.

Вторая часть посвящена анализу данных уже проведенного эксперимента, включающего информацию о посещаемости платформы, количестве заказов и выручке по группам.

Однако стоит упомянуть, что эксперимент включает несколько ключевых этапов подготовки, которые остаются за кадром:

1. Определение ключевых метрик и формулировка гипотез. Мы выбираем метрики, которые наиболее точно отражают цели эксперимента (например, конверсия, выручка или средний чек), и формулируем гипотезы на основе данных и бизнес-задач.
2. Выбор способа рандомизации и определение параметров выборки. Пользователи распределяются между группами А и В с использованием определенного метода рандомизации (например, по user ID или cookie).
3. Расчет необходимого размера выборки. Это важно для обеспечения статистической значимости результатов. Расчет выполняется с учетом мощности теста, уровня допустимой ошибки и предполагаемого эффекта.
4. Запуск эксперимента и сбор данных. Фиксируются действия пользователей, показатели тестируемых групп, а данные записываются в базу для последующего анализа.
5. Проверка валидности эксперимента. На этом этапе мы оцениваем корректность рандомизации, равномерность распределения пользователей между группами и отсутствие значимых аномалий в данных.
6. Расчет результатов и принятие решений. После завершения теста проводится анализ данных, проверяется статистическая значимость различий между группами, чтобы принять решение о внедрении изменений.


Мы находимся на финальном этапе — анализе результатов эксперимента. Наша задача — применить методы статистического анализа для проверки значимости различий между тестовой и контрольной группами. Мы также уделим внимание возможным ошибкам при тестировании, интерпретации результатов и оценке их влияния на ключевые метрики бизнеса.

**Задачи второй части проекта**
- Провести анализ данных, полученных в ходе A/B теста.
- Проверить статистическую значимость различий между группами.
- Интерпретировать результаты теста и сформулировать обоснованное решение о внедрении изменений.

**Описание данных.**

Информация по гипотезам: `hypothesis.csv`
- `Hypothesis` — краткое описание гипотезы;
- `Reach` — охват пользователей по 10-бальной шкале;
- `Impact` — влияние на пользователей по 10-бальной шкале;
- `Confidence` — уверенность в гипотезе по 10-бальной шкале;
- `Efforts` — затраты ресурсов на проверку гипотезы по 10-бальной шкале. Чем больше значение Efforts, тем дороже проверка гипотезы.


Информация по заказам: `orders.csv.`
- `transactionId` — идентификатор заказа;
- `visitorId` — идентификатор пользователя, совершившего заказ;
- `date` — дата, когда был совершён заказ;
- `revenue` — выручка заказа;
- `group` — группа A/B-теста, в которую попал заказ.


Информация по заказам: `visitors.csv`.
- `date` — дата;
- `group` — группа A/B-теста;
- `visitors` — количество пользователей в указанную дату в указанной группе A/B-теста

### Применим фреймворк ICE для приоритизации гипотез. Отсортируем их по убыванию приоритета.

ICE (Impact, Confidence, Effort) оценивает гипотезы по влиянию на бизнес (Impact), уровню уверенности в успехе (Confidence) и трудозатратам на реализацию (Effort). Каждая гипотеза получает баллы по этим критериям на основе экспертной оценки, и затем рассчитывается итоговый показатель:

**Impact (Влияние)**
- Оценивает, насколько сильно гипотеза может повлиять на ключевую метрику (например, выручку, конверсию, средний чек).
- Значение задается экспертно на основе предыдущего опыта, результатов исследований или аналогичных внедрений.
- Если гипотеза может увеличить выручку на 10% или более, ей дают высокий балл (например, 8-10).
- Если влияние ожидается минимальным (например, 1-2%), балл будет низким (1-3).
- В случае отсутствия данных используется интуитивная оценка на основе аналогий с другими гипотезами.

**Confidence (Уверенность)**
- Оценивает вероятность того, что гипотеза действительно принесет ожидаемый эффект.


Значение формируется на основе:
- Анализа данных (например, результатов прошлых тестов, исследования пользователей).
- Обоснованности гипотезы (есть ли доказательства, подтверждающие её эффективность).
- Если гипотеза опирается на A/B-тесты или аналитические исследования, её Confidence будет выше (например, 7-10).
- Если гипотеза основана на предположении без подкрепляющих данных, её Confidence будет низким (1-4).

**Effort (Усилие)**
- Оценивает, сколько ресурсов (времени, денег, человеческих усилий) потребуется на реализацию гипотезы.
- Баллы присваиваются в зависимости от сложности задачи:
- Низкие затраты: гипотезу можно внедрить за день-два или с минимальными усилиями — низкий балл (1-3).
- Высокие затраты: гипотеза требует длительной проработки, больших ресурсов — высокий балл (7-10).
- Значения Effort задаются путем консультаций с командой разработки или других исполнителей, которые оценивают трудозатраты.

```python
# Подсчитаем показатель ICE для гипотез
hypothesis['ice'] = (hypothesis['impact'] * hypothesis['confidence'] / hypothesis['efforts']).round(2)
hypothesis.sort_values(by='ice', ascending= False)
```

### Примените фреймворк RICE для приоритизации гипотез. Отсортируйте их по убыванию приоритета.

**Reach (Охват)**
- Оценивает, сколько пользователей будет затронуто гипотезой (например, процент трафика или количество пользователей, которые почувствуют эффект).


Например:
- Если гипотеза охватит 80% пользователей, Reach = 8.
- Если гипотеза затронет только 10% пользователей, Reach = 2.

```python
# Подсчитаем показатель RICE для гипотез
hypothesis['rice'] = hypothesis['impact'] * hypothesis['confidence']  * hypothesis['reach'] / hypothesis['efforts']
hypothesis.sort_values(by='rice', ascending= False)
```

### Построим график кумулятивной выручки по группам.

```python
from matplotlib.ticker import FuncFormatter

# Зададим функцию для форматирования больших чисел на оси Y
def human_format(x, _):
    if x >= 1_000_000:
        return f'{x / 1_000_000:.1f} M'  # Миллионы
    elif x >= 1_000:
        return f'{x / 1_000:.0f} k'      # Тысячи
    else:
        return str(int(x))

def draw_arrow(
      ax,
      tail_position,
      head_position,
      invert=True,
      radius=0.5,
      color=color_three,
      lw=0.5,
      head_width=4,
      head_length=8,
      tail_width=0.5
   ):
    # Настройки стрелки
    kw = dict(arrowstyle=f"Simple,tail_width={tail_width},head_width={head_width},head_length={head_length}", color=color, lw=lw)
    
    # Установка стиля соединения для стрелки в зависимости от параметра invert
    if invert:
        connectionstyle = f"arc3,rad=-{radius}"
    else:
        connectionstyle = f"arc3,rad={radius}"
    
    # Создание стрелки
    a = FancyArrowPatch(
        tail_position, head_position,
        connectionstyle=connectionstyle,
        **kw
    )
    
    # Добавление стрелки на график
    ax.add_patch(a)
    
# Создание графика
fig, ax = plt.subplots()
    
# График группы A
plt.plot(data['date'], data['revenue_a_cum'], label='A', color=color_two, linewidth=3)

# График группы B
plt.plot(data['date'], data['revenue_b_cum'], label='B', color=color_one, linewidth=3)

# Настройка оси Y
plt.gca().yaxis.set_major_formatter(FuncFormatter(human_format))

# Настройка оси X (вывод меток через одну дату)
plt.xticks(data['date'][::6])  # Настройка меток на оси Х

# Добавление заголовков и подзаголовков
title = '<ГРАФИК КУМУЛЯТИВНОЙ ВЫРУЧКИ>\n\nМы можем наблюдать изменения метрики для групп <TEST(B)> и <CONTROL(A)>'
fig_text(
    x=0.085, y=1.09, s=title,
    highlight_textprops=[
        {"color": '#263238FF', "fontsize": 18, "weight": "bold"},
        {"color": color_one, "weight": "bold"},
        {'color': color_two, "weight": "bold"}
    ]
)

# Подпись внизу
fig_text(
    x=0.60, y=0.01, s='<Источник данных: BEST E-COMMERCE PLATFORM>',
    highlight_textprops=[
        {"color": '#90A4AEFF', "fontsize": 12, "weight": "normal", 'style': 'italic'}
    ]
)

# Добавление текста с описанием
fig_text(
    x=0.32, y=0.61, s='<РЕЗКИЙ СКАЧОК! АНОМАЛИИ?>',
    highlight_textprops=[
        {"color": color_three,  "weight": "bold"}
    ]
)

# Добавление стрелки
highlight_date = pd.Timestamp('2019-08-18')
highlight_value = 3000000  # Замените на ваше значение

tail_position = (pd.Timestamp('2019-08-10'), 3000000 )
head_position = (highlight_date, highlight_value - 20,)  # Сместим стрелку вверх для наглядности

# Применяем функцию для построения стрелки
draw_arrow(ax, tail_position, head_position, tail_width=1, head_length=10, radius=0.1, color=color_three,)

# Подписи осей
plt.xlabel('Дата')
plt.ylabel('Сумма выручки')
plt.show()
```

### Посчитаем статистическую значимость различий в среднем количестве заказов на посетителя между группами по «сырым» данным.

- Примем уровень статистической значимости alpha = 0.05;
- Нулевая гипотеза H0, показатели между тестовой и контрольной группой не имеют статистически значимых различий.
- Альтернативная гипотзеа H1, группы имеют статистически значимые отличия.


```python
# Для подсчёта среднего количества заказов на посетителя сформируем выборки, в которых 
# для посетителей сделавших хотя бы один заказ будет указано количество заказов, а для остальных нули

sample_a = (pd.concat([transactions_by_a['transactions'],
                       pd.Series(0, index=np.arange(data['visitors_a'].sum() - len(transactions_by_a['transactions'])),
                                 name = 'transactions')], axis=0))

sample_b = (pd.concat([transactions_by_b['transactions'],
                      pd.Series(0, index=np.arange(data['visitors_b'].sum() - len (transactions_by_b['transactions'])),
                               name = 'transactions')], axis=0))


# Зададим уровень статистичкой значимости
alpha = 0.05

# Для оценки статистической значимости применим непараметрический критерий Манна-Уитни
_, pvalue = st.mannwhitneyu(sample_a, sample_b)

rate  = (sample_b.mean() /  sample_a.mean()-1)

if pvalue < alpha:
    print("Отклоняем нулевую гипотезу: распределение пользователей между группами отличается.")
else:
    print("Не удалось отклонить нулевую гипотезу: распределение пользователей равномерное.")
    
print(f"Вероятность получить такое или более экстремальное значение {round(pvalue,5)}")

print(f"Отличие между группами составляет  почти {round(rate * 100, 0)} %")
```

### Подсчитаем статистическую значимость в конверсии посетителя в покупателя (Z-test для пропорций)

- Примем уровень статистической значимости alpha = 0.05;
- Нулевая гипотеза H0, показатели между тестовой и контрольной группой не имеют статистически значимых различий.
- Альтернативная гипотзеа H1, группы имеют статистически значимые отличия.

```python
from statsmodels.stats.proportion import proportions_ztest
# Шаг 1: Уникальные покупатели (сделали хотя бы один заказ)
unique_buyers_a = orders[orders['group'] == 'A']['visitorid'].nunique()
unique_buyers_b = orders[orders['group'] == 'B']['visitorid'].nunique()

# Шаг 2: Общее количество посетителей
visitors_a = visitors[visitors['group'] == 'A']['visitors'].sum()
visitors_b = visitors[visitors['group'] == 'B']['visitors'].sum()

# Шаг 3: Пропорции (конверсии)
conversion_a = unique_buyers_a / visitors_a
conversion_b = unique_buyers_b / visitors_b

# Шаг 4: Z-тест для пропорций
successes = [unique_buyers_a, unique_buyers_b]  # Уникальные покупатели
samples = [visitors_a, visitors_b]  # Общее количество посетителей

z_stat, p_value = proportions_ztest(successes, samples)

print(f"P-value: {p_value:.3f}")

# Дополнительно: относительный прирост конверсии
relative_increase = (conversion_b / conversion_a - 1) * 100
print(f"Относительный прирост конверсии группы B к группе A: {relative_increase:.2f}%")
```

Аналогичные расчёты можно выполнить по формуле, учитывая что при нулевой гипотезе разность истинных средних равны 0.

```python
# Комбинированная пропорция успехов
p_combined = (unique_buyers_a + unique_buyers_b) / (visitors_a + visitors_b)
# разница пропорций
difference = conversion_b - conversion_a
# считаем статистику в ст.отклонениях стандартного нормального распределения
z_value = difference / np.sqrt(p_combined * (1 - p_combined) * (1/visitors_a + 1/visitors_b))

# задаем стандартное нормальное распределение (среднее 0, ст.отклонение 1)
distr = st.norm(0, 1)  
p_value = (1 - distr.cdf(abs(z_value))) * 2
print(f"Z-статистика: {z_value:.3f}")
print('p-значение: ', round(p_value,3))

if p_value < alpha:
    print('Отвергаем нулевую гипотезу: между долями есть значимая разница')
else:
    print(
        'Не получилось отвергнуть нулевую гипотезу, нет оснований считать доли разными'
    ) 
```

Мы зафиксировали статистически значимое отличие между тестовой и контрольной группой по показателю конверсиии из посетителя в покупателя на «сырых данных». Тестовая группа опережает контрольную на 15 %. Для чистоты эксперимента выполним такую же проверку и на «очищенных» данных.

#### Посчитаем статистическую значимость с применением бутстрапа.

```python
# Функция для бутстрапа
def bootstrap(data1, data2, iterations=10000):
    boot_means_diff = [] # Список для хранения разности средних
    for i in range(iterations): # Запускаем цикл на заданное количество итераций
        # Случайная выборка с возвращением для группы 1
        sample1 = np.random.choice(data1, size=len(data1), replace=True)
        # Случайная выборка с возвращением для группы 2
        sample2 = np.random.choice(data2, size=len(data2), replace=True)
        # Разность средних значений случайных выборок
        boot_means_diff.append(np.mean(sample2) - np.mean(sample1))
        # Преобразуем список в массив numpy для удобства анализа
    return np.array(boot_means_diff)

# Данные по средним чекам из групп
groupA = orders[orders['group'] == 'A']['revenue']
groupB = orders[orders['group'] == 'B']['revenue']

# Выполняем бутстрап
boot_diff = bootstrap(groupA, groupB)

# Рассчитываем p-value
observed_diff = np.mean(groupB) - np.mean(groupA)
pvalue = p_value = np.mean(np.abs(boot_diff) >= np.abs(observed_diff))


# Интерпретируем результаты
if pvalue < alpha:
    print("Отклоняем нулевую гипотезу: распределение пользователей между группами отличается.")
else:
    print("Не удалось отклонить нулевую гипотезу:" 
          "различия в среднем чеке между группами не являются статистически значимыми.")
    
print(f"Вероятность получить такое или более экстремальное значение {round(pvalue,5)}")

print(f"Отличие между группами составляет {round(observed_diff, 0)}")
```

Различие в p-value связано с тем, что:
- Манн-Уитни оценивает, как распределения перекрываются, и более устойчив к выбросам. Если данные двух групп сильно пересекаются, тест может дать высокое p-value, даже если средние заметно различаются.
- Бутстрап напрямую оценивает разницу в средних. Если выбросы искажают средние значения, это может привести к снижению точности оценки вероятности (p-value).

### Посчитаем статистическую значимость различий в среднем количестве заказов на посетителя между группами по «очищенным» данным. 

```python
# Для подсчёта среднего количества заказов на посетителя сформируем выборки с исключением «аномальных пользователей»
sample_a_filtered =  pd.concat(
    [transactions_by_a[~transactions_by_a['visitors'].isin(abnormal_users)]['transactions'],
    pd.Series(
        0,
        index=np.arange(
            data['visitors_a'].sum() - len(transactions_by_a['transactions'])
        ), name='orders')], axis=0)
     
sample_b_filtered =  pd.concat(
    [transactions_by_b[~transactions_by_b['visitors'].isin(abnormal_users)]['transactions'],
     pd.Series(
         0,
         index=np.arange(
             data['visitors_b'].sum() - len(transactions_by_b['transactions'])
       ), name='orders')], axis=0)

# Зададим уровень статистичкой значимости
alpha = 0.05

# Для оценки статистической значимости применим непараметрический критерий Манна-Уитни
_, pvalue = st.mannwhitneyu(sample_a_filtered, sample_b_filtered)

rate  = (sample_b_filtered.mean() /  sample_a_filtered.mean()-1)

if pvalue < alpha:
    print("Отклоняем нулевую гипотезу: распределение пользователей между группами отличается.")
else:
    print("Не удалось отклонить нулевую гипотезу: распределение пользователей равномерное.")
    
print(f"Вероятность получить такое или более экстремальное значение {round(pvalue,5)}")

print(f"Отличие между группами составляет  почти {round(rate * 100, 0)} %")
```

На очищенных данных отличие по среднему количеству заказов между тестовой и контрольной группой стало еще более заметным, а p-value стало еще более маленький. 

### Посчитаем статистическую значимость различий в конверсии из посетителя  в покупателя между группами по «очищенным» данным. 

```python
# Шаг 1: Уникальные покупатели (сделали хотя бы один заказ)
unique_buyers_a = orders.query("group == 'A' and visitorid not in @abnormal_users")['visitorid'].nunique()
unique_buyers_b = orders.query("group == 'B' and visitorid not in @abnormal_users")['visitorid'].nunique()

# Шаг 2: Общее количество посетителей
visitors_a = visitors[visitors['group'] == 'A']['visitors'].sum()
visitors_b = visitors[visitors['group'] == 'B']['visitors'].sum()

# Шаг 3: Пропорции (конверсии)
conversion_a = unique_buyers_a / visitors_a
conversion_b = unique_buyers_b / visitors_b

# Шаг 4: Z-тест для пропорций
successes = [unique_buyers_a, unique_buyers_b]  # Уникальные покупатели
samples = [visitors_a, visitors_b]  # Общее количество посетителей

z_stat, p_value = proportions_ztest(successes, samples)

print(f"P-value: {p_value:.3f}")

# Дополнительно: относительный прирост конверсии
relative_increase = (conversion_b / conversion_a - 1) * 100
print(f"Относительный прирост конверсии группы B к группе A: {relative_increase:.2f}%")
if p_value < alpha:
    print('Отвергаем нулевую гипотезу: между долями есть значимая разница')
else:
    print(
        'Не получилось отвергнуть нулевую гипотезу, нет оснований считать доли разными'
    ) 
```

На очищенных данных отличие по конверсии между тестовой и контрольной группой стало еще более заметным, а p-value стало еще более маленький.

### Посчитаем статистическую значимость различий в среднем чеке заказа между группами по «очищенным» данным. 

```python
# Применяем критерий Манна - Уитни
pvalue = st.mannwhitneyu(orders.query("group == 'A' and visitorid not in @abnormal_users")['revenue'],  
                       orders.query("group == 'B' and visitorid not in @abnormal_users")['revenue'])[1]

# Подсчитываем разницу между средними значениями по группам.
rate = (orders.query("group == 'B' and visitorid not in @abnormal_users")['revenue'].mean()
        / orders.query("group == 'A' and visitorid not in @abnormal_users")['revenue'].mean() - 1)

# Интерпретируем результаты
if pvalue < alpha:
    print("Отклоняем нулевую гипотезу: распределение пользователей между группами отличается.")
else:
    print("Не удалось отклонить нулевую гипотезу:"
          "различия в среднем чеке между группами не являются статистически значимыми.")
    
print(f"Вероятность получить такое или более экстремальное значение {round(pvalue,3)}")

print(f"Отличие между группами составляет  почти {round(rate * 100, 2)} %")
```

Результаты теста по очищенным данным свидетельствуют о том что разница в среднем чеке между группами незначительна, если исключить из исследования выбросы, то она состави менее 1%.

```python
# Функция для бутстрапа
def bootstrap(data1, data2, iterations=10000):
    boot_means_diff = []
    for i in range(iterations):
        sample1 = np.random.choice(data1, size=len(data1), replace=True)
        sample2 = np.random.choice(data2, size=len(data2), replace=True)
        boot_means_diff.append(np.mean(sample2) - np.mean(sample1))
    return np.array(boot_means_diff)

# Данные по средним чекам из групп
groupA = orders.query("group == 'A' and visitorid not in @abnormal_users")['revenue']
groupB = orders.query("group == 'B' and visitorid not in @abnormal_users")['revenue']

# Выполняем бутстрап
boot_diff = bootstrap(groupA, groupB)

# Рассчитываем p-value
observed_diff = np.mean(groupB) - np.mean(groupA)
p_value  = np.mean(np.abs(boot_diff) >= np.abs(observed_diff))

# Гистограмма распределения бутстрапированных разностей средних
sns.histplot(boot_diff, bins=50, kde=True, color=color_three, label='Bootstrap Distribution')

# Добавление наблюдаемой разницы средних
plt.axvline(observed_diff, color='red', linestyle='--', label=f'Observed Difference ({observed_diff:.2f})')

# Добавление линий для 2.5% и 97.5% квантилей (интервал)
ci_low, ci_high = np.percentile(boot_diff, [2.5, 97.5])
plt.axvline(ci_low, color=color_one, linestyle='--', label=f'2.5% Quantile ({ci_low:.2f})')
plt.axvline(ci_high, color=color_one, linestyle='--', label=f'97.5% Quantile ({ci_high:.2f})')
plt.text(x=observed_diff, y=plt.ylim()[1]*0.9, 
         s=f'P-value: {p_value:.3f}', color='red', fontsize=12)

# Подписи
# Добавление заголовков и подзаголовков
title = (
    '<БУТСТРЭП РАСПРЕДЕЛЕНИЕ>\n\n'
    'На графике отображены линиями <значения доверительного интервала> и <наблюдаемое различие>'
)
fig_text(
    x=0.086, y=1.03, s=title,
    highlight_textprops=[
        {"color": '#263238FF', "fontsize": 18, "weight": "bold"},
        {"color": color_one, "weight": "bold"},
        {'color': 'red', "weight": "bold"}
    ]
)

# Подпись внизу
fig_text(
    x=0.60, y=0.01, s='<Источник данных: BEST E-COMMERCE PLATFORM>',
    highlight_textprops=[
        {"color": '#90A4AEFF', "fontsize": 12, "weight": "normal", 'style': 'italic'}
    ]
)
plt.xlabel('Mean Difference' )
plt.ylabel('Frequency')
plt.show()
# Интерпретируем результаты
if p_value < alpha:
    print("Отклоняем нулевую гипотезу: распределение пользователей между группами отличается.")
else:
    print("Не удалось отклонить нулевую гипотезу:" 
          "различия в среднем чеке между группами не являются статистически значимыми.")
    
print(f"Вероятность получить такое или более экстремальное значение {round(p_value,5)}")

print(f"Отличие между группами составляет {round(observed_diff, 0)}")
```

Результаты бутстрэп теста по очищенным данным свидетельствуют о том что разница в среднем чеке между группами незначительна, а вероятность получить такое или более экстремальное значение составляет более 90 %.

### Смоделируем проблему подглядывания (peeking problem)

- Используем данные с кумулятивными значениями по дням.
- Для каждого дня выполняем Mann-Whitney U-тест на сравнение средних чеков в группах.
- Сохраняем значения p-value для каждого дня, чтобы потом построить график.

```python
# Сортируем данные по дате
orders = orders.sort_values(by='date')

# Уникальные даты
unique_dates = orders['date'].unique()

# Список для хранения p-value
p_values = []

# Цикл по дням
for current_date in unique_dates:
    # Данные за текущий день и предыдущие дни (кумулятивно)
    groupA = orders[(orders['group'] == 'A') & (orders['date'] <= current_date)]['revenue']
    groupB = orders[(orders['group'] == 'B') & (orders['date'] <= current_date)]['revenue']
    
    # Выполняем Mann-Whitney U-тест
    _, p_value = st.mannwhitneyu(groupA, groupB, alternative='two-sided')
    p_values.append(p_value)

    
    
    
# Визуализация p-value
fig, ax = plt.subplots()
plt.plot(unique_dates, p_values, label='P-value', linewidth=3, color = color_three)
plt.axhline(0.05, color=color_one, linestyle='--', label='α = 0.05')

#Добавление заголовков и подзаголовков
title = (
    '<Динамика P-value при ежедневном тестировании средних чеков>\n\n'
    'Мы можем наблюдать изменения показателя p-value отностительно уровня <статистической значимости>'
)
fig_text(
    x=0.08, y=1.09, s=title,
    highlight_textprops=[
        {"color": '#263238FF', "fontsize": 18, "weight": "bold"},
        {"color": color_one, "weight": "bold"}
    ]
)

# Добавление текста с описанием
fig_text(
    x=0.40, y=0.42, s='<Здесь значение p-value почти> \n<перешагнуло уровень>\n'
                      '<статистической значимости>',
    highlight_textprops=[
        {"color": 'black', 'alpha' : 0.5, "weight": "bold"},
        {"color": 'black', 'alpha' : 0.5, "weight": "bold"},
        {"color": 'black', 'alpha' : 0.5, "weight": "bold"}
    ]
)

# Добавление стрелки
highlight_date = pd.Timestamp('2019-08-08')
highlight_value = 0.1  # Замените на ваше значение

tail_position = (pd.Timestamp('2019-08-17'), 0.2301 )
head_position = (highlight_date, highlight_value)  # Сместим стрелку вверх для наглядности

# Применяем функцию для построения стрелки
draw_arrow(ax, tail_position, head_position, tail_width=1, head_length=10, invert=False, radius=0.1,)

plt.xlabel('Дата')
plt.ylabel('P-value')


plt.show()
```
